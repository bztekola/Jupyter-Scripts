* This script looks at different aspects of the Fab process, such as work order count, machine use by time, machine yield by part, reworks and more.
* The aim here is to develope a metric or set of metrics that will encapsulate the Fab experience and convey all progress and shortcomings.
* So far this script only contains structured data with charts, but through ongoing colloboration between Stats and Fab we will turn these stats into a measure of progress modeled after the well known [Overall Equipment Effectivness Metric](https://en.wikipedia.org/wiki/Overall_equipment_effectiveness)

```python
elapsed_quantile = 0.90

fab = processes[processes['process group'] == 'fabricate'].copy()

fab.drop(['process group','process order'], 1, inplace = True)
if fab[fab['process break timestamp'].isnull() == False].empty:
    fab.drop(['process break timestamp'], 1, inplace = True)
    
fab.drop(fab[fab['time start'] == '0000-00-00 00:00:00'].index, inplace = True)
fab.drop(fab[fab['time end'] == '0000-00-00 00:00:00'].index, inplace = True)    

fab['time elapsed'] = fab['time end'] - fab['time start']
fab['time elapsed (seconds)'] = fab['time elapsed'].dt.days/24/60/60 + fab['time elapsed'].dt.seconds

fab['qty out'] = np.where(fab['qty out'] > fab['qty in'], fab['qty in'], fab['qty out'])
fab['machine yield'] = fab['qty out']/fab['qty in']

print('you had {} nulls in your fab table before removing them\n'.format(len(fab[fab.isnull().any(1)])))
fab.dropna(inplace = True)

fab['year and month start'] = pd.to_datetime(fab['time start'].dt.year.map(str) + '-' + fab['time start'].dt.month.map(str))
fab['year and month start'] = [str(x)[:7] for x in fab['year and month start']]

print('count of all lines: {:,.0f}'.format(len(fab)))
print('{:,.0f}% of lines have an elapsed time under or at {}'.format(elapsed_quantile * 100, fab['time elapsed'].quantile(elapsed_quantile)))
print('{:,.0f}% of lines have an elapsed time over {}'.format((1 - elapsed_quantile) * 100, fab['time elapsed'].quantile(1 - elapsed_quantile)))
print('keep only what\'s under the {:,.0f}th percentile because anything over {} is most likely an error'.format(elapsed_quantile * 100, fab['time elapsed'].quantile(elapsed_quantile)))

fab_drop = fab[fab['time elapsed'] >= fab['time elapsed'].quantile(elapsed_quantile)].sort_values('time elapsed', ascending = False)

fab.drop(fab[fab['time elapsed'] >= fab['time elapsed'].quantile(elapsed_quantile)].index, inplace = True)
```

# [Overall Equipment Effectivness Metric](https://en.wikipedia.org/wiki/Overall_equipment_effectiveness)
* OEE is calculated with the formula (Availability) times (Performance) times (Quality)
* Availability = operating time / scheduled time
    * where operating time is scheduled_time - scheduled_downtime - unscheduled_downtime (e.g. breakdowns)
* Performance (Productivity) = (Parts Produced * Ideal Cycle Time) / Operating time
* Quality = (Units produced - Defective units) / (Units produced)

### Availability
* Operating Time / Scheduled Time
* Note that the Availability Metric is a pure measurement of Uptime that is designed to exclude the effects of Quality, Performance, and Scheduled Downtime Events.

```python
hours_per_day = 6
days_per_week = 5
weeks_per_month = 4
scheduled = hours_per_day * days_per_week * weeks_per_month

availability = fab_by_month[['total time elapsed (hours)']].copy()
availability['availability %'] = availability['total time elapsed (hours)']/scheduled

print('If we assume that Fab\'s machine can run {} hours per day, {} days per week, \nwe have {} hours per month.'.format(hours_per_day, days_per_week, scheduled))
print('Based on this scheduled running time, Fab is running their machine at {:,.0f}% \ncapacity, on average each month.'.format(availability['availability %'].mean() * 100))
```

```python
fig, ax = plt.subplots(figsize = (20,5))

availability['availability %'].plot(color = 'purple',
                                    ax = ax,
                                    kind = 'bar',
                                    alpha = 0.7,
                                    edgecolor = 'black')
ax.plot([availability['availability %'].mean()] * len(availability),
       color = 'red',
       ls = '--')
ax.set_xticks(np.arange(len(availability)))
ax.set_xticklabels(availability.index,
                  rotation = 45)
ax.set_title('Machine Utilization % Based on a Scheduled {} hours per Month'.format(scheduled),
            fontsize = 25)
ax.set_xlabel('')
ax.set_ylim(0,1.50)
vals = ax.get_yticks()
ax.set_yticklabels(['{:,.0f}%'.format(x * 100) for x in vals])
ax.legend(['Average % per Month','Utilization %'],
         fontsize = 15)
plt.show()
```

```python
fab_revenue = pd.read_sql(
'''
SELECT
AVG(A.revenue) AS 'avg monthly fab revenue'
FROM
(SELECT
DATE_FORMAT(o.date_purchased, '%Y-%m') AS 'year and month',
SUM((op.products_quantity - op.products_quantity_free) * op.products_price) AS revenue
FROM orders_products op
JOIN orders o ON op.orders_id = o.orders_id
WHERE op.part_id IN (SELECT part_id FROM skus WHERE bom_type = 'pnp')
AND DATE(o.date_purchased) BETWEEN '2017-01-01' AND ' '''+ date_end +''' '
GROUP BY DATE_FORMAT(o.date_purchased, '%Y-%m')) A
''', db)
```

```python
print('The average monthly Fab revenue since Jan 2017 is ${:,.0f}.'.format(fab_revenue['avg monthly fab revenue'][0]))

rev_per_unit = fab_revenue['avg monthly fab revenue'][0]/(availability['availability %'].mean() * 100)
print('So 1 "unit" of capacity is equal to ${:,.0f}.'.format(rev_per_unit))

additional_per_month = (1 - availability['availability %'].mean()) * rev_per_unit
print('Which means we could be making ${:,.0f} more per month if we run the machine at full capacity.'.format(additional_per_month))
print('Which is ${:,.0f} more per year.'.format(additional_per_month * 12))
```

### Performance
* (Parts Produced * Ideal Cycle Time) / Operating time
* Note that the Performance Metric is a pure measurement of speed that is designed to exclude the effects of Quality and Availability.
* Because we don't log downtime, scheduled or otherwise, this metric will be the same as the Availability metric.
* So let's skip this one for now

### Quality
* (Units produced - Defective units) / (Units produced)
* Note that the Quality portion of the OEE Metric represents the Good Units produced as a percentage of the Total Units Started

```python
# these are the groups under which we will measure rejects
groups = ['prep','processStockSubtract','qa','secondary','test'] # and "rework", but do this separately

quality = processes[processes['process group'].isin(groups)].copy()

quality.drop(quality[quality['time start'] == '0000-00-00 00:00:00'].index, inplace = True)
quality.drop(quality[quality['time end'] == '0000-00-00 00:00:00'].index, inplace = True)
quality['time start'] = pd.to_datetime(quality['time start'])
quality['time end'] = pd.to_datetime(quality['time end'])

quality['year and month start'] = pd.to_datetime(quality['time start'].dt.year.map(str) + '-' + quality['time start'].dt.month.map(str))
quality['year and month start'] = [str(x)[:7] for x in quality['year and month start']]

quality.rename(columns = {'qty in':'total qty'}, inplace = True)
```

```python
fig, ax = plt.subplots(figsize =(20,5))

quality_by_month = quality.groupby('year and month start').agg({'total qty':'sum',                                             'qty reject':'sum'})
quality_by_month['quality %'] = (quality_by_month['total qty'] - quality_by_month['qty reject'])/quality_by_month['total qty']

quality_by_month['quality %'].plot(color = 'green',
                                   ax = ax,
                                   kind = 'bar',
                                   alpha = 0.7,
                                   edgecolor = 'black')
ax.plot([quality_by_month['quality %'].mean()] * len(quality_by_month),
       color = 'red',
       ls = '--')
ax.set_xticks(np.arange(len(quality_by_month)))
ax.set_xticklabels(quality_by_month.index,
                  rotation = 45)
ax.set_title('Quality % Based on a Scheduled {} hours per Month'.format(scheduled),
            fontsize = 25)
ax.set_xlabel('')
ax.set_ylim(0.90,1.05)
vals = ax.get_yticks()
ax.set_yticklabels(['{:,.0f}%'.format(x * 100) for x in vals])
ax.legend(['Average % per Month','Quality %'],
         fontsize = 15)
plt.show()
```

# Applying OEE to real life
* The whole point of the OEE metric is to take a bunch of small metrics and mash them together to create one big metric that gives you an accurate description of the state of everything.
* We can take the general framework of the OEE metric and tailor it to our needs.
* Obviously not every step of the Fab experience is equally important, so we can establish a rating system by order of importance, and weight each OEE-step.
* Below is an example of this, the actual weights will be up to you, I just chose some based on what I know about Fab and what I think you want.

```python
avail_rate = 5/10 
quality_rate = 4/10
rework_rate = 1/10

first = availability.loc[date_end[:7], 'availability %']
second = quality_by_month.loc[date_end[:7], 'quality %']
third = (1 - reworks_by_month.loc[date_end[:7], 'rework reject %'])

print('your OEE metric for ' + pretty_cur + ' is:')
print('{:,.1f}%'.format(((avail_rate * first) + (quality_rate * second) + (rework_rate * third)) * 100))

print('\nthe break down is: ')
print('availability: {:,.1f}%'.format(first * 100))
print('quality: {:,.1f}%'.format(second * 100))
print('rework success: {:,.1f}%'.format(third * 100))
```

# Data sheet for Fab production meeting

```python
data_sheet_date_end = '2018-02'
data_sheet_write = 'yes'

#===================================================================================================================

wo_data_sheet = wo.groupby('year and month created').agg({'wo id':'count',
                                                          'part id':'nunique',
                                                          'wo qty':'sum',
                                                          'wo qty out':'sum'}).rename(columns = {'wo id':'wo count',
                                                                                                'part id':'count of unique PNs'})
wo_data_sheet['wo yield %'] = wo_data_sheet['wo qty out']/wo_data_sheet['wo qty']
wo_data_sheet.drop(['wo qty','wo qty out'], 1, inplace = True)
print(wo_data_sheet.index.min())

wo_data_sheet_mean = pd.DataFrame(wo_data_sheet.mean()).reset_index().rename(columns = {0:'monthly avg','index':'wo data'})
print(wo_data_sheet_mean)

#===================================================================================================================

fab_data_sheet = fab[fab['year and month start'] <= data_sheet_date_end].groupby('year and month start').agg({'qty in':'sum',
                                                                                                              'qty out':'sum',
                                                                                                              'time elapsed (seconds)':'sum',
                                                                                                              'wo id':'count'}).rename(columns = {'wo id':'wo count',
                                                                                                                                                 'time elapsed (seconds)':'time elapsed'})
fab_data_sheet['machine yield'] = fab_data_sheet['qty out']/fab_data_sheet['qty in']
fab_data_sheet['qty per wo'] = fab_data_sheet['qty in']/fab_data_sheet['wo count']
fab_data_sheet['time per wo'] = fab_data_sheet['time elapsed']/fab_data_sheet['wo count']
fab_data_sheet.drop(['qty in','qty out','wo count'], 1, inplace = True)

fab_data_sheet_mean = pd.DataFrame(fab_data_sheet.mean()).rename(columns = {0:'monthly avg','index':'fabrication data'})

print('\n' + fab_data_sheet.index.min())

def time(x):
    m, s = divmod(x, 60)
    h, m = divmod(m, 60)
    d = h // 24
    h2 = h - 24 * d
    return '%i day(s) %i:%i:%i' % (d, h2, m, s)

fab_data_sheet_mean['monthly avg'].loc['time elapsed'] = time(fab_data_sheet_mean['monthly avg'].loc['time elapsed'])
fab_data_sheet_mean['monthly avg'].loc['time per wo'] = time(fab_data_sheet_mean['monthly avg'].loc['time per wo'])
fab_data_sheet_mean.reset_index(inplace = True)
fab_data_sheet_mean.rename(columns = {'index':'fabrication data'}, inplace = True)

print(fab_data_sheet_mean)

#===================================================================================================================

reworks_data_sheet = reworks[reworks['year and month start'] <= data_sheet_date_end].groupby('year and month start').agg({'qty fixed':'sum',
                                                                                             'total qty':'sum',
                                                                                             'time elapsed':'sum',
                                                                                             'part id':'nunique',
                                                                                             'wo id':'count'}).rename(columns = {'part id':'count of unique PNs',
                                                                                                                                'wo id':'count of wos containing reworks'})
reworks_data_sheet['rework success %'] = reworks_data_sheet['qty fixed']/reworks_data_sheet['total qty']
reworks_data_sheet['time per rework session'] = reworks_data_sheet['time elapsed']/reworks_data_sheet['count of wos containing reworks']
reworks_data_sheet.drop(['qty fixed','total qty','time elapsed'], 1, inplace = True)

print('\n' + reworks_data_sheet.index.min())

reworks_data_sheet_mean = pd.DataFrame(reworks_data_sheet.mean()).reset_index().rename(columns = {0:'monthly avg','index':'rework data'})

print(reworks_data_sheet_mean)

#===================================================================================================================

print('\n' + availability.index.min())
print(availability['availability %'].mean())

#===================================================================================================================

if data_sheet_write == 'yes':
    
    writer = pd.ExcelWriter('Beeps Metric Overhaul Data Sheet for March 07, 2018 Meeting.xlsx')
    
    startrow = 2
    
    wo_data_sheet_mean.to_excel(writer, 'data', startrow = startrow, index = False)
    fab_data_sheet_mean.to_excel(writer, 'data', startrow = startrow, startcol = len(wo_data_sheet_mean.columns) + 2, index = False)
    reworks_data_sheet_mean.to_excel(writer, 'data', startrow = startrow, startcol = len(wo_data_sheet_mean.columns) + len(fab_data_sheet_mean.columns) + 4, index = False)
    
    writer.save()
```
