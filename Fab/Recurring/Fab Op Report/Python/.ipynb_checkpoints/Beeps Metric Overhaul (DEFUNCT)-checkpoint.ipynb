# Get Fab data sans Rework data
* we query twice
* main_01 is where process_class = 'Fabricate' and GROUP BY wo_id
* main_02 is where process_class != 'Fabicate' and no GROUP BY
* look [here](https://volcano.adafruit.com/volcano/ada_plm.php#/Production/WorkOrders/17/WorkOrderProfile?sub=0) at wo_id = 17
* in PLM there is only one entry for when the process is 'Fabricate'
* but in SQL there are several entries and in these entries spa_qty_in is x1
* this makes me think that some sort of fabricate-testing process was going on which caused these mutliple lines
* there are many other entries in the db that are like this
* for the sake of this analysis I don't care about any of this
* but if I leave all of this in then when I join other tables to processes, I will get multuple rows for when process_class = 'Fabricate'
* and this will mess me up
* so these two queries, concatenated into one data frame, will solve this problem for me

```python
main_01 = pd.read_sql(
'''
SELECT

p1.wo_id,
wo.wo_id_string,
wo.sku_id,
k.part_id,
wo.wo_date_created,
wo.wo_date_scheduled,
wo.wo_date_active,
wo.wo_date_completed,
wo.wo_qty,
wo.wo_qty_out,
wo.wo_status,
wo.wo_substatus,
p1.process_id,
p1.process_name,
p1.process_class,
CASE WHEN p1.qty_in < p1.qty_out THEN wo.wo_qty ELSE p1.qty_in END AS qty_in,
p1.qty_out,
CASE WHEN p1.qty_in < p1.qty_out THEN wo.wo_qty ELSE p1.qty_in END - p1.qty_out AS qty_reject,
p1.process_order,

CASE WHEN p1.qty_in < p1.qty_out THEN wo.wo_qty ELSE p1.qty_in END AS spa_qty_in,
p1.qty_out AS spa_qty_out,
CASE WHEN p1.qty_in < p1.qty_out THEN wo.wo_qty ELSE p1.qty_in END - p1.qty_out AS spa_qty_reject,
p1.time_start,
p1.time_end,
p2.spa_time_start,
p2.spa_time_end,
p2.spa_admin

FROM processes p1

JOIN sp_actions p2 ON p1.process_id = p2.process_id
JOIN work_orders wo ON p1.wo_id = wo.wo_id

JOIN
(SELECT
sku_id,
part_id
FROM skus
GROUP BY sku_id) k ON wo.sku_id = k.sku_id

WHERE p1.process_name = 'Fabricate'
GROUP BY p1.wo_id
ORDER BY p1.wo_id, p1.process_order
''', db)

#=====

main_02 = pd.read_sql(
'''
SELECT

p1.wo_id,
wo.wo_id_string,
wo.sku_id,
k.part_id,
wo.wo_date_created,
wo.wo_date_scheduled,
wo.wo_date_active,
wo.wo_date_completed,
wo.wo_qty,
wo.wo_qty_out,
wo.wo_status,
wo.wo_substatus,
p1.process_id,
p1.process_name,
p1.process_class,
CASE WHEN p1.qty_in < p1.qty_out THEN wo.wo_qty ELSE p1.qty_in END AS qty_in,
p1.qty_out,
CASE WHEN p1.qty_in < p1.qty_out THEN wo.wo_qty ELSE p1.qty_in END - p1.qty_out AS qty_reject,
p1.process_order,

p2.spa_qty + p2.spa_qty_reject AS spa_qty_in,
p2.spa_qty AS spa_qty_out,
p2.spa_qty_reject,
p1.time_start,
p1.time_end,
p2.spa_time_start,
p2.spa_time_end,
p2.spa_admin

FROM processes p1

JOIN sp_actions p2 ON p1.process_id = p2.process_id
JOIN work_orders wo ON p1.wo_id = wo.wo_id

JOIN
(SELECT
sku_id,
part_id
FROM skus
GROUP BY sku_id) k ON wo.sku_id = k.sku_id

WHERE p1.process_name != 'Fabricate'
ORDER BY p1.wo_id, p1.process_order
''', db)

main = pd.concat([main_01, main_02])
```

# Get Rework data
* we will deal with these metrics separately from total OEE

```python
reworks = pd.read_sql(
'''
SELECT

wo.wo_id,
wo.wo_id_string,
wo.sku_id,
k.part_id,
wo.wo_date_created,
wo.wo_date_scheduled,
wo.wo_date_active,
wo.wo_date_completed,
wo.wo_qty,
wo.wo_qty_out,
wo.wo_status,
wo.wo_substatus,
p.process_id,
CASE WHEN p.process_class = 'processRework' THEN 'Rework' END AS process_name,
p.process_class,
p.qty_in,
p.qty_out,
p.qty_in - p.qty_out AS qty_reject,
p.process_order,
sp.spa_qty + sp.spa_qty_reject AS spa_qty_in,
sp.spa_qty AS spa_qty_out,
sp.spa_qty_reject,
p.time_start,
p.time_end,
sp.spa_time_start,
sp.spa_time_end,
sp.spa_admin

FROM processes p
JOIN work_orders wo ON wo.wo_id = p.rework_id
JOIN sp_actions sp ON p.process_id = sp.process_id
JOIN
(SELECT
sku_id,
part_id
FROM skus
GROUP BY sku_id) k ON wo.sku_id = k.sku_id
''', db)

reworks['process_order'] = np.where(reworks['process_order'] == 0, 6, 0)
```

# Concatenate Fab and Rework data

```python
fab = pd.concat([main, reworks])
```

### Fill in time_start for reworks
* reworks currently only have an spa_time_start
* we need them to have a time_start so we can use a groupby later on
* reworks have the maximum process_order, that is, I set the process order to be 6 earlier on in this script
* so if we sort by wo_id, and then by process_order, and then front fill any NaNs we should be good

```python
fab.sort_values(['wo_id','process_order'], inplace = True)

fab['time_start'] = fab['time_start'].fillna(method = 'ffill')
fab['time_end'] = fab['time_end'].fillna(method = 'ffill')
```

# Clean up
* keep only work orders that are completed
* Remove where wo_date_created is in current month
* add is_rework columns for future groupbys
* convert strings to datetimes

```python
fab['time_start'] = pd.to_datetime(fab['time_start'])
fab['time_end'] = pd.to_datetime(fab['time_end'])

fab['wo_date_created'] = pd.to_datetime(fab['wo_date_created'])
fab['wo_date_scheduled'] = pd.to_datetime(fab['wo_date_scheduled'])

fab = fab[fab['wo_status'] == 'completed']
fab = fab[fab['wo_date_created'] <= pd.to_datetime(end_time)]

fab['is_rework'] = np.where(fab['process_class'] == 'processRework', 1, 0)
fab['wo_created_date_format'] = pd.to_datetime((fab['wo_date_created'].dt.year).map(str) + '-' + (fab['wo_date_created'].dt.month).map(str))
```

# Metrics to measure success

[Overall Equipment Effectiveness](https://en.wikipedia.org/wiki/Overall_equipment_effectiveness)
* OEE is calculated with the formula (Availability) times (Performance) times (Quality)
* Availability = operating time / scheduled time
    * where operating time is scheduled_time - scheduled_downtime - unscheduled_downtime (e.g. breakdowns)
* Performance (Productivity) = (Parts Produced * Ideal Cycle Time) / Operating time
* Quality = (Units produced - Defective units) / (Units produced)

# Availability
* Availability = operating time / scheduled time
* or Loading = scheduled Time / calendar Time

```python
avail = fab.copy()
avail = avail[avail['time_start'] < pd.to_datetime(start_time)]
avail = avail[avail['process_class'] == 'processFabricate']
avail.drop_duplicates('wo_id', keep = 'first', inplace = True)
avail = avail[['wo_id','wo_id_string','sku_id','part_id','time_start','time_end','spa_admin']]

avail['time_start'] = avail['time_start'].astype(dt.timedelta)
avail['time_end'] = avail['time_end'].astype(dt.timedelta)
avail['run_time'] = avail['time_end'] - avail['time_start']
avail['run_time_in_min'] = avail['run_time'].dt.seconds//60.0

# week begins on monday and ends on sunday
avail['week'] = avail['time_start'].dt.week
avail['year'] = avail['time_start'].dt.year
avail['day_start'] = avail['time_start'].dt.weekday_name
avail['day_end'] = avail['time_end'].dt.weekday_name

print ('WO ID dupes: {:,.0f}'.format(len(avail.index[avail['wo_id'].duplicated()])))
```

```python
print ('count of weekend run-days: {:,.0f}'.format(len(avail.index[(avail['day_start'].isin(['Saturday','Sunday'])) | (avail['day_end'].isin(['Saturday, Sunday']))])))
```

### availability per week
* if availability percentage exceeds 100%, replace it with 100%
* we want to look at this by 5 day work week
* any availability percentage over 100% suggests that the machines ran over the weekend

### Here we define "Availability" to be actual total run time per week divided by the theoretical maximum run time per week
* this theoretical max time is up to for debate
* see "pick and place options" above

```python
avail_by_week = avail.groupby(['year','week']).agg({'run_time_in_min':'sum'})
avail_by_week['availability'] = avail_by_week['run_time_in_min']/time

avail_by_week_over = avail_by_week.copy()
print ('count of weeks above 100%: {:,.0f}'.format(len(avail_by_week_over.index[avail_by_week_over['availability'] > 1])))

avail_by_week['availability'] = np.where(avail_by_week['availability'] > 1, 1, avail_by_week['availability'])
avail_by_week.tail()
```

* we want to look at this by 5 day work week
* any performance percentage over 100% suggests that the machines ran over the weekend

```python
perf = fab.copy()
perf = perf[perf['time_start'] < pd.to_datetime(start_time)]
perf = perf[perf['process_class'] == 'processFabricate']
perf.drop_duplicates('wo_id', keep = 'first', inplace = True)
perf = perf[['wo_id','wo_id_string','sku_id','part_id','time_start','time_end','wo_qty', 'spa_qty_in']]

perf['time_start'] = perf['time_start'].astype(dt.timedelta)
perf['time_end'] = perf['time_end'].astype(dt.timedelta)
perf['run_time'] = perf['time_end'] - perf['time_start']
perf['run_time_in_min'] = perf['run_time'].dt.seconds//60.0
perf['spa_qty_in'] = [float(x) for x in perf['spa_qty_in']]
perf['run_time_in_min'] = perf['run_time_in_min'].replace(0,1)

# week begins on monday and ends on sunday
perf['week'] = perf['time_start'].dt.week
perf['year'] = perf['time_start'].dt.year
perf['day'] = perf['time_start'].dt.weekday_name

print ('WO ID dupes: {:,.0f}'.format(len(perf.index[perf['wo_id'].duplicated()])))
```

### get parts per min per sku
* this is total units, not just good units

```python
parts_per_min = perf.groupby(['sku_id']).agg({'run_time_in_min':'sum','wo_qty':'sum'}).reset_index()
parts_per_min['min_per_part'] = parts_per_min['run_time_in_min']/parts_per_min['wo_qty']

perf_02 = pd.merge(perf, parts_per_min[['sku_id','min_per_part']], on = 'sku_id')

perf_02['time_to_produce_total_parts'] = perf_02['wo_qty'] * perf_02['min_per_part']
```

### Here we define "Performance" as the actual total time to Fab per week divided by the theoretical maximum time per week
* see "pick and place options" above for the max time

```python
perf_by_week = perf_02.groupby(['year','week']).agg({'time_to_produce_total_parts':'sum'})
perf_by_week['performance'] = perf_by_week['time_to_produce_total_parts']/time

perf_by_week_over = perf_by_week.copy()
print ('count of weeks above 100%: {:,.0f}'.format(len(perf_by_week_over.index[perf_by_week_over['performance'] > 1])))

perf_by_week['performance'] = np.where(perf_by_week['performance'] > 1, 1, perf_by_week['performance'])

perf_by_week.tail()
```

```python
perf_by_week['performance'].nlargest(n = 2)
```

```python
fig = plt.figure(figsize = (20,5))
ax = fig.add_subplot(1,1,1)

x = np.arange(0, len(perf_by_week.index))
y = perf_by_week['performance']

ax.plot(x, y)
ax.set_xticks([])
vals = ax.get_yticks()
ax.set_yticklabels(['{:,.0f}%'.format(x * 100) for x in vals])

plt.title('Performance and Productivity Over Time', fontsize = 15)
plt.ylabel('Performance and Productivity', fontsize = 15)
plt.xlabel('Weeks', fontsize= 15)
plt.legend(['Performance and Productivity'], loc = 'best', fontsize = 15)
plt.grid()
plt.show()
```

# Total Quality
* Quality = (Units produced - defective units) / (Units produced)

```python
qual = fab.copy()
qual = qual[qual['time_start'] < pd.to_datetime(start_time)]
qual = qual[qual['process_class'] == 'processFabricate']
qual.drop_duplicates('wo_id', keep = 'first', inplace = True)
qual = qual[['wo_id','wo_id_string','sku_id','spa_qty_in','wo_qty', 'wo_qty_out','time_start']]

qual['week'] = qual['time_start'].dt.week
qual['year'] = qual['time_start'].dt.year
qual['day'] = qual['time_start'].dt.weekday_name

print ('WO ID dupes: {:,.0f}'.format(len(qual.index[qual['wo_id'].duplicated()])))

qual_by_week = qual.groupby(['year','week']).agg({'wo_qty':'sum','wo_qty_out':'sum'})
qual_by_week['total_quality'] = qual_by_week['wo_qty_out']/qual_by_week['wo_qty']

qual_by_week.tail()
```

```python
fig = plt.figure(figsize = (20,5))
ax = fig.add_subplot(1,1,1)

x = np.arange(0, len(qual_by_week.index))
y = qual_by_week['total_quality']

ax.plot(x, y)
ax.set_xticks([])
vals = ax.get_yticks()
ax.set_yticklabels(['{:,.0f}%'.format(x * 100) for x in vals])

plt.title('Total Quality Over Time', fontsize = 15)
plt.ylabel('Total Quality', fontsize = 15)
plt.xlabel('Weeks', fontsize= 15)
plt.legend(['Total Quality'], loc = 'best', fontsize = 15)
plt.grid()
plt.show()
```

# Rework Quality
### measure this separately from total OEE
* there are some intricacies here
* say you Fab x100 units
* x25 do not pass and are rejected
* someone attempts to rework only x20 and is successful
* this would be 20/20 = 100% rework percentage
* NOT 20/25 = 80%

```python
oee_rework = fab.copy()
oee_rework = oee_rework[oee_rework['time_start'] < pd.to_datetime(start_time)]
oee_rework = oee_rework[oee_rework['process_class'] == 'processRework']

oee_rework['week'] = oee_rework['time_start'].dt.week
oee_rework['year'] = oee_rework['time_start'].dt.year
oee_rework['day'] = oee_rework['time_start'].dt.weekday_name

print ('WO ID dupes: {:,.0f}'.format(len(oee_rework.index[oee_rework['wo_id'].duplicated()])))
print ('dupes are ok here!')

# use spa_qty because that's how many units were brought in to rework
# this is not the total number of units that were rejected and need to be reworked
# this is the number of units taken to be rework and then reworked
reworks_by_week = oee_rework.groupby(['year','week']).agg({'spa_qty_in':'sum','spa_qty_out':'sum'})
reworks_by_week['rework_success'] = reworks_by_week['spa_qty_out']/reworks_by_week['spa_qty_in']
reworks_by_week.fillna(0, inplace = True)

reworks_by_week.tail()
```

```python
fig = plt.figure(figsize = (20,5))
ax = fig.add_subplot(1,1,1)

x = np.arange(0, len(reworks_by_week.index))
y = reworks_by_week['rework_success']

ax.plot(x, y)
ax.set_xticks([])
vals = ax.get_yticks()
ax.set_yticklabels(['{:,.0f}%'.format(x * 100) for x in vals])

plt.title('Rework Success Over Time', fontsize = 15)
plt.ylabel('Rework Success', fontsize = 15)
plt.xlabel('Weeks', fontsize= 15)
plt.legend(['Rework Success'], loc = 'best', fontsize = 15)
plt.grid()
plt.show()
```

# OEE Total
* [Overall Machine Efficiency](https://en.wikipedia.org/wiki/Overall_equipment_effectiveness)
* OEE is calculated with the formula (Availability) times (Performance) times (Quality)

```python
dates_main = fab.copy()
dates_main['year'] = dates_main['time_start'].dt.year
dates_main['week'] = dates_main['time_start'].dt.week
dates_main = dates_main[dates_main['time_start'] < start_time]

dates = dates_main.groupby(['year','week']).agg({'wo_id':'count'}).drop('wo_id',1)
```

```python
oee_total_01 = pd.merge(dates, avail_by_week[['availability']], how = 'left', left_index = True, right_index = True)
oee_total_02 = pd.merge(oee_total_01, perf_by_week[['performance']], how = 'left', left_index = True, right_index = True)
oee_total = pd.merge(oee_total_02, qual_by_week[['total_quality']], how = 'left', left_index = True, right_index = True)

oee_total['oee'] = oee_total['availability'] * oee_total['performance'] * oee_total['total_quality']

oee_total_format = oee_total.copy()
oee_total_format['oee'] = ['{:,.0f}%'.format(x * 100) for x in oee_total_format['oee']]

oee_total_format.tail()
```

```python
print ('OEE percentage greater than or equal to 100%: {:,.0f}'.format(len(oee_total.index[oee_total['oee'] >= 1])))
```

```python
fig = plt.figure(figsize = (20,5))
ax = fig.add_subplot(1,1,1)

x = np.arange(0, len(oee_total.index))
y = oee_total['oee']

ax.plot(x, y)
ax.set_xticks([])
vals = ax.get_yticks()
ax.set_yticklabels(['{:,.0f}%'.format(x * 100) for x in vals])

plt.title('Overall Machine Efficiency (OEE) Over Time', fontsize = 15)
plt.ylabel('Overall Machine Efficiency (OEE)', fontsize = 15)
plt.xlabel('Weeks', fontsize= 15)
plt.legend(['Overall Machine Efficiency (OEE)'], loc = 'best', fontsize = 15)
plt.grid()
plt.show()
```

### View Availability, Performance, and Quality over recent weeks

```python
fig = plt.figure(figsize = (20,5))
ax = fig.add_subplot(1,1,1)

oee_recent = oee_total.copy()
oee_recent = oee_recent.tail(10)

oee_recent[['availability','performance','total_quality']].plot(kind = 'bar', ax = ax)

vals = ax.get_yticks()
ax.set_yticklabels(['{:,.0f}%'.format(x * 100) for x in vals])
ax.set_xticklabels([])

plt.title('OEE Component Breakdown in Recent Weeks', fontsize = 15)
plt.ylabel('Overall Machine Efficiency (OEE)', fontsize = 15)
plt.xlabel('Recent Weeks', fontsize= 15)
plt.legend(['Availability','Performance','Total Quality'], loc = 'best')
plt.grid()
plt.show()
```
