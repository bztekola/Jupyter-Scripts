### To see if the times between processes and sp_actions tables differ

```python
a = pd.read_sql(
'''
SELECT
process_id,
TIMESTAMPDIFF(MINUTE, time_start, time_end) AS p_time 
FROM processes
WHERE process_group IN ('test','prep')
AND time_end != '0000-00-00 00:00:00'
GROUP BY process_id
''', db)

col_fix(a)

b = pd.read_sql(
'''
SELECT
process_id,
SUM(TIMESTAMPDIFF(MINUTE, spa_time_start, spa_time_end)) AS sp_time
FROM sp_actions
WHERE process_id IN '''+ str(tuple(a['process id'].tolist())) +'''
GROUP BY process_id
''', db)

col_fix(b)

c = pd.merge(a, b, how = 'left', on = 'process id')

c.info()
c.dropna(inplace = True)

c['diff (min)'] = np.abs(c['p time'] - c['sp time'])

c.sort_values('diff (min)', ascending = False, inplace = True)
```

### Get data for CSV

```python
pd.read_sql(
'''
SELECT
DISTINCT process_group
FROM processes
''', db)
```

```python
super_main = pd.read_sql(
'''
SELECT
wo.wo_id,
k.part_id,
pd.products_name,
TIMESTAMPDIFF(MINUTE, spa_time_start, spa_time_end) AS time_elapsed_in_minutes, 
sp.spa_qty,
process_group

FROM sp_actions sp

JOIN processes p ON sp.process_id = p.process_id
AND DATE(time_start) BETWEEN ' '''+ s +''' ' AND ' '''+ e +''' ' 
AND p.wo_id != 0 # to exclude reworks
AND process_group IN ('test','prep','secondary') # get only these groups
AND time_end != '0000-00-00 00:00:00' # to exclude open processes

JOIN work_orders wo ON p.wo_id = wo.wo_id
JOIN skus k ON wo.sku_id = k.sku_id
JOIN products_description pd ON k.part_id = pd.part_id
''', db)

col_fix(super_main)

# get the time elapsed over all steps
total_time = super_main.groupby('part id')[['time elapsed in minutes']].sum()

# get the qty made over only one of the steps
qty_main = super_main.groupby(['part id','process group'])[['spa qty']].sum().unstack(1)
qty_main.columns = qty_main.columns.droplevel(0)

# take the max qty made over all steps considered
q = dict(qty_main.max(1))


# put them together
main = super_main[['part id','products name']].drop_duplicates(['part id','products name'])
main = main.set_index('part id').join(total_time).reset_index()
main['qty made'] = main['part id'].map(q)

# get time per part
main['min per'] = np.round(main['time elapsed in minutes']/main['qty made'], 2)
main.sort_values('time elapsed in minutes', ascending = False, inplace = True)

# get the PLM link
l1 = 'https://volcano.adafruit.com/volcano/ada_plm.php#/Parts/'
l2 = '/PartProfile'
main['link'] = l1 + main['part id'].astype(str) + l2
```

### Check it

```python
a1 = super_main['time elapsed in minutes'].sum()
a2 = main['time elapsed in minutes'].sum()

if a1 != a2:
    raise ValueError('mismatch')
    
b1 = np.sum(list(q.values()))
b2 = main['qty made'].sum()

if b1 != b2:
    raise ValueError('mismatch')
```

### By part id and group

```python
by_part = super_main.groupby(['part id','process group'])[['time elapsed in minutes']].sum().unstack(1).fillna(0)
by_part.columns = by_part.columns.droplevel(0)
by_part['total time'] = by_part.sum(1)

l1 = 'https://volcano.adafruit.com/volcano/ada_plm.php#/Parts/'
l2 = '/PartProfile'
by_part['link'] = l1 + by_part.index.astype(str) + l2
by_part.reset_index(inplace = True)
by_part.sort_values('total time', ascending = False, inplace = True)

for col in by_part.columns:
    try:
        by_part[col] = [int(np.round(x)) for x in by_part[col]]
    except:
        pass

display(by_part.head())

c_dfs = pd.DataFrame()
cols = by_part.columns.tolist()
for w in ['part id','link']:
    cols.remove(w)
```

### Confidence intervals per part

```python
c = jb_conf(by_part)
sample_size = c.loc['sample size'][0]

c = c.loc[:'upper'].copy()
c.drop('part id',1,inplace = True)
```

### Avg total hours per workday

```python
days = 60 # 90 days in quarter; that's 60 workdays at 5 days per week
minutes = 60
c2 = c * sample_size / days / minutes

for col in c2.columns:
    c2[col] = [int(np.round(x)) for x in c2[col]]
    
print('avg time (hours) per workday')
c2
```

### To Excel

```python
if close_workbook == 'yes':
    
    t = 'Previous 90 days ending {} - Time to Test and Prep Fab Parts'.format(e)
    
    writer = pd.ExcelWriter(csv_path + t + '.xlsx', engine = 'xlsxwriter')
    
    sht_title = 'All data'
    main.to_excel(writer, sht_title, startrow = 2, startcol = 0, index = False)
    worksheet = writer.sheets[sht_title]
    text = 'All data by part from {} to {}'.format(s,e)
    worksheet.write(0,0,text)

    sht_title = 'Time by part and group'
    by_part.to_excel(writer, sht_title, startrow = 2, startcol = 0, index = False)    
    worksheet = writer.sheets[sht_title]
    text = 'Time elapsed (in minutes) per part per group'.format(s,e)
    worksheet.write(0,0,text)
    
    sht_title = 'Avg time per workday'
    c2.to_excel(writer, sht_title, startrow = 2, startcol = 0, index = True)
    worksheet = writer.sheets[sht_title]
    text = 'Avg time (in hours) per workday per group'
    worksheet.write(0,0,text)
    
    writer.save()
```
