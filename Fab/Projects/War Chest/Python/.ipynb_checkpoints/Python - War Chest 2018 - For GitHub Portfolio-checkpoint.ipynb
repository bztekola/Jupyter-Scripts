### Get Fab and Kitting skus

```python
skus_main = pd.read_sql(
'''
SELECT
sku_date_modified,
sku_id,
k.part_id,
pd.products_name,
bom_type,
p.products_dont_sell_alone,
p.products_carrot_only,
p.products_meta_type,
p.products_coming_soon,
p.discontinue_status,
p.products_status,
p.products_stripes,
p.products_discontinued
FROM skus k
JOIN products_description pd ON k.part_id = pd.part_id
JOIN parts p ON k.part_id = p.part_id
WHERE bom_type IN ('kitted','pnp') # get just pnp and kitted
AND sku_status IN ('working','pending')
''', db)

col_fix(skus_main)

# if there is more than one sku id then put them on one line
df = pd.DataFrame(skus_main.groupby('part id')['sku id'].apply(lambda x: ','.join(x.map(str))))
skus_main['all working or pending sku ids'] = skus_main['part id'].map(dict(zip(df.index.to_series(), df['sku id'])))
```

```python
# keep the most recent sku whether its pending or working
skus_main.sort_values(['part id','sku date modified'], inplace = True)
skus_main.drop_duplicates('part id', keep = 'last', inplace = True)
```

### Mask product names

```python
import urllib.request

# get some random words
word_url = "http://svnweb.freebsd.org/csrg/share/dict/words?view=co&content-type=text/plain"
response = urllib.request.urlopen(word_url)
long_txt = response.read().decode()
words = long_txt.splitlines()
words = ['The ' + x.title() + ' Product' for x in words]

skus_main['products name'] = skus_main['products name'].str.replace('/',' ')
d = list(skus_main['products name'].unique())
jump = int(np.floor(len(words)/len(d)))
random_words = words[::jump][:len(d)]

e = dict(zip(random_words, d))

for k,v in e.items():
    skus_main['products name'] = skus_main['products name'].replace(v,k)
```

### Get the combos that these PNs are contained in

```python
ptc_main = pd.read_sql(
'''
SELECT
part_id,
contains_part_id,
pts_quantity
FROM products_to_stuff
WHERE part_id > 0
''', db)

col_fix(ptc_main)
```

```python
ptc_clean = ptc_main[ptc_main['contains part id'].isin(skus_main['part id'])].copy()
ptc_clean.rename(columns = {'part id':'contained in',
                           'contains part id':'part id'}, inplace = True)
ptc_clean['(contained in, pts quantity)'] = list(zip(ptc_clean['contained in'], ptc_clean['pts quantity']))

ptc_temp = pd.DataFrame(ptc_clean.groupby('part id')['(contained in, pts quantity)'].apply(lambda x: '; '.join(x.map(str))))
ptc_temp.reset_index(inplace = True)
```

### Get all sales data

```python
sales_super_main = pd.read_sql(
'''
SELECT
DATE(date_purchased) AS date_purchased,
DATE_FORMAT(date_purchased, '%Y-%m') AS year_and_month,
op.orders_id,
part_id,
products_quantity AS qty_total, 
products_price
FROM orders_products op
JOIN orders o ON op.orders_id = o.orders_id
AND orders_status NOT IN (8,9,10,11,12,14,15)
AND DATE(o.date_purchased) <= ' '''+ date_end +''' '

WHERE op.part_id IN '''+ str(tuple(skus_main['part id'])) +'''
OR op.part_id IN '''+ str(tuple(ptc_clean['contained in'])) +'''
''', db)

col_fix(sales_super_main)

# change to datetime
sales_super_main['date purchased'] = pd.to_datetime(sales_super_main['date purchased'])

d1 = sales_super_main['date purchased'].min().date()
d2 = sales_super_main['date purchased'].max().date()

print('data is from {} to {}'.format(d1,d2))
```

### Make a copy

```python
sales_main = sales_super_main.copy()
```

### Create columns in sales data

```python
# umbrella part id
# forecast this PN
sales_main['umbrella part id'] = sales_main['part id'].map(dict(zip(ptc_clean['contained in'], ptc_clean['part id'])))
sales_main['umbrella part id'].fillna(sales_main['part id'], inplace = True)

# qty multiplier
sales_main['qty multiplier'] = sales_main['part id'].map(dict(zip(ptc_clean['contained in'], ptc_clean['pts quantity'])))
sales_main['qty multiplier'].fillna(1, inplace = True)

# modified qty total
# forecast this qty
sales_main['mod qty total'] = sales_main['qty total'] * sales_main['qty multiplier']    

# revenue
sales_main['revenue'] = sales_main['qty total'] * sales_main['products price']

for col in ['umbrella part id','qty multiplier','mod qty total']:
    sales_main[col] = [int(x) for x in sales_main[col]]

# ptc info
sales_main['(contained in, pts quantity)'] = sales_main['umbrella part id'].map(dict(zip(ptc_temp['part id'], ptc_temp['(contained in, pts quantity)'])))
    
n = sales_main[(sales_main.isnull().any(1)) & (sales_main['(contained in, pts quantity)'].isnull() == False)]
print('{:,.0f} null(s)'.format(len(n)))
```

### Flag parts to exclude
Either they have no sales data or possess at least one of the attributes in the function below

```python
s1 = set(sales_main['umbrella part id'])
s2 = set(skus_main['part id'])
no_sales = list(s2 - s1)
```

```python
def exclude_flag(x):
    if (x['products dont sell alone'] == 1
        or x['products carrot only'] == 1
        or x['products meta type'] == 'Base Product'
        or x['products coming soon'] == 1
        or x['discontinue status'] == 'Discontinued'
        or (x['products status'] == 1 and x['products stripes'] != 0)
        or x['part id'] in no_sales):
        res = 'yes'
    else:
        res = 'no'
        
    return res
        
skus_main['exclude'] = skus_main.apply(exclude_flag, axis = 1)

n = skus_main[skus_main.isnull().any(1)]
print('{:,.0f} null(s)'.format(len(n)))
```

### Find and drop outliers
Use Part ID, not Umbrella Part ID

```python
for_outliers = sales_clean.groupby('part id').agg({'qty total':['mean','std','count']})
for_outliers.columns = for_outliers.columns.droplevel(0)
for_outliers['phi'] = np.where(for_outliers['count'] < 30, stats.t.ppf(0.95, for_outliers['count'] - 1), 1.96)
for_outliers['lower'] = for_outliers['mean'] - 3 * for_outliers['std']
for_outliers['upper'] = for_outliers['mean'] + 3 * for_outliers['std']
for_outliers.dropna(inplace = True)

for col in ['lower','upper']:
    sales_clean['outlier ' + col] = sales_clean['part id'].map(dict(zip(for_outliers.index.to_series(), for_outliers[col])))
    
sales_clean['outlier'] = np.where(np.logical_or(sales_clean['qty total'] < sales_clean['outlier lower'], sales_clean['qty total'] > sales_clean['outlier upper']), 'yes','no')
print('proportion of outliers')
print(sales_clean['outlier'].value_counts()/sales_clean['outlier'].value_counts().sum())

sales_clean.drop(sales_clean[sales_clean['outlier'] == 'yes'].index, inplace = True)
sales_clean.reset_index(drop = True, inplace = True)
```

### Determine popularity tiers
Use umbrella part id and mod qty total

```python
print('popularity tiers are based on data from %s to %s' % (three_months_ago, date_end))

n = int(np.round((pd.to_datetime(date_end) - pd.to_datetime(three_months_ago)).days/30))

print('which is about {} months'.format(n))
```

```python
# first get all sales data for the last 3 months
a = sales_clean[sales_clean['date purchased'].between(three_months_ago, date_end)]

# get the top percentile of revenue for the key parts
b = a.groupby(['bom type','umbrella part id'])[['revenue','qty total']].sum().reset_index()
key1 = b.groupby(['bom type'])[['revenue']].quantile(0.85)
key2 = b.groupby(['bom type'])[['qty total']].quantile(0.85)

dict_ = {}
ls = []
for index, row in b.iterrows():
    if row['revenue'] >= key1.loc[row['bom type']]['revenue'] and row['qty total'] >= key2.loc[row['bom type']]['qty total']:
        dict_[row['umbrella part id']] = '01 - key'
        
# now from the main df drop all key parts
c = a.drop(a[a['umbrella part id'].isin(list(dict_.keys()))].index)

# for the rest of the popularity tiers we'll use the mod qty total
d = c.groupby(['bom type','umbrella part id'])[['mod qty total']].sum().reset_index()
qty = d.groupby('bom type')[['mod qty total']].describe()
qty.columns = qty.columns.droplevel(0)

# flag all other popularity tiers
for index, row in d.iterrows():
    if row['mod qty total'] >= qty.loc[row['bom type']]['75%']:
        ret = '02 - high'
    elif row['mod qty total'] >= qty.loc[row['bom type']]['50%'] and row['mod qty total'] < qty.loc[row['bom type']]['75%']:
        ret = '03 - med'
    else:
        ret = '04 - low'
    
    dict_[row['umbrella part id']] = ret
    
# map popularity tiers to main df    
sales_clean['popularity tier'] = sales_clean['umbrella part id'].map(dict_)    

# any part id not sold in the last three months, label with "not enough data"
sales_clean['popularity tier'].fillna('05 - not enough data', inplace = True)
```

```python
check = sales_clean.drop_duplicates('umbrella part id').groupby(['popularity tier','bom type'])[['popularity tier']].count().unstack(1)
check['row total'] = check.sum(1)
check.loc['column total'] = check.sum()
```

### Prepare data for forecasting

```python
for_forecast = pd.DataFrame()

for pn in list(set(sales_clean['umbrella part id'])):
    
    # get qty per month
    df1 = sales_clean[sales_clean['umbrella part id'] == pn].groupby('year and month')[['mod qty total']].sum()

    # fill in months where no one bought anything
    # make the final year and month the date_end of this script
    d1 = df1.index.min()
    d2 = date_end[:7]
    
    dates = pd.date_range(d1, d2, freq = 'MS')
    
    # make it so that every PN has at least 6 months of data
    n = 6
    l = len(dates)
    if l < n:
        new_d1 = pd.to_datetime(d1) - pd.DateOffset(months = (n-l))
        new_d1 = str(new_d1)[:7]
        dates = pd.date_range(new_d1, d2, freq = 'MS')
    
    dates = pd.DataFrame(dates, columns = ['year and month']).set_index('year and month')
    # join them
    df2 = dates.join(df1).fillna(0) 
    df2['umbrella part id'] = pn
    df2.reset_index(inplace = True)
    
    # create one big dataframe, for the forecast
    for_forecast = for_forecast.append(df2, ignore_index = True)

# map popularity tier and bom
for col in ['bom type','popularity tier']:
    for_forecast[col] = for_forecast['umbrella part id'].map(dict(zip(sales_clean['umbrella part id'], sales_clean[col])))
    
# rename these for a layman to understand
for_forecast.rename(columns = {'umbrella part id':'part id',
                              'mod qty total':'qty total'}, inplace = True)
```

```python
fab_csv = '/Users/jarad/fake Folder/Fab/Projects/War\ Chest\ 2018/R/Forecasts\ for\ 10\ -\ Oct\ onwards/Fab\ War\ Chest\ 2018\ Forecasts\ for\ 2018-10\ to\ 2019-02\ .xlsx'
fab_csv = fab_csv.replace('\\','')
fab_key = pd.read_excel(fab_csv, sheet_name = '01 - key')
fab_high = pd.read_excel(fab_csv, sheet_name = '02 - high')
fab_med = pd.read_excel(fab_csv, sheet_name = '03 - medium')

fab_models = pd.concat([fab_key, fab_high, fab_med], sort = False)
fab_models.drop(fab_models[fab_models['forecast month'] != ym].index, inplace = True)
fab_models.reset_index(drop = True, inplace = True)
```

```python
fab_results = fab_models.groupby(['popularity tier','in interval'])[['in interval']].count().unstack(1)
fab_results.columns = fab_results.columns.droplevel(0)
fab_results.loc['column total'] = fab_results.sum()
fab_results['% yes'] = fab_results['yes']/fab_results.sum(1)
print('For Fab: Was the actual qty sold within the estimation interval?')
fab_results.format_(['n0','n0','p2'])
```
