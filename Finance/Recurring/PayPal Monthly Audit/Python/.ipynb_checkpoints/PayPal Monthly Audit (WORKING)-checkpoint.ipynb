### Get db data

```python
os = pd.read_sql(
'''
SELECT
o.orders_id,
os.orders_status_name
FROM orders o
JOIN orders_status os ON o.orders_status = os.orders_status_id
WHERE o.payment_module_code = 'paypalwpp'
''', db)

o_main = pd.read_sql(
'''
SELECT
DATE(date_purchased) AS date_purchased,
o.orders_id,
ot.value AS db_order_total,
0 AS 'deleted'
FROM orders o
LEFT JOIN orders_total ot ON o.orders_id = ot.orders_id
AND ot.class = 'ot_total'
WHERE payment_module_code = 'paypalwpp'
''', db)

od_main = pd.read_sql(
'''
SELECT
DATE(date_purchased) AS date_purchased,
orders_id,
order_total AS db_order_total,
1 AS 'deleted'
FROM orders_deleted od
WHERE od.payment_module_code = 'paypalwpp'
''', db)

for df in [os, o_main, od_main]:
    col_fix(df)
```

```python
ot = pd.read_sql(
'''
SELECT
DATE_FORMAT(o.date_purchased, '%Y-%m') As year_and_month,
SUM(ot.value) AS monthly_total
FROM orders_total ot
JOIN orders o ON ot.orders_id = o.orders_id
AND DATE(o.date_purchased) BETWEEN ' '''+ date_start +''' ' AND ' '''+ date_end +''' '
AND orders_status NOT IN (8,9,10,11,12,14,15)
AND payment_method != 'Replacement Order'
WHERE ot.class = 'ot_subtotal'
GROUP BY DATE_FORMAT(o.date_purchased, '%Y-%m')
''', db)
col_fix(ot)
```

### Structure it

```python
# get date columns
for df in [o_main, od_main]:
    df['date purchased'] = pd.to_datetime(df['date purchased'])
    df['year and month purchased'] = [str(x)[:7] for x in df['date purchased']]

# make one dataframe    
orders_main = pd.concat([o_main, od_main], sort = True)
orders_main.reset_index(drop = True, inplace = True)

# map orders status
orders_main['orders status'] = orders_main['orders id'].map(dict(zip(os['orders id'],os['orders status name'])))
orders_main['orders status'] = orders_main['orders status'].str.lower()

# get dupes
dupes = orders_main[orders_main['orders id'].duplicated()]
print('%i oid dupe(s)\ndrop them' % len(dupes))

# store and drop them
orders_main_dupes = orders_main[orders_main['orders id'].isin(dupes['orders id'].tolist())].sort_values('orders id')
orders_main.drop_duplicates('orders id', inplace = True)

# get nulls
nulls = orders_main[(orders_main.isnull().any(1)) & (orders_main['deleted'] != 1)]
print('\n%i null(s)' % len(nulls))
```

### Get PayPal db data

```python
pp_super_main = pd.read_sql(
'''
SELECT
# for some reason you must subtract 4 hours from this datetime to convert it into PST time
DATE(DATE_SUB(payment_date, INTERVAL 4 HOUR)) AS date_payment,
TIME(DATE_SUB(payment_date, INTERVAL 4 HOUR)) AS time_payment,
parent_txn_id,
txn_id,
order_id AS orders_id,
payment_status,
txn_type,
mc_currency AS 'currency',
mc_gross AS 'pp order total',
mc_fee AS 'pp fee'
FROM paypal
''', db)

col_fix(pp_super_main)
```

### Structure it

```python
# make a copy
pp_main = pp_super_main.copy()

# dates
pp_main['year and month payment'] = [str(x) for x in pp_main['date payment']]
pp_main['date payment'] = pd.to_datetime(pp_main['date payment'], errors = 'coerce')

# make this nice
pp_main['payment status'] = pp_main['payment status'].str.lower()
pp_main['payment status'] = np.where(pp_main['payment status'].str.contains('pending'), 'pending', pp_main['payment status'])
pp_main['payment status'] = pp_main['payment status'].str.replace('canceled_reversal', 'canceled reversal')

# and this too
pp_main['txn type'] = pp_main['txn type'].str.replace('_',' ')

# get and remove dupes
dupes = pp_main[pp_main.duplicated()]
print('%i dupe(s)\nremove them' % len(dupes))
pp_main.drop_duplicates(inplace = True)

# get nulls
nulls1 = pp_main[pp_main.isnull().any(1)]
print('\n%i null(s)\nremove them' % len(nulls1))
pp_main.dropna(inplace = True)

# merge
pp_main = pd.merge(pp_main, orders_main, how = 'left', on = 'orders id', copy = False)

# restrict by date
# recall that this "date" is EST
pp_main = pp_main[pp_main['date payment'].between(date_start, date_end)].copy()

# get nulls, not including orders status
cols = pp_main.columns.tolist()
cols.remove('orders status')

nulls2 = pp_main[pp_main[cols].isnull().any(1)]
print('\n%i null(s)' % len(nulls2))
```

### Reconcile database tables

```python
df = pp_main.groupby(['orders id','payment status'], as_index = False)[['pp order total','db order total']].sum()

# paypal shows refunds and reversals as negative numbers
# we show them as positive
# cast all to positive here
for col in df.columns:
    try:
        df[col] = np.abs(df[col])
    except:
        pass
    
# get mismatches by OID    
mismatch = df[np.abs(df['pp order total'] - df['db order total']) > 1].copy()
mismatch['deleted'] = mismatch['orders id'].map(dict(zip(orders_main['orders id'], orders_main['deleted'])))

# map date purchased
mismatch['year and month purchased'] = mismatch['orders id'].map(dict(zip(pp_main['orders id'], pp_main['year and month purchased'])))

# get mismatches by payment status
mismatch2 = mismatch.groupby('payment status').sum().drop(['orders id','deleted'], 1)
mismatch2['difference'] = mismatch2['db order total'] - mismatch2['pp order total']
mismatch2.sort_values('difference', inplace = True)
```

### Fix currency conversions

```python
# make a copy
csv_main = csv_super_main.copy()

# view the counts
print(csv_main['currency'].value_counts())

# get all ids
ids = csv_main['parent txn id'][csv_main['currency'] != 'usd'].unique().tolist()

# extract them
for_fix = csv_main[csv_main['parent txn id'].isin(ids)].copy()

# remove from csv_main
csv_main.drop(csv_main[csv_main['parent txn id'].isin(ids)].index, inplace = True)

# this unfolds all "types" so you can see which types, other than "currency conv" are associated
# you want to discard these "currency conv" labels and use whatever the other label is
df = for_fix.groupby(['parent txn id','type'])[['txn id']].count().unstack(1)
df.columns = df.columns.droplevel(0)
df.reset_index(inplace = True)
df.drop('general currency conversion', 1, inplace = True)

d = {}
for i in range(len(df)):
    try:
        l = df.iloc[[i]].dropna(axis = 1).drop('parent txn id', 1).columns.values[0] 
    except:
        l = 'general payment' # if there is ONLY "currency conv", then call it a "general payment"
        
    d[df.iloc[i]['parent txn id']] = l
    
# now get all USD
fixed = for_fix[for_fix['currency'] == 'usd'].copy()

# check it
if len(fixed) == len(df):
    print('\nmatch')
else:
    print('\nmismatch')

# map the fixed type    
fixed['type'] = fixed['parent txn id'].map(d)    

# now add it back into the main data
csv_main = pd.concat([csv_main, fixed], sort = True)    
```

```python
amount_types = ['express checkout payment',
                'general payment',
                'mobile payment',
                'website payment']

amounts_main = csv_main_pst[(csv_main_pst['balance impact'] == 'credit')
                          & (csv_main_pst['type'].isin(amount_types))].copy()

amounts = amounts_main.groupby('year and month payment pst').agg({'csv gross':['count','sum']})
amounts.columns = amounts.columns.droplevel(0)
amounts.rename(columns = {'count':'payments received','sum':'amount received'}, inplace = True)
```

```python
# note that here we use csv_main, not csv_main_pst
refunds_main = csv_main[(csv_main['balance impact'] != 'memo')
                      & (csv_main['type'] == 'payment refund')
                      & (csv_main['date payment pst'].between(pd.to_datetime(date_start).date(), pd.to_datetime(date_end).date()))].copy()

refunds = refunds_main.groupby('year and month payment pst').agg({'orders id':'nunique','csv gross':'sum','csv fee':'sum'})
refunds.columns = ['refunds sent','amount refunded','refunded fees']
refunds['refunded fees'] = refunds['refunded fees'] * -1
```

```python
payment_types = pd.read_sql(
'''
SELECT
DATE_FORMAT(date_purchased, '%Y-%m') AS 'year and month',
payment_method AS 'payment type'
FROM orders
WHERE DATE(date_purchased) BETWEEN ' '''+ date_start +''' ' AND ' '''+ date_end +''' '
AND orders_status NOT IN (8,9,10,11,12,14,15)
''', db)
```

```python
df = payment_types.groupby(['year and month','payment type'])[['payment type']].count().unstack(1).fillna(0)
df.columns = df.columns.get_level_values(1)
df.sort_values(df.index[-1], ascending = False, axis = 1,inplace = True)
df = df.iloc[-1,:5]

fig, ax = plt.subplots(figsize = (20,5))

x = range(len(df))
y1 = df

ax.bar(x, y1,
       color = [list(colors.values())[i] for i in np.arange(1, len(colors))],
       edgecolor = 'black',
       lw = 2,
       alpha = 0.7,
       width = 0.70)

ax.set_xlabel(calendar.month_abbr[int(date_end[-5:-3])] + ' ' + date_end[:4],
             labelpad = 20)
ax.set_title('Top 5 Payment Types by Count',
            y = 1.02)

vals = ax.get_yticks()
ax.set_yticklabels('{:,.0f}'.format(x/1000) for x in vals)
ax.set_ylabel('thousands')

ax.set_xticks(x)
ax.set_xticklabels(df.index, fontsize = 20)
ax.grid(color = colors['color04'], alpha = 0.2)
ax.set_facecolor(colors['color01'])

plt.savefig(ax.get_title(), bbox_inches = 'tight')
```
