### Get subscriptions data

```python
subs1_main = pd.read_sql(
'''
SELECT
*
FROM subscriptions
WHERE subscriptions_type = 'adabox'
''', db)

col_fix(subs1_main)
subs1_main['gift'] = np.where(subs1_main['gifter customers id'] > 0, 1,0)

subs2_main = pd.read_sql(
'''
SELECT
*
FROM subscriptions_history
''', db)

col_fix(subs2_main)
```

### Check them

```python
one = set(subs1_main['subscriptions id'])
two = set(subs2_main['subscriptions id'])

print('count of unique subscription ids:\n')
print('%i in subscriptions table' % len(one))
print('%i in subscriptions_history table' % len(two))

if len(one) > len(two):
    print('\nsubscription ids in subscriptions table but not in subscriptions_history:' , one - two)
else:
    print('\nsubscription ids in subscriptions_history but not in subscriptions table:' , two - one)
```

### Structure it

```python
# fix the first
ls1 = ['subscriptions id','customers id','date created','subscriptions status','gift']
subs1 = subs1_main[ls1].copy()
subs1.rename(columns = {'subscriptions status':'umbrella status',
                       'date created':'date and time created'}, inplace = True)

# then the second
ls2 = ['subscriptions id','orders id','action','comments','timestamp','subscriptions status']
subs2 = subs2_main[ls2].copy()
subs2.rename(columns = {'subscriptions status':'line status',
                        'timestamp':'date and time transaction'}, inplace = True)

# merge them
subs_super_main = pd.merge(subs2, subs1, how = 'left', on = ['subscriptions id'])

# lowercase
ls = ['action','comments','line status','umbrella status']
for col in ls:
    subs_super_main[col] = subs_super_main[col].str.lower()
    
# sort it    
subs_super_main.sort_values(['subscriptions id','date and time transaction'], inplace = True)    
```

### Clean it

```python
print('years containing lines where subscriptions id equals zero:\n')

zeros = subs_super_main[subs_super_main['subscriptions id'] == 0]
years = set(zeros['date and time transaction'].map(str).str[:4])
print(years)

print('\n"actions" of these lines:\n')
print(zeros['action'].value_counts())

print('\nexclude these lines')

subs_main = subs_super_main[~subs_super_main['subscriptions id'].isin([0,5708])].copy()

# adabox008 was shipped out in july 2018 when it should have gone out in june 2018
# if we don't change this then when we group by quarter our numbers will be weird
# for super-simplicity, assume any transaction of any kind made between july 1 and july 15, 2018 should be counted as being in june 2018

for col in ['created','transaction']:
    
    subs_main['date and time ' + col] = np.where(subs_main['date and time ' + col].between('2018-07-01','2018-07-15'),
                                                 np.datetime64('2018-06-01T12:00'),
                                                 subs_main['date and time ' + col])    
    
    subs_main['date ' + col] = [x.date() for x in subs_main['date and time ' + col]]
    subs_main['date ' + col] = pd.to_datetime(subs_main['date ' + col])
    subs_main['year and month ' + col] = [str(x)[:7] for x in subs_main['date and time ' + col]]    
    subs_main['year and quarter ' + col] = subs_main['date ' + col].dt.year.map(str) + '-0' + subs_main['date ' + col].dt.quarter.map(str)
```

### Restrict by quarter

```python
subs_main = subs_main[subs_main['year and quarter transaction'] <= quarter_end].copy()
```

### Churn

```python
# we need to group the main dataframe in two ways in order to get all the data we need for churn

# first, go by the subscriptons_history_id level to get subs created and shipments sent
df1 = subs_main.groupby([report_type + ' transaction','action'])[['subscriptions id']].count().unstack(1).fillna(0)
df1.columns = df1.columns.droplevel(0)

new_cols1 = ['create shipment order',
            'create subscription',
            'redeem gift subscription',
            'activate subscription']

old_cols1 = df1.columns.tolist()
df1 = df1[new_cols1]
df1['total gained'] = df1.sum(1) - df1['create shipment order']

# then, go by the subscriptions_id level to get all types of churn, voluntary or involuntary
# if we drop dupes and keep the last then we are keeping the most recent entry
# recall that "umbrella status" is the "subscriptions status" from the "subscriptions" table
df2 = subs_main.drop_duplicates('subscriptions id', keep = 'last').groupby([report_type + ' transaction','umbrella status'])[['subscriptions id']].count().unstack(1).fillna(0)
df2.columns = df2.columns.droplevel(0)

new_cols2 = ['cancelled','expired','payment invalid']
old_cols2 = df2.columns.tolist()

df2 = df2[new_cols2]
df2['total lost'] = df2.sum(1)

# check it
# first, sum the total count of subscriptions created
one = df1['create subscription'].sum()

# then, sum the count of all types of cancellations
two = df2['total lost'].sum()

# take the difference
three = one - two

# this difference should equal the CURRENT count of active subs
# note that "summary" was created up above
if three == summary['active']:
    print('match\n')
else:
    print('mismatch\n')
    
# make on dataframe for churn    
churn = df1.join(df2)

# get churn per period
churn['churn rate'] = churn['total lost']/(churn['total gained'].cumsum() - churn['total lost'].shift(1).cumsum())

# rearrange cols
cols = churn.columns.tolist()
for x in ['total gained','total lost','churn rate']:
    cols.remove(x)
cols = cols + ['total gained','total lost','churn rate']
churn = churn[cols]

churn['churn rate'].fillna(churn['total lost']/churn['total gained'], inplace = True)

print(churn.iloc[1:].mean())
churn
```

### Churn chart

```python
pretty_dates = churn.index.tolist()
pretty_dates = [x[:4] + '\nQ' + x[-1] for x in pretty_dates]

fig, ax = plt.subplots(figsize = (20,5))

x = np.arange(len(churn))
y1 = churn['total gained']
y2 = churn['total lost']
y3 = churn['churn rate']

width = 0.40

ax.bar(x, y1, width = width, label = y1.name.title(), color = adablue01, edgecolor = 'white')
ax.bar(x + width, y2, width = width, label = y2.name.title(), color = adapink, edgecolor = 'white')

ax.set_yticklabels(['{:,.0f}'.format(x) for x in ax.get_yticks()])
ax.set_title('AdaBox Subs Gained, Lost, and Churn Rate')
ax.set_xticks(x + width/2)
ax.set_ylabel('Count')
ax.set_xticklabels(pretty_dates)
ax.legend(bbox_to_anchor = (-0.1,1))

ax2 = ax.twinx()
ax2.plot(x, y3, label = y3.name.title(), color = adablue02, lw = 5, ls = '--', marker = 'o', mew = 8)
max_ = np.max(ax2.get_yticks()) + 0.05
ax2.set_ylim(0,max_)
ax2.set_yticklabels(['{:,.0f}%'.format(x*100) for x in ax2.get_yticks()])
ax2.legend(bbox_to_anchor = (-0.1,0.5))
ax2.set_ylabel('Churn Rate', rotation = 270, labelpad = 30)

plt.savefig(ax.get_title(), bbox_inches = 'tight')
plt.show()
```

### Part cost and revenue

```python
# get all relevant OIDs
oids = list(set(subs_main['orders id']))
oids.remove(0)

# get part data
parts = pd.read_sql(
'''
SELECT
orders_id,
op.part_id,
pd.products_name,
products_price
FROM orders_products op
JOIN products_description pd ON op.part_id = pd.part_id
WHERE orders_id IN '''+ str(tuple(oids)) +'''
''', db)

col_fix(parts)

# check it for dupes
dupes = np.sum(parts['orders id'].duplicated())
print('%i dupe(s)' % dupes)

# get cost data from my all_cost.py
cost = all_cost[all_cost['part id'].isin(parts['part id'].unique())].copy()
cost.drop(['type','products stripe of','products stripe qty'], 1, inplace = True)

# adabox008 is not in my all_cost because it's marked as "coming soon", a flag that i exclude
# so create it here
add_on = pd.DataFrame({'part id':3761,
                      'products name':'AdaBox008 - Octo',
                      'sku id':12013,
                      'cost per unit':36.71},
                      index = [0])

# add it in
cost = pd.concat([cost, add_on], sort = False)

# create some columns
cost['short name'] = cost['products name'].str.split(' ').str[0]

# map part id via orders id
for col in ['products price','part id']:
    subs_main[col] = subs_main['orders id'].map(dict(zip(parts['orders id'], parts[col])))

# map the cost info via part id
cols = cost.columns.tolist()
cols.remove('part id')

for col in cols:
    subs_main[col] = subs_main['part id'].map(dict(zip(cost['part id'], cost[col])))
```

### Shipping cost and revenue

```python
# get shipping revenue
shipping_revenue = pd.read_sql(
'''
SELECT
ot1.orders_id,
ot1.value + IFNULL(ot2.value, 0) AS shipping_revenue
FROM orders_total ot1
LEFT JOIN orders_total ot2 ON ot1.orders_id = ot2.orders_id
AND ot2.class = 'ot_ddp'
WHERE ot1.class = 'ot_shipping' 
AND ot1.orders_id IN '''+ str(tuple(oids)) +'''
GROUP BY ot1.orders_id
''', db)

col_fix(shipping_revenue)

# map it
subs_main['shipping revenue'] = subs_main['orders id'].map(dict(zip(shipping_revenue['orders id'], shipping_revenue['shipping revenue'])))

# get UPS cost
ups_cost = pd.read_sql(
'''
SELECT
orders_id,
SUM(netAmount) AS shipping_charge
FROM ups_billing
WHERE orders_id IN '''+ str(tuple(oids)) +'''
GROUP BY orders_id
''', db)

col_fix(ups_cost)

# get DHL cost
dhl_cost = dhl_main[['orders id','shipping charge']][dhl_main['orders id'].isin(oids)].copy()

# join the two
shipping_cost = pd.concat([ups_cost, dhl_cost], sort = False)

# map it
subs_main['shipping charge'] = subs_main['orders id'].map(dict(zip(shipping_cost['orders id'], shipping_cost['shipping charge'])))

# create the summary2
summary2 = subs_main.groupby([report_type + ' transaction','short name'])[['products price','shipping revenue','cost per unit','shipping charge']].sum()

# count gifts
g = subs_main[subs_main['gift'] == 1].groupby([report_type + ' transaction','short name'])[['orders id']].nunique()
g.columns = ['count of gifts']

# join them
summary2 = summary2.join(g)

# count all boxes sent out
b = subs_main.groupby([report_type + ' transaction','short name'])[['orders id']].nunique()
b.columns = ['total sent']

# join them
summary2 = summary2.join(b)

# create some columns
summary2['total revenue'] = summary2[['products price','shipping revenue']].sum(1)
summary2['total cost'] = summary2[['cost per unit','shipping charge']].sum(1)
summary2['gross profit'] = summary2['total revenue'] - summary2['total cost']
summary2['margin'] = 1 - (summary2['total cost']/summary2['total revenue'])

# just until we get shipping data
# make the shipping cost of adabox008 the same as adabox007
ship_avg = list((summary2['shipping charge']/summary2['total sent']).values)
ship_avg = ship_avg[-2]

summary2.loc['2018-02','shipping charge'] = (ship_avg * summary2.loc['2018-02','total sent']).values[0]

summary2 = summary2.reset_index().set_index(report_type + ' transaction')
summary2['revenue per box'] = summary2['total revenue']/summary2['total sent']
summary2['cost per box'] = summary2['total cost']/summary2['total sent']

print('shipping avg estimation for AdaBox008: ${:,.2f}\n'.format(ship_avg))

fmt = ['m2','m2','m2','m2','n0','n0','m2','m2','m2','p2','m2','m2']

m = pd.DataFrame(summary2.mean())
m.columns = ['mean']
m = m.T
m = m.format_(fmt)
m = m.T
print(m)

summary2.format_([0] + fmt)
```
