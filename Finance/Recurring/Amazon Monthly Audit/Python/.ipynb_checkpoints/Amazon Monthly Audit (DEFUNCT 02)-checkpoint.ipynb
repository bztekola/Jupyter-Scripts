```python
import sys
sys.path.insert(0, '/Users/jarad/fake_folder/Python Libraries')

from jb_libraries import *
%matplotlib inline

import glob
import re

my_colors = {'face color':'#146eb4',
             'bar 01':'white',
             'bar 02':'#000000',
             'bar 03':'#ff9900'}

date_start = '2017-06-01'
date_end = '2018-06-30'

slide_path = r'/Users/jarad/fake_folder/Finance/Recurring/Amazon Monthly Audit/Docs/'
csv_path = r'/Users/jarad/fake_folder/Finance/Recurring/Amazon Monthly Audit/Docs/Statements/'
```

### Get database data

```python
lapwa_log_main = pd.read_sql(
'''
SELECT
order_id AS 'orders id',
LOWER(action) AS 'action',
order_reference_id AS 'order reference id',
amount AS 'lapwa log amount'
FROM lapwa_log
''', db)

orders_main = pd.read_sql(
'''
SELECT
DATE(o.date_purchased) AS 'date purchased',
DATE_FORMAT(o.date_purchased, '%Y-%m') AS 'year and month purchased',
o.orders_id AS 'orders id',
ot.value AS 'ot amount',
os.orders_status_name AS 'orders status'
FROM orders o 
JOIN orders_total ot ON o.orders_id = ot.orders_id
AND ot.class = 'ot_total'
JOIN orders_status os ON o.orders_status = os.orders_status_id
WHERE o.payment_module_code = 'lapwa'
''', db)

orders_deleted = pd.read_sql(
'''
SELECT
orders_id AS 'orders id'
FROM orders_deleted
''', db)
```

### Get CSV data

```python
all_files = glob.glob(csv_path + '/*.csv')
frame = pd.DataFrame()
list_ = []

for file_ in all_files:
    df = pd.read_csv(file_, index_col = None, header = 0)
    list_.append(df)
    
csv_main = pd.concat(list_)    

# clean column headers
columns_list = []
for y in csv_main.columns:
    columns = '_'.join(x.lower() for x in re.findall('[A-Z][^A-Z]*', y))
    columns_list.append(columns)
    
csv_main.columns = columns_list 

# fix column headers
csv_main.columns = [x.replace('_',' ') for x in csv_main.columns]
csv_main.columns = [x.lstrip().rstrip() for x in csv_main.columns]

# convert to datetime and get rid of time units
csv_main['transaction posted date'] = pd.to_datetime(csv_main['transaction posted date'])

csv_main['posted date'] = pd.to_datetime(csv_main['transaction posted date']).dt.date
csv_main['posted date'] = pd.to_datetime(csv_main['posted date'])
csv_main['year and month posted'] = pd.to_datetime(csv_main['transaction posted date'].dt.year.map(str) + '-' + csv_main['transaction posted date'].dt.month.map(str))
csv_main['year and month posted'] = [str(x)[:7] for x in csv_main['year and month posted']]

# convert to numeric
csv_main['transaction amount'] = csv_main['transaction amount'].str.replace(',','')
csv_main['transaction amount'] = pd.to_numeric(csv_main['transaction amount'])

csv_main['transaction percentage fee'] = pd.to_numeric(csv_main['transaction percentage fee'])
csv_main['transaction fixed_fee'] = pd.to_numeric(csv_main['transaction fixed fee'])
csv_main['total transaction fee'] = pd.to_numeric(csv_main['total transaction fee'])

csv_main['net transaction amount'] = csv_main['net transaction amount'].str.replace(',','')
csv_main['net transaction amount'] = pd.to_numeric(csv_main['net transaction amount'])

# rename
csv_main.rename(columns = {'amazon order reference id':'order reference id',
                          'transaction type':'action',
                          'transaction amount':'csv amount',
                          'total transaction fee':'csv transaction fee',
                          'net transaction amount':'csv net amount'}, inplace = True)

# keep only the columns you want
csv = csv_main.copy()
csv = csv[[
     'transaction posted date', 
     'posted date',
     'year and month posted',
     'amazon transaction id',
     'action',
     'order reference id',
     'csv amount',
     'csv transaction fee',
     'csv net amount']]

# change the wording of this label
csv['action'] = np.where(csv['action'] == 'authorization', 'authorize', csv['action'])

# lower case
csv['action'] = [x.lower() for x in csv['action']]

# dupes
print('dupe count before removal: ', np.sum(csv.duplicated()))
csv.drop_duplicates(inplace = True)
print('dupe count after removal:', np.sum(csv.duplicated()))

print('if a line in the CSV has no order reference id, then the actions are:', csv['action'][(csv['order reference id'].isnull())].unique())
```

```python
data_main = pd.merge(lapwa_log_main,
                    orders_main,
                    how = 'left',
                    on = 'orders id').merge(csv,
                                           how = 'outer',
                                           on = ['order reference id','action'])

data_main['deleted'] = np.where(data_main['orders id'].isin(orders_deleted['orders id'].tolist()),
                                                           'yes',
                                                           'no')
data_main['date purchased'] = pd.to_datetime(data_main['date purchased'])
data_main = data_main[data_main['date purchased'].between(pd.to_datetime(date_start), pd.to_datetime(date_end))]

data_main['action'].unique()
```

### Auth versus Capture audit
* if an auth is successful, and orders_status != "void", then it should have a corresponding capture amount
* if not, then we auth'd it, gave away the product, and did not collect any payment
* jan/feb 2017 was the time period containing the already-known "gift certificates" bug

```python
#by_oid = data_main.groupby(['year and month purchased','orders id','action'])[['action']].count().unstack(2)
by_oid = data_main.groupby(['year and month purchased','orders id','orders status','action'])[['lapwa log amount']].sum().unstack(3)
by_oid.columns = by_oid.columns.droplevel(0)
by_oid[(by_oid['capture'].isnull())]
```

### Capture and Authorize Audit 

```python
for action in ['authorize','capture']:
    df = data_main[data_main['action'] == action].groupby('year and month purchased')[['ot amount','csv amount']].sum().copy()
    df.fillna(0, inplace = True)
    df['difference'] = np.abs(df['ot amount'] - df['csv amount'])
    df['m2m'] = df['ot amount'].pct_change()
    print(action + '\n')
    print(df.format_(['m0','m0','m0','p2']).replace('nan%',''))
    print('\nmean')
    print(df.mean())
    print('\n')
```

```python
approval = data_main[data_main['action'].isin(['authorize','capture'])].groupby(['year and month purchased','action'])[['orders id']].count().unstack(1).fillna(0)
approval.columns = approval.columns.droplevel(0)
approval['approval %'] = approval['capture']/approval['authorize']
print(approval.format_(['n0','n0','p2']))
print('\n')
print(approval.mean())
```

```python
fig1 = plt.figure(figsize = (20,5))
ax = fig1.add_subplot(1,1,1)

capture_for_plot = data_main[data_main['action'] == 'capture'].groupby('year and month purchased')[['ot amount']].sum()

x = range(len(capture_for_plot))
y1 = capture_for_plot['ot amount']
y2 = [capture_for_plot['ot amount'].mean()] * len(capture_for_plot)

# plot it
plot01 = ax.bar(x, y1, 
                color = 'white', 
                label = 'Monthly Capture Amount Totals', 
                edgecolor = 'black')
plot02, = ax.plot(x, y2, 
                  '--', 
                  linewidth = 5, 
                  label = 'Monthly Average', 
                  color = my_colors['bar 03'])

# x ticks
ax.set_xticks(x)
ax.set_xticklabels([calendar.month_abbr[int(str(x)[-2:])] + '\n' + x[:4] for x in capture_for_plot.index],
                  rotation = 0,
                  fontsize = 20,
                  fontweight = 'bold',
                  fontname = 'Arial')

# y ticks
#ax.set_ylim(0, max(y1) + 100000)
vals = ax.get_yticks()
ax.set_yticklabels(['${:,.0f}'.format(x/1000) for x in vals], 
                   fontweight = 'bold', 
                   fontname = 'Arial', 
                   fontsize = 20)
ax.set_ylabel('thousands', fontsize = 20, fontweight = 'bold')

# labels and other
ax.set_title('Amazon Revenue', fontsize = 30, fontweight = 'bold', fontname = 'Arial', y = 1.02)
ax.set_facecolor(my_colors['face color'])
ax.legend(handles = [plot02], fontsize = 20, loc = 'upper left')
ax.grid()

plt.savefig(slide_path + ax.get_title(), bbox_inches = 'tight')
```

### Refund Audit

```python
refund_for_plot = csv[(csv['action'] == 'refund')
                     & (csv['posted date'].between(date_start, date_end))].groupby('year and month posted')[['csv amount']].sum() * -1
refund_for_plot.loc['mean'] = refund_for_plot.mean()
refund_for_plot['m2m'] = refund_for_plot.pct_change()
refund_for_plot.loc['mean','m2m'] = np.nan

print(refund_for_plot.format_(['m0','p2']).replace('nan%',''))
```

```python
fig1 = plt.figure(figsize = (20,5))
ax = fig1.add_subplot(1,1,1)

x = range(len(refund_for_plot.iloc[:-1]))
y1 = refund_for_plot['csv amount'].iloc[:-1]
y2 = [refund_for_plot['csv amount'].mean()] * len(refund_for_plot.iloc[:-1])

# plot it
plot01 = ax.bar(x, y1, 
                color = 'white', 
                label = 'Monthly Refund Amount Totals', 
                edgecolor = 'black')
plot02, = ax.plot(x, y2, 
                  '--', 
                  linewidth = 5, 
                  label = 'Monthly Average', 
                  color = my_colors['bar 03'])

# x ticks
ax.set_xticks(x)
ax.set_xticklabels([calendar.month_abbr[int(str(x)[-2:])] + '\n' + x[:4] for x in refund_for_plot.index[:-1]],
                  rotation = 0,
                  fontsize = 20,
                  fontweight = 'bold',
                  fontname = 'Arial')

# y ticks
vals = ax.get_yticks()
ax.set_yticklabels(['${:,.1f}'.format(x/1000) for x in vals], 
                   fontweight = 'bold', 
                   fontname = 'Arial', 
                   fontsize = 20)
ax.set_ylabel('thousands', fontsize = 20, fontweight = 'bold')

# labels and other
ax.set_title('Monthly Refund Amounts', fontsize = 30, fontweight = 'bold', fontname = 'Arial', y = 1.02)
ax.set_facecolor(my_colors['face color'])
ax.legend(handles = [plot02], fontsize = 20, loc = 'upper left')
ax.grid()
plt.savefig(slide_path + ax.get_title(), bbox_inches = 'tight')
```

### Fees

```python
fees_by_month = csv[csv['posted date'].between(date_start, date_end)].groupby(['year and month posted','action'])[['csv transaction fee']].sum().unstack(1).fillna(0)
fees_by_month.columns = fees_by_month.columns.droplevel(0)
fees_by_month['total'] = fees_by_month.sum(1)
fees_by_month.loc['mean'] = fees_by_month.mean()
fees_by_month = fees_by_month * -1
fees_by_month['total m2m'] = fees_by_month['total'].pct_change()
fees_by_month.loc['mean','total m2m'] = np.nan
fees_by_month.format_(['m0'] * (len(fees_by_month.columns) - 1) + ['p2']).replace('nan%','')
```

```python
payment_types = pd.read_sql(
'''
SELECT
DATE_FORMAT(date_purchased, '%Y-%m') AS 'year and month',
payment_method AS 'payment type'
FROM orders
WHERE DATE(date_purchased) BETWEEN ' '''+ date_start +''' ' AND ' '''+ date_end +''' '
AND orders_status != 9
''', db)
```

```python
payment_types_by_month = payment_types.groupby(['year and month','payment type'])[['payment type']].count().unstack(1).fillna(0)
payment_types_by_month.columns = payment_types_by_month.columns.get_level_values(1)

payment_types_by_month.sort_values(payment_types_by_month.index[-1], 
                                   ascending = False, 
                                   axis = 1,
                                   inplace = True)
fig, ax = plt.subplots(figsize = (20,5))

x = range(len(payment_types_by_month.iloc[-1].head()))
y1 = payment_types_by_month.iloc[-1].head()

ax.bar(x, y1,
       color = [x for x in list(my_colors.values())[1:]],
       edgecolor = 'white',
       width = 0.5)

# x ticks
ax.set_xticks(x)
ax.set_xticklabels(payment_types_by_month.iloc[-1].head().index,
                  rotation = 0,
                  fontsize = 20,
                  fontweight = 'bold',
                  fontname = 'Arial')

# y ticks
vals = ax.get_yticks()
ax.set_yticklabels(['{:,.0f}'.format(x/1000) for x in vals], 
                   fontweight = 'bold', 
                   fontname = 'Arial', 
                   fontsize = 20)
ax.set_ylabel('thousands', fontsize = 20, fontweight = 'bold')

# labels and other
ax.set_title('Top 5 Payment Types by Count', fontsize = 30, fontweight = 'bold', fontname = 'Arial', y = 1.02)
ax.set_facecolor(my_colors['face color'])

plt.savefig(slide_path + ax.get_title(), bbox_inches = 'tight')
```
