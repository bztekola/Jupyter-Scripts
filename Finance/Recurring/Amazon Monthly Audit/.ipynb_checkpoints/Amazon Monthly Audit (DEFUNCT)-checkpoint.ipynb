```python
path = r'/Users/jarad/fake_folder/Finance/Recurring/Amazon Monthly Audit/Statements/'
all_files = glob.glob(path + '/*.csv')
frame = pd.DataFrame()
list_ = []

for file_ in all_files:
    df = pd.read_csv(file_, index_col = None, header = 0)
    list_.append(df)
    
csv_raw = pd.concat(list_)    

# clean column headers
columns_list = []
for y in csv_raw.columns:
    columns = '_'.join(x.lower() for x in re.findall('[A-Z][^A-Z]*', y))
    columns_list.append(columns)
    
csv_raw.columns = columns_list  

# convert to datetime and get rid of time units
csv_raw['transaction_posted_date'] = pd.to_datetime(csv_raw['transaction_posted_date'])

csv_raw['posted_date'] = pd.to_datetime(csv_raw['transaction_posted_date']).dt.date
csv_raw['posted_date'] = pd.to_datetime(csv_raw['posted_date'])
csv_raw['posted_date_format'] = pd.to_datetime(csv_raw['transaction_posted_date'].dt.year.map(str) + '-' + csv_raw['transaction_posted_date'].dt.month.map(str))
csv_raw['posted_date_format'] = [str(x)[:7] for x in csv_raw['posted_date_format']]

# convert to numeric
csv_raw['transaction_amount'] = csv_raw['transaction_amount'].str.replace(',','')
csv_raw['transaction_amount'] = pd.to_numeric(csv_raw['transaction_amount'])

csv_raw['transaction_percentage_fee'] = pd.to_numeric(csv_raw['transaction_percentage_fee'])
csv_raw['transaction_fixed_fee'] = pd.to_numeric(csv_raw['transaction_fixed_fee'])
csv_raw['total_transaction_fee'] = pd.to_numeric(csv_raw['total_transaction_fee'])

csv_raw['net_transaction_amount'] = csv_raw['net_transaction_amount'].str.replace(',','')
csv_raw['net_transaction_amount'] = pd.to_numeric(csv_raw['net_transaction_amount'])

# rename
csv_raw.rename(columns = {'amazon_order_reference_id':'order_ref'}, inplace = True)

# keep only the columns you want
csv = csv_raw.copy()
csv = csv[[
     'transaction_posted_date', 
     'posted_date',
     #'posted_date_and_time',
     'posted_date_format',
     'transaction_type',
     'order_ref',
     'transaction_amount',
     #'transaction_fixed_fee',
     #'transaction_percentage_fee',
     'total_transaction_fee',
     'net_transaction_amount']]

# change the wording of this label
csv['transaction_type'] = np.where(csv['transaction_type'] == 'Authorization', 'Authorize', csv['transaction_type'])
```

# Authorized audit
* This section compares the amount that we have in our db with the amount that Amazon captures
* If we have \$10 and Amazon captures \$20, then Amazon has captured too much and they owe us
* If we have \$10 and Amazon captures \$0, then we are giving away out product for free

```python
auth_audit = pd.merge(orders[['date_purchased','date_format','transaction_type','orders_id','order_ref','o_total']][orders['transaction_type'] == 'Authorize'],
                      orders[['transaction_type','orders_id','order_ref','o_total']][orders['transaction_type'] == 'Capture'],
                      how = 'left',   # this left join discards any instances of double auth
                                      # like when the first auth is declined or something
                                      # here we only care about matching what we Authed and what Amazon captured
                      on = ['orders_id','order_ref'])
auth_audit.drop(['transaction_type_x','transaction_type_y'], 1, inplace = True)
auth_audit.fillna(0, inplace = True)
auth_audit['difference'] = auth_audit['o_total_x'] - auth_audit['o_total_y']
auth_audit['date_purchased'] = pd.to_datetime(auth_audit['date_purchased'])
auth_audit = auth_audit[auth_audit['date_purchased'].between(pd.to_datetime(date_start), pd.to_datetime(date_end))]
```

```python
auth_audit_by_month = auth_audit.copy()
auth_audit_by_month = auth_audit_by_month.groupby('date_format', as_index = False)[['difference']].sum()
auth_audit_by_month
```

```python
ax = auth_audit.groupby('date_format')[['difference']].sum().plot(kind = 'bar', 
                                                                  figsize = (20,5),
                                                                  grid = True,
                                                                  rot = 45,
                                                                  color = 'red',
                                                                  alpha = 0.3)

vals = ax.get_yticks()
ax.set_yticklabels(['${:,.0f}'.format(x) for x in vals])

ax.set_xlabel('Year and Month', fontsize = 15)
ax.set_ylabel('\$$ Difference', fontsize = 15)
ax.set_title('Dollar amount that was Authorized but never Captured', fontsize = 15)
plt.legend([])
plt.show()
```

# Capture Audit
* Here we audit the amount that Amazon captured (listed in the CSV download from Seller Central) with what we have as captured in the db
* If the capture_amount_difference is small (less than \$50) then we'll consider the difference the result of a refund

```python
capture = pd.merge(csv[csv['transaction_type'] == 'Capture'],
                   orders[['transaction_type','orders_id','o_total','order_ref']][orders['transaction_type'] == 'Capture'],
                   how = 'left',
                   on = ['order_ref','transaction_type'])

capture['capture_amount_difference'] = capture['o_total'] - capture['transaction_amount']
capture['capture_amount_difference'].fillna(0, inplace = True)
```

```python
capture[capture['capture_amount_difference'] != 0].groupby('posted_date_format', as_index = False)[['capture_amount_difference']].sum()
```

```python
capture_by_month = capture[capture['posted_date'].between(date_start, date_end)].groupby('posted_date_format')[['transaction_amount']].sum()
capture_by_month['% change'] = capture_by_month['transaction_amount'].pct_change()

print(capture_by_month['transaction_amount'].mean())

capture_by_month.format_(['m0','p2'])
```
