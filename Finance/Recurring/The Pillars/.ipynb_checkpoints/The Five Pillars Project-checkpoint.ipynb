# Get and clean QuickBooks data

```python
# get all CSVs
path = r'/Users/jarad/Desktop/Desktop/The Five Pillars Project/QB CSVs/'
all_files = glob.glob(path + '/*.csv')
frame = pd.DataFrame()
list_ = []

for file_ in all_files:
    df = pd.read_csv(file_)
    list_.append(df)
    
qb_main = pd.concat(list_)   

# make a copy
qb = qb_main.copy()

# drop columns of all nulls
qb.dropna(how = 'all', axis = 1, inplace = True)

# distinguish between income and expenses
qb.rename(columns = {'Unnamed: 4':'main cat',
                     'Unnamed: 5':'sub cat',
                     'Unnamed: 3':'income/expense'}, inplace = True)

# fill out income/expense
qb['income/expense'].fillna(method = 'ffill',inplace = True)

# drop "totals" rows
qb['main cat'].fillna('', inplace = True)
qb = qb[~qb['main cat'].str.contains('Total')]
qb['main cat'].replace('', np.nan, inplace = True)
qb['main cat'].fillna(method = 'ffill', inplace = True)

qb['sub cat'].fillna('', inplace = True)
qb = qb[~qb['sub cat'].str.contains('Total')]
qb['sub cat'].replace('', np.nan, inplace = True)
qb['sub cat'].fillna(method = 'ffill', inplace = True)
qb['sub cat'].fillna(qb['main cat'], axis = 0, inplace = True)

# drop columns
qb.drop(['Unnamed: 0','Unnamed: 1','Unnamed: 2','Num','Split','Balance'], 1, inplace = True)

# clean strings
main_fix = qb['main cat'].str.split('·', expand = True)
main_fix[1] = main_fix[1].fillna(main_fix[0])
main_fix.drop(0, 1, inplace = True)
main_fix.rename(columns = {1:'main cat'}, inplace = True)

# merge clean strings
qb.drop('main cat', 1, inplace = True)
qb = pd.merge(qb, main_fix, left_index = True, right_index = True, copy = False)

# clean strings
sub_fix = qb['sub cat'].str.split('·', expand = True)
sub_fix[1] = sub_fix[1].fillna(sub_fix[0])
sub_fix.drop(0, 1, inplace = True)
sub_fix.rename(columns = {1:'sub cat'}, inplace = True)

# merge clean strings
qb.drop('sub cat', 1, inplace = True)
qb = pd.merge(qb, sub_fix, left_index = True, right_index = True, copy = False)

# fill nan with main cat
qb['sub cat'] = np.where(qb['sub cat'].isnull(), qb['main cat'], qb['sub cat'])

# fix numbers
qb['Amount'] = qb['Amount'].str.replace(',','')
qb['Amount'] = pd.to_numeric(qb['Amount'])

# get rid of where amount is null
qb = qb[qb['Amount'].isnull() == False]

# fil string nulls
qb['Name'].fillna('', inplace = True)

# fix memo columns
qb['Memo'].fillna('',inplace = True)

# fix date column and add another
qb = qb[qb['Date'].isnull() == False]
qb['Date'] = pd.to_datetime(qb['Date'])
qb['year and month'] = pd.to_datetime(qb['Date'].dt.year.map(int).map(str) + '-' + qb['Date'].dt.month.map(str))
qb['year and month'] = [str(x)[:7] for x in qb['year and month']]

# fix headers
qb.columns = [x.lower() for x in qb.columns]

# fix more
qb['main cat'] = [x.strip() for x in qb['main cat']]
qb['main cat'] = [x.lower() for x in qb['main cat']]

qb['sub cat'] = [x.strip() for x in qb['sub cat']]
qb['sub cat'] = [x.lower() for x in qb['sub cat']]

qb['income/expense'] = [x.lower() for x in qb['income/expense']]
qb['type'] = [x.lower() for x in qb['type']]
qb['memo'] = [x.lower() for x in qb['memo']]

# organize columns
qb = qb[['income/expense','date', 'year and month','main cat','sub cat','type','name','amount','memo']]

def string_clean(x):
    if x == '401kcontribution':
        return '401k contribution'
    elif x == 'healthcontribution':
        return 'health contribution'
    elif x == 'visioncontribution':
        return 'vision contribution'
    elif x == 'dentalcontribution':
        return 'dental contribution'
    else:
        return x
    
qb['main cat'] = qb['main cat'].apply(string_clean)    

if qb[qb.isnull().any(1)].empty:
    print ('no nulls!')
else:
    print ('you have nulls')
```

```python
def main_cat_rename(x):
    if x['main cat'] == 'insurance, bus':
        return 'business insurance'
    
    elif x['main cat'] == 'telephone, bus':
        return 'business telephone'
    
    elif x['main cat'] == 'contractors':
        return 'contractor wages'    
    
    elif x['main cat'] == 'pension administration - 401k':
        return 'pension administration'
    
    elif x['main cat'] == 'travel- car svc':
        return 'travel: car'
    
    elif x['main cat'] == 'transportation, local':
        return 'local transportation'
    
    elif x['main cat'] == 'meals & entertn':
        return 'meals and entertainment'
    
    elif x['main cat'] == 'travel, bus  (long distance)':
        return 'transportation, long distance'
    
    elif x['main cat'] == 'legal-prof fees':
        return 'legal professional fees'
    
    elif x['main cat'] == 'health contribution':
        return 'health insurance contribution'
    
    elif x['sub cat'] == 'kraft  (commissary)':
        return 'snacks'
    
    elif x['name'] == 'Lyft':
        return 'car service'
    
    elif x['name'] == 'MIT':
        return 'r&d'
    
    elif x['name'] == 'Lynda.com, Inc.':
        return 'education'
    
    elif x['sub cat'] == 'software (saas)':
        return 'SaaS'
    
    elif x['main cat'] == 'freight in':
        return 'total freight'
    
    else:
        return x['main cat']
    
qb['main cat'] = qb.apply(main_cat_rename, axis = 1)    
```

# Create data frame for m2m comparison

```python
# create previous month's date
m2m_compare_start = str((pd.to_datetime(current_year_and_month) - pd.DateOffset(months = 1)).date())[:7]
m2m_compare_start

# get this month and last month
now = qb.copy()
now = now[(now['income/expense'] == 'expense') & (qb['year and month'] >= m2m_compare_start)]

# add a new column for this chart only
def five_pillars_cat(x):
    for key, value in exp_dict.items():
        if x['main cat'] in value:
            return key
        
now['five pillars cat'] = now.apply(five_pillars_cat, axis = 1)

# groupby new column        
now_by_type = now.groupby(['year and month','five pillars cat'])[['amount']].sum()    
now_by_type = now_by_type.unstack(0)
now_by_type.columns = now_by_type.columns.get_level_values(1)
now_by_type.sort_values(current_year_and_month, ascending = False, inplace = True)

# create m2m change
now_by_type['M2M Change'] = now_by_type[current_year_and_month]/now_by_type[m2m_compare_start] - 1

# get other stats
now_by_type['% of Expense'] = now_by_type[current_year_and_month]/cur_exp
now_by_type['% of Income'] = now_by_type[current_year_and_month]/cur_income

# total change
cur_m2m_change = now_by_type[current_year_and_month].sum()/now_by_type[m2m_compare_start].sum() - 1

# percent of income
cur_percent_of_income = now_by_type['% of Income'].sum()
```

# Create data frame for averages

```python
avg_main = qb.copy()

avg_main['five pillars cat'] = avg_main.apply(five_pillars_cat, axis = 1)

avg = avg_main.groupby(['year and month','five pillars cat'])[['amount']].sum().unstack(1)
avg.columns = avg.columns.get_level_values(1)

avg = pd.DataFrame(avg.mean()).rename(columns = {0:'Monthly Mean'})
avg = pd.merge(avg,
              avg_main[avg_main['year and month'] == current_year_and_month].groupby('five pillars cat')[['amount']].sum(),
              left_index = True,
              right_index = True,
              copy = False)

avg.sort_values('amount', ascending = False, inplace = True)
```

```python
retail = pd.read_sql(
'''
SELECT
DATE_FORMAT(o.date_purchased, '%Y-%m') AS 'year and month',
SUM((op.products_quantity - op.products_quantity_free) * op.products_price) AS 'retail revenue'
FROM orders_products op
JOIN orders o ON op.orders_id = o.orders_id
WHERE op.part_id NOT IN (1205,2481,2684,2904,2931,2877,2893,2905,2929,3126,3075,3358,3359,3360,2907,2908,3074)
AND DATE(o.date_purchased) BETWEEN '2017-01-01' AND '2017-10-31'
AND o.orders_status != 9
AND o.orders_status != 10
AND o.payment_method != 'Replacement Order'
AND o.orders_reseller = 0
AND o.orders_super_reseller = 0
AND o.payment_module_code != 'purchaseOrder'
GROUP BY DATE_FORMAT(o.date_purchased, '%Y-%m')
''', db)

retail_by_month = retail.groupby('year and month', as_index = False)[['retail revenue']].sum()
```

```python
all_revenue = pd.read_sql(
'''
SELECT
DATE_FORMAT(o.date_purchased, '%Y-%m') AS 'year and month',
SUM((op.products_quantity - op.products_quantity_free) * op.products_price) AS 'all revenue'
FROM orders_products op
JOIN orders o ON op.orders_id = o.orders_id
WHERE DATE(o.date_purchased) BETWEEN '2017-01-01' AND '2017-10-31'
AND o.orders_status != 9
AND o.orders_status != 10
AND o.payment_method != 'Replacement Order'
GROUP BY DATE_FORMAT(o.date_purchased, '%Y-%m')
''', db)

all_rev_by_month = all_revenue.groupby('year and month', as_index = False)[['all revenue']].sum()
```

```python
# isolate just marketing expenses
just_marketing = qb[qb['main cat'].isin(exp_dict['Marketing and Biz Dev'])].groupby('year and month', as_index = False)[['amount']].sum()

# create a row for Oct 2017
just_marketing = pd.concat([just_marketing, pd.DataFrame({'year and month':'2017-10','amount':65000}, index = [0])])
```

```python
# merge retail rev, all rev, and just marketing
by_month = pd.merge(retail_by_month, just_marketing, on = 'year and month', copy = False)
by_month = pd.merge(by_month, all_rev_by_month, on = 'year and month', copy = False)
by_month = pd.merge(by_month, just_ad_spend[['year and month','ROAS']], on = 'year and month', copy = False)

# rename columns
by_month.rename(columns = {'amount':'marketing cost'}, inplace = True)

# calculate 
by_month['marketing as % of retail revenue'] = by_month['marketing cost']/by_month['retail revenue']

# reorder columns
by_month = by_month[['year and month','marketing cost','retail revenue','all revenue','marketing as % of retail revenue','ROAS']]
```
