# Get CSVs

```python
path = r'/Users/jarad/fake_folder/Finance/Recurring/Authorize.net Audit/Docs/Authorize.net Statements'
all_files = glob.glob(path + '/*.txt')
frame = pd.DataFrame()
list_ = []

for file_ in all_files:
    df = pd.read_csv(file_, error_bad_lines = False, sep = '\t')
    list_.append(df)
    
auth_raw_main = pd.concat(list_)    

auth_raw_main['Submit Date/Time'] = pd.to_datetime(auth_raw_main['Submit Date/Time'])
auth_raw_main['Year and Month'] = pd.to_datetime(auth_raw_main['Submit Date/Time'].dt.year.map(str) + '-' + auth_raw_main['Submit Date/Time'].dt.month.map(str))
auth_raw_main['Year and Month'] = [str(x)[:7] for x in auth_raw_main['Year and Month']]

auth_raw_main['Settlement Date/Time'] = pd.to_datetime(auth_raw_main['Settlement Date/Time'])
auth_raw_main['Year and Month Settlement'] = pd.to_datetime(auth_raw_main['Settlement Date/Time'].dt.year.map(str) + '-' + auth_raw_main['Settlement Date/Time'].dt.month.map(str))
auth_raw_main['Year and Month Settlement'] = [str(x)[:7] for x in auth_raw_main['Year and Month Settlement']]
```

```python
auth_raw = auth_raw_main.copy()

# drop dupes
auth_raw.drop_duplicates(['Transaction ID','Submit Date/Time'], inplace = True)

# lowercase
auth_raw.columns = [x.lower() for x in auth_raw.columns]
auth_raw['action code'] = [x.lower() for x in auth_raw['action code']]

# drop columns which contain all nulls
auth_raw.dropna(how = 'all', axis = 1, inplace = True)
```

# Set transaction fee change

```python
auth_raw['transaction fee'] = np.where(auth_raw['year and month'] <= '2017-04', 0.05, 0.10)
```

# Get auth data from db

```python
auth01 = pd.read_sql(
'''
SELECT
time,
order_id AS 'orders id',
LOWER(authorization_type) AS 'db auth type',
transaction_id AS 'transaction id',
amount AS 'auth db order total'
FROM authorizenet
WHERE order_id IN '''+ str(tuple(oids['orders id'])) +'''
''', db)

auth01['table'] = 'authorizenet' 


auth02 = pd.read_sql(
'''
SELECT
time,
orders_id AS 'orders id',
LOWER(authorization_type) AS 'db auth type',
transaction_id AS 'transaction id',
0.00 AS 'auth db order total'
FROM authorizenet_zero
WHERE orders_id IN '''+ str(tuple(oids['orders id'])) +'''
''', db)

auth02['table'] = 'authorizenet_zero' 

# concat
auth_db_main = pd.concat([auth01, auth02])

# convert to date time
auth_db_main['time'] = pd.to_datetime(auth_db_main['time'])
```

```python
auth_db = auth_db_main.copy()
```

# Get orders total data 

```python
# from 2016 onwards the only payment module codes for credit cards are authorizenet_aim and authorizenet_cim

ot = pd.read_sql(
'''
SELECT
DATE(o.date_purchased) AS date,
DATE_FORMAT(o.date_purchased, '%Y-%m') AS 'year and month',
ot.orders_id AS 'orders id',
ot.value AS 'ada order total'
FROM orders_total ot
JOIN orders o ON ot.orders_id = o.orders_id
WHERE ot.orders_id IN '''+ str(tuple(oids['orders id'])) +'''
AND ot.class = 'ot_total'
''', db)

ot['date'] = pd.to_datetime(ot['date'])
```

# Check the transaction ids from db

### Nulls

```python
if auth_db[auth_db['transaction id'].isnull()].empty:
    print('no nulls')
else:
    print('check your transaction ids')
```

### Zeros

```python
if auth_db[auth_db['transaction id'] == 0].empty:
    print('no zeros')
else:
    print('check your transaction ids')

# get zeros    
zeros = auth_db[auth_db['transaction id'] == 0]

# show them
print('count of transaction ids that are zeros: {:,.0f}'.format(len(zeros.index)))

# remove them
auth_db = auth_db[auth_db['transaction id'] != '0']
```

### Blanks

```python
if auth_db[auth_db['transaction id'] == ''].empty:
    print('transaction id is not blank')
else:
    print('you have blank transaction ids!!')

# get blanks
blank = auth_db[auth_db['transaction id'] == '']

# show them
print('count of blank transaction ids: {:,.0f}'.format(len(blank.index)))

# remove them
auth_db = auth_db[auth_db['transaction id'] != '']

# show them again
print('count of blank transaction ids after removal: {:,.0f}'.format(len(auth_db.index[auth_db['transaction id'] == ''])))

# convert to int
auth_db['transaction id'] = pd.to_numeric(auth_db['transaction id'], downcast = 'integer')
```

# Quick note on auth db transaction ids
* The authorizenet table shows the last authorization and the capture, even if there have been several auth attempts
* The authorizenet_zero table shows every auth attemp
* Because the authorizenet table behaves this way, the transaction id belonging to the captured line is also given to the final auth attempt
* We have to exclude this final auth attemp from the authorizenet table in order to reconcile the db with the CSV

```python
auth_db = auth_db.sort_values(['orders id','db auth type','transaction id']).drop_duplicates('transaction id', keep = 'last')
```

# Merge to create the audit

```python
audit = pd.merge(auth_db,
                 auth_raw[['transaction id','total amount','response code','action code','year and month','transaction fee']],
                 how = 'left',
                 on = 'transaction id').merge(ot[['orders id','ada order total']], how = 'left', on = 'orders id')

# tidy up after the join    
audit['ada order total'] = np.where(audit['action code'].isin(['auth_only','void','expired']), 0, audit['ada order total'])

audit['ada order total'] = [np.round(x, 2) for x in audit['ada order total']]
audit['auth db order total'] = [np.round(x, 2) for x in audit['auth db order total']]
audit['total amount'] = [np.round(x, 2) for x in audit['total amount']]
```

# Find mismatches

### A review on action codes
***
* AUTH_CAPTURE (Authorize and Capture): This is the default transaction type. When using AIM, if the x_type field is not sent to us, the type will default to AUTH_CAPTURE. Simple Checkout uses AUTH_CAPTURE only. The Virtual Terminal defaults to AUTH_CAPTURE unless you select a different transaction type. With an AUTH_CAPTURE transaction, the process is completely automatic. The transaction is submitted to your processor for authorization and, if approved, is placed in your Unsettled Transactions with the status Captured Pending Settlement. The transaction will settle at your next batch. Settlement occurs every 24 hours, within 24 hours of your Transaction Cut-off Time.
***
* AUTH_ONLY (Authorization Only): When an AUTH_ONLY transaction is submitted to Authorize.Net, we send the transaction to your processor for authorization. If approved, the transaction is placed in your Unsettled Transactions with a status of Authorized/Pending Capture.
***
* PRIOR_AUTH_CAPTURE (Prior Authorization Capture): PRIOR_AUTH_CAPTURE transactions are used to capture funds for an existing transaction with a status of Authorized/Pending Capture, and can only be submitted from a website or billing application.

### In summary:
* We only really care about mismatches where the action code equals prior_auth_capture
***
### Also:
* If the count of mismatches is not too high, we will assume that they are manual refunds
* There is currently no whole-hearted way to verify all refunds since so manu are manually entered

```python
mismatches = audit.copy()
mismatches = mismatches[mismatches['action code'].isin(['prior_auth_capture','auth_capture'])]
mismatches = mismatches[np.abs(mismatches['total amount'] - mismatches['ada order total']) > 1]
```

```python
mismatches_prior_auth_capture = mismatches.copy()
mismatches_prior_auth_capture = mismatches_prior_auth_capture[mismatches_prior_auth_capture['action code'] == 'prior_auth_capture']
mismatches_prior_auth_capture['diff'] = mismatches_prior_auth_capture['auth db order total'] - mismatches_prior_auth_capture['ada order total']
mismatches_prior_auth_capture
```

### Credit card revenue

```python
ot_by_month = ot.groupby('year and month')[['ada order total']].sum()
ot_by_month['% change'] = ot_by_month['ada order total'].pct_change()
ot_by_month.format_(['m0','p2'])
```

```python
types = pd.read_sql(
'''
SELECT
DATE_FORMAT(date_purchased, '%Y-%m') AS 'year and month',
order_total AS 'order total',
payment_method AS 'payment method'
FROM orders
WHERE DATE(date_purchased) BETWEEN ' '''+ date_start +''' ' AND ' '''+ date_end +''' '
AND orders_status != 9
AND orders_status != 10
''', db)
```

```python
types_by_month = types.groupby(['year and month','payment method'])[['order total']].sum().unstack(1).fillna(0)
types_by_month.columns = types_by_month.columns.get_level_values(1)
types_by_month['total'] = types_by_month.sum(1)
types_by_month.format_(['m0','m0','m0','m0','m0','m0','m0','m0','m0','m0','m0','m0',])
```

```python
types_by_month['Credit Card']/types_by_month['total']
```

```python
types_breakdown = types_by_month.iloc[-1:,:-1]
types_breakdown = types_breakdown.T.sort_values(date_end[:7], ascending = False).T

fig = plt.figure(figsize = (20,5))
ax = fig.add_subplot(1,1,1)

types_breakdown.plot(kind = 'bar',
                     edgecolor = 'black',
                     ax = ax)
ax.set_xlim(-0.3, 0.3)
ax.set_xticks([])
ax.set_xlabel(calendar.month_abbr[int(date_end[5:-3])] + ' ' + date_end[:4],
                  fontname = 'Arial',
                  fontsize = 20,
                  fontweight = 'bold')
vals = ax.get_yticks()
ax.set_yticklabels(['${:,.0f}'.format(x/1000) for x in vals],
                  fontname = 'Arial',
                  fontsize = 20,
                  fontweight = 'bold')
ax.set_ylabel('Thousands',
                  fontname = 'Arial',
                  fontsize = 30,
                  fontweight = 'bold')
ax.set_title('Order Totals by Payment Type',
                  fontname = 'Arial',
                  fontsize = 30,
                  fontweight = 'bold')
ax.set_facecolor(colors['color03'])
ax.legend(fontsize = 15)
plt.show()
```

### Number of transactions

```python
auth_raw_by_month = auth_raw[auth_raw['year and month settlement'].between(date_start[:7], date_end[:7])].groupby('year and month settlement')[['transaction id']].count().rename(columns = {'transaction id':'transaction count'})
auth_raw_by_month['% change'] = auth_raw_by_month['transaction count'].pct_change()
auth_raw_by_month.format_(['n0','p2'])
```

```python
fig = plt.figure(figsize = (20,5))
ax = fig.add_subplot(1,1,1)

x = np.arange(len(auth_raw_by_month))
y = auth_raw_by_month['transaction count']
y2 = [auth_raw_by_month['transaction count'].mean()] * 13

plot01, = ax.plot(x, y,
                   '--o',
                   markersize = 10,
                   linewidth = 5,
                   color = colors['color01'],
                   label = 'Count')
plot02, = ax.plot(x,y2,
                    color = 'white',
                    linewidth = 5,
                    label = 'Monthly Avg')
ax.set_xticks(x)
ax.set_xticklabels([calendar.month_abbr[int(str(x)[-2:])] + ' ' + str(x)[:4] for x in auth_raw_by_month.index],
                  fontname = 'Arial',
                  fontsize = 15,
                  fontweight = 'bold')
vals = ax.get_yticks()
ax.set_yticklabels(['{:,.0f}'.format(x) for x in vals],
                  fontname = 'Arial',
                  fontsize = 20,
                  fontweight = 'bold')
ax.set_title('Count of Transactions Over Time',
                  fontname = 'Arial',
                  fontsize = 30,
                  fontweight = 'bold')
ax.legend(handles = [plot01,plot02],
         fontsize = 15)
ax.set_facecolor(colors['color03'])
```

### Transaction fees

```python
fees_by_month = auth_raw[auth_raw['year and month settlement'].between(date_start[:7], date_end[:7])].groupby('year and month settlement')[['transaction fee']].sum()
fees_by_month['% change'] = fees_by_month['transaction fee'].pct_change()
fees_by_month.format_(['m0','p2'])
```

```python
fig1 = plt.figure(figsize = (20,5))
ax = fig1.add_subplot(1,1,1)

# plots
fees_by_month.plot(kind = 'bar', 
                    ax = ax, 
                    use_index = False,
                    edgecolor = 'white',
                    color = colors['color01'])

x = range(len(fees_by_month.index))
y_mean = [fees_by_month.mean().values[0]] * 13
plot02, = ax.plot(x, y_mean, color = 'white', linewidth = 5, label = 'Monthly Avg')

ax.set_xticklabels([calendar.month_abbr[int(x[-2:])] + ' ' + x[:4] for x in fees_by_month.index],
                  rotation = 0,
                  fontsize = 15,
                  fontweight = 'bold',
                  fontname = 'Arial')

ax.set_ylim(0, max(fees_by_month['transaction fee'] + 100))
vals = ax.get_yticks()
ax.set_yticklabels(['${:,.0f}'.format(x) for x in vals],
                   fontsize = 20,
                   fontweight = 'bold',
                   fontname = 'Arial')
ax.set_title('Transaction Fees',
              fontsize = 30,
              fontweight = 'bold',
              fontname = 'Arial')

ax.legend(['Average','Monthly Fee'],
         fontsize = 15)

ax.set_facecolor(colors['color03'])
plt.show()
```

### Fees as % of revenue

```python
fees_as_perc = pd.merge(ot_by_month[['ada order total']],
                       fees_by_month[['transaction fee']],
                       left_index = True,
                       right_index = True)

fees_as_perc['fees as % of revenue'] = fees_as_perc['transaction fee']/fees_as_perc['ada order total']
fees_as_perc.format_(['m0','m0','p2'])
```

### Approval %
* 1: approved
* 2: declined
* 3: error
* 4: action required (typically used for AFDS transactions that are held for review)

```python
approval = auth_raw[auth_raw['year and month settlement'].between(date_start[:7], date_end[:7])].groupby(['year and month settlement','response code'])[['transaction id']].count().unstack(1).fillna(0)
approval.columns = approval.columns.get_level_values(1)
approval['approval %'] = approval[1]/approval[[1,2,3]].sum(1)
```
