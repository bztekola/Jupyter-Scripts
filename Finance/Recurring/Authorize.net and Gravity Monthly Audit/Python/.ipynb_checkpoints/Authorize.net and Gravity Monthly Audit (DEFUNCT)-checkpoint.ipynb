# Get CSVs

```python
path = r'/Users/jarad/fake_folder/Finance/Recurring/Authorize.net and Gravity Monthly Audit/Docs/Authorize.net Statements'
all_files = glob.glob(path + '/*.txt')
frame = pd.DataFrame()
list_ = []

for file_ in all_files:
    df = pd.read_csv(file_, error_bad_lines = False, sep = '\t', low_memory = False)
    list_.append(df)
    
auth_raw_main = pd.concat(list_)    

# lowercase
auth_raw_main.columns = [x.lower() for x in auth_raw_main.columns]
auth_raw_main['action code'] = [x.lower() for x in auth_raw_main['action code']]

# change to string for later stuff
auth_raw_main['transaction id'] = [str(x) for x in auth_raw_main['transaction id']]

# drop columns which contain all nulls
auth_raw_main.dropna(how = 'all', axis = 1, inplace = True)

# get date columns
auth_raw_main['submit date/time'] = pd.to_datetime(auth_raw_main['submit date/time'])
auth_raw_main['year and month submit'] = [str(x)[:7] for x in auth_raw_main['submit date/time']]

auth_raw_main['settlement date/time'] = pd.to_datetime(auth_raw_main['settlement date/time'])
auth_raw_main['date settlement'] = [x.date() for x in auth_raw_main['settlement date/time']]
auth_raw_main['year and month settlement'] = [str(x)[:7] for x in auth_raw_main['settlement date/time']]

# set transaction fee change
# it changed from $0.05 to $0.10 on 2017-04-01
auth_raw_main['transaction fee'] = np.where(auth_raw_main['year and month submit'] <= '2017-04', 0.05, 0.10)

# change column name
auth_raw_main.rename(columns = {'total amount':'csv total',
                               'response code':'csv response code'}, inplace = True)

# clean this up
auth_raw_main['action code'] = [x.replace('_',' ') for x in auth_raw_main['action code']]
```

```python
auth_raw = auth_raw_main.copy()

# drop dupes
auth_raw.drop_duplicates(['transaction id','submit date/time'], inplace = True)

# slim it down
auth_raw = auth_raw[['csv response code',
                      'transaction id',
                      'invoice number',
                      'csv total',
                      'action code',
                      'customer id',
                      'year and month submit',
                      'date settlement',
                      'year and month settlement',
                      'transaction fee']]

print([x for x in auth_raw['action code'].unique()])
```

```python
auth_db_01 = pd.read_sql(
'''
SELECT
order_id AS 'orders id',
response_code AS 'db response code',
LOWER(authorization_type) AS 'db action code',
transaction_id AS 'transaction id'
FROM authorizenet
WHERE transaction_id >= '''+ str(auth_raw['transaction id'].min()) +'''
''', db)

auth_db_01['db action code'] = [x.replace('_',' ') for x in auth_db_01['db action code']]

#=====================================================================================

auth_db_02 = pd.read_sql(
'''
SELECT
orders_id AS 'orders id',
response_code AS 'db response code',
LOWER(authorization_type) AS 'db action code',
transaction_id AS 'transaction id'
FROM authorizenet_zero
WHERE transaction_id >= '''+ str(auth_raw['transaction id'].min()) +'''
''', db)

auth_db_02['db action code'] = [x.replace('_',' ') for x in auth_db_02['db action code']]

#=====================================================================================

auth_db_main = pd.concat([auth_db_01, auth_db_02])
auth_db_main.drop_duplicates(inplace = True)
```

```python
audit_01 = pd.merge(auth_raw,
                    auth_db_main,
                    how = 'left',
                    on = 'transaction id')   

# usually the OID is buried in the so called "invoice numer" so we extract it here
# this is always the case so be careful using this for anything super important
for index, row in audit_01.iterrows():
    try:
        if math.isnan(row['orders id']):
            if len(row['invoice number'].split('-')[0]) == 7:
                audit_01.loc[index, 'orders id'] = row['invoice number'].split('-')[0]
    except:
        pass
    
audit_01['orders id'].fillna(0, inplace = True)
audit_01['orders id'] = [int(x) for x in audit_01['orders id']]
audit_01['customer id'].fillna(0, inplace = True)
```

```python
ot = pd.read_sql(
'''
SELECT
DATE(o.date_purchased) AS 'date purchased',
DATE_FORMAT(o.date_purchased, '%Y-%m') AS 'year and month purchased',
o.orders_id AS 'orders id',
ot.value AS 'ot total'
FROM orders o
JOIN orders_total ot ON o.orders_id = ot.orders_id
AND ot.class = 'ot_total'
# one month behind the "date start" that we set way up at the start of this script
WHERE DATE(o.date_purchased) >= ' '''+ str((pd.to_datetime(date_start) - pd.DateOffset(months = 1)).date()) +''' '
AND o.payment_method IN ('Credit Card', 'Subscription','Purchase Order')
''', db)

#=====================================================================================

od = pd.read_sql(
'''
SELECT
orders_id AS 'orders id'
FROM orders_deleted
''', db)

#=====================================================================================

fake_orders = pd.read_sql(
'''
SELECT
orders_id AS 'orders id'
FROM orders
WHERE orders_fake = 1
''', db)
```

```python
audit = pd.merge(audit_01,
                 ot,
                 how = 'left',
                 on = 'orders id')

def ot_total_fix(x):
    if x['action code'] == 'auth only':
        return 0
    else:
        return x['ot total']
    
audit['ot total'] = audit.apply(ot_total_fix, axis = 1)

audit['csv total'] = [np.round(x, 2) for x in audit['csv total']]
audit['ot total'] = [np.round(x, 2) for x in audit['ot total']]

audit.drop_duplicates(['orders id','transaction id'], inplace = True)
audit['ot total'].fillna(0, inplace = True)
audit['customer id'].fillna(0, inplace = True)

audit['deleted'] = np.where(audit['orders id'].isin(od['orders id'].tolist()), 'yes','no')
audit['fake order'] = np.where(audit['orders id'].isin(fake_orders['orders id'].tolist()), 'yes','no')

audit['db response code'].fillna(0, inplace = True)
audit['db action code'].fillna(audit['action code'], inplace = True)

audit = audit[audit['year and month settlement'] <= date_end[:7]]
```

```python
response_codes_dict = {1:'approved',
                       2:'declined',
                       3:'error',
                       4:'action required'}

audit['response code'] = audit['csv response code'].map(response_codes_dict)
```

```python
nulls = audit[(audit.isnull().any(1))
       & (audit['deleted'] == 'no')
       & (audit['orders id'] > 0)
       & (audit['action code'].isin(['auth capture']))
       & (audit['response code'] == 'approved')]
if nulls.empty:
    print('no nulls where action code is "auth capture"')
else:
    print('you have %i null(s) where action code is "auth capture"' % len(nulls))
```

### Approval Percentage
* from email with subject line "Quick question on benchmarks from Adafruit Industries"
* If you are attempting to take a transaction download and calculate it into an approval percentage there are a few steps. 

 

1.       Count all Response Code 1 transactions =COUNTIF(Range,"1")

2.       Count then Minus all Action Code void and Expired transactions from the previous steps results =COUNTIF(Range,"Void") , =COUNTIF(Range,"Expired")

3.       Count all 2 transactions =COUNTIF(Range,"2")

4.       Add the totals from step 2 and 3 to get the total transaction count

5.       Divide step 3 by step 4 to get the decline percentage

6.       1 minus step 5 results should give you the approval percentage listed in Transaction Statistics



```python
df1 = audit[(audit['csv response code'] == 1)
          & (~audit['action code'].isin(['expired','void']))].groupby('year and month settlement')[['transaction id']].count()
df1.columns = ['approved']

df2 = audit[audit['csv response code'] == 2].groupby('year and month settlement')[['transaction id']].count()
df2.columns = ['not approved']

df3 = df1.join(df2)
df3['total transaction count'] = df3.sum(1)

df3['approval %'] = df3['approved']/df3['total transaction count']

approval_by_month = df3.copy()

m = pd.DataFrame(approval_by_month.mean())
m.columns = ['mean']
m = m.T
m = m.format_(['n0'] * 3 + ['p2'])
m = m.T
print(m)
approval_by_month.format_(['n0'] * 3 + ['p2'])
```

### Fraud and Chargebacks
* From Kelly in email with subject line "CSP Monthly Chargeback/Fraud Report"
*****
* Fraud - Void (orders_status = 15):
* fraud we stopped
*****
* Fraud - Confirmed (orders_status = 9):
* fraud we did not stop in time and which often result in chargebacks being filed later on.

```python
osh_super_main = pd.read_sql(
'''
SELECT
osh.orders_status_history_id,
DATE(osh.date_added) AS date_added,
DATE_FORMAT(osh.date_added, '%Y-%m') AS year_and_month,
osh.orders_id,
o.order_total,
os.orders_status_name,
o.payment_method
FROM orders_status_history osh
JOIN orders_status os ON osh.orders_status_id = os.orders_status_id
JOIN orders o ON osh.orders_id = o.orders_id
WHERE DATE(osh.date_added) BETWEEN ' '''+ date_start +''' ' AND ' '''+  date_end +''' '
''', db)

col_fix(osh_super_main)

osh_main = osh_super_main.copy()
osh_main.sort_values(['orders id','orders status history id'], inplace = True)
osh_main.drop_duplicates('orders id', keep = 'last', inplace = True)
```

```python
ls = ['Fraud - Void','Fraud - Confirmed'] # fraud stopped, fraud not stopped
f = osh_main[osh_main['orders status name'].isin(ls)].copy()

d = {ls[0]:'fraud stopped',
    ls[1]:'fraud not stopped'}

for k,v in d.items():
    f['orders status name'] = f['orders status name'].replace(k,v)

f2 = f.groupby(['year and month','orders status name'])[['order total']].sum().unstack(1).fillna(0)
f2.columns = f2.columns.droplevel(0)
f2['total fraud'] = f2.sum(1)
f2['stopped as % of total'] = f2['fraud stopped']/f2['total fraud']

m = pd.DataFrame(f2.mean())
m.columns = ['mean']
m = m.T
m = m.format_(['m0','m0','m0','p2'])
m = m.T
print(m)

f2.format_(['m0','m0','m0','p2'])
```
