# Get data

```python
data_main = pd.read_sql(
'''
SELECT
DATE(o.date_purchased) AS date,
o.orders_id AS 'orders id',
customers_id AS name,
ot.value AS 'subtotal'
FROM orders_total ot
JOIN orders o ON ot.orders_id = o.orders_id
WHERE o.customers_id IN '''+ str(tuple(id_list)) +'''
AND ot.class = 'ot_subtotal'
AND o.orders_id NOT IN (SELECT orders_id FROM orders_products WHERE part_id = 1205)
''', db)

data_main['name'] = [str(x) for x in data_main['name']]
data_main['date'] = pd.to_datetime(data_main['date'])

for key, value in resellers.items():
    data_main['name'] = data_main['name'].str.replace(str(value[0]), key)
    
arrow_data = pd.read_sql(
'''
SELECT
DATE(o.date_purchased) AS date,
o.orders_id AS 'orders id',
'Arrow' AS name,
ot.value AS 'subtotal'
FROM orders_total ot
JOIN orders o ON ot.orders_id = o.orders_id
WHERE o.customers_id IN '''+ str(tuple(arrow_ids)) +'''
AND ot.class = 'ot_subtotal'
AND o.orders_id NOT IN (SELECT orders_id FROM orders_products WHERE part_id = 1205)
''', db)

arrow_data['date'] = pd.to_datetime(arrow_data['date'])    

data = pd.concat([data_main, arrow_data])
```

# Create fake company for illustration
* good one sided paired t-test calculator [here](http://www.socscistatistics.com/tests/ttestdependent/Default2.aspx)

```python
fake_before = np.array(np.random.randint(10, 15, 5))
fake_after = np.array(np.random.randint(20, 30, 5))

print(np.mean(fake_before))
print(np.mean(fake_after))
print('')
print(np.std(fake_before))
print(np.std(fake_after))
print('')
print(fake_before)
print(fake_after)
```

```python
subs = pd.DataFrame({'subtotals before':fake_before,
             'subtotals after':fake_after})

subs[['subtotals before','subtotals after']].format_(['m0','m0'])
```

```python
# H0: before - after = 0
# H1: before - after < 0
fake_results = {}
fake_test = stats.ttest_ind(fake_before, fake_after)
fake_test_stat = fake_test[0]
fake_pvalue = fake_test[1]/2

df = (len(fake_before) + len(fake_after)) - 2
fake_crit_value = stats.t.ppf(0.05, df)

if fake_test_stat <= fake_crit_value:
    fake_res = 'yes, reports make a difference' # reject the null
else:
    fake_res = 'no, reports did not make a difference' # fail to reject null

fake_results['fake company'] = [len(fake_before), 
                                 len(fake_after),
                                 fake_test_stat, 
                                 fake_crit_value,
                                 fake_res,
                                 np.mean(fake_before),
                                 np.mean(fake_after)]

pd.DataFrame.from_items(fake_results.items(),
                       orient = 'index',
                       columns = ['count before',
                                  'count after',
                                  'test statistic',
                                  'critical value',
                                  'result',
                                  'aov before',
                                  'aov after'])
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)

pd.DataFrame({'before':fake_before}).plot(kind = 'density', label = 'before', ax = ax)
pd.DataFrame({'after':fake_after}).plot(kind = 'density', label = 'after', ax = ax)
plt.xticks([])
plt.title('Fake Company AOV')
plt.legend()
```

# Analyze

```python
resellers
```

```python
results = {}
for key, value in resellers.items():
    
    df = data.copy()
    df = df[df['name'] == key]
    
    if key == 'Arrow': 
        
        # arrow has ordered more after the reports than before
        # as a result our arrays are not the same length
        # here we reverse the order of the other resellers
        # we count how many orders arrow had before the reports
        # and measure the AOV of the same number orders after the reports
        # not perfect but ok for now
        
        # before
        df_before = df.copy()
        df_before = df_before[df_before['date'] <= value[1]]
        order_count = len(df_before)

        before = np.array(df_before['subtotal'])

        # after
        df_after = df.copy()
        df_after = df_after[df_after['date'] > value[1]]
        df_after.sort_values('date', ascending = True, inplace = True)
        df_after.reset_index(drop = True, inplace = True)
        df_after = df_after.iloc[:order_count]

        after = np.array(df_after['subtotal'])

    else:
    
        # after
        df_after = df.copy()
        df_after = df_after[df_after['date'] > value[1]]
        order_count = len(df_after)

        after = np.array(df_after['subtotal'])

        # before
        df_before = df.copy()
        df_before = df_before[df_before['date'] <= value[1]]
        df_before.sort_values('date', ascending = False, inplace = True)
        df_before.reset_index(drop = True, inplace = True)
        df_before = df_before.iloc[:order_count]

        before = np.array(df_before['subtotal'])
    
    # we compare before and after for each reseller
    # this is right tailed paired t-test
    
    # null: mean_before - mean_after = 0
    # alt: mean_before - mean_after < 0
    
    if np.abs(np.std(before) - np.std(after)) < 1:
        equal = True
    else:
        equal = False
    
    test = stats.ttest_ind(before, after, equal_var = equal)
    test_statistic = test[0]
    p_value = test[1]/2
    
    # this is for the right tail
    df = ((len(df_before) + len(df_after)) - 2)
    crit_value = stats.t.ppf(0.05, df)
    
    if test_statistic <= crit_value:
        stat_res = 'yes, reports made a difference'
    else:
        stat_res = 'no, reports did not make a difference'
    
    fig = plt.figure()
    ax = fig.add_subplot(1,1,1)
    df_after['subtotal'].plot(kind = 'density', label = 'after', ax = ax)
    df_before['subtotal'].plot(kind = 'density', label = 'before', ax = ax)
    plt.title(key)
    plt.legend()

    results[key] = [len(df_before), 
                    test_statistic, 
                    crit_value, 
                    p_value, 
                    stat_res,
                    df_before['subtotal'].mean(),
                    df_after['subtotal'].mean(),
                    df_before['subtotal'].std()]
    
results = pd.DataFrame.from_items(results.items(),
                                   orient = 'index',
                                   columns = ['order count',
                                             'test statistic',
                                             'critical value',
                                             'p value',
                                             'stat result',
                                             'aov before',
                                             'aov after',
                                             'aov before sigma'])    

results.sort_values('stat result', ascending = False, inplace = True)
results
```

```python
results_for_report = results.copy()
results_for_report = results_for_report[['order count','aov before','aov after','stat result']]
results_for_report.rename(columns = {'order count before':'number of orders considered'}, inplace = True)
results_for_report.format_([0,'m0','m0',0])
```
