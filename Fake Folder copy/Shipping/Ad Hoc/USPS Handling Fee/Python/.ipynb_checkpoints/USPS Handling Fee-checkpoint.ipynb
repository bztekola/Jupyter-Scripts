### Get data

```python
ces_main = pd.read_sql(
'''
SELECT 
DATE(date_purchased) AS date_purchased,
DATE_FORMAT(date_purchased, '%Y-%m') AS year_and_month,
o.orders_id,
order_total,
issue_type
FROM cs_error_reports c
LEFT JOIN orders o ON o.orders_id = c.orders_id
AND DATE(date_purchased) BETWEEN ' '''+ date_start +''' ' AND ' '''+ date_end +''' '
AND shipping_module_code = 'usps'
''', db)

col_fix(ces_main)
ces_main.dropna(inplace = True)
ces_main['issue type'] = ces_main['issue type'].str.replace('_',' ')

d = ces_main[ces_main['orders id'].duplicated()]
print('{:,.0f} dupes'.format(len(d)))
ces_main.drop_duplicates(inplace = True)
```

```python
ces_main.groupby('issue type')[['order total']].sum().sort_values('order total').plot(kind = 'barh',
                                                                                      grid = True, 
                                                                                      zorder = 5, 
                                                                                      edgecolor = 'black')
```

### Isolate specific issue_types

```python
set(ces_main['issue type'])
```

```python
ls = ['damaged','lost']
ces = ces_main[ces_main['issue type'].isin(ls)].copy()

by_month = ces.groupby('year and month')[['order total']].sum()
m = by_month.mean().values[0]
s = by_month.std().values[0]
n = len(by_month)
phi = stats.t.ppf(0.95, n-1)

l = m - phi * (s/np.sqrt(n))
u = m + phi * (s/np.sqrt(n))

print('95% confidence interval for avg order total per month for orders that were ' + ' or '.join(ls))
pd.DataFrame(columns = ['lower','mean','upper'], data = [[l,m,u]]).format_(['m0'] * 3)
```

### Just USPS First Class

```python
st = '2018-01-01'
en = '2018-10-31'

ot_main = pd.read_sql(
'''
SELECT
*
FROM orders_total
WHERE class IN ('ot_shipping','ot_subtotal')
AND orders_id IN
    (SELECT 
    orders_id 
    FROM orders
    WHERE DATE(date_purchased) BETWEEN ' '''+ date_start +''' ' AND ' '''+ date_end +''' '
    AND shipping_module_code = 'usps'
    AND orders_status = 3)
''', db)

col_fix(ot_main)
```

```python
n = len(set(ot_main['orders id']))
print('{:,.0f} OIDs that were shipped via USPS from {} to {}'.format(n, st, en))
```

```python
oids = ot_main[ot_main['title'].str.contains('First-Class', case = False)]['orders id'].tolist()
df = ot_main[ot_main['orders id'].isin(oids)]

print('{:,.0f} OIDs that were shipped via USPS First-Class'.format(len(oids)))
print('or {:,.2f}% of all USPS orders'.format(len(oids)/n * 100))

by_oid = df.groupby(['orders id','class']).agg({'value':'sum'}).unstack(1)
by_oid.columns = by_oid.columns.droplevel(0)
```

```python
col = ot_main[(~ot_main['title'].str.contains('First-Class', case = False))
       & (ot_main['class'] == 'ot_shipping')]['title'].str.split('(', expand = True)[2]
set(col)
```

```python
print('subtotal percentiles of USPS First-Class')
by_oid['ot_subtotal'].describe(percentiles = np.arange(0,1.01,0.01)).tail()
```

```python
m = by_oid['ot_shipping'].mean()
s = by_oid['ot_shipping'].std()
n = len(by_oid)
phi = 1.96
l = m - phi * (s/np.sqrt(n))
u = m + phi * (s/np.sqrt(n))

print('95% confidence interval for avg USPS First-Class shipping charge per order')
pd.DataFrame(columns = ['lower','mean','upper'], data = [[l,m,u]]).format_(['m2'] * 3)
```
