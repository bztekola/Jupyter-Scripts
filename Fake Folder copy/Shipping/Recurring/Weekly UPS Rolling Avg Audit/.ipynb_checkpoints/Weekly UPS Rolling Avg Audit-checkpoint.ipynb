# Get data

```python
# get max date where there is a basis value
# use this as your 2017 date limit

ups_max_date = pd.read_sql(
'''
SELECT
MAX(transaction_date) As 'max date'
FROM ups_billing
WHERE basis_value != 0
''', db)
```

```python
ups_max_date
```

```python
ups_main = pd.read_sql(
'''
SELECT
DATE(transaction_date) AS date,
DATE_FORMAT(transaction_date, '%Y-%m') AS 'year and month',
basis_value AS 'gross charge'
FROM ups_billing
WHERE DATE(transaction_date) <= ' '''+ str(ups_max_date['max date'].values[0]) +''' '
''', db)

ups_main['date'] = pd.to_datetime(ups_main['date'])
ups_main['year'] = ups_main['date'].dt.year
ups_main['week'] = ups_main['date'].dt.week
```

```python
ups_main[::-1]
```

```python
# group by year and week
ups_group = ups_main.groupby(['year','week'], as_index = False)[['gross charge']].sum()
```

```python
# fill in the weeks where no UPS shipments ocurred
# but remember that 2017 is not finished yet
# therefore you need to get the max week that corrsponds to your max date, but only for 2017
max_week = ups_main['week'][(ups_main['year'] == 2017)].max()

ups = pd.DataFrame()
for year in ups_group['year'].unique():
    if year != 2017:
        weeks = pd.DataFrame({'week':range(1,54)})
    else:
        weeks = pd.DataFrame({'week':range(1, int(max_week) - 1)})
        
    df = ups_group[ups_group['year'] == year]
    df = pd.merge(weeks, df, how = 'left', on = 'week')

    df['year'].fillna(year, inplace = True)
    df['gross charge'].fillna(0,inplace = True)
    
    ups = ups.append(df, ignore_index = True)
    
ups = ups[['year','week','gross charge']]    
```

# Get 52 week rolling mean without any ramp ups

```python
ups['rolling mean'] = ups['gross charge'].rolling(window = 52).mean()
ups.dropna(inplace = True)
```

```python
colors = {
'color01':'#ffe8ff',
'color02':'#fea3a0',
'color03':'#49224d',
'color04':'#015a6c',
'color05':'#073a60'
}
```

```python
fig = plt.figure(figsize = (20,5))
ax = fig.add_subplot(1,1,1)

x = np.arange(len(ups))
y = ups['rolling mean']

ax.plot(x, y, linewidth = 8, color = colors['color02'])

ax.set_xticks(np.arange(0, len(ups), len(ups)/4))
ax.set_xticklabels([int(x) for x in ups['year'].unique()],
                  fontname = 'Arial',
                  fontsize = 20,
                  fontweight = 'bold')
ax.set_xlabel('Year',
              fontname = 'Arial',
              fontsize = 20,
              fontweight = 'bold')

vals = ax.get_yticks()
ax.set_yticklabels(['${:,.0f}'.format(x) for x in vals],
                  fontname = 'Arial',
                  fontsize = 20,
                  fontweight = 'bold')
ax.set_ylabel('Rolling Avg',
                  fontname = 'Arial',
                  fontsize = 20,
                  fontweight = 'bold')

ax.set_title('UPS 52 Weekly Rolling Avg without any Ramp Ups',
                  fontname = 'Arial',
                  fontsize = 20,
                  fontweight = 'bold')

ax.set_facecolor(colors['color04'])
ax.grid()
plt.show()
```

```python
just_2017 = ups.copy()
just_2017 = just_2017[just_2017['year'] == 2017]

fig = plt.figure(figsize = (20,5))
ax = fig.add_subplot(1,1,1)

x = np.arange(len(just_2017))
y = just_2017['rolling mean']

ax.plot(x, y, linewidth = 8, color = colors['color02'])

ax.set_xticks([])
ax.set_xlabel('Jan 2017 to Now',
              fontname = 'Arial',
              fontsize = 20,
              fontweight = 'bold')

vals = ax.get_yticks()
ax.set_yticklabels(['${:,.0f}'.format(x) for x in vals],
                  fontname = 'Arial',
                  fontsize = 20,
                  fontweight = 'bold')
ax.set_ylabel('Rolling Avg',
                  fontname = 'Arial',
                  fontsize = 20,
                  fontweight = 'bold')

ax.set_title('UPS 52 Weekly Rolling Avg without any Ramp Ups\nJust 2017',
                  fontname = 'Arial',
                  fontsize = 20,
                  fontweight = 'bold')

ax.set_facecolor(colors['color04'])
ax.grid()
plt.show()
```

# Avg costs

```python
basis = pd.read_sql(
'''
SELECT
orders_id,
billed_weight,
basis_value,
incentiveAmount,
netAmount
FROM ups_billing
WHERE orders_id IN (SELECT orders_id FROM ups_billing WHERE charge_description = 'Ground Residential')
AND DATE(transaction_date) BETWEEN '2017-11-11' AND '2017-12-11'
AND orders_id != 0
''', db)
```

```python
by_oid = basis.groupby('orders_id')[['basis_value']].sum()
by_oid['$ off at 10%'] = by_oid['basis_value'] * 0.10
by_oid['$ off at 7.10%'] = by_oid['basis_value'] * 0.071
by_oid['diff'] = by_oid['$ off at 10%'] - by_oid['$ off at 7.10%']
```

```python
by_oid['diff'].describe()
```

```python
pd.read_sql(
'''
SELECT
DISTINCT charge_description
FROM ups_billing
WHERE (charge_description LIKE 'worldwide%' OR charge_description LIKE 'ww%')
AND DATE(transaction_date) BETWEEN '2017-11-11' AND '2017-12-11'
AND orders_id != 0
''', db)
```

```python
ww_basis = pd.read_sql(
'''
SELECT
orders_id,
billed_weight,
basis_value,
incentiveAmount,
netAmount
FROM ups_billing
WHERE orders_id IN (SELECT orders_id FROM ups_billing WHERE charge_description LIKE 'worldwide%' OR charge_description LIKE 'ww%')
AND DATE(transaction_date) BETWEEN '2017-11-11' AND '2017-12-11'
AND orders_id != 0
''', db)
```

```python
ww_by_oid = ww_basis.groupby('orders_id')[['basis_value']].sum()
ww_by_oid['$ off at 20%'] = ww_by_oid['basis_value'] * 0.2
ww_by_oid['$ off at 5%'] = ww_by_oid['basis_value'] * 0.05
ww_by_oid['diff'] = ww_by_oid['$ off at 20%'] - ww_by_oid['$ off at 5%']
```

```python
ww_by_oid['diff'].describe()
```

# DHL and UPS

```python
inter = pd.read_sql(
'''
SELECT
DATE(date_purchased) AS date,
orders_id,
shipping_module_code AS service
FROM orders
WHERE DATE(date_purchased) BETWEEN '2017-03-01' AND '2017-12-11'
AND shipping_module_code IN ('upsxml','dhlexpress')
AND delivery_country NOT IN ('united states','united states ','us')
''', db)
```

```python
by_inter_service = inter.groupby(['date','service'])[['orders_id']].count().unstack(1).fillna(0)
by_inter_service.columns = by_inter_service.columns.get_level_values(1)

fig = plt.figure(figsize = (20,5))
ax = fig.add_subplot(1,1,1)

#=====

by_inter_service.plot(figsize = (20,5),
               ax = ax,
               linewidth = 5,
               color = [colors['color02'],colors['color03']])

ax.set_xticks([])
ax.set_xlabel('Since March 2017',
            fontname = 'Arial',
            fontweight = 'bold',
            fontsize = 20)

vals = ax.get_yticks()
ax.set_yticklabels(['{:,.0f}'.format(x) for x in vals],
            fontname = 'Arial',
            fontweight = 'bold',
            fontsize = 20)
ax.set_ylabel('Package Counts',
            fontname = 'Arial',
            fontweight = 'bold',
            fontsize = 20)

ax.set_title('UPS and DHL\nInternational Only',
            fontname = 'Arial',
            fontweight = 'bold',
            fontsize = 20)

ax.legend(['DHL','UPS'], fontsize = 20)
ax.grid()
ax.set_facecolor(colors['color04'])
```
