### Get ship_log data

```python
sl1 = pd.read_sql(
'''
SELECT
DATE(shipped_date) AS date_shipped,
orders_id,
shippers_id
FROM ship_log
WHERE orders_id BETWEEN '''+ str(min_oid) +''' AND '''+ str(max_oid) +'''
''', db)
col_fix(sl1)
sl1['partial id'] = 0

sl2 = pd.read_sql(
'''
SELECT
DATE(psl_shipped_date) AS date_shipped,
psl_parent_id AS orders_id,
psl_partial_id AS partial_id,
psl_shippers_id AS shippers_id
FROM partial_ship_log
WHERE psl_parent_id BETWEEN '''+ str(min_oid) +''' AND '''+ str(max_oid) +'''
''', db)
col_fix(sl2)
```

```python
sl1.drop(sl1[sl1['orders id'].isin(sl2['orders id'].tolist())].index, inplace = True)
sl1.reset_index(drop = True, inplace = True)
```

```python
ship_log = pd.concat([sl1, sl2], sort = False)
```

### Structure it

```python
# dates and days
ship_log['date shipped'] = pd.to_datetime(ship_log['date shipped'])
ship_log['quarter shipped'] = ship_log['date shipped'].dt.quarter
ship_log['day name shipped'] = ship_log['date shipped'].dt.weekday_name
ship_log['day type shipped'] = np.where(ship_log['day name shipped'].isin(['Saturday','Sunday']), 'weekend','weekday')

# OIDs and names
ship_log['umbrella orders id'] = ship_log['orders id'].map(str) + '-' + ship_log['partial id'].map(str)
ship_log['shippers name'] = ship_log['shippers id'].map(dict(zip(names['shippers id'], names['shippers name'])))

# exclude orders shipped after our date_end
ship_log.drop(ship_log[ship_log['date shipped'] > date_end].index, inplace = True)
ship_log.reset_index(drop = True, inplace = True)
```

### Get orders data

```python
orders_main = pd.read_sql(
'''
SELECT
date_purchased AS date_and_time_purchased,
DATE(date_purchased) AS date_purchased,
orders_id,
IF(orders_fake = 1, 'yes','no') AS fake_order,
IF(orders_po_number IN ('','n/a'), 'no','yes') AS po,
IF(orders_reseller = 0 AND orders_super_reseller = 0, 'non reseller','reseller/super') AS customer,
IF(orders_id IN (SELECT orders_id FROM subscriptions_history), 'yes','no') AS adabox,
LOWER(orders_status_name) AS final_orders_status
FROM orders o
JOIN orders_status os ON o.orders_status = os.orders_status_id
WHERE orders_id BETWEEN '''+ str(min_oid) +''' AND '''+ str(max_oid) +'''
''', db)

col_fix(orders_main)
```

### Structure it

```python
# dates and days
orders_main['date purchased'] = pd.to_datetime(orders_main['date purchased'])
orders_main['year and quarter purchased'] = jb_dates(orders_main['date purchased'], 'year and quarter')

orders_main['quarter purchased'] = orders_main['date purchased'].dt.quarter
orders_main['day name purchased'] = orders_main['date purchased'].dt.weekday_name
orders_main['day type purchased'] = np.where(orders_main['day name purchased'].isin(['Saturday','Sunday']), 'weekend','weekday')

# label "normal" vs "special" shipments
oids = ship_log[ship_log['partial id'] == 0]['orders id'].tolist()

orders_main['order type'] = np.where((orders_main['customer'] == 'non reseller')
                            & (orders_main['fake order'] == 'no')
                            & ((orders_main['po'] == 'no') | ((orders_main['po'] == 'yes') & (orders_main['orders id'].isin(oids))))
                            & (orders_main['adabox'] == 'no'),
                            'regular','special')
```

# Shipper Averages per day

```python
est_date_start = '2016-10-01'
est_date_end = '2018-10-31'
```

```python
avg_per_day = for_avg[(for_avg.index.get_level_values(0) >= est_date_start) 
                    & (for_avg.index.get_level_values(0) <= est_date_end)].groupby(level = ['day type','order type']).mean()

avg_per_day.loc[('weekday','regular'), 'unique shipper count'] = 9
avg_per_day.loc[('weekend','regular'), 'unique shipper count'] = 4

avg_per_day.loc[('weekday','special'), 'unique shipper count'] = 4
avg_per_day.loc[('weekend','special'), 'unique shipper count'] = 0

avg_per_day['shipments per shipper'] = avg_per_day['shipment count']/avg_per_day['unique shipper count']

# assume any shipper who works an 8 hour shift actually ships for 6 hours
# this is 5 hours * 60 minutes = 360 minutes 
# so this is minutes per shipment
hrs_worked = 5
avg_per_day['min per shipment'] = (hrs_worked*60)/avg_per_day['shipments per shipper']

# make these manual changes based on the info angel gave in the email with subject line "Shipping and the upcoming holidays"


avg_per_day.format_(['n0','n0','n0','n0','n2'])
```

### 95% confidence intervals for count of orders per day during Q4

```python
m = gr1.groupby(level = [1,2]).mean()
s = gr1.groupby(level = [1,2]).std()
n = gr1.groupby(level = [1,2]).count()
phi = stats.t.ppf(0.95, n-1)

lower = m - phi * (s/np.sqrt(n))
lower.columns = ['lower']
upper = m + phi * (s/np.sqrt(n))
upper.columns = ['upper']

m.columns = ['mean']

est_count = lower.join(m).join(upper)
est_count.index.names = ['day type','order type']

est_count.format_(['n0'] * 3)
```

# For shipping: Black Friday and Cyber Monday only

```python
cm = ['2016-11-28','2017-11-27']
bf = ['2016-11-25','2017-11-24']

hol_ls = cm + bf

hol1 = orders_main[orders_main['date purchased'].isin(hol_ls)].copy()
hol1['holiday'] = np.where(hol1['date purchased'].isin(cm), 'cyber monday', 'black friday')
hol1['year purchased'] = hol1['date purchased'].dt.year
```

### 95% confidence intervals for count of BF/CM orders

```python
hol2 = hol1.groupby(['year purchased','holiday'])[['orders id']].count().unstack(1)
hol2.columns = hol2.columns.droplevel(0)
hol2.index.names = ['order count']

ls = []
for col in hol2.columns:
    df = hol2[col]
    m = df.mean()
    s = df.std()
    n = len(df)
    phi = stats.t.ppf(0.95, n-1)
    lower = m - phi * (s/np.sqrt(n))
    upper = m + phi * (s/np.sqrt(n))
    ls.append([col,lower,m,upper])
    
hol_conf = pd.DataFrame(ls, columns = ['holiday','lower','mean','upper'])
hol_conf.set_index('holiday', inplace = True)

hol_conf.format_(['n0','n0','n0'])
```

### Structure it

```python
# remove partial data from non-partial data
op1.drop(op1[op1['orders id'].isin(op2['orders id'].tolist())].index, inplace = True)
op1.reset_index(drop = True, inplace = True)

# join the two
op_main = pd.concat([op1,op2], sort = True)
op_main['umbrella orders id'] = op_main['orders id'].map(str) + '-' + op_main['partial id'].map(str)
```

### Get gathers data

```python
g1 = pd.read_sql(
'''
SELECT
DATE(time_in) AS date_gathered,
orders_id,
0 AS partial_gather_id,
gathered_by_id,
g.gather_batch
FROM gathers g
LEFT JOIN orders o ON g.gather_batch = o.gather_batch
WHERE DATE(time_in) >= ' '''+ date_start +''' '
AND DATE(time_out) <= ' '''+ date_end +''' '
''', db)

col_fix(g1)
g1['partial id'] = 0
g1['gather type'] = 'regular'
g1['partial id'] = 0

n = g1[g1.isnull().any(1)]
print('{:,.0f} nulls or {:,.2f}% of total lines'.format(len(n), len(n)/len(g1) * 100))
g1.dropna(inplace = True)
g1.reset_index(drop = True, inplace = True)

#=====

g2 = pd.read_sql(
'''
SELECT
DATE(pg_time_in) AS date_gathered,
partial_gather_id,
0 AS gather_batch,
pg_gathered_by_id AS gathered_by_id
FROM partial_gathers
WHERE DATE(pg_time_in) >= ' '''+ date_start +''' '
AND DATE(pg_time_out) <= ' '''+ date_end +''' '
''', db)

col_fix(g2)
g2['gather type'] = 'special'

g3 = pd.read_sql(
'''
SELECT
ps_parent_id AS orders_id,
ps_partial_id AS partial_id,
partial_gather_id
FROM partial_shipments
WHERE ps_parent_id BETWEEN '''+ str(min_oid) +''' AND '''+ str(max_oid) +'''
''', db)

col_fix(g3)
```

```python
cols = g3.columns.tolist()
cols.remove('partial gather id')

for c in cols:
    g2[c] = g2['partial gather id'].map(dict(zip(g3['partial gather id'], g3[c])))

n = g2[g2.isnull().any(1)]

print('{:,.0f} nulls or {:,.2f}% of total lines'.format(len(n), len(n)/len(g2) * 100))
g2.dropna(inplace = True)
g2.reset_index(drop = True, inplace = True)

gathers_main = pd.concat([g1,g2], sort = False)

for col in gathers_main.columns:
    try:
        gathers_main[col] = [int(x) for x in gathers_main[col]]
    except:
        pass

gathers_main['umbrella orders id'] = gathers_main['orders id'].map(str) + '-' + gathers_main['partial id'].map(str)
gathers_main['qty total'] = gathers_main['umbrella orders id'].map(dict(zip(op_main['umbrella orders id'], op_main['qty total'])))

for col in gathers_main.columns:
    try:
        gathers_main[col] = [int(x) for x in gathers_main[col]]
    except:
        pass
    
gathers_main['gatherer name'] = gathers_main['gathered by id'].map(dict(zip(names['shippers id'], names['shippers name'])))
gathers_main['date gathered'] = pd.to_datetime(gathers_main['date gathered'])
gathers_main['quarter'] = gathers_main['date gathered'].dt.quarter
gathers_main['day name'] = gathers_main['date gathered'].dt.weekday_name
gathers_main['day type'] = np.where(gathers_main['day name'].isin(['Saturday','Sunday']), 'weekend','weekday')        

n = gathers_main[gathers_main.isnull().any(1)]

print('{:,.0f} nulls or {:,.2f}% of total lines'.format(len(n), len(n)/len(gathers_main) * 100))
gathers_main.dropna(inplace = True)
gathers_main.reset_index(drop = True, inplace = True)
```

```python
# shipment count
a = ship_log_g.groupby(['date purchased','order type']).agg({'umbrella orders id':'count'})
a.columns = ['shipment count']

# total qty and unique gatherer count
b = gathers_main.groupby(['date gathered','gather type']).agg({'qty total':'sum','gathered by id':'nunique'})
b.rename(columns = {'gathered by id':'unique gatherer count'}, inplace = True)

for x in [a,b]:
    x.fillna(0, inplace = True)
    x.index.set_names(['date','order type'], inplace = True)
        
for_avg2 = a.join(b)
for_avg2.fillna(1, inplace = True) # if it was a weekend and no one gathered then make it x1 gatherer

for col in for_avg2.columns:
    for_avg2[col] = [int(x) for x in for_avg2[col]]    
    
for_avg2.reset_index(inplace = True)
for_avg2['date'] = pd.to_datetime(for_avg2['date'])
for_avg2['quarter'] = for_avg2['date'].dt.quarter
for_avg2['day name'] = for_avg2['date'].dt.weekday_name
for_avg2['day type'] = np.where(for_avg2['day name'].isin(['Saturday','Sunday']), 'weekend','weekday')    
for_avg2.drop(for_avg2[for_avg2['quarter'] != 4].index, inplace = True)    

for_avg2['qty per gatherer'] = for_avg2['qty total']/for_avg2['unique gatherer count']
for_avg2['qty per gatherer'] = [int(x) for x in for_avg2['qty per gatherer']]

for_avg2['qty per shipment'] = for_avg2['qty total']/for_avg2['shipment count']

for_avg2.drop(for_avg2[for_avg2['quarter'] != 4].index, inplace = True)
for_avg2.reset_index(drop = True, inplace = True)

for_avg2.head()
```

# Reseller QA (WORKING)

```python
qa_main = pd.read_sql(
'''
SELECT
DATE(timestamp) AS date_qa,
LOWER(admin) AS admin,
orders_id,
pop_id,
qa_qty
FROM gather_qa g
JOIN orders_products op ON g.orders_products_id = op.orders_products_id
WHERE orders_id BETWEEN '''+ str(min_oid) +''' AND '''+ str(max_oid) +'''
''', db)

col_fix(qa_main)
```

```python
qa1 = qa_main.groupby(['date qa','orders id'], as_index = False).agg({'qa qty':'sum'})

cols = orders_main.columns.tolist()
cols.remove('orders id')

for c in cols:
    qa1[c] = qa1['orders id'].map(dict(zip(orders_main['orders id'], orders_main[c])))

# order count and qty
qa2 = qa1.groupby(['date purchased','day type purchased','order type']).agg({'orders id':'count','qa qty':'sum'})
qa2.rename(columns = {'orders id':'order count'}, inplace = True)
qa3 = qa2.groupby(level = [1,2]).mean()
qa3.rename(columns = {'order count':'avg order count','qa qty':'avg qa qty'}, inplace = True)
qa3.index.names = ['day type','order type']

for_admin = qa_main.copy()
for col in ['order type','day type purchased']:
    for_admin[col] = for_admin['orders id'].map(dict(zip(orders_main['orders id'], orders_main[col])))
    
# employees    
g1 = for_admin.groupby(['date qa','day type purchased','order type'])[['admin']].nunique()
qa4 = g1.groupby(level = [1,2]).mean()
qa4.columns = ['unique admin count']
qa4.index.names = ['day type','order type']

# join
qa5 = qa3.join(qa4)
```
