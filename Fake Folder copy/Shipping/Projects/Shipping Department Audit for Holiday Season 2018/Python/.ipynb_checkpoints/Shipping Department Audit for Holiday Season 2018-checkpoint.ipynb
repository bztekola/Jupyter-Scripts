### Get the first and last OIDs within our data range
We can't search by date alone because partials make this weird

```python
oid = pd.read_sql(
'''
SELECT
MIN(orders_id) AS min_oid,
MAX(orders_id) AS max_oid
FROM orders
WHERE DATE(date_purchased) BETWEEN ' '''+ date_start +''' ' AND ' '''+ date_end +''' '
''', db)

col_fix(oid)

min_oid = oid['min oid'].values[0]
max_oid = oid['max oid'].values[0]
```

### Shipping data

```python
sl1 = pd.read_sql(
'''
SELECT
DATE(shipped_date) AS date_shipped,
orders_id,
shippers_id
FROM ship_log
WHERE orders_id BETWEEN '''+ str(min_oid) +''' AND '''+ str(max_oid) +'''
''', db)
col_fix(sl1)
sl1['partial id'] = 0

sl2 = pd.read_sql(
'''
SELECT
DATE(psl_shipped_date) AS date_shipped,
psl_parent_id AS orders_id,
psl_partial_id AS partial_id,
psl_shippers_id AS shippers_id
FROM partial_ship_log
WHERE psl_parent_id BETWEEN '''+ str(min_oid) +''' AND '''+ str(max_oid) +'''
''', db)
col_fix(sl2)

sl1.drop(sl1[sl1['orders id'].isin(sl2['orders id'].tolist())].index, inplace = True)
sl1.reset_index(drop = True, inplace = True)

ship_log = pd.concat([sl1, sl2], sort = False)
ship_log.reset_index(drop = True, inplace = True)
ship_log['umbrella orders id'] = ship_log['orders id'].map(str) + '-' + ship_log['partial id'].map(str)
```

### Names

```python
names = pd.read_sql(
'''
SELECT
shippers_id AS id,
LOWER(shippers_username) AS name
FROM shippers
''', db)

col_fix(names)

ship_log['shippers name'] = ship_log['shippers id'].map(dict(zip(names['id'], names['name'])))
ship_log['shippers name'].fillna('no name', inplace = True)
ship_log.drop('shippers id',1,inplace = True)
```

### Orders data

```python
orders_main = pd.read_sql(
'''
SELECT
date_purchased AS date_and_time_purchased,
DATE(date_purchased) AS date_purchased,
orders_id,
IF(orders_fake = 1, 'yes','no') AS fake_order,
IF(orders_po_number IN ('','n/a'), 'no','yes') AS po,
IF(orders_reseller = 0 AND orders_super_reseller = 0, 'non reseller','reseller/super') AS customer,
IF(orders_id IN (SELECT orders_id FROM subscriptions_history), 'yes','no') AS adabox,
LOWER(orders_status_name) AS final_orders_status
FROM orders o
JOIN orders_status os ON o.orders_status = os.orders_status_id
WHERE orders_id BETWEEN '''+ str(min_oid) +''' AND '''+ str(max_oid) +'''
''', db)

col_fix(orders_main)
```

```python
oids = ship_log[ship_log['partial id'] == 0]['orders id'].tolist()

orders_main['order type'] = np.where((orders_main['customer'] == 'non reseller')
                            & (orders_main['fake order'] == 'no')
                            & ((orders_main['po'] == 'no') | ((orders_main['po'] == 'yes') & (orders_main['orders id'].isin(oids))))
                            & (orders_main['adabox'] == 'no'),
                            'regular','special')

cols = orders_main.columns.tolist()
cols.remove('orders id')
for c in cols:
    ship_log[c] = ship_log['orders id'].map(dict(zip(orders_main['orders id'], orders_main[c])))
    
for col in ['date shipped','date purchased']:
    ship_log[col] = pd.to_datetime(ship_log[col])
```

### Deleted orders

```python
deleted = pd.read_sql(
'''
SELECT
orders_id
FROM orders_deleted
''', db)

col_fix(deleted)

ship_log.drop(ship_log[ship_log['orders id'].isin(deleted['orders id'].tolist())].index, inplace = True)
ship_log.reset_index(drop = True, inplace = True)
```

### Flag holidays

```python
bf = [dt.datetime(2016,11,25), dt.datetime(2017,11,24), dt.datetime(2018,11,23)]
cm = [dt.datetime(2016,11,28), dt.datetime(2017,11,27), dt.datetime(2018,11,26)]

def holiday(x):
    if x['date purchased'] in bf:
        ret = 'black friday'
    elif x['date purchased'] in cm:
        ret = 'cyber monday'
    else:
        ret = 'no holiday'
    
    return ret

ship_log['holiday'] = ship_log.apply(holiday, axis = 1)

ship_log.groupby([ship_log['date purchased'].dt.year,'holiday'])[['orders id']].nunique().unstack(1)
```

### Qty gathered

```python
op1 = pd.read_sql(
'''
SELECT
orders_id,
SUM(products_quantity) AS qty_total
FROM orders_products
WHERE orders_id BETWEEN '''+ str(min_oid) +''' AND '''+ str(max_oid) +'''
GROUP BY orders_id
''', db)

col_fix(op1)
op1['partial id'] = 0

op2 = pd.read_sql(
'''
SELECT
pop_parent_id AS orders_id,
pop_partial_id AS partial_id,
SUM(pop_quantity) AS qty_total
FROM partial_orders_products
WHERE pop_parent_id BETWEEN '''+ str(min_oid) +''' AND '''+ str(max_oid) +'''
GROUP BY pop_parent_id, pop_partial_id
''', db)

col_fix(op2)

# remove partial data from non-partial data
op1.drop(op1[op1['orders id'].isin(op2['orders id'].tolist())].index, inplace = True)

# join the two
op_main = pd.concat([op1,op2], sort = True)
op_main['umbrella orders id'] = op_main['orders id'].map(str) + '-' + op_main['partial id'].map(str)
op_main.reset_index(drop = True, inplace = True)

ship_log['qty total'] = ship_log['umbrella orders id'].map(dict(zip(op_main['umbrella orders id'], op_main['qty total'])))

n = ship_log[ship_log.isnull().any(1)]
print('{:,.0f} nulls before dropping'.format(len(n)))
ship_log.dropna(inplace = True)
```

### Gatherers

```python
g1 = pd.read_sql(
'''
SELECT
time_out,
time_in,
orders_id,
gathered_by_id,
g.gather_batch
FROM gathers g
LEFT JOIN orders o ON g.gather_batch = o.gather_batch
WHERE orders_id BETWEEN '''+ str(min_oid) +''' AND '''+ str(max_oid) +'''
''', db)

col_fix(g1)

g1['partial id'] = 0
g1['gather type'] = 'regular'
```

```python
g2 = pd.read_sql(
'''
SELECT
pg_time_out AS time_out,
pg_time_in AS time_in,
ps_parent_id AS orders_id,
pg_gathered_by_id AS gathered_by_id,
0 AS gather_batch,
ps.ps_partial_id AS partial_id
FROM partial_gathers pg
LEFT JOIN partial_shipments ps ON ps.ps_partial_id = pg.partial_gather_id
WHERE ps_parent_id BETWEEN '''+ str(min_oid) +''' AND '''+ str(max_oid) +'''
''', db)

col_fix(g2)
g2['gather type'] = 'special'
```

```python
# remove partial data from non-partial data
g1.drop(g1[g1['orders id'].isin(g2['orders id'].tolist())].index, inplace = True)

# join the two
g_main = pd.concat([g1,g2], sort = True)
g_main['umbrella orders id'] = g_main['orders id'].map(str) + '-' + g_main['partial id'].map(str)
g_main.reset_index(drop = True, inplace = True)

# get names
g_main['gatherers name'] = g_main['gathered by id'].map(dict(zip(names['id'], names['name'])))
g_main['gatherers name'].fillna('no name', inplace = True)
g_main.drop('gathered by id',1,inplace = True)

# map it
for col in ['gather batch','time in','time out','gatherers name']:
    ship_log[col] = ship_log['umbrella orders id'].map(dict(zip(g_main['umbrella orders id'], g_main[col])))
    
n = ship_log[ship_log.isnull().any(1)]
print('nulls account for {:,.2f}% of total lines\ndrop them'.format(len(n)/len(ship_log) * 100))
ship_log.dropna(inplace = True)
ship_log.reset_index(drop = True, inplace = True)    
```

### Check avg

```python
avg = ship_log.groupby('date purchased')[['orders id']].count().mean()[0]

avg_check = pd.read_sql(
'''
SELECT
AVG(A.count) AS avg_per_day
FROM
(SELECT
COUNT(*) AS count
FROM orders
WHERE DATE(date_purchased) BETWEEN ' '''+ date_start +''' ' AND ' '''+ date_end +''' '
GROUP BY DATE(date_purchased)) A
''', db)

col_fix(avg_check)

print('x{:,.0f} orders per day on average from {} to {}\norders table says x{:,.0f}'.format(avg,
                                                                                            date_start,
                                                                                            date_end,
                                                                                            avg_check['avg per day'].values[0]))
```

### Add some data

```python
for col in ['shipped','purchased']:
    
    ship_log['day name ' + col] = ship_log['date ' + col].dt.weekday_name.str.lower()
    ship_log['day type ' + col] = np.where(ship_log['day name ' + col].isin(['saturday','sunday']), 'weekend','weekday')
    ship_log['year ' + col] = ship_log['date ' + col].dt.year
    
    for col2 in ['year and month','year and quarter']:
        ship_log[col2 + ' ' + col] = jb_dates(ship_log['date ' + col], col2)
```

### Remove some data
