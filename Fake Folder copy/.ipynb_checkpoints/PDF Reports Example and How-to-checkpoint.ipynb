```python
# import libraries
import pandas as pd
import numpy as np

# create data
cols = ['hi','ok','no']
data = [[1,2,3],[4,5,6],[7,8,np.nan]]
df1 = pd.DataFrame(columns = cols, data = data)
```

```python
# use this to load in your HTML txt template, which has placeholders for the dict values
from jinja2 import Environment, FileSystemLoader
env = Environment(loader = FileSystemLoader('.'))
template = env.get_template('myreport.html')

# create dict for HTML placeholders
template_vars = {'title1' : 'title1',
                 'title2' : 'title2',
                 'title3' : 'title3',
                 'df1': df1.to_html(index = False),
                 'words':'hi there<br>i am writing some fake words for you!!<br>'}

# map data to placeholders
html_out = template.render(template_vars)

# use weasyprint to turn HTML to PDF
from weasyprint import HTML
HTML(string = html_out).write_pdf('jb test report.pdf', stylesheets = ["CSS Stylesheets/Recurring Reports Stylesheet.css"]) # with stylesheet, THIS WORKS
```
