```python
combo_cost_by_sku = pd.read_sql(
'''
SELECT
pts.part_id AS 'part id',
pts.contains_part_id AS 'contains part id',
pd.products_name AS 'part name',
pts.pts_quantity * IF(k.sku_outsourced_assembly = 1, pm.consigned_latest_cost, pm.latest_cost) AS 'latest cost',
k.sku_status AS 'sku status'

FROM products_to_stuff pts
JOIN skus k ON pts.contains_part_id = k.part_id
JOIN products_manufacturing pm ON k.sku_id = pm.sku_id
JOIN products_description pd ON pts.part_id = pd.part_id

WHERE pts.part_id != 0
''', db)

combo_cost_by_part = combo_cost_by_sku[combo_cost_by_sku['sku status'] == 'working'].groupby(['part id','part name'], as_index = False)[['latest cost']].sum()

combo_cost_by_part['sku id'] = 0
combo_cost_by_part['sku status'] = 'working'
combo_cost_by_part['part type'] = 'combo'
combo_cost_by_part['outsourced assembly'] = 'no'

#===================================================================================================================

not_combo_cost_main = pd.read_sql(
'''
SELECT
k.part_id AS 'part id',
k.sku_id AS 'sku id',
pd.products_name AS 'part name',
IF(k.sku_outsourced_assembly = 1, pm.consigned_latest_cost, pm.latest_cost) AS 'latest cost',
IF(k.sku_outsourced_assembly = 1, 'yes','no') AS 'outsourced assembly',
k.sku_status AS 'sku status'

FROM products_manufacturing pm
JOIN skus k ON pm.sku_id = k.sku_id
JOIN parts p ON k.part_id = p.part_id
JOIN products_description pd ON k.part_id = pd.part_id

WHERE p.products_stripes = 0 # do stripes separately
''', db)

not_combo_cost_main['part type'] = 'part'

#===================================================================================================================

stripes_cost_main = pd.read_sql(
'''
SELECT
A.part_id AS 'part id',
k.sku_id AS 'sku id',
pd.products_name AS 'part name',
A.latest_cost AS 'latest cost',
A.sku_status AS 'sku status',
IF(k.sku_outsourced_assembly = 1, 'yes','no') AS 'outsourced assembly'

FROM

(SELECT
p.part_id,
p.products_stripe_of,
p.products_stripe_qty * IF(k.sku_outsourced_assembly = 1, pm.consigned_latest_cost, pm.latest_cost) AS latest_cost,
k.sku_status
FROM parts p
JOIN skus k ON p.products_stripe_of = k.part_id
JOIN products_manufacturing pm ON k.sku_id = pm.sku_id
WHERE p.products_stripes = 1) A

JOIN skus k ON A.part_id = k.part_id
JOIN products_description pd ON A.part_id = pd.part_id

''', db)

stripes_cost_main['part type'] = 'stripe'

#===================================================================================================================

# put it all together
cost = pd.concat([combo_cost_by_part,
                 not_combo_cost_main,
                 stripes_cost_main])

# check for nulls
if cost[cost.isnull().any(1)].empty:
    print('no nulls')
else:
    print('YOU HAVE NULLS')

# check for dupe sku ids    
dupes = cost[cost['sku id'].duplicated()
             & (cost['sku status'] == 'working')
             & (cost['sku id'] != 0)
             & (cost['part type'] != 'stripe')]

if cost[cost['sku id'].isin(dupes['sku id'].tolist())].empty:
    print('no dupe sku ids amongst working skus')
else:
    print('YOU HAVE DUPE SKU IDS')
```
