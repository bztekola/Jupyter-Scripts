### Find current max OID

```python
oid_info = pd.read_sql(
'''
SELECT
MIN(orders_id) AS min_oid,
MAX(orders_id) AS max_oid
FROM orders
''', db)

col_fix(oid_info)

oid_info
```

### Get orders data

```python
orders = pd.read_sql(
'''
SELECT
DATE(date_purchased) AS date_purchased,
orders_id,
LOWER(orders_status_name) AS orders_status_name,
LOWER(payment_method) AS payment_method
FROM orders o
JOIN orders_status os ON o.orders_status = os.orders_status_id
ORDER BY orders_id DESC
''', db)
col_fix(orders)

orders_deleted = pd.read_sql(
'''
SELECT
DATE(date_purchased) AS date_purchased,
orders_id,
'deleted' AS orders_status_name,
'deleted' AS payment_method
FROM orders_deleted o
''', db)

col_fix(orders_deleted)
```

```python
for df in [orders, orders_deleted]:
    df['date purchased'] = pd.to_datetime(df['date purchased'])
    df['year and month'] = [str(x)[:7] for x in df['date purchased']]
    df['year'] = df['date purchased'].dt.year
```

### Join orders and orders_deleted

```python
o_main = pd.concat([orders, orders_deleted], sort = True)
```

### Check out dupes

```python
dupe_oids = o_main[o_main['orders id'].duplicated()]['orders id'].tolist()
dupes = o_main[o_main['orders id'].isin(dupe_oids)].sort_values('orders id')

dupes['orders status name'].value_counts()
```

```python
o_main.sort_values(['orders id','orders status name'], inplace = True)
o_main.drop_duplicates('orders id', keep = 'last', inplace = True) 
```

### Check out non sequential OIds

```python
o_main['orders id2'] = o_main['orders id'].shift(-1)
x = np.sum((o_main['orders id'] - o_main['orders id2']) != 1)

if x == 0:
    print('order numbers are sequential\n')
else:
    print('order numbers have gaps\n')
    
print('{:,.0f} actual orders, deleted or otherwise, regardless of order status'.format(len(o_main)))
print('max order id is {}'.format(np.max(o_main['orders id'])))    
```

### By year

```python
by_year = o_main.groupby(['year','orders status name'])[['orders id']].count().unstack(1).fillna(0)
by_year.columns = by_year.columns.droplevel(0)

by_year['row total'] = by_year.sum(1)
by_year['row total (sans deleted)'] = by_year.sum(1) - by_year[['row total','deleted']].sum(1)
by_year.loc['column total'] = by_year.sum()

by_year.format_(['n0'] * len(by_year.columns))
```

### How many are missing?

```python
min_ = o_main['orders id'].min()
max_ = o_main['orders id'].max()

s1 =  set(o_main['orders id'].tolist()) # the OIDs we have
s2 = set(np.arange(min_, max_)) # an idealized list of OIDs that we should have

s3 = set(s2-s1)
```

```python
a = by_year.loc['column total','row total'] # total number of orders, deleted or otherwise, no dupes
b = len(s3) # number of missing OIDs
c = 1020 # first OID ever was 1021, so there are 1020 OIDs before this one

# the sum of a,b,c should equal the current max OID number
# in other words the expression below should equal zero
oid_info['max oid'][0] - (a + b + c)
```

```python
by_month = o_main.groupby('date purchased')[['orders id']].count()
by_month.columns = ['order count']
by_month.plot(figsize = (20,5))
```

```python
date = '2017-11-01'
conf = jb_conf(by_month['order count'][by_month.index >= date])

print('95% confidence interval for order count per day since {}'.format(date))
conf.format_(['n0'] * 5)
```

```python
print('days until two millionth order')
by_month['running sum'] = by_month['order count'].cumsum()

goal = 2000000
left_to_go = goal - by_month.iloc[-1]['running sum']

days_to_go = left_to_go/conf[['lower','mean','upper']]
days_to_go.format_(['n0'] * 3)
```

```python
now = by_month.iloc[-1].name
for d in days_to_go.values[0]:
    d2 = str((now + pd.DateOffset(days = d)).date())
    print(d2)
```
