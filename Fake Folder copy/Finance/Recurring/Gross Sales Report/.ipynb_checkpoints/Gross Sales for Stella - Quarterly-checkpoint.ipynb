```python
import sys
sys.path.insert(0, '/Users/jarad')

import calendar

from db2 import *
import numpy as np
import pandas as pd
import xlsxwriter
```

```python
date_start = '2017-07-01'
date_end = '2017-09-30'

title = '2017 - Q3 - Gross Sales.xlsx'

cur_date_name = '2017 - Q3'
```

```python
ot = pd.read_sql(
'''
SELECT
DATE(o.date_purchased) AS date,
ot.orders_id,
ot.value,
ot.class
FROM orders_total ot
JOIN orders o ON ot.orders_id = o.orders_id
WHERE DATE(o.date_purchased) BETWEEN ' '''+ date_start +''' ' AND ' '''+ date_end +''' '
''', db)
```

```python
ot['date'] = pd.to_datetime(ot['date'])
ot['month number'] = ot['date'].dt.month
ot['month'] = ot['date'].dt.strftime('%b')
ot['year'] = ot['date'].dt.year
```

```python
gross = ot.groupby(['month number','month','year','orders_id','class']).agg({'value':'sum'}).unstack(level = [0,1,2,3]).T.reset_index().drop('level_0', 1)
gross.fillna(0, inplace = True)
```

```python
gross.rename(columns = {'ot_check_fee':'Check Fee',
                        'ot_coupon':'Coupon',
                        'ot_ddp':'Export Fee',
                        'ot_gv':'Gift Certificate',
                        'ot_refund':'Refund',
                        'ot_shipping':'Shipping',
                        'ot_tax':'Tax',
                        'ot_total':'Total',
                        'ot_wiretransfer_fee':'Wire Transfer Fee',
                        '':'Misc.',
                        'orders_id':'Orders ID',
                        'ot_subtotal':'Subtotal',
                        'month':'Month',
                        'year':'Year'}, inplace = True)
```

```python
gross['Month'].value_counts()
```

# Get refunds and merge

```python
ot_refunds = pd.read_sql(
'''
SELECT
ot.orders_id AS 'Orders ID',
value AS Refund
FROM orders_total ot
JOIN orders o ON ot.orders_id = o.orders_id
WHERE class = 'ot_refund'
AND DATE(o.date_purchased) BETWEEN ' '''+ date_start +''' ' AND ' '''+ date_end +''' '
''', db)

cs_refunds = pd.read_sql(
'''
SELECT
orders_id AS 'Orders ID',
amount_refunded AS Refund
FROM cs_error_reports
WHERE orders_id IN '''+ str(tuple(gross['Orders ID'])) +'''
AND amount_refunded != 0
''', db)

refunds = pd.concat([ot_refunds, cs_refunds])

gross.drop('Refund', 1, inplace = True)
gross = pd.merge(gross, refunds, how = 'left', on = 'Orders ID' )
gross['Refund'].fillna(0, inplace = True)
```

```python
gross.head()
```

```python
gross[['Coupon','Gift Certificate','Refund']] = gross[['Coupon','Gift Certificate','Refund']].apply(lambda x: x * -1)

# for adabox shipments, ot_subtotal = 60 while ot_total = 0
# here we forget the provided ot_total column and make our own
gross.drop('Total',1,inplace = True)
gross = gross[['Year','Month','month number','Orders ID','Subtotal','Coupon','Gift Certificate','Refund','Check Fee','Wire Transfer Fee','Export Fee','Tax','Shipping']]
gross['Total'] = gross.iloc[:,4:15].sum(1)
gross = gross[['Year','Month','month number','Orders ID','Subtotal','Coupon','Gift Certificate','Refund','Check Fee','Wire Transfer Fee','Export Fee','Tax','Shipping','Total']]
```

```python
gross_format = gross.copy()
gross_format = gross_format[['Year','Month','Orders ID','Subtotal','Coupon','Gift Certificate','Refund','Check Fee','Wire Transfer Fee','Export Fee','Tax','Shipping','Total']]
```

```python
gross_format.head()
```

```python
by_month = gross.copy()

by_month = by_month.groupby(['Year','month number','Month']).sum().reset_index().drop('month number', 1)
by_month = by_month[['Year','Month','Subtotal','Coupon','Gift Certificate','Refund','Check Fee','Wire Transfer Fee','Export Fee','Tax','Shipping','Total']]
```

```python
by_month = by_month.append(by_month.sum(), ignore_index = True)

by_month['Year'][3] = 'Totals'
by_month.set_value(3,'Month', '')
```

```python
writer = pd.ExcelWriter(title, engine='xlsxwriter')

gross_format.to_excel(writer, sheet_name = cur_date_name, startrow=7, startcol=1, index=False, header=True)
by_month.to_excel(writer, sheet_name = cur_date_name, startrow=7, startcol=16, index=False, header=True)

workbook = writer.book
worksheet1 = writer.sheets[cur_date_name]

numbers_format = workbook.add_format()
numbers_format.set_num_format('##,##0.00')

worksheet_title_format = workbook.add_format({
        'bold': 1,
        'font_size': 20})

table_title = workbook.add_format({
        'bold': 1,
        'border': 1,
        'align': 'center',
        'valign': 'vcenter',
        'bg_color': 'gray',
        'font_size': 18,
        'font_color': 'white'})

border = workbook.add_format({
        'top':6,
        'num_format':'##,##0.00'})

worksheet1.write('A1', cur_date_name + ' Gross Sales', worksheet_title_format)
worksheet1.merge_range('B7:N7', 'By Orders ID', table_title)
worksheet1.merge_range('Q7:AB7', 'Totals', table_title)

worksheet1.write('Q12', by_month.loc[3]['Year'], border)
worksheet1.write('R12', by_month.loc[3]['Month'], border)
worksheet1.write('S12', by_month.loc[3]['Subtotal'], border)
worksheet1.write('T12', by_month.loc[3]['Coupon'], border)
worksheet1.write('U12', by_month.loc[3]['Gift Certificate'], border)
worksheet1.write('V12', by_month.loc[3]['Refund'], border)
worksheet1.write('W12', by_month.loc[3]['Check Fee'], border)
worksheet1.write('X12', by_month.loc[3]['Wire Transfer Fee'], border)
worksheet1.write('Y12', by_month.loc[3]['Export Fee'], border)
worksheet1.write('Z12', by_month.loc[3]['Tax'], border)
worksheet1.write('AA12', by_month.loc[3]['Shipping'], border)
worksheet1.write('AB12', by_month.loc[3]['Total'], border)

worksheet1.set_column('A:AB', 15)
worksheet1.set_column('F:N', None, numbers_format)
worksheet1.set_column('S:AB', None, numbers_format)

worksheet1.write('A3', 'No types of orders have been excluded from this report', worksheet_title_format)
worksheet1.write('A4', 'The subtotal here (Col S) may be different than what you see in PLM. This is because of the way PLM is calculated. This is normal. For more specifics plz email Jarad@adafruit.com.')

writer.save()
```
