### Adjust back for sampling
With this API you cannot turn off sampling, and there is no way to know what the sample size is (sampling is explained [here](https://support.google.com/analytics/answer/2637192?hl=en)). So let's just create our own correction. 

For Jan 2019, GA website total unique searches was 390,485, API says 223,352. 

API / website = 223,352 / 390,485 = 0.57198

This is our sampling correction.

```python
correction = 0.57198

for col in ['search results','refinements']:
    main[col] = main[col] * (1 + correction)
```

### Rate the similarity of strings
* Use Jaccard Index: https://en.wikipedia.org/wiki/Jaccard_index
* Or use https://stackoverflow.com/questions/48117508/how-to-find-a-similar-substring-inside-a-large-string-with-a-similarity-score-in?noredirect=1&lq=1

```python
def jacc(a,b):
    s1 = set(a)
    s2 = set(b)
    inter = s1.intersection(s2)
    union = s1.union(s2)
    return len(inter)/len(union)
```

### Find min score to use as threshold
Seems safe to say that any score over 50% is a good match

```python
test_words = ['arduino','raspberry pi','feather']
for t in test_words:
    ls = []
    words = list(set(main[main['search keyword'].str.contains(t)]['search keyword']))    
    for w in words:
        score = jacc(t, w)
        ls.append(score)
        
    print('\ntest word:  %s' % t)
    conf = jb_conf(pd.DataFrame(ls)).loc[:'upper'].rename(columns = {0:'95% confidence interval'})
    display(conf.format_(['p2']))
    print(words[:25])
```

### Get part categories to use as top words

```python
categories = pd.read_sql(
'''
SELECT
DISTINCT LOWER(categories_name) AS categories_name
FROM categories_description
''', db)

col_fix(categories)
```

### Rate all search keywords against category names

```python
start = dt.datetime.now()

top_words = list(categories['categories name'])
all_words = list(set(main['search keyword'][main['search keyword'] != '']))

ls = []
for a in all_words:
    score = 0
    for t in top_words:
        inter_score = jacc(a,t)
        if inter_score > score:
            score = inter_score
            t2 = t
    ls.append([t2,a,score])
    
end = dt.datetime.now()

print('runtime: {}'.format(end-start))

cols = ['top word','search keyword','score']
scores = pd.DataFrame(data = ls, columns = cols)
```

### By top word and keyword

```python
N = 10
top_words_dict = {}
search_words_dict = {}
for_avg = []

main2 = main.copy()
for col in main2.columns:
    try:
        main2[col] = main2[col].str.title()
    except:
        pass
    
main2['account name'] = main2['account name'].str.replace('Www.Adafruit.Com','www.Adafruit.com')    
   
for account_name in list(set(main2['account name']))[::-1]:

#=========================
# by top word
#=========================
    
    a1 = main2[main2['account name'] == account_name]
    a2 = a1.groupby(['account name','top word'])[['search results']].sum()
    a2.sort_values('search results', ascending = False, inplace = True)
    a3 = a2.iloc[:10].copy()
    
    total1 = a1['search results'].sum()
    
    a3['% of account name'] = a3['search results']/total1
    
    top_words_dict[account_name] = a3
    
#=========================
# by keyword search
#=========================    

    top_words = list(a3.index.get_level_values(1))
    [for_avg.append(x) for x in top_words]
    
    b1 = main2[(main2['account name'] == account_name)
             & (main2['top word'].isin(top_words))]
    b2 = b1.groupby(['account name','top word','search keyword'])[['search results']].sum()
    b3 = b2['search results'].groupby(level = [0,1], group_keys = False)
    b4 = b3.apply(lambda x: x.sort_values(ascending = False).head(N))
    b5 = pd.DataFrame(b4)    
    
    total2 = a2.reset_index()
    total2.rename(columns = {'search results':'total'}, inplace = True)    
    
    b6 = b5.reset_index().merge(total2, on = ['account name','top word'])
    b6['% of "top words"'] = b6['search results']/b6['total']
    b6.sort_values(['total','search results'], ascending = [False,False], inplace = True)
    b6.drop('total', 1, inplace = True)   
    b7 = b6.set_index(['account name','top word','search keyword'])
    
    search_words_dict[account_name] = b7
    
    print(account_name)
    fmt = ['n0','p1']
    display(top_words_dict[account_name].head().format_(fmt))
    display(search_words_dict[account_name].head().format_(fmt))
    print('\n')
    
for_avg = list(set(for_avg))
```
