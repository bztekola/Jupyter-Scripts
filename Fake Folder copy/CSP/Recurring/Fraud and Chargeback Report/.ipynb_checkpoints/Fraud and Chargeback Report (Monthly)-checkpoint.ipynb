### Get total revenue data

```python
ds = pd.read_sql(
'''
SELECT
DATE_FORMAT(timestamp, '%Y-%m') AS year_and_month,
SUM(d_value) AS total_revenue
FROM daily_stats
WHERE d_class = 'd_all'
AND timestamp BETWEEN ' '''+ date_start +''' ' AND ' '''+ date_end +''' '
GROUP BY year_and_month
''', db)

col_fix(ds)
ds.set_index('year and month', inplace = True)
```

### Fraud confirmed, fraud stopped, and returns

```python
dict_ = {}

ls = ['fraud - confirmed',
      'fraud - void']

for x in ls:
    df = fraud_main[fraud_main['orders status name'] == x].groupby('year and month added').agg({'orders id':'nunique',
                                                                                                'order total':'sum'})
    df.rename(columns = {'orders id':'unique OID count',
                        'order total':x + ' total'}, inplace = True)
    df = df.join(ds)
    df[x + ' as % of total revenue'] = df[x + ' total']/df['total revenue']
    
    dict_[x] = df
```

### All fraud

```python
a = dict_['fraud - confirmed'][['fraud - confirmed total']]
b = dict_['fraud - void'][['fraud - void total']]
c = ds

fmt = ['m0','m0','m0','m0','p2','p2']

all_fraud = a.join(b)
all_fraud['total fraud'] = all_fraud.sum(1)
all_fraud = all_fraud.join(c)
all_fraud['total fraud as % of total revenue'] = all_fraud['total fraud']/all_fraud['total revenue']
all_fraud['% of fraud stopped'] = all_fraud['fraud - void total']/all_fraud['total fraud']
all_fraud.format_(fmt)
```

### Plot it

```python
fig, ax = plt.subplots(figsize = (20,5))

cols = ['fraud - confirmed total','fraud - void total']

all_fraud[cols].plot(ax = ax,
                     kind = 'bar',
                     stacked = True,
                     edgecolor = 'white',
                     lw = 3,
                     color = colors,
                     zorder = 5)

ax.set_xticklabels([jb_dates(x, 'year and month xticks') for x in all_fraud.index],rotation = 0)
ax.set_yticklabels(['${:,.0f}'.format(x/1000) for x in ax.get_yticks()])
ax.set_ylabel('Thousands', fontsize = 20)
ax.set_xlabel('')
ax.legend([x.title() for x in cols], bbox_to_anchor = (0.7,-0.15), edgecolor = 'black', fontsize = 25)
ax.set_title('Fraud Over Time\nwith % of Fraud Stopped', y = 1.02)
ax.grid(color = 'white', alpha = 0.75, zorder = 0)

for i in range(len(all_fraud)):
    x = i
    y = all_fraud.iloc[i]['fraud - confirmed total'] + 1000
    wrt = all_fraud.iloc[i]['% of fraud stopped']
    wrt2 = '{:,.2f}%'.format(wrt * 100)
    ax.annotate(wrt2, xy = (x,y), ha = 'center', fontsize = 20, color = 'white', zorder = 10)
    
plt.show()
```

### 95% confidence intervals per month

```python
main_ls = []
for col in all_fraud.columns:
    df = all_fraud[col]
    m = df.mean()
    s = df.std()
    n = len(df)
    phi = stats.t.ppf(0.95, n-1)
    l = m - phi * (s/np.sqrt(n))
    u = m + phi * (s/np.sqrt(n))
    
    ls = [col,l,m,u]
    main_ls.append(ls)
```

```python
fmt = ['m0','m0','m0','m0','p2','p2']

conf = pd.DataFrame(data = main_ls, columns = ['data','lower','mean','upper'])
conf.set_index('data', inplace = True)
conf = conf.T
conf.format_(fmt)
```
