### Get orders status history data

```python
osh1 = pd.read_sql(
'''
SELECT
DATE(date_added) AS date_added,
orders_id,
0 AS partial_id,
CONCAT(orders_id, '-', 0) AS umbrella_orders_id,
orders_billing_id
FROM orders_status_history
WHERE orders_id NOT IN (SELECT psh_parent_id FROM partial_status_history)
AND orders_billing_id IN '''+ str(tuple(bs_include)) +'''
AND orders_status_id NOT IN '''+ str(tuple(os_exclude)) +'''
''', db)

col_fix(osh1)
```

```python
osh2 = pd.read_sql(
'''
SELECT
DATE(psh_date_added) AS date_added,
psh_parent_id AS orders_id,
psh_partial_id AS partial_id,
CONCAT(psh_parent_id, '-', psh_partial_id) AS umbrella_orders_id,
psh_billing_id AS orders_billing_id
FROM partial_status_history
WHERE psh_billing_id IN '''+ str(tuple(bs_include)) +'''
AND psh_partial_status_id NOT IN '''+ str(tuple(os_exclude)) +'''
''', db)

col_fix(osh2)
```

### Merge together

```python
osh_main = pd.concat([osh1, osh2], sort = False)
osh_main.reset_index(drop = True, inplace = True)

osh_main['date added'] = pd.to_datetime(osh_main['date added'])
osh_main['year and month'] = jb_dates(osh_main['date added'], 'year and month')

osh_main['billing status name'] = osh_main['orders billing id'].map(dict(zip(bs['bs status id'], bs['bs status name'])))
osh_main.drop('orders billing id', 1, inplace = True)
```

### Get orders total data

```python
ot1 = pd.read_sql(
'''
SELECT
CONCAT(orders_id, '-', 0) AS umbrella_orders_id,
value AS umbrella_orders_total
FROM orders_total 
WHERE orders_id NOT IN (SELECT pt_parent_id FROM partial_total)
AND class = 'ot_total'
''', db)

col_fix(ot1)
```

```python
ot2 = pd.read_sql(
'''
SELECT
CONCAT(pt_parent_id, '-', pt_partial_id) AS umbrella_orders_id,
pt_value AS umbrella_orders_total
FROM partial_total
WHERE pt_class = 'pt_total'
''', db)

col_fix(ot2)
```

```python
ot_main = pd.concat([ot1, ot2], sort = False)
ot_main.reset_index(drop = True, inplace = True)

osh_main['umbrella orders total'] = osh_main['umbrella orders id'].map(dict(zip(ot_main['umbrella orders id'], ot_main['umbrella orders total'])))
```

### Remove duplicates
* Some (umbrella OIDs) + (billing status) combos are dupes, like [OID 998982](https://volcano.adafruit.com/volcano/Order_Status.php?oid=998982), which was billed twice and paid three times, despite it not being made of partials.
* Below we take the earliest date some umbrella OID was marked as "billed", likewise for "payment received"

```python
# do it
osh_main.sort_values(['umbrella orders id','billing status name','date added'], inplace = True)
osh_main.drop_duplicates(['umbrella orders id','billing status name'], keep = 'first', inplace = True)

# check it
a = osh_main.groupby(['umbrella orders id','billing status name'])[['billing status name']].count().unstack(1)
a.columns = a.columns.droplevel(0)

dupes = a[a.sum(1) > 2]

if len(dupes) != 0:
    raise ValueError('You still have dupes')
```

### Remove paid but not billed
* Some umbrella OIDs were paid without being billed, like [OID 1950471](https://volcano.adafruit.com/volcano/Order_Status.php?oid=1950471).
* Remove these

```python
a = osh_main.groupby(['umbrella orders id','billing status name'])[['billing status name']].count().unstack(1)
a.columns = a.columns.droplevel(0)

b = a[(a['payment received'] == 1) & (a['billed'].isnull())].index.tolist()

print('{:,.0f} umbrella OIDs were paid but not billed'.format(len(b)))

osh_main.drop(osh_main[osh_main['umbrella orders id'].isin(b)].index, inplace = True)
osh_main.reset_index(drop = True, inplace = True)
```

### Unpaid Rate
*****

* Teddy wants a ratio of unpaid to paid (but I think he really means unpaid to billed), which can also be called something like the "unpaid rate".
* We are going to get the current unpaid rate, then get past unpaid rates, then create a benchmark, then compare the two.
* Any umbrella OID which was billed within 30 days of some end date we choose is not applicable since we only care about outstanding debt which is more than 30 days old. 
* So let's make a time window which starts one year + 30 days from now and ends 30 days prior to now.
* Then we'll slide the window back every 30 days in order to get historical ratios

```python
now = dt.datetime.now()
ls = []

for i in np.arange(1,13):
    
    #=========================
    # create time window
    #=========================
    
    d = int(30*i)
    start = str((now - pd.DateOffset(years = 1, days = d)).date())
    end = str((now - pd.DateOffset(days = d)).date())
    
    valid_for = pd.to_datetime(end) + pd.DateOffset(days = 30)

    #=========================    
    # get all umbrella OIDs that have been billed within our time window
    #=========================
    
    oids = osh_main[(osh_main['billing status name'] == 'billed')
                  & (osh_main['date added'].between(start, end))]['umbrella orders id'].tolist()
    
    #=========================
    # get the data belonging to these umbrella OIDs
    #=========================    
    
    osh = osh_main[osh_main['umbrella orders id'].isin(oids)].copy()

    #=========================    
    # sum up billed and paid
    # groupbys alphabetize, so "billed" comes first, "payment received" comes second
    #=========================
    
    billed, paid = osh.groupby('billing status name')[['umbrella orders total']].sum().values

    #=========================    
    # get the unpaid rate, which is the complement of the paid rate, so is 1 - (paid / billed)
    #=========================
    
    unpaid_rate = (1 - paid/billed).item()
    
    #=========================    
    # store results
    #=========================
    
    ls.append([start,end,valid_for,billed.item(),paid.item(),unpaid_rate])
    
# create dataframe    
cols = ['date start','date end','stat valid for','billed','paid','unpaid rate']
ratios = pd.DataFrame(data = ls, columns = cols)

# get confidence interval for avg rate per month
mean = ratios['unpaid rate'].mean()
stand_dev = ratios['unpaid rate'].std()
n = len(ratios)
alpha = 0.05
t = stats.t.ppf(1-alpha/2, n-1)

lower = mean - t * (stand_dev/np.sqrt(n))
upper = mean + t * (stand_dev/np.sqrt(n))

cols = ['lower','mean','upper']
conf = pd.DataFrame(data = [[lower,mean,upper]], columns = cols).T
conf.columns = ['unpaid rate']

print('95% confidence interval for unpaid rate')
display(conf.format_(['p2']))

# add some comments
def comments(x):
    val = x['unpaid rate']
    
    if np.logical_and(val >= conf.loc['lower'].item(), val <= conf.loc['upper'].item()):
        ret = 'normal'
    else:
        if val > conf.loc['upper'].item():
            ret = 'HIGHER than avg'
        else:
            ret = 'lower than avg'
            
    return ret
            
ratios['comments'] = ratios.apply(comments, axis = 1)

fmt = [0,0,0,'m0','m0','p2',0]
display(ratios.format_(fmt))
```

```python
from scipy.interpolate import interp1d

fig, ax = plt.subplots()

chart = ratios.set_index('date end')
chart = chart.iloc[::-1]

x = range(len(chart))
y = chart['unpaid rate'].values

x_smooth = np.linspace(np.min(x), np.max(x), 500)
f = interp1d(x, y, kind = 'quadratic')
y_smooth = f(x_smooth)

ax.plot(x_smooth, y_smooth, lw = 3)

for v in conf.values:
    ax.axhline(v, ls = '--', color = 'purple', alpha = 0.50)
    
ax.set_yticklabels('{:,.2f}%'.format(x * 100) for x in ax.get_yticks())

xticks = chart.index.tolist()
ax.set_xticks(range(len(xticks)))
ax.set_xticklabels(xticks, rotation = 45)
ax.set_xlabel('Year Ending')

ax.grid(alpha = 0.25)

ax.set_title('Unpaid Rate\nwith 95% Confidence Interval for avg rate per month')

plt.show()
```
