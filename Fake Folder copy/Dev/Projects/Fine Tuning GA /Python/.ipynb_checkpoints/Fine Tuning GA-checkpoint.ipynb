# Proportion of Orders that are Tracked and OID Revenue
* From email with subject line "Fine Tuning GA: Proportion of Orders that are Tracked and OID Revenue"
* Hand-check [here](https://analytics.google.com/analytics/web/#/report/conversions-ecommerce-sales-performance-beta/a7723544w14744224p15556579/), on GA

### GA data

```python
ga_rev = get_ga(date_start, date_end, ['date','transactionId'], ['transactionRevenue'])

ga_rev.rename(columns = {'transaction id':'orders id',
                         'transaction revenue':'ga subtotal',
                         'date':'ga date'}, inplace = True)

for x in ['orders id','ga subtotal']:
    ga_rev[x] = pd.to_numeric(ga_rev[x])
```

### Find dupes

```python
oids = ga_rev[ga_rev['orders id'].duplicated()]['orders id'].tolist()
dupes = ga_rev[ga_rev['orders id'].isin(oids)].sort_values('orders id')

print('{:,.0f} dupes'.format(len(oids)))
print('or {:,.2f}% of total lines'.format(len(oids)/len(ga_rev) * 100))
print('remove them')

ga_rev.drop(ga_rev[ga_rev['orders id'].isin(oids)].index, inplace = True)
```

```python
dupes[dupes['orders id'].isin(oids[:5])]
```

### orders_total data

```python
ot_main = pd.read_sql(
'''
SELECT
DATE(date_purchased) AS date_purchased,
ot.orders_id,
ot.value AS db_subtotal,
orders_reseller AS reseller,
orders_super_reseller AS super_reseller,
LOWER(orders_po_number) AS po,
orders_fake AS fake,
IF(o.orders_id IN (SELECT orders_id FROM subscriptions_history), 1, 0) AS adabox

FROM orders_total ot

JOIN orders o ON ot.orders_id = o.orders_id
AND DATE(date_purchased) BETWEEN ' '''+ date_start +''' ' AND ' '''+ date_end +''' '

WHERE class = 'ot_subtotal'
''', db)

col_fix(ot_main)

ot_main['po'] = ot_main['po'].str.strip()
ls = ['','n/a']
ot_main['po'] = np.where(ot_main['po'].isin(ls), 0, 1)
```

### Audit

```python
# merge
audit_main = pd.merge(ot_main, ga_rev, how = 'left', on = 'orders id')

# find nulls
nulls = audit_main[audit_main.isnull().any(1)]
print('{:,.0f} nulls, which is to say the count of OIDs not in GA'.format(len(nulls)))
print('or {:,.2f}% of total lines'.format(len(nulls)/len(audit_main) * 100))

# make columns
audit_main['year and month'] = [str(x)[:7] for x in audit_main['date purchased']]
audit_main['customer type'] = np.where(audit_main[['reseller','super reseller','po','fake','adabox']].sum(1) > 0, 
                                       'reseller/super/po/fake/adabox',
                                       'regular customer')
audit_main['in ga'] = np.where(audit_main['ga subtotal'].isnull(), 'no', 'yes')
```

### Group by GA

```python
by_ga = audit_main.groupby(['in ga','customer type'])[['customer type']].count().unstack(1)
by_ga.columns = by_ga.columns.droplevel(0)
by_ga['row total'] = by_ga.sum(1)
by_ga.loc['column total'] = by_ga.sum()

if len(ot_main) == by_ga.loc['column total','row total']:
    print('match')
else:
    print('mismatch')
    
by_ga.format_(['n0'] * 3)
```

```python
by_ga.div(len(ot_main)).format_(['p2'] * 3)
```

### OID mismatches

```python
mismatch = audit_main[np.abs(audit_main['db subtotal'] - audit_main['ga subtotal']) > 0]

print('{:,.0f} mismatches'.format(len(mismatch)))
print('or {:,.2f}% of total lines'.format(len(mismatch)/sum(audit_main['in ga'] == 'yes') * 100))

diff = mismatch['db subtotal'].sum() - mismatch['ga subtotal'].sum()
print('\ntotal difference between db and ga subtotal: ${:,.0f}'.format(diff))
print('or ${:,.0f} per mismatch order'.format(diff/len(mismatch)))
```
