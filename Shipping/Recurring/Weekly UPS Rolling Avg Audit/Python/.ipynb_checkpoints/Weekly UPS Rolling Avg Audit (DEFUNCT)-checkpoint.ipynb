```python
date_end = str(dt.datetime.now().date())
one_year_ago = str((dt.datetime.now() - pd.DateOffset(years = 1)).date())
two_months_ago = str((dt.datetime.now() - pd.DateOffset(months = 2)).date())

# the saturday two weeks ago, if you're running this regularly on mondays
week_end = '2018-11-10'
```

### Get main

```python
ups_super_main = pd.read_sql(
'''
SELECT
orders_id AS 'orders id',
DATE(transaction_date) AS date,
DATE_FORMAT(transaction_date, '%Y-%m') AS 'year and month',
IF(account_number = '000071EY05', '71', 'A8') AS account,
charge_description AS 'charge description',
basis_value AS 'gross charge',
netAmount AS 'net charge'
FROM ups_billing
''', db)

ups_super_main['date'] = pd.to_datetime(ups_super_main['date'], errors = 'coerce')
```

```python
pd.read_sql(
'''
SELECT
*
FROM orders_status
ORDER BY orders_status_id
''', db)
```

```python
orders_main = pd.read_sql(
'''
SELECT
orders_id AS 'orders id',
DATE(date_purchased) AS 'date purchased',
delivery_country AS 'delivery country raw'
FROM orders
WHERE shipping_module_code = 'upsxml'
AND orders_status NOT IN (8,9,14,15) # exclude only fraud and voids, since we pay shipping for returns, replacements, etc.
''', db)

orders_main['delivery country raw'] = [x.lower() for x in orders_main['delivery country raw']]
orders_main['delivery country raw'] = orders_main['delivery country raw'].str.strip()
orders_main['region'] = np.where(orders_main['delivery country raw'] == 'united states', 'domestic','international')
orders_main['date purchased'] = pd.to_datetime(orders_main['date purchased'])
```

```python
print('delivery regions as % of total')
print('from %s to %s' % (ups_super_main['date'].min().date(), ups_super_main['date'].max().date()))
orders_main['region'].value_counts()/orders_main['region'].value_counts().sum()
```

### Clean up UPS data

```python
# create dates so we can fill in where we didn't ship any UPS
dates = pd.DataFrame(pd.date_range(ups_super_main['date'].min(), ups_super_main['date'].max()))
dates.columns = ['date']

ups_main = pd.merge(dates, ups_super_main, how = 'left', on = 'date')

# create date columns
ups_main['year and month'] = [str(x)[:7] for x in ups_main['date']]

# make it so that the week ends on saturday, just like UPS
ups_main['week'] = ups_main['date'].dt.week.shift(-2).ffill()
ups_main['year'] = ups_main['date'].dt.year

# get the end date of each year/week
week_ending = ups_main.groupby(['year','week'], as_index = False)[['date']].max()
week_ending.columns = ['year','week','week ending']

ups_main = pd.merge(ups_main, week_ending, on = ['year','week'])
ups_main['week ending'] = [x.date() for x in ups_main['week ending']]
```

```python
# sum by week
ups_by_week = ups_main.groupby('week ending')[['gross charge']].sum()

# get 52 week rolling avg as if there were never any ramp ups
ups_by_week['rolling avg'] = ups_by_week['gross charge'].rolling(window = 52).mean()

ups_by_week.index = pd.to_datetime(ups_by_week.index)
```

```python
# 6/25/2018 is the beginning of the new rate structure (see email "UPS: Updated Pricing Agreement")
# so our rolling avg starts over here
# but make it 6/23, the saturday before 6/25, for simplicity

updated_avg = {}
df = ups_by_week.loc['2018-06-23':]

for i in np.arange(1, len(df) + 1):
    new_avg = df.iloc[:i]['gross charge'].values
    new_avg = np.mean(new_avg)
    ix = df.iloc[i-1].name
    updated_avg[ix] = new_avg
    
ups_by_week['current rolling avg'] = ups_by_week.index.to_series().map(updated_avg)    

# from email with subject line "Quick question on ramp up during w/e 5/27/2018"
# john from UPS says rolling avg w/e 6/02/2018 was $81,552
# db says 88,371
# so MAPE is 8.36%
# a MAPE of <= 10% is expected
mape = 0.0836
print('MAPE is {:,.2f}%'.format(mape * 100))

ups_by_week['current rolling avg +/-'] = ups_by_week['current rolling avg'] * mape
```

```python
ups_by_week[ups_by_week.index <= week_end].iloc[::-1].dropna().drop(['rolling avg','gross charge'],1).format_(['m0','m0']).head(3)
```

### Plot it

```python
for_chart = ups_by_week.copy()
for_chart.drop(['current rolling avg','current rolling avg +/-'],1,inplace = True)

fig, axes = plt.subplots(nrows = 2, ncols = 1, figsize = (10,10))

for_chart[for_chart.index < week_end].plot(ax = axes[0],
                                                 grid = True,
                                                 title = 'UPS gross charge from the database, along with the 52 week rolling average\nfrom all time')

for_chart[(for_chart.index >= one_year_ago)
            & (for_chart.index <= week_end)].plot(ax = axes[1],
                                                             grid = True,
                                                             title = 'Database gross charge within the last year',
                                                             color = ['green','purple'])
axes[0].set_xlabel('')
axes[0].set_xticklabels('')
axes[1].set_xlabel('')
for ax in axes:
    ax.set_yticklabels(['${:,.0f}'.format(x) for x in ax.get_yticks()])
    
plt.savefig('ups weekly gross charges', bbox_inches = 'tight')
plt.show()
```

### Get outgoing orders only, for later calcs

```python
# return only outgoing data that is both in "orders" and "ups_billing"
outgoing_main = ups_super_main[ups_super_main['orders id'].isin(orders_main['orders id'])].copy()

# map delivery region
outgoing_main['region'] = outgoing_main['orders id'].map(dict(zip(orders_main['orders id'], orders_main['region'])))

# create date columns
outgoing_main['year and month'] = [str(x)[:7] for x in outgoing_main['date']]
outgoing_main['week'] = outgoing_main['date'].dt.week
outgoing_main['year'] = outgoing_main['date'].dt.year

# get "week ending"
outgoing_main = pd.merge(outgoing_main, week_ending, on = ['year','week'])
outgoing_main['week ending'] = [x.date() for x in outgoing_main['week ending']]
```

```python
# prepare for weekly summary
by_oid = outgoing_main.groupby(['week ending','region','orders id'], as_index = False).agg({'gross charge':'sum',
                                                                                            'net charge':'sum'})
by_oid['week ending'] = pd.to_datetime(by_oid['week ending'])
```

```python
# find outliers by UPS net charge, by region
# we want averages per week, outliers will inflate these averages
outlier_dict = {}
outlier_ls = []
for region in ['domestic','international']:
    df = by_oid[by_oid['region'] == region]
    mn = df['net charge'].mean()
    s = df['net charge'].std()
    outlier_dict[region] = [mn + 3*s]
    
    for index, row in df.iterrows():
        if row['net charge'] > mn + 3*s:
            outlier_ls.append(row['orders id'])
            
print(outlier_dict)
print('{:,.2f}% of total OIDs are outliers'.format(len(outlier_ls)/len(by_oid) * 100))
print('that\'s {:,.0f} out of {:,.0f}'.format(len(outlier_ls), len(by_oid)))
print('remove them')

by_oid.drop(by_oid[by_oid['orders id'].isin(outlier_ls)].index, inplace = True)
```

```python
# the true first ramp up date for this data is 2017-07-08
# but as our plotting window moves we must update this initial date

ramp_up_start = [#'2017-07-08',
                #'2017-09-24',
#                 one_year_ago,
#                '2017-10-22',
                '2017-11-25',
                '2018-01-13',
                '2018-03-31',
                '2018-05-05']

ramp_up_end = [#'2017-09-02',
#               '2017-10-22',
               '2017-12-09',
               '2018-03-03',
               '2018-04-21',
               '2018-05-26'] # final ramp up end date as of 6/11/2018

ups_yellow = '#FCB900'
ups_brown = '#301504'
#ups_facecolor = '#eae7e5'
ups_facecolor = 'white'

fig, axes = plt.subplots(figsize = (20,10), nrows = 2, ncols = 1)

region_ls = ['domestic','international']

for i in range(len(region_ls)):

    df1 = by_oid[(by_oid['region'] == region_ls[i])
               & (by_oid['week ending'] >= one_year_ago)].groupby('week ending').agg({'net charge':'sum',
                                                                                      'gross charge':'sum',
                                                                                      'orders id':'count'})

    df1.rename(columns = {'orders id':'oid count'}, inplace = True)

    df2 = by_oid[(by_oid['region'] == region_ls[i])
               & (by_oid['week ending'] >= one_year_ago)].groupby('week ending').agg({'net charge':'mean'})
    df2.rename(columns = {'net charge':'avg net per order'}, inplace = True)

    df3 = df1.join(df2)    
    df3 = df3[['gross charge','net charge','oid count','avg net per order']]
    df3 = df3.loc[:week_end]
    
    df3.index = [str(x)[:10] for x in df3.index]
    
    df3['for axis'] = range(len(df3))
    
    x = np.arange(0,len(df3))
    y1 = df3['avg net per order']
    
    axes[i].plot(x, y1, color = ups_brown, lw = 6)
    
    min_y = df3['avg net per order'].min() - df3['avg net per order'].min()/20
    max_y = df3['avg net per order'].max() + df3['avg net per order'].max()/20
    axes[i].set_ylim(min_y, max_y)    
    
    axes[i].set_title(region_ls[i].title(), fontsize = 25, y = 1.02)
    axes[i].set_xlabel('')
    axes[i].set_yticklabels(['${:,.2f}'.format(x) for x in axes[i].get_yticks()], fontsize = 15)
    axes[i].legend().set_visible(False)
        
    plt.suptitle('Avg Charge per UPS package\nwhere yellow represents a ramp up period', y = 1.02, fontsize = 25)
       
    print(region_ls[i])
    print(df3.drop('for axis',1).tail(3).iloc[::-1].format_(['n0'] * 3 + ['n2']))
    print('\n')   
    
    axes[i].set_xticks(x)
    axes[i].set_xlim(0,len(df3) - 1)
    axes[0].set_xticklabels('')
    axes[1].set_xticklabels([x for x in df3.index], rotation = 45, ha = 'right')
    
    for m,n in zip(ramp_up_start, ramp_up_end):
        axes[i].axvline(df3.loc[m,'for axis'], color = 'black')
        axes[i].axvline(df3.loc[n,'for axis'], color = 'black')    
        axes[i].fill_between([df3.loc[m,'for axis'], df3.loc[n,'for axis']], max_y, color = ups_yellow)    

    ramp_up = len(df1.loc[:'2018-06-30']) - 1
    axes[i].axvline(ramp_up, color = 'black', ls = '--')
    
plt.savefig('avg charge per package', bbox_inches = 'tight')
plt.show()        
```

### Confidence interval

```python
a = by_oid[(by_oid['region'] == 'international') & (by_oid['week ending'] > two_months_ago)]

m = a['net charge'].mean()
s = a['net charge'].std()
n = 8 # two months is 8 weeks
phi = stats.t.ppf(0.95, n - 1)

lower = m - phi * (s/np.sqrt(n))
upper = m + phi * (s/np.sqrt(n))

print('95% confidence interval for avg UPS international charge per order:')
print('Lower: ${:,.2f}\nMean: ${:,.2f}\nUpper: ${:,.2f}\n'.format(lower, m, upper))
print('Sample size is 8 weeks')
```
