```python
import sys
sys.path.insert(0,'/Users/jarad/fake_folder')

from jb_libraries import *
%matplotlib inline

#===================================================================================================================

def orders_to_cost_function(date_start, date_end):

    A = pd.read_sql(
    '''
    SELECT
    DATE(date_purchased) AS 'date',
    orders_id AS 'orders id'
    FROM orders
    WHERE DATE(date_purchased) BETWEEN '2017-01-01' AND '2018-01-31'
    ''', db)

    B = pd.read_sql(
    '''
    SELECT
    op.orders_id AS 'orders id',
    op.part_id AS 'main part id',
    pts.part_id AS 'sub part id',
    pts.contains_part_id AS 'contains part id'

    FROM orders_products op

    LEFT JOIN products_to_stuff pts ON op.part_id = pts.part_id

    WHERE op.orders_id IN '''+ str(tuple(A['orders id'])) +'''
    ''', db)

    B['sub part id'].fillna(B['main part id'], inplace = True)
    B['contains part id'].fillna(B['main part id'], inplace = True)

    C = pd.read_sql(
    '''
    SELECT
    otb.orders_id AS 'orders id',
    k.part_id AS 'cost part id',
    otb.sku_id AS 'sku id',
    IF(k.sku_outsourced_assembly = 1, pm.consigned_latest_cost, pm.latest_cost) AS 'latest cost'
    FROM orders_to_barcodes otb
    JOIN products_manufacturing pm ON otb.sku_id = pm.sku_id
    JOIN skus k ON otb.sku_id = k.sku_id
    WHERE otb.orders_id IN '''+ str(tuple(A['orders id'])) +'''
    ''', db)

    orders_to_cost_main = pd.merge(A,
                                 B,
                                 how = 'left',
                                 on = 'orders id').merge(C,
        #                                                how = 'left',
                                                        left_on = ['orders id','contains part id'],
                                                        right_on = ['orders id','cost part id'])

    # note the commented-out left join here
    # orders_to_barcodes is a historical record of every sku that was scanned for an order
    # so if a shipper doesn't scan a sku it won't be in otb

    # on the other hand we want to connect products_to_stuff and orders_to_barcodes
    # but products_to_stuff is NOT historical, meaning if we add or subtract some sku from some pack
        # then pts will reflect this change

    # if we left join, we have nulls either because pts changed or a shipper didn't scan a sku
    # it's more accurate to inner join and exclude the otb nulls
    # this way pts can change all it wants to, and otb will remain a historical record
    # from Jan 2017 to Jan 2018, only ~3.5% of all scanned skus had nulls
    # meaning ~3.5% of skus were not scanned by accident

    orders_to_cost_main['main part id'] = [int(x) for x in orders_to_cost_main['main part id']]
    orders_to_cost_main['sub part id'] = [int(x) for x in orders_to_cost_main['sub part id']]
    orders_to_cost_main['contains part id'] = [int(x) for x in orders_to_cost_main['contains part id']]
    orders_to_cost_main['cost part id'] = [int(x) for x in orders_to_cost_main['cost part id']]
    orders_to_cost_main['sku id'] = [int(x) for x in orders_to_cost_main['sku id']]

    if orders_to_cost_main[orders_to_cost_main.isnull().any(1)].empty:
        print('no nulls')
    else:
        print('you have nulls!!')

    orders_to_cost = orders_to_cost_main.groupby(['orders id','main part id'], as_index = False)[['latest cost']].sum()    
```

```python
orders_to_cost_function('2018-01-01','2018-01-03')
```
