```python
import sys
sys.path.insert(0,'/Users/jarad')

import pandas as pd
import numpy as np
from db2 import *
import glob

import datetime as dt

pd.set_option('display.max_columns', 500)
pd.set_option('display.max_rows', 100)

import matplotlib.pyplot as plt
%matplotlib inline

def str_to_num(x):
    try:
        return x.str.extract('(\d+)', expand = False)
    except:
        return x
```

```python
july_main = pd.read_csv(r'/Users/jarad/Desktop/july.csv', header = 2)
july_main['month'] = 7

aug_main = pd.read_csv(r'/Users/jarad/Desktop/aug.csv', header = 2)
aug_main['month'] = 8

sep_main = pd.read_csv(r'/Users/jarad/Desktop/sep.csv', header = 2)
sep_main['month'] = 9

csv = pd.concat([july_main,aug_main,sep_main])
```

```python
csv = csv[~csv['Keyword status'].str.contains('Total')]
csv.iloc[:,3:] = csv.iloc[:,3:].apply(str_to_num)
csv.iloc[:,3:] = csv.iloc[:,3:].astype(float)

csv.columns = [x.lower() for x in csv.columns]
csv.columns = [x.strip() for x in csv.columns]
```

# Group by month

```python
# conv value / cost
# cost / conv value
```

```python
by_month = csv.groupby('month')[['all conv. value','impr.','cost','conv. value','conversions','clicks']].sum()
by_month['conv value / cost'] = by_month['conv. value']/by_month['cost']
by_month['cost / conv. value'] = by_month['cost']/by_month['conv. value']

by_month
```

```python
ax = by_month[['conv value / cost','cost / conv. value']].plot(kind = 'bar', 
                                                               figsize = (20,5), 
                                                               width = 0.15, 
                                                               rot = 45,
                                                              grid = True)


vals = ax.get_yticks()
ax.set_yticklabels(['{:,.0f}%'.format(x * 100) for x in vals])

ax.set_title('Conversion Value and Costs', fontsize = 15)
ax.set_xticklabels(['Jul','Aug','Sep'], fontsize = 15)
ax.set_xlabel('Month', fontsize = 15)
ax.set_ylabel('Percentages', fontsize = 15)

plt.show()
```

# Group by keywords

```python
by_keyword = csv.groupby('keyword')[['all conv. value','impr.','cost','conv. value','conversions','clicks']].sum()
by_keyword['conv value / cost'] = by_keyword['conv. value']/by_keyword['cost']
by_keyword['cost / conv. value'] = by_keyword['cost']/by_keyword['conv. value']

by_keyword.sort_values('conv value / cost', ascending = False).head(10)
```

# Top peformers over all three months

```python
by_keyword[by_keyword['conv value / cost'] >= 2.5].sort_values('conv value / cost', ascending = False).reset_index()
```

# Top performers for Sep

```python
sep = csv.copy()
sep = sep[sep['month'] == 9]
sep[sep['conv. value / cost'] >= 2.5].sort_values('conv. value / cost', ascending = False)
```
