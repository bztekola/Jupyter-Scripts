### Settings

```python
date_start = '2018-11-01'
```

### Get BOA data

```python
boa_main = pd.read_csv('/Users/jarad/Fake Folder/Desktop/Budget for Python/Budget csv main.csv')
boa_main.columns = boa_main.columns.str.lower()

# make a copy
boa = boa_main.copy()

# structure
boa['posted date'] = pd.to_datetime(boa['posted date'])
boa['year and month'] = jb_dates(boa['posted date'], 'year and month')

boa['amount'] = pd.to_numeric(boa['amount'])
boa['payee'] = boa['payee'].str.lower()

ls = ['ny',
     'new york',
     'astoria',
     '[^a-z]',
     'xxxxx']

for x in ls:
    boa['payee'] = boa['payee'].str.replace(x,' ')

boa['payee'] = boa['payee'].str.replace('\s+',' ')
boa['payee'] = boa['payee'].str.strip()

ls = ['transfer',
     'payment',
     'chase credit crd des epay id indn monica e bushnell co id web',
     'beginning balance as of']
for x in ls:
    boa.drop(boa[boa['payee'].str.contains(x)].index, inplace = True)
    boa.reset_index(drop = True, inplace = True)
           
# restrict by date        
boa.drop(boa[boa['posted date'] < date_start].index, inplace = True)
boa.reset_index(drop = True, inplace = True)
```

### Get chase data

```python
chase_main = pd.read_csv('/Users/jarad/Fake Folder/Desktop/Budget for Python/mona chase.csv', header = None)
chase_main[0] = chase_main[0].str.lower()
chase_main.drop(chase_main[chase_main[0].str.contains('nyp')].index, inplace = True)
chase_main.reset_index(drop = True, inplace = True)

chase = chase_main.copy()

# fix dates
chase['posted date'] = chase[0].str[:5]
for index, row in chase.iterrows():
    
    d = row['posted date']
    
    if '01/' in d:
        yr = '/2019'
    else:
        yr = '/2018'
    chase.loc[index, 'posted date'] = d + yr
    
chase['posted date'] = pd.to_datetime(chase['posted date'], errors = 'coerce')
chase['year and month'] = jb_dates(chase['posted date'], 'year and month')
chase.drop(chase[chase['posted date'] < date_start].index, inplace = True)
chase.reset_index(drop = True, inplace = True)

# fix text
ls = ['ny',
     'new york',
     'astoria',
     '[^a-z]',
     'xxxxx']

chase['payee'] = chase[0]
for x in ls:
    chase['payee'] = chase['payee'].str.replace(x,' ')   
chase['payee'] = chase['payee'].str.replace('\s+',' ')
chase['payee'] = chase['payee'].str.strip()

# drop these strings
ls = ['payment thank you web']
for x in ls:
    chase.drop(chase[chase['payee'] == x].index, inplace = True)
    chase.reset_index(drop = True, inplace = True)
       
# fix amounts
for_amount = chase[0].str.split(' ', expand = True)
for_amount.fillna(method = 'ffill', axis = 1, inplace = True)
col = for_amount.columns[-1]
chase['amount'] = for_amount[col]
chase['amount'] = chase['amount'].str.replace(',','')
chase['amount'] = pd.to_numeric(chase['amount'])
# make purchases negative, refunds positive
chase['amount'] = chase['amount'] * -1

# fix columns
chase.drop(0, axis = 1, inplace = True)
chase['card'] = 'chase'
```

### Concat

```python
main = pd.concat([boa, chase], sort = False)
main.reset_index(drop = True, inplace = True)

date_end = str(main['posted date'].max().date())

if main.isnull().any(1).sum() > 0:
    raise ValueError('u have nulls')

# label income and expense    
main['income/expense'] = np.where(main['amount'] < 0, 'expense','income')

# clean up names
ls = ['grubhub',
     'instacart',
     'key food',
     'audible',
     'withdrwl',
     'lyft',
     'deposit',
     'google',
     'jcpenney',
     'fruit mar']

for x in ls:
    for index, row in main.iterrows():
        if x in row['payee']:
            main.loc[index, 'payee'] = x
        else:
            pass
```

### Single out refunds and returns

```python
main2 = main.groupby(['payee','income/expense'])[['amount']].sum().unstack(1).fillna(0)
main2.columns = main2.columns.droplevel(0)
main2 = main2[(main2 != 0).all(1)]
main2['total'] = main2.sum(1)

display(main2.format_(['m0'] * 3))

main2_d = main2[['total']].T.to_dict('list')
```

### By income/expense

```python
by_ie = main.groupby('income/expense')[['amount']].sum()
by_ie.loc['net'] = by_ie.sum()

ls = ['income','expense','net']
by_ie = by_ie.loc[ls]

inc_v, exp_v, net_v = by_ie['amount'].values.flatten().tolist()

print('From {} to {}'.format(date_start, date_end))
print('Net comes to ${:,.0f} per month on average.'.format(net_v/4))
print('Expense as a % of income: {:,.2f}%'.format(np.abs(exp_v)/inc_v * 100))
display(by_ie.format_(['m0']))
```

```python
ax = by_ie['amount'].map(np.abs).plot(kind = 'bar',
                                      color = ['green','purple','blue'],
                                      edgecolor = 'black',
                                      lw = 3,
                                      alpha = 0.75)

ax.set_xticklabels(ax.get_xticklabels(), rotation = 0)
ax.set_xlabel('')

ax.set_yticklabels('${:,.0f}'.format(x) for x in ax.get_yticks())

ax.set_title('Totals')

for i in range(len(by_ie)):
    val = by_ie.iloc[i].item()
    if i < 2:
        c = 'white'
    else:
        c = 'black'
    w = '-${:,.0f}'.format(np.abs(val)) if val < 0 else '${:,.0f}'.format(val)
    ax.annotate(w,
               xy = (i, 10000),
               fontsize = 12,
               ha = 'center',
               color = c)

plt.show()
```

### By payee
Excluding income and rent

```python
d1 = main['posted date'].min().date()
d2 = main['posted date'].max().date()

m = int(np.ceil((d2 - d1).days/30))

print('{} to {} is about {} months'.format(d1,d2,m))
```

```python
exp = main[(main['income/expense'] == 'expense')
         & (main['payee'] != 'check')].groupby(['payee','card'])[['amount']].sum().unstack(1).fillna(0)

exp.columns = exp.columns.droplevel(0)
for col in exp.columns:
    exp[col] = exp[col].map(int)
    
exp['total'] = exp.sum(1)    
exp = exp[['total']]   

# adjust for refunds and returns
for k,v in main2_d.items():
    if k in exp.index.tolist():
        exp.loc[k, 'total'] = main2.loc[k,'total']

# add some columns        
exp.sort_values('total', inplace = True)
exp['% of total'] = exp['total']/exp['total'].sum()
exp['running %'] = exp['% of total'].cumsum()

exp['avg per {} months'.format(m)] = exp['total']/m

fmt = ['m0','p2','p2','m0']
display(exp.format_(fmt).head(25))
```

```python
def my_autopct(pct):
    return ('%.1f%%' % pct) if pct > 2 else ''

fig, ax = plt.subplots(figsize = (10,10))

chart = exp
data = np.abs(chart['total'])
labs = chart.index.tolist()

wedges, labels, texts = ax.pie(data,
                              labels = labs,
                              rotatelabels = True,
                              autopct = my_autopct,
                              pctdistance = 0.85)

for i in range(len(wedges)):
    wedges[i].set_edgecolor('black')

for t in texts:
    t.set_color('white')
    t.set_fontsize(15)
    
ax.set_title('Pie Chart of Expenses\nExcluding Rent'.format(date_start, date_end), y = 1.2)

plt.show()
```
