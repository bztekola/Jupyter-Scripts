```python
def str_to_num(x):
    try:
        return x.str.extract('(\d+)', expand = False)
    except:
        return x
```

```python
reports.iloc[:,3:] = reports.iloc[:,3:].apply(str_to_num)
reports.iloc[:,3:] = reports.iloc[:,3:].apply(pd.to_numeric)

reports.columns = [x.lower() for x in reports.columns]
reports.columns = [x.strip() for x in reports.columns]

reports.rename(columns = {'search keyword':'keyword'}, inplace = True)
```

```python
adwords.iloc[:,0:2] = adwords.iloc[:,0:2].apply(str_to_num)
adwords.iloc[:,0:2] = adwords.iloc[:,0:2].apply(pd.to_numeric)

adwords.iloc[:,4:11] = adwords.iloc[:,4:11].apply(str_to_num)
adwords.iloc[:,4:11] = adwords.iloc[:,4:11].apply(pd.to_numeric)

adwords.columns = [x.lower() for x in adwords.columns]
adwords.columns = [x.strip() for x in adwords.columns]

adwords = adwords.iloc[:, 2:]
```
