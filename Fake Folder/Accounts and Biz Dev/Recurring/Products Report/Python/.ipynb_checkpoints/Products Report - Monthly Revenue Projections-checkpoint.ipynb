```python
import warnings
import itertools
import statsmodels.api as sm
```

### Set end time

```python
db_end_date = '2017-08-31'
db_start_date = str((pd.to_datetime(db_end_date) - pd.DateOffset(years = 5)).date())
```

### Get revenue for current month to check past month's projection

```python
check = pd.read_sql(
'''
SELECT
DATE_FORMAT(o.date_purchased, '%Y-%m') AS date_format,
SUM((op.products_quantity - op.products_quantity_free) * op.products_price) AS revenue
FROM orders o 
JOIN orders_products op ON o.orders_id = op.orders_id
WHERE DATE(o.date_purchased) BETWEEN '2017-08-01' AND '2017-08-31'
AND op.part_id NOT IN '''+ str(tuple(always_exclude)) +'''
AND o.orders_status != 9
AND o.orders_status != 10
AND o.payment_method != 'Replacement Order'
GROUP BY date_format
''', db)

check['revenue'] = check['revenue'].apply(lambda x: '{:,.0f}'.format(x))
check
```

### Get data for the last 5 years

```python
data_raw = pd.read_sql(
'''
SELECT
DATE_FORMAT(o.date_purchased, '%Y-%m') AS date_format,
SUM((op.products_quantity - op.products_quantity_free) * op.products_price) AS revenue
FROM orders o 
JOIN orders_products op ON o.orders_id = op.orders_id
WHERE DATE(o.date_purchased) BETWEEN ' '''+ db_start_date +''' ' AND ' '''+ db_end_date +''' '
AND op.part_id NOT IN '''+ str(tuple(always_exclude)) +'''
GROUP BY date_format
''', db)

data = data_raw.copy()
```

```python
data['date'] = data['date_format'].map(str) + '-01'
data['date'] = pd.to_datetime(data['date'])
data.set_index('date', inplace = True)
data.drop('date_format', 1, inplace = True)
```

### Plot current data

```python
fig = plt.figure(figsize = (15,10))
fig.subplots_adjust(hspace=.5)

ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)

data['revenue'].plot(ax = ax1)
ax1.set_title('Revenue Plot')
ax1.grid()

data['revenue'].plot(kind = 'density', ax = ax2)
ax2.set_title('Revenue Density')
ax2.grid()

plt.show()
```

### Get test and training set

```python
test_start_number = int(np.round(len(data.index) * 0.80))
test_start_date = pd.to_datetime(data.index[test_start_number])
```

### Find best ARIMA model

```python
# Define the p, d and q parameters
p = d = q = range(0, 3)

# Generate all different combinations of p, d and q triplets
pdq = list(itertools.product(p, d, q))

# Generate all different combinations of seasonal p, d and q triplets
seasonal_pdq = [(x[0], x[1], x[2], 12) for x in list(itertools.product(p, d, q))]

# specify to ignore warning messages
warnings.filterwarnings('ignore')

dict_ = {}
i = 0
for param in pdq:
    for param_seasonal in seasonal_pdq:
        try:
            mod = sm.tsa.statespace.SARIMAX(data['revenue'],
                                            order = param,
                                            seasonal_order = param_seasonal,
                                            enforce_stationarity = True,
                                            enforce_invertibility = True)
            results = mod.fit()
            test = results.get_prediction(start = test_start_date, dynamic = False)
            
            y_forecasted = test.predicted_mean
            y_truth = data['revenue'][test_start_number:]
            mse = ((y_forecasted - y_truth) ** 2).mean()
            rmse = np.sqrt(mse)
            
            dict_[i] = [param, param_seasonal, results.aic, mse, rmse]
            i = i + 1
        except:
            continue
            
model_search = pd.DataFrame.from_items(dict_.items(),
                       orient = 'index',
                       columns = ['params','seasonal_params','AIC','MSE','RMSE'])
```

### View models with lowest AIC and lowest RMSE

```python
quantile = 0.35

model_search[(model_search['RMSE'] <= model_search['RMSE'].quantile(quantile)) & (model_search['AIC'] <= model_search['AIC'].quantile(quantile))].sort_values('RMSE')
```

### Create model

```python
#(2,0,2)(2,2,0)
#(2,2,2)(2,1,0)
#(1,0,2)(2,2,0)
```

```python
model = sm.tsa.statespace.SARIMAX(data['revenue'],
                                order = (2, 2, 2),
                                seasonal_order = (2, 1, 0, 12),
                                enforce_stationarity = True,
                                enforce_invertibility = True)

results = model.fit()
print(results.summary())
```

### Plot residuals

```python
results.plot_diagnostics(figsize=(15, 12))
plt.show()
```

### Validate forecast

```python
test = results.get_prediction(start = test_start_date, dynamic = True)
test_ci = test.conf_int()
```

```python
fig = plt.figure(figsize = (20,5))
ax = fig.add_subplot(1,1,1)

plot01, = ax.plot(data['revenue'], label = 'Actual')
plot02, = ax.plot(test.predicted_mean, label = 'Predicted')

ax.fill_between(test_ci.index,
                test_ci.iloc[:, 0],
                test_ci.iloc[:, 1], color='k', alpha=.2)

vals = ax.get_yticks()
ax.set_yticklabels(['${:,.0f}'.format(x) for x in vals])

plt.title('TEST SET\nActual and Predicted Revenue', fontsize = 15)
plt.ylabel('Revenue', fontsize = 15)
plt.xlabel('Year and Month', fontsize = 15)
plt.legend(handles = [plot01, plot02], loc = 'best', fontsize = 12)
plt.grid()
plt.show()
```

### Forecast

```python
pred = results.get_forecast(steps=10)
pred_ci = pred.conf_int()

fig = plt.figure(figsize = (20,5))
ax = fig.add_subplot(1,1,1)

plot01, = ax.plot(data['revenue'], label = 'Actual')
plot02, = ax.plot(pred.predicted_mean, label = 'Forecasted')

ax.fill_between(pred_ci.index,
                pred_ci.iloc[:, 0],
                pred_ci.iloc[:, 1], color='k', alpha=.2)

vals = ax.get_yticks()
ax.set_yticklabels(['${:,.0f}'.format(x) for x in vals])

plt.title('FORECAST\nActual and Forecasted Revenue', fontsize = 15)
plt.ylabel('Revenue', fontsize = 15)
plt.xlabel('Year and Month', fontsize = 15)
plt.legend(handles = [plot01, plot02], loc = 'best', fontsize = 12)
plt.grid()
plt.show()
```

### Final forecast

```python
forecast_df = pd.DataFrame({'forecast':pred.predicted_mean})
forecast_df.index.rename('date_format', inplace = True)
forecast_df = pd.merge(forecast_df, pd.DataFrame({'lower':pred_ci.iloc[:, 0]}), left_index = True, right_index = True, copy = False)
forecast_df = pd.merge(forecast_df, pd.DataFrame({'upper':pred_ci.iloc[:, 1]}), left_index = True, right_index = True, copy = False)
forecast_df = forecast_df[['lower','forecast','upper']]
```

### Test

```python
change_test = pd.read_sql(
'''
SELECT
DATE_FORMAT(o.date_purchased, '%Y-%m') AS date_format,
MONTH(o.date_purchased) AS month,
SUM((op.products_quantity - op.products_quantity_free) * op.products_price) AS revenue
FROM orders_products op
JOIN orders o ON op.orders_id = o.orders_id
WHERE DATE(o.date_purchased) BETWEEN '2013-01-01' AND ' '''+ db_end_date +''' '
AND op.part_id NOT IN '''+ str(tuple(always_exclude)) +'''
AND o.orders_status != 9
AND o.orders_status != 10
AND o.payment_method != 'Replacement Order'
GROUP BY date_format
''', db)
```

```python
cur_month = change_test[change_test['month'] == 9]
cur_forecast = forecast_df['forecast'][forecast_df.index == '2017-09-01'].values[0]
est_change = (cur_forecast - cur_month['revenue'].iloc[-1])/cur_month['revenue'].iloc[-1]

lower = cur_month['revenue'].pct_change().mean() - 1.96 * cur_month['revenue'].pct_change().std()
upper = cur_month['revenue'].pct_change().mean() + 1.96 * cur_month['revenue'].pct_change().std()
```

```python
if np.logical_and(est_change >= lower, est_change <= upper):
    print ('The estimated YoY change is: {:,.02}%.\nAn acceptable YoY change will be between {:,.0f}% and {:,.0f}%.\nTherefore, the forecast is reasonable'.format(est_change * 100, lower * 100, upper * 100))
else:
    print ('The estimated YoY change is: {:,.02}%.\nAn acceptable YoY change will be between {:,.0f}% and {:,.0f}%.\nTherefore, the forecast is not reasonable'.format(est_change * 100, lower * 100, upper * 100))
```
