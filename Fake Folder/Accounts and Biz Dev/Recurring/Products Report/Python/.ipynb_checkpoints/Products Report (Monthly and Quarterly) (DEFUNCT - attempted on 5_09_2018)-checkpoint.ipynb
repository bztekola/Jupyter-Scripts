### get main data

```python
# get main data
op_main = pd.read_sql(
'''
SELECT
DATE(o.date_purchased) AS date_purchased,
op.orders_id,
op.part_id,
op.products_quantity AS 'total qty',
op.products_quantity_free AS 'qty free',
op.products_price,

CASE 
WHEN o.orders_reseller = 0 AND o.orders_super_reseller = 0 THEN 'non reseller'
WHEN o.orders_reseller = 1 AND o.orders_super_reseller = 0 THEN 'reseller'
WHEN o.orders_super_reseller = 1 THEN 'super reseller'
ELSE 'you missed one' END AS 'customer'

FROM orders_products op
JOIN orders o ON op.orders_id = o.orders_id
WHERE DATE(o.date_purchased) BETWEEN ' '''+ date_start +''' ' AND ' '''+ date_end +''' '
AND o.orders_status NOT IN (9,10,11,12,14,15)
AND o.payment_method != 'Replacement Order'
''', db)

op_main.columns = [x.replace('_',' ') for x in op_main.columns]

op_main['date purchased'] = pd.to_datetime(op_main['date purchased'])
op_main['year and quarter'] = op_main['date purchased'].dt.year.map(str) + '-Q' + op_main['date purchased'].dt.quarter.map(str)
op_main['year and month'] = [str(x)[:7] for x in op_main['date purchased']]

partner = pd.read_sql(
'''
SELECT
orders_id AS 'orders id'
FROM orders_products
WHERE part_id IN
(SELECT
part_id
FROM products_description
WHERE products_name LIKE '%partner%')
''', db)

op_main['customer'] = np.where(op_main['orders id'].isin(partner['orders id'].tolist()),
                               'partnership',
                                op_main['customer'])
```

### fix prices and costs

```python
# particle price fix
particle = [3450,3451,3452,3453,3455,3454,3234,3233,2798,2725,2721,2723,2724,3051,3457,2799,3233,3234,2722]
op_main['products price'] = np.where(op_main['part id'].isin(particle), 1.00, op_main['products price'])

# before about may 5, 2018 we priced stripes differently than we do now
# if someone bought a stripe before this date then assume that the cost is such that the profit margin
    # is equal to the mean_stripe_margin
op_main['cost per unit'] = op_main['part id'].map(dict(zip(all_cost['part id'], all_cost['cost per unit'])))
op_main['cost per unit'] = np.where((op_main['part id'].isin(cost03['part id'].tolist())) & (op_main['date purchased'] < '2018-05-04'),
                                   op_main['products price'] * (1 - mean_stripe_margin), # from part_cost.py
                                   op_main['cost per unit']) # from part_cost.py

# fix adabox prices and cost   
op_main['products price'] = np.where(op_main['part id'] == 3067, 60.0, op_main['products price'])
op_main['cost per unit'] = np.where(op_main['part id'] == 3067, 28.59, op_main['cost per unit']) # avg cost per box over all boxes

# fix partnership costs
op_main['cost per unit'] = np.where(op_main['customer'] == 'partnership', 0, op_main['cost per unit'])

# remove these parts
op_main.drop(op_main[op_main['part id'].isin([2843, 1408, 3792])].index, inplace = True)
print('%i part(s) that were not free but which have price equal to zero' % len(op_main[(op_main['qty free'] == 0) & (op_main['products price'] == 0)]))
print('remove these')
op_main.drop(op_main[((op_main['qty free'] == 0) & (op_main['products price'] == 0))].index, inplace = True)

op_main['qty bought'] = op_main['total qty'] - op_main['qty free']
op_main['revenue'] = op_main['qty bought'] * op_main['products price']
op_main['total cost'] = op_main['total qty'] * op_main['cost per unit']
op_main['gross profit'] = op_main['revenue'] - op_main['total cost']

print('%i null(s) before removing' % np.sum(op_main.isnull().any(1)))
op_main.dropna(inplace = True)
print('%i dupes(s) before removing' % np.sum(op_main.duplicated()))
op_main.drop_duplicates(inplace = True)

not_free = op_main[op_main['qty free'] == 0].copy()
print('mean profit margin without freebies: {:,.1f}%'.format(np.mean(not_free['gross profit']/not_free['revenue']) * 100))
```

### part names

```python
part_names = pd.read_sql(
'''
SELECT
part_id AS 'part id',
LOWER(products_name) AS 'part name'
FROM products_description
''', db)

op_main['part name'] = op_main['part id'].map(dict(zip(part_names['part id'], part_names['part name'])))

print('%i null(s)' % np.sum(op_main.isnull().any(1)))
print('%i dupes(s)' % np.sum(op_main.duplicated()))
```

### categories

```python
op_main['category'] = op_main['part id'].map(dict(zip(all_cats_df['part id'], all_cats_df['cat name'])))
op_main['category'] = np.where(op_main['part id'].isin(cost03['part id'].tolist()), 'stripe', op_main['category'])

print('%i null(s)' % np.sum(op_main.isnull().any(1)))
print('%i dupes(s)' % np.sum(op_main.duplicated()))
```

### boms

```python
boms = pd.read_sql(
'''
SELECT
sku_id AS 'sku id',
part_id AS 'part id',
bom_type AS 'bom',
sku_date_modified AS 'sku date modified'
FROM skus
WHERE part_id != 0
''', db)

boms = boms.sort_values(['part id','sku date modified']).drop_duplicates('part id', keep = 'last')

op_main['bom'] = op_main['part id'].map(dict(zip(boms['part id'], boms['bom'])))
op_main['bom'] = np.where(op_main['part id'].isin(cost02['part id'].tolist()), 'combo', op_main['bom']) # from part_cost.py
op_main['bom'] = np.where(op_main['orders id'].isin(partner['orders id'].tolist()), 'partnership', op_main['bom'])
op_main['bom'].fillna('resale - no labor', inplace = True)
print('%i part(s) with "unspecified" bom - replace this/these' % len(op_main['part id'][op_main['bom'] == 'unspecified'].unique()))
op_main['bom'] = op_main['bom'].str.replace('unspecified','resale - no labor')

print('%i null(s)' % np.sum(op_main.isnull().any(1)))
print('%i dupes(s)' % np.sum(op_main.duplicated()))
```

### new products

```python
new_prods = pd.read_sql(
'''
SELECT
part_id AS 'part id',
DATE(products_date_added) AS 'date added'
FROM parts
''', db)

new_prods['date added'] = pd.to_datetime(new_prods['date added'])

new_prods['year and month added'] = [str(x)[:7] for x in new_prods['date added']]
new_prods['year and quarter added'] = new_prods['date added'].dt.year.map(str) + '-Q' + new_prods['date added'].dt.quarter.map(str)

cur_new_prods = new_prods[new_prods['date added'] >= cur_per_start].drop('date added', 1).drop_duplicates()

op_main = pd.merge(op_main,
                      new_prods,
                      on = 'part id',
                      how = 'left',
                      copy = False)

op_main['new product'] = np.where(op_main['date added'].between(cur_per_start, cur_per_end),1,0)

op_main['new product over time'] = np.where((op_main['year and month added'] == op_main['year and month'])
                                              | (op_main['year and quarter added'] == op_main['year and quarter']),1,0)

print('%i null(s)' % np.sum(op_main.isnull().any(1)))
print('%i dupes(s)' % np.sum(op_main.duplicated()))
```

# Compare with PLM [Daily Stats](https://volcano.adafruit.com/volcano/ada_plm.php?zenAdminID=c2d3ee839d38936bafca30cd28172680#/Stats/Sales)

```python
ds = pd.read_sql(
'''
SELECT
DATE_FORMAT(timestamp, '%Y-%m') AS 'year and month',
SUM(d_value) AS 'PLM revenue'
FROM daily_stats
WHERE d_class = 'd_all'
AND DATE(timestamp) BETWEEN ' '''+ date_start +''' ' AND ' '''+ date_end +''' '
GROUP BY DATE_FORMAT(timestamp, '%Y-%m')
''', db)

ds.set_index('year and month', inplace = True)
```

```python
particle_check = pd.read_sql(
'''
SELECT
DATE_FORMAT(o.date_purchased, '%Y-%m') AS 'year and month',
SUM((op.products_quantity - op.products_quantity_free) * op.products_price) AS 'particle revenue'
FROM orders_products op
JOIN orders o ON op.orders_id = o.orders_id
AND DATE(o.date_purchased) BETWEEN ' '''+ date_start +''' ' AND ' '''+ date_end +''' '
AND o.orders_status NOT IN (9,10,11,12,14,15)
AND o.payment_method != 'Replacement Order'
WHERE op.part_id IN '''+ str(tuple(particle)) +'''
GROUP BY DATE_FORMAT(o.date_purchased, '%Y-%m')
''', db)

particle_check.set_index('year and month', inplace = True)

particle_fix_for_check = op_main[op_main['part id'].isin(particle)].groupby('year and month')[['revenue']].sum().rename(columns = {'revenue':'fixed particle revenue'})
```

```python
refunds01 = pd.read_sql(
'''
SELECT
DATE_FORMAT(o.date_purchased, '%Y-%m') AS 'year and month',
ot.orders_id AS 'orders id',
ot.value AS 'ot rev'
FROM orders_total ot
JOIN orders o ON ot.orders_id = o.orders_id
AND DATE(o.date_purchased) BETWEEN ' '''+ date_start +''' ' AND ' '''+ date_end +''' '
AND o.orders_status NOT IN (9,10,11,12,14,15)
AND o.payment_method != 'Replacement Order'
WHERE ot.class = 'ot_subtotal'
''', db)
```

```python
refunds02 = pd.read_sql(
'''
SELECT
o.orders_id AS 'orders id',
SUM((op.products_quantity - op.products_quantity_free) * op.products_price) AS 'op rev'
FROM orders o
JOIN orders_products op ON o.orders_id = op.orders_id
WHERE DATE(o.date_purchased) BETWEEN ' '''+ date_start +''' ' AND ' '''+ date_end +''' '
AND o.orders_status NOT IN (9,10,11,12,14,15)
AND o.payment_method != 'Replacement Order'
GROUP BY op.orders_id
''', db)
```

```python
refunds_main = refunds01.merge(refunds02, on = 'orders id')
refunds_main['refunds'] = refunds_main['ot rev'] - refunds_main['op rev']
refunds = refunds_main.groupby('year and month')[['refunds']].sum()
```

```python
check = op_main.groupby('year and month')[['revenue']].sum().rename(columns = {'revenue':'prod report revenue'}).join(ds).join(particle_check).join(particle_fix_for_check).join(refunds)
check['fixed particle revenue'] = check['fixed particle revenue'] * -1
check = check[['prod report revenue',
               'fixed particle revenue',
               'particle revenue',
               'refunds',
               'PLM revenue']]
check['difference'] = check['PLM revenue'] - (check.sum(1) - check['PLM revenue'])
check.reset_index(inplace = True)

def diff(x):
    if (x['difference'] > -1) and (x['difference'] < 1):
        return 0
    else:
        return x['difference']

check['difference'] = check.apply(diff, axis = 1)
```

```python
op_main2 = op_main[op_main['date purchased'].between(cur_per_start, date_end)]
#op_main2['orders'] = op_main2.groupby('orders id')[['orders id']].cumcount()
#op_main2['orders'] = np.where(op_main2['orders'] > 1, 0, op_main2['orders'])  

total_rev = np.sum(op_main2['revenue'][(op_main2['date purchased'].between(cur_per_start, cur_per_end))])

if close_workbook == 'yes':

    workbook = xlsxwriter.Workbook(workbook_title + '.xlsx',
                                   {'nan_inf_to_errors': True})

    tabs = ['overview',
           'all customers',
           'resellers',
           'super resellers',
           'manufacturing',
           'new products',
           'by category',
           'by bom',
           'notes on revenue totals']

#===== formats
#==========================================================================================

    colors = {
    '1':'#343635',
    '2':'#2e4874',
    '3':'#7eaba4',
    '4':'#928c85',
    '5':'#347c83',
    '6':'#ebe9f2'
    } # http://www.color-hex.com/color-palette/57101

    title = workbook.add_format({'font_size':25,
                                 'font_name':'Arial (Bold)'})

    sub_title = workbook.add_format({'font_size':15,
                                     'font_name':'Arial (Bold)'})


    col_names = workbook.add_format({'font_name':'Arial (Bold)',
                                     'font_color':'white',
                                     'valign':'vcenter',
                                     'align':'center',
                                     'bg_color':colors['3'],
                                     'bottom':1,
                                     'top':1,
                                     'left':1,
                                     'right':1})

    money = workbook.add_format({'num_format':'$#,##0'})
    percent = workbook.add_format({'num_format':'0.00%'})
    number = workbook.add_format({'num_format':'#,##0'})

    for tab in tabs:
        workbook.add_worksheet(tab) # create each tab

    my_worksheets = {}
    for sht in workbook.worksheets():
        my_worksheets[sht.get_name()] = sht # create dict like tab_name:worksheet_instance

#===== same formatting
#==========================================================================================
    for k, v in my_worksheets.items():
        sht = my_worksheets[k]
        sht.write(0, 0,
                  k.title(),
                  title)
        sht.write(1, 0,
                  workbook_title,
                  sub_title)

#===== overview
#==========================================================================================    
    fmts = {'revenue':money, 
            'revenue %':percent,
            'gross profit':money,
            'gross profit %':percent,
            'orders':number, 
            'orders %':percent}

    sht = my_worksheets['overview']

    start_row = 4
    start_col = 0

    for x in ['revenue','gross profit','orders']:

        sht.write(start_row, start_col, x.title(), sub_title)

        df = op_main2.groupby([time,'customer'])[[x]].sum().unstack(1).fillna(0)
        df.columns = df.columns.droplevel(0)
        cols = ['non reseller','reseller','super reseller','partnership']
        df = df[cols]
        df['total'] = df.sum(1)
        df.reset_index(inplace = True)

        for i in range(len(df.columns)):
            sht.write(start_row + 1, start_col + i,
                      df.columns[i],
                      col_names)

            sht.set_column(start_col + i,
                           start_col + i,
                           np.max([len(str(x)) for x in df.iloc[:, i]] + [len(str(x)) for x in df.columns]) + 1)

            for j in range(len(df)):
                sht.write(start_row + 2 + j, start_col + i,
                         df.iloc[j,i],
                         fmts[x])

        start_row = start_row + len(df) + 2

        df2 = df.set_index(time).pct_change()
        df2.loc['YoY'] = df.set_index(time).pct_change(periods = 12).iloc[-1]
        df2.reset_index(inplace = True)
        df2.replace(np.nan, '', inplace = True)
        df2.replace(np.inf, '', inplace = True)
        df2.replace(-np.inf, '', inplace = True)

        for i in range(len(df2.columns)):
            sht.write(start_row + 1, start_col + i,
                      df2.columns[i],
                      col_names)

            for j in range(len(df2)):
                sht.write(start_row + 2 + j, start_col + i,
                         df2.iloc[j,i],
                         fmts[x + ' %'])

        start_row = start_row + len(df2) + 2

        df3 = pd.DataFrame(df.mean(), columns = [report_type + ' avg']).reset_index()

        for i in range(len(df3.columns)):
            sht.write(start_row + 1, start_col + i,
                      df3.columns[i],
                      col_names)

            for j in range(len(df3)):
                sht.write(start_row + 2 + j, start_col + i,
                         df3.iloc[j,i],
                         fmts[x])

        start_col = start_col + len(df.columns) + 2
        start_row = 4

#===== all customers, resellers, super resellers, manufacturing
#==========================================================================================        

    for_charts = {}
    for_charts_ot = {}

    dict_ = {'all customers':op_main2['customer'].isin(['non reseller','reseller','super reseller','partnership']),
            'resellers':op_main2['customer'].isin(['reseller']),
            'super resellers':op_main2['customer'].isin(['super reseller']),
            'manufacturing':op_main2['bom'] == 'pnp'}

    for k, v in dict_.items():

        fmts = {'part id':None,
                'part name':None,
                 cur_pretty:money,
                 prev_pretty:money,
                '% of New Products':percent,
                '% of total':percent,
                '% change':percent,
                '% of ' + k.title():percent,
                'margin':percent,
                'revenue':money,
                'gross profit':money,
                'total qty':number,
                time:None}

        sht = my_worksheets[k]

        df1 = op_main2[(op_main2['date purchased'].between(cur_per_start, cur_per_end))
                        & (dict_[k])].groupby(['part id','part name'], as_index = False)[['revenue','gross profit']].sum()
        df1['% of ' + k.title()] = df1['revenue']/df1['revenue'].sum()
        df1['% of total'] = df1['revenue']/total_rev
        df1['margin'] = df1['gross profit']/df1['revenue']
        df1.rename(columns = {'revenue':cur_pretty}, inplace = True)

        df2 = op_main2[(op_main2['date purchased'] >= prev_per_start)
                       & (op_main2['date purchased'] < prev_per_end)
                       & (dict_[k])].groupby(['part id','part name'], as_index = False)[['revenue']].sum()
        df2.rename(columns = {'revenue':prev_pretty}, inplace = True)

        df3 = pd.merge(df1,
                      df2,
                      how = 'left',
                      on = ['part id','part name']).drop('gross profit',1)
        df3['% change'] = df3[cur_pretty]/df3[prev_pretty] - 1
        cols = ['part id','part name',prev_pretty,cur_pretty,'% change','% of ' + k.title(),'% of total','margin']
        df3 = df3[cols].sort_values(cur_pretty, ascending = False).head(25).fillna(0)    

        for_charts[k] = df3

        start_row = 4
        start_col = 0

        sht.write(start_row, start_col, 'Top 25 Parts by Revenue', sub_title)

        for i in range(len(df3.columns)):
            sht.write(start_row + 1, start_col + i,
                     df3.columns[i],
                     col_names)

            sht.set_column(start_col + i,
                       start_col + i,
                       np.max([len(str(x)) for x in df3.iloc[:, i]] + [len(str(x)) for x in df3.columns]) + 1)

            for j in range(len(df3)):
                sht.write(start_row + 2 + j, start_col + i,
                         df3.iloc[j,i],
                         fmts[df3.columns[i]])

        ot = op_main2[dict_[k]].groupby(time, as_index = False)[['revenue','gross profit','total qty']].sum()
        ot['margin'] = ot['gross profit']/ot['revenue']
        cols = [time,'revenue','gross profit','margin','total qty']
        ot = ot[cols]

        for_charts_ot[k] = ot

        start_col = start_col + len(df3.columns) + 1

        sht.write(start_row, start_col, 'Totals Over Time', sub_title)

        for i in range(len(ot.columns)):
            sht.write(start_row + 1, start_col + i,
                     ot.columns[i],
                     col_names)

            sht.set_column(start_col + i,
                           start_col + i,
                           np.max([len(str(x)) for x in ot.iloc[:, i]] + [len(str(x)) for x in ot.columns]) + 1)

            for j in range(len(ot)):
                sht.write(start_row + 2 + j, start_col + i,
                         ot.iloc[j,i],
                         fmts[ot.columns[i]])

        start_row = start_row + len(ot) + 2

        avg = pd.DataFrame(ot.mean(), columns = [report_type + ' avg']).reset_index().rename(columns = {'index':'type'})

        for i in range(len(avg.columns)):
            sht.write(start_row + 1, start_col + i,
                     avg.columns[i],
                     col_names)

            sht.set_column(start_col + i,
                           start_col + i,
                           np.max([len(str(x)) for x in avg.iloc[:, i]] + [len(str(x)) for x in avg.columns]) + 1)

            for j in range(len(avg)):
                sht.write(start_row + 2 + j, start_col + i,
                         avg.iloc[j,i],
                         fmts[avg.loc[j]['type']])

        start_row = 4
        start_col = 0

#===== new products
#==========================================================================================        

    sht = my_worksheets['new products']    

    new = op_main2[op_main2['new product'] == 1].groupby(['part id','part name'], as_index = False)[['revenue','gross profit']].sum().sort_values('revenue', ascending = False).rename(columns = {'revenue':cur_pretty})
    new['% of New Products'] = new[cur_pretty]/new[cur_pretty].sum()
    new['% of total'] = new[cur_pretty]/total_rev
    new['margin'] = new['gross profit']/new[cur_pretty]
    new.drop('gross profit', 1, inplace = True)
    new['part id'] = [int(x) for x in new['part id']]

    start_row = 4
    start_col = 0

    sht.write(start_row, start_col, 'All New Products by Revenue', sub_title)

    fmts = {'part id':None,
            'part name':None,
             cur_pretty:money,
             prev_pretty:money,
            '% of New Products':percent,
            '% of total':percent,
            '% change':percent,
            'margin':percent,
            'revenue':money,
            'gross profit':money,
            'total qty':number,
            time:None}

    for i in range(len(new.columns)):
        sht.write(start_row + 1, start_col + i,
                 new.columns[i],
                 col_names)

        sht.set_column(start_col + i,
                   start_col + i,
                   np.max([len(str(x)) for x in new.iloc[:, i]] + [len(str(x)) for x in new.columns]) + 1)

        for j in range(len(new)):
            sht.write(start_row + 2 + j, start_col + i,
                     new.iloc[j,i],
                     fmts[new.columns[i]])

    new_ot = op_main2[op_main2['new product over time'] == 1].groupby(time, as_index = False)[['revenue','gross profit','total qty']].sum()
    new_ot['margin'] = new_ot['gross profit']/new_ot['revenue']
    new_ot = new_ot[[time,'revenue','gross profit','margin','total qty']]

    start_col = start_col + len(new.columns) + 1

    sht.write(start_row - 1, start_col, '*parts which were released and sold in the same period')
    sht.write(start_row, start_col, 'Totals Over Time', sub_title)

    for i in range(len(new_ot.columns)):
        sht.write(start_row + 1, start_col + i,
                 new_ot.columns[i],
                 col_names)

        sht.set_column(start_col + i,
                       start_col + i,
                       np.max([len(str(x)) for x in new_ot.iloc[:, i]] + [len(str(x)) for x in new_ot.columns]) + 1)

        for j in range(len(new_ot)):
            sht.write(start_row + 2 + j, start_col + i,
                     new_ot.iloc[j,i],
                     fmts[new_ot.columns[i]])

    start_row = start_row + len(new_ot) + 2

    new_avg = pd.DataFrame(new_ot.mean(), columns = [report_type + ' avg']).reset_index().rename(columns = {'index':'type'})

    for i in range(len(new_avg.columns)):
        sht.write(start_row + 1, start_col + i,
                 new_avg.columns[i],
                 col_names)

        sht.set_column(start_col + i,
                       start_col + i,
                       np.max([len(str(x)) for x in new_avg.iloc[:, i]] + [len(str(x)) for x in new_avg.columns]) + 1)

        for j in range(len(new_avg)):
            sht.write(start_row + 2 + j, start_col + i,
                     new_avg.iloc[j,i],
                     fmts[new_avg.loc[j]['type']])

#===== by category, by bom
#==========================================================================================        

    cat_and_bom_charts = {}

    fmts = {cur_pretty:money,
           prev_pretty:money,
           '% change':percent,
           '% of total':percent,
           'margin':percent,
           'category':None,
           'bom':None}

    for col in ['category','bom']:

        sht = my_worksheets['by ' + col]

        df1 = op_main2[op_main2['date purchased'].between(cur_per_start, cur_per_end)].groupby(col, as_index = False)[['revenue','gross profit']].sum()
        df1['% of total'] = df1['revenue']/total_rev
        df1['margin'] = df1['gross profit']/df1['revenue']
        df1.rename(columns = {'revenue':cur_pretty}, inplace = True)

        df2 = op_main2[(op_main2['date purchased'] >= prev_per_start)
                       & (op_main2['date purchased'] < prev_per_end)].groupby(col, as_index = False)[['revenue']].sum()
        df2.rename(columns = {'revenue':prev_pretty}, inplace = True)

        if len(df1) >= len(df2):
            how = 'left'
        else:
            how = 'right'

        df3 = pd.merge(df1,
                      df2,
                      how = how,
                      on = col).drop('gross profit',1).fillna(0)
        df3['% change'] = df3[cur_pretty]/df3[prev_pretty] - 1
        cols = [col,cur_pretty,prev_pretty,'% change','% of total','margin']
        df3 = df3[cols].sort_values(cur_pretty, ascending = False)

        cat_and_bom_charts['by ' + col] = df3

        start_row = 4
        start_col = 0

        if col == 'category':
            wrt = 'Categories'
        else:
            wrt = 'BOMs'

        sht.write(start_row, start_col, 'All ' + wrt + ' by Revenue', sub_title)

        for i in range(len(df3.columns)):
            sht.write(start_row + 1, start_col + i,
                     df3.columns[i],
                     col_names)

            sht.set_column(start_col + i,
                       start_col + i,
                       np.max([len(str(x)) for x in df3.iloc[:, i]] + [len(str(x)) for x in df3.columns]) + 1)

            for j in range(len(df3)):
                sht.write(start_row + 2 + j, start_col + i,
                         df3.iloc[j,i],
                         fmts[df3.columns[i]])

#        if col == 'bom':
#            unspec = op_main2[['part id','part name']][op_main2['bom'] == 'unspecified'].drop_duplicates('part id').sort_values('part id', ascending = False)        
#            start_col =  start_col + len(df3.columns) + 1
#            sht.write(start_row, start_col,
#                     '*parts with "unpecified" BOMs')

#            for i in range(len(unspec.columns)):
#                sht.write(start_row + 1, start_col + i,
#                         unspec.columns[i],
#                         col_names)

#                sht.set_column(start_col + i,
#                           start_col + i,
#                           np.max([len(str(x)) for x in unspec.iloc[:, i]] + [len(str(x)) for x in unspec.columns]) + 1)

#                for j in range(len(unspec)):
#                    sht.write(start_row + 2 + j, start_col + i,
#                             unspec.iloc[j,i])

#===== notes on revenue totals
#==========================================================================================        

    sht = my_worksheets['notes on revenue totals']

    start_row = 4
    start_col = 0

    for i in range(len(check.columns)):
        sht.write(start_row + 1, start_col + i,
                 check.columns[i],
                 col_names)

        sht.set_column(start_col + i,
                   start_col + i,
                   np.max([len(str(x)) for x in check.iloc[:, i]] + [len(str(x)) for x in check.columns]) + 1)

        for j in range(len(check)):
            sht.write(start_row + 2 + j, start_col + i,
                     check.iloc[j,i],
                     money)

    text01 = 'The revenue in the Products Report is calculated a little differently than in PLM Daily Stats.\n'
    text02 = 'This worksheet shows you how to get from the Products Report revenue all the way to the Daily Stats revenue.\n'
    text03 = '\n'
    text04 = 'First, start with the Products Report revenue (col B).\n'
    text05 = 'Then subtract out the Particle Revenue that we fix in the Report (ask jarad@adafruit.com why this is so) (col C).\n'
    text06 = 'Then add back in the Particle revenue that\'s in the database (col D).\n'
    text07 = 'Account for manual refunds that are not reflected in Daily Stats (col E).\n'
    text08 = 'Check out the "difference" column (col G), which is "PLM revenue" minus the sum of all the columns before "PLM revenue".\n'
    text09 = '\nThere! You have now gone from the Products Report revenue all the way to Daily Stats!'
    text10 = '\n\nHowever, if the "difference" column is much greater than zero, this means that am order has been processed and Daily Stats has not yet updated, so to speak'
    text11 = 'Don\'t worry! This is totally normal and will all balance out in the next month!'
    text = ''.join([text01,text02,text03,text04,text05,text06,text07,text08,text09,text10,text11])

    options = {'width': 750,
               'height': 300}

    sht.insert_textbox(start_row + len(check) + 3,
                      start_col,
                      text,
                      options)

#===== close
#==========================================================================================        

    workbook.close()
```
