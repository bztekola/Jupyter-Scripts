### Get stacky data

```python
stacky_main = pd.read_sql(
'''
SELECT
stacky_part_id AS part_id,
stacky_link_part_id AS stacky
FROM stacky
''', db)

col_fix(stacky_main)
```

```python
s1 = set(stacky_main['part id'])
s2 = set(stacky_main['stacky'])
s3 = s1.intersection(s2)

print('{:,.0f} parts are main parts, {:,.0f} parts are stackies, and {:,.0f} are both.'.format(len(s1) - len(s3),
                                                                                               len(s2) - len(s3),
                                                                                               len(s3)))
if s3.issubset(s1) and s3.issubset(s2):
    print('\nhowever, every part that is "both" is a subset of "main part" and "stacky part"')
    
    # to conteract any duplicate tuples created in the section below, remove the intersection
    for c in ['part id','stacky']:
        stacky_main.drop(stacky_main[stacky_main[c].isin(s3)].index, inplace = True)
        stacky_main.reset_index(drop = True, inplace = True)
    
else:
    raise ValueError('there are instances where a main_part is a stacky and that same stacky is not a main_part')
```

### Create a list of tuples like (part id, stacky)

```python
stacky_tuples = list(zip(stacky_main['part id'], stacky_main['stacky']))

l1 = len(set(stacky_tuples))
l2 = len(stacky_tuples)

if l1/l2 == 1:
    print('no dupes')
else:
    raise ValueError('stack tuples have dupes')
```

```python
s = len(set(stacky_main['part id']))
t = total_parts['count'].item()

print('{:,.1f}% of parts, or {:,.0f} out of {:,.0f} parts have stackies.'.format(s/t * 100, s, t))
```

### Get orders data

```python
op_main = pd.read_sql(
'''
SELECT
op.orders_id,
op.part_id
FROM orders_products op
JOIN orders o ON op.orders_id = o.orders_id
AND o.orders_status IN (3,13) # shipped, no shipment necessary
AND DATE(o.date_purchased) BETWEEN ' '''+ date_start +''' ' AND ' '''+ date_end +''' '
''', db)

col_fix(op_main)
```

### Get how many stackies each order chose

```python
import itertools
s = dt.datetime.now()

d1 = {}
d2 = {}
for oid in set(op_main['orders id']):
        
    df = op_main[op_main['orders id'] == oid]
    ls = df['part id'].tolist()
    combo = set(list(itertools.combinations(ls, 2)))
    
    intersect = list(combo.intersection(stacky_tuples))
    total_stackies = len(intersect)
    total_parts = len(df)
    percent_stackies = total_stackies/total_parts
    
    d1[oid] = [total_parts,
              total_stackies,
              percent_stackies]
    
    for i in intersect:
        if i in d2:
            val = d2[i] + 1
            d2[i] = val
        else:
            d2[i] = 1
    
e = dt.datetime.now()
print(e-s)
```

### Structure orders results
If A comes with B and C and D as a stacky, and D comes with B and C, and someone orders A, B, C, and D, my loop will count this as (A,B), (A,C), (A,D), (D,B), (D,C) = 5. But the customer only bought 4 parts. Right now, there is no way to know the exact part/stacky combination, and in this case the stacky_percent will be > 100%. So below I normalize the stack_percent to be between zero and one, and I then "recount" the number of stackies as a function of the stacky_percent and count_of_parts.

```python
# create a dataframe from dict
oid_results = pd.DataFrame.from_dict(d1, orient = 'index', columns = ['count of parts',
                                                                      'count of stackies',
                                                                      'percent stackies'])
# structure it
oid_results.reset_index(inplace = True)
oid_results.rename(columns = {'index':'orders id'}, inplace = True)

# get the count of changes I'll make
n = np.sum(oid_results['percent stackies'] > 1)
print('{:,.0f} percentages will be altered.'.format(n))

# use this function to fix the percentages, as explained above
def fix(df):
    if df['percent stackies'] > 1:
        df['count of stackies'] = df['count of parts'] - 1
        df['percent stackies'] = df['count of stackies']/df['count of parts']
    return df

oid_results = oid_results.apply(fix, axis = 1)
```

### How many orders contain X stackies?

```python
a = pd.DataFrame(oid_results['count of stackies'].value_counts())
a.reset_index(inplace = True)
a.rename(columns = {'index':'stackies per order','count of stackies':'count of orders'}, inplace = True)
a['% of total order count'] = a['count of orders']/a['count of orders'].sum()
a.sort_values(['count of orders','stackies per order'], ascending = [False, True], inplace = True)

fmt = ['n0','n0','p1']
a.format_(fmt)
```

### Structure part and stacky results

```python
# turn dict in dataframe and structure it
stacky_results = pd.DataFrame.from_dict(d2, orient = 'index', columns = ['part/stacky count'])
stacky_results.sort_values('part/stacky count', ascending = False, inplace = True)
stacky_results.reset_index(inplace = True)
stacky_results.rename(columns = {'index':'part/stacky'}, inplace = True)

# splir part/stacky tuple into two separate columns
stacky_results[['part', 'stacky']] = pd.DataFrame(stacky_results['part/stacky'].tolist(),
                                                             index = stacky_results.index)
stacky_results.drop('part/stacky',1,inplace = True)

# reorder columns
stacky_results = stacky_results[['part','stacky','part/stacky count']]

# get the total count of orders per part using the part id of the part/stacky combo
total_count = pd.DataFrame(op_main.groupby('part id').size(), columns = ['order count']).reset_index()
stacky_results['count of orders containing part'] = stacky_results['part'].map(dict(zip(total_count['part id'], total_count['order count'])))

# get proportion of stackies per count of orders
stacky_results['percent of total'] = stacky_results['part/stacky count']/stacky_results['count of orders containing part']

# get all part names
all_parts = pd.read_sql(
'''
SELECT
part_id,
LOWER(products_name) AS products_name
FROM products_description
''', db)

col_fix(all_parts)

# map some part names
for col in ['part','stacky']:
    stacky_results['temp'] = stacky_results[col].map(dict(zip(all_parts['part id'], all_parts['products name'])))
    stacky_results[col] = 'PN ' + stacky_results[col].map(str) + ': ' + stacky_results['temp']
    stacky_results.drop('temp',1,inplace = True)
```

```python
top_p = 0.05

ls1 = ['part/stacky count','percent of total']
ls2 = ['count','percent']

for i in range(2):
    n = stacky_results[ls1[i]].quantile(1-top_p)
    stacky_results['top {:,.0f}% by {}'.format(top_p*100, ls2[i])] = np.where(stacky_results[ls1[i]] >= n, 'X','')
```

```python
fmt = [0,0,'n0','n0','p1',0]
stacky_results.format_(fmt).head()
```

### Send to Excel

```python
if write_csv == 'yes':
       
    # create workbook object
    workbook_title = 'Stackies as of %s' % date_end
    workbook  = xlsxwriter.Workbook(csv_path + '/' + workbook_title + '.xlsx')
    
    # setup formats
    colors = ['#343635',
              '#2e4874',
              '#7eaba4',
              '#928c85',
              '#347c83',
              '#bfb9d6']

    title = workbook.add_format({'font_size':25,
                                 'font_name':'Arial (Bold)'})

    subtitle = workbook.add_format({'font_size':15,
                                    'font_name':'Arial (Bold)'})


    col_names = workbook.add_format({'font_name':'Arial (Bold)',
                                     'font_color':'white',
                                     'valign':'vcenter',
                                     'align':'center',
                                     'bg_color':colors[2],
                                     'bottom':1,
                                     'top':1,
                                     'left':1,
                                     'right':1})

    money = workbook.add_format({'num_format':'$#,##0','align':'center','valign':'center'})
    money2 = workbook.add_format({'num_format':'$#,##0.00','align':'center','valign':'center'})
    percent = workbook.add_format({'num_format':'0.0%','align':'center','valign':'center'})
    number = workbook.add_format({'num_format':'#,##0','align':'center','valign':'center'})
    dummy_left = workbook.add_format({'font_color':'black'})
    dummy_center = workbook.add_format({'font_color':'black','align':'center','valign':'center'})
    
    # create worksheet object
    worksheet = workbook.add_worksheet('Stacky Data')
    
    # instantiate worksheet object
    sht = worksheet
        
    # write workbook title
    sht.write(0,0,workbook_title,title)
        
    # write date range of data
    m = int(np.floor((pd.to_datetime(date_end) - pd.to_datetime(date_start)).days/30))
    sht.write(1,0,'Data is from %s to %s (about %i months).' % (date_start, date_end, m))
    
    # write proportion of orders with stackies
    n1 = np.sum(oid_results['count of stackies'] > 0)
    n2 = len(oid_results)
    p = n1/n2
    sht.write(2,0,'{:,.1f}% of orders, or {:,.0f} out of {:,.0f} orders chose stackies.'.format(p*100,n1,n2))
    
    # write stacky data   
    df = stacky_results.copy()
    cols = df.columns.tolist()
    
    fmts = {'part':dummy_left,
            'stacky':dummy_left,
            'part/stacky count':number,
            'count of orders containing part':number,
            'percent of total':percent,
            'top {:,.0f}% by count'.format(top_p*100):dummy_center,
            'top {:,.0f}% by percent'.format(top_p*100):dummy_center}
    
    start_row = 4
    start_col = 0
    
    # write column headers
    for i in range(len(cols)):
        sht.write(start_row,
                  start_col + i,
                  cols[i].title(),
                  col_names)
        
        # change column width
        len1 = [len(str(x)) for x in df.iloc[:, i]]
        len2 = [len(df.columns[i])]
        len3 = np.max(len1 + len2)
        
        sht.set_column(start_col + i,
                       start_col + i,
                       len3 + 5)
        
        # write row data
        for j in range(len(df)):
            sht.write(start_row + 1 + j,
                      start_col + i,
                      df.iloc[j,i],
                      fmts[cols[i]])
            
    # add autofilter
    worksheet.autofilter(start_row,
                         start_col,
                         start_row + len(df),
                         start_col + len(df.columns) - 1)

    workbook.close()    
```
