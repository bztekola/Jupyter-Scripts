### Db data

```python
os = pd.read_sql(
'''
SELECT
o.orders_id,
os.orders_status_name
FROM orders o
JOIN orders_status os ON o.orders_status = os.orders_status_id
WHERE o.payment_module_code = 'paypalwpp'
''', db)

col_fix(os)
```

```python
o_main = pd.read_sql(
'''
SELECT
DATE_FORMAT(date_purchased, '%Y-%m') AS 'year and month purchased',
DATE(date_purchased) AS 'date purchased',
orders_id AS 'orders id',
order_total AS 'db order total',
0 AS 'deleted'
FROM orders
WHERE payment_module_code = 'paypalwpp'
''', db)

o_main['date purchased'] = pd.to_datetime(o_main['date purchased'])

#=====================================================================================

od_main = pd.read_sql(
'''
SELECT
DATE_FORMAT(date_purchased, '%Y-%m') AS 'year and month purchased',
DATE(date_purchased) AS 'date purchased',
orders_id AS 'orders id',
order_total AS 'db order total',
1 AS 'deleted'
FROM orders_deleted od
WHERE od.payment_module_code = 'paypalwpp'
''', db)

od_main['date purchased'] = pd.to_datetime(od_main['date purchased'])

#=====================================================================================

orders_main = pd.concat([o_main, od_main])
orders_main.reset_index(drop = True, inplace = True)
orders_main['orders status'] = orders_main['orders id'].map(dict(zip(os['orders id'],os['orders status name'])))

print('%i oid dupe(s)' % np.sum(orders_main['orders id'].duplicated()))
print('drop the dupes\n')

dupes = orders_main[orders_main['orders id'].duplicated()]
orders_main_dupes = orders_main[orders_main['orders id'].isin(dupes['orders id'].tolist())].sort_values('orders id')

orders_main.drop_duplicates('orders id', inplace = True)
```

```python
pp_db_super_main = pd.read_sql(
'''
SELECT
DATE_FORMAT(payment_date, '%Y-%m') AS 'year and month payment',
DATE(payment_date) AS 'payment date',
TIME(payment_date) AS 'time',
parent_txn_id AS 'parent transaction id',
txn_id AS 'transaction id',
order_id AS 'orders id',
LOWER(payment_status) AS 'pp status',
mc_currency AS 'currency',
mc_gross AS 'pp order total',
mc_fee AS 'pp fee'
FROM paypal
''', db)

for col in ['parent transaction id','transaction id']:
    pp_db_super_main[col] = [x.lower() for x in pp_db_super_main[col]]
    
pp_db_main = pp_db_super_main.copy()
    
pp_db_main['payment date'] = pd.to_datetime(pp_db_main['payment date'], errors = 'coerce')
pp_db_main['pp net'] = pp_db_main['pp order total'] - pp_db_main['pp fee']

pp_db_main.drop_duplicates(inplace = True)  

#=========================================================================================================

status_fix = {'canceled_reversal':'cancelled reversal',
             'pending (echeck)':'pending',
             'pending (none)':'pending',
             'pending (other)':'pending',
             'pending (regulatoryreview)':'pending'}
 
pp_db_main['jb status'] = pp_db_main['pp status'].map(status_fix)
pp_db_main['jb status'].fillna(pp_db_main['pp status'], inplace = True)
pp_db_main['jb status'] = np.where(pp_db_main['jb status'] == '', 'no status', pp_db_main['jb status'])

print('%i line(s) where trans id is missing' % len(pp_db_main[pp_db_main['transaction id'] == '']))
print('remove them')
no_trans = pp_db_main[pp_db_main['transaction id'] == '']
pp_db_main.drop(pp_db_main[pp_db_main['transaction id'] == ''].index, inplace = True)

#====================================================================================================

# sometimes a single transaction contains multiple steps, like if someone buys something, then wants a reversal, then
    # changes their mind and reverses the reversal
# each step has a trans id
# those steps have one parent trans id
# if a transaction has only one step then we make the parent trans id equal to the trans id
pp_db_main['parent transaction id'] = np.where(pp_db_main['parent transaction id'] == '',
                                               pp_db_main['transaction id'],
                                               pp_db_main['parent transaction id'])

#=========================================================================================================

print(pp_db_main.groupby('jb status')[['orders id']].count().sort_values('orders id', ascending = False))

pp_db_main['date purchased'] = pp_db_main['orders id'].map(dict(zip(orders_main['orders id'], orders_main['date purchased'])))

print('\nmin and max dates where "date purchased" is null')
print(pp_db_main['payment date'][pp_db_main['date purchased'].isnull()].min())
print(pp_db_main['payment date'][pp_db_main['date purchased'].isnull()].max())

pp_db_main['year and month purchased'] = [str(x)[:7] for x in pp_db_main['date purchased']]
```

```python
path = r'/Users/jarad/fake_folder/Finance/Recurring/PayPal Monthly Audit/Docs/Monthly CSV Statements/'
all_files = glob.glob(path + '/*.CSV')
frame = pd.DataFrame()
list_ = []

for file_ in all_files:
    df = pd.read_csv(file_, error_bad_lines = False)
    list_.append(df)
    
csv_super_main = pd.concat(list_)  

csv_super_main.drop_duplicates(inplace = True)
csv_super_main.reset_index(inplace = True, drop = True)
csv_super_main.columns = [x.lower() for x in csv_super_main.columns]
csv_super_main.columns = [x.strip() for x in csv_super_main.columns]
csv_super_main.rename(columns = {'gross':'csv gross',
                                 'fee':'csv fee',
                                 'net':'csv net',
                                 'date':'payment date'}, inplace = True)

# csv is in pacific time
csv_super_main['payment date'] = pd.to_datetime(csv_super_main['payment date'])
csv_super_main.rename(columns = {'payment date':'payment date pst'}, inplace = True)
csv_super_main['year and month payment pst'] = [str(x)[:7] for x in csv_super_main['payment date pst']]
csv_super_main['date and time payment pst'] = pd.to_datetime((csv_super_main['payment date pst'].map(str)).str[:10] + ' ' + csv_super_main['time'].map(str))

# date and time in CSV download is pacific time, so 3 hours behind us, add 3 hours here to make is EST
csv_super_main['date and time payment est'] = csv_super_main['date and time payment pst'] + pd.DateOffset(hours = 3)
csv_super_main['year and month payment est'] = [str(x)[:7] for x in csv_super_main['date and time payment est']]
csv_super_main['payment date est'] = csv_super_main['date and time payment est'].dt.date

csv_super_main['reference txn id'].fillna(csv_super_main['transaction id'], inplace = True)

to_number = ['csv gross',
             'csv fee',
             'csv net',
             'shipping and handling amount',
             'insurance amount',
             'sales tax']

for col in to_number:
    try:
        csv_super_main[col] = csv_super_main[col].str.replace(',','')
        csv_super_main[col] = pd.to_numeric(csv_super_main[col])
    except:
        pass
csv_super_main['csv fee'] = csv_super_main['csv gross'] - csv_super_main['csv net']        

# sometimes a single transaction contains multiple steps, like if someone buys something, then wants a reversal, then
    # changes their mind and reverses the reversal
# each step has a trans id
# those steps have one parent trans id
# if a transaction has only one step then we make the parent trans id equal to the trans id
csv_super_main.rename(columns = {'reference txn id':'parent transaction id'}, inplace = True)
csv_super_main['parent transaction id'].fillna(csv_super_main['transaction id'], inplace = True)    

ls = ['parent transaction id',
      'transaction id',
      'from email address',
      'to email address',
      'status',
      'type',
      'balance impact']
for col in ls:
    csv_super_main[col] = csv_super_main[col].str.lower()        

csv_super_main['from email address'].fillna('someone_else@jb_email.com', inplace = True)
csv_super_main['to email address'].fillna('someone_else@jb_email.com', inplace = True)
csv_super_main['transaction is from'] = np.where(csv_super_main['from email address'] == 'paypal@adafruit.com', 'adafruit','someone else')    
```

### Fix lines that involve currency conversions
* hint: this was a total nightmare, lol

```python
# get parent ids for "general curreny conversion"
parent_id = csv_super_main['parent transaction id'][(csv_super_main['type'] == 'general currency conversion')
                                                  | (csv_super_main['currency'] != 'USD')].tolist()
cur_conv_super_main = csv_super_main[csv_super_main['parent transaction id'].isin(parent_id)].sort_values(['parent transaction id','date and time payment est']).copy()
cur_conv_main = cur_conv_super_main.copy()

print('%i lines are this "type"' % len(cur_conv_main))
print('that\'s {:,.2f}% of {:,.0f} total lines'.format(len(cur_conv_main)/len(csv_super_main) * 100, len(csv_super_main)))

print('\nhere are the currencies counts:\n')
print(cur_conv_main.groupby('currency').agg({'csv gross':['count','sum']}).sort_values([('csv gross','sum')]))

#===== first fix

# if a "parent transaction id" has only two lines and both types are "general currency conversions", 
# then that type should be "general payment" where "currency" = "USD"
# so first get all double "parent transaction id"

first_fix = cur_conv_main.groupby('parent transaction id').filter(lambda x: x['parent transaction id'].count() == 2).sort_values('parent transaction id')

# double check that every type is "general current conversion"
if len(set(first_fix['type']) - set(['general currency conversion'])) == 0: # meaning an empty set
    print('\nfor "first fix", you got them all')
else:
    print('\nfor "first_fix", you missed something')
    
# drop all but "USD" and replace type with "general payment"    
first_fix = first_fix[first_fix['currency'] == 'USD'].copy()
first_fix['type'] = first_fix['type'].str.replace('general currency conversion','general payment')

# drop these from "cur_conv_main"
cur_conv_main = cur_conv_main[~cur_conv_main['parent transaction id'].isin(first_fix['parent transaction id'])].copy()

#===== second fix

second_fix = cur_conv_main.copy()

# fix this real quick
# this isn't perfect, we're changing not-USD to USD, but these sub payments do not have a corresponding currency
# conversion, at any rate, there aren't that many and the amount is not that much
second_fix['currency'] = np.where((second_fix['type'] == 'subscription payment') & (second_fix['currency'] != 'USD'), 'USD', second_fix['currency'])

# fill these "currency conversion" types with nulls
second_fix['type'] = np.where(second_fix['type'] == 'general currency conversion', np.nan, second_fix['type'])

# sort by this stuff to push the non-null types up to the top
second_fix.sort_values(['parent transaction id','date and time payment est','type'], inplace = True)

# fill forward with true type
second_fix['type'].fillna(method = 'ffill', inplace = True)

# drop all but USD
second_fix = second_fix[second_fix['currency'] == 'USD'].copy()

#===== concat the two fixes together

final_fix = pd.concat([first_fix, second_fix])

#===== check

# final_fix sum should equal the original 
if final_fix['csv gross'].sum() == csv_super_main[(csv_super_main['parent transaction id'].isin(parent_id))
                                                  & ((csv_super_main['currency'] == 'USD') | (csv_super_main['type'] == 'subscription payment'))]['csv gross'].sum():
    print('\nmatch')
else:
    print('\nmismatch')
    
# make a copy of super_main so you can check it
# from the copy, remove all pre-fixed parent ids and add back in the fixed versions
csv_main = csv_super_main.copy()
csv_main = csv_main[~csv_main['parent transaction id'].isin(parent_id)].copy()
csv_main = pd.concat([csv_main, final_fix])

# check totals of fixed df against original df
if csv_main['csv gross'].sum() == csv_super_main[(csv_super_main['currency'] == 'USD') 
                                               | (csv_super_main['type'] == 'subscription payment')]['csv gross'].sum():
    print('match')
else:
    print('mismatch')
```
