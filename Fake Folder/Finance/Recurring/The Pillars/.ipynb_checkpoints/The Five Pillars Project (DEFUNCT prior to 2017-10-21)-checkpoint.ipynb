# Sales - get data

```python
sales = pd.read_sql(
'''
SELECT
DATE_FORMAT(o.date_purchased, '%Y-%m') AS 'Year and Month',
SUM((op.products_quantity - op.products_quantity_free) * op.products_price) AS Revenue
FROM orders o
JOIN orders_products op ON o.orders_id = op.orders_id
WHERE o.orders_status != 9
AND o.orders_status != 10
AND o.payment_method != 'Replacement Order'
AND DATE(o.date_purchased) BETWEEN ' '''+ date_start +''' ' AND ' '''+ date_end +''' '
GROUP BY DATE_FORMAT(o.date_purchased, '%Y-%m') 
''', db)
```

```python
sales_format = sales.copy()
sales_format['Revenue'] = ['${:,.0f}'.format(x) for x in sales_format['Revenue']]
sales_format
```

### Sales - plot

```python
ax = sales.set_index('Year and Month').plot(kind = 'bar', figsize = (20,5), rot = 45, color = 'pink')

vals = ax.get_yticks()
ax.set_yticklabels(['${:,.0f}'.format(x) for x in vals])

ax.grid()
ax.legend('')
plt.ylabel('Revenue ($)', fontsize = 15)
plt.xlabel('Year and Month', fontsize = 15)
plt.title('Revenue by Month', fontsize = 15)
plt.show()
```

### Sales - YTD

```python
ytd = pd.read_sql(
'''
SELECT
SUM(CASE WHEN YEAR(o.date_purchased) = 2015 THEN (op.products_quantity - op.products_quantity_free) * op.products_price END) AS '2015',
SUM(CASE WHEN YEAR(o.date_purchased) = 2016 THEN (op.products_quantity - op.products_quantity_free) * op.products_price END) AS '2016',
SUM(CASE WHEN YEAR(o.date_purchased) = 2017 THEN (op.products_quantity - op.products_quantity_free) * op.products_price END) AS '2017'
FROM orders o
JOIN orders_products op ON o.orders_id = op.orders_id
WHERE o.orders_status != 9
AND o.orders_status != 10
AND o.payment_method != 'Replacement Order'
AND YEAR(o.date_purchased) IN (2015,2016,2017)
AND MONTH(o.date_purchased) IN (1,2,3,4,5,6,7)
''', db)
```

```python
ytd = ytd.T
ytd.reset_index(inplace = True)
ytd.rename(columns = {0:'YTD Revenue Comparison','index':'Year'}, inplace = True)
```

# Rent

```python
rent_raw = excel.parse('rent')

rent = rent_raw.copy()
rent.reset_index(inplace = True)
rent.dropna(how = 'all', inplace = True)
rent.dropna(axis = 1, how = 'all', inplace = True)
rent.drop(['level_0','level_1','Num','Split','Balance'], 1, inplace = True)
rent.dropna(how = 'any', inplace = True)

rent['Year and Month'] = pd.to_datetime(rent['Date'].dt.year.map(str) + '-' + rent['Date'].dt.month.map(str))
rent['Year and Month'] = [str(x)[:7] for x in rent['Year and Month']]
rent.rename(columns = {'Amount':'Rent'}, inplace = True)
```

### Rent by floor and by month

```python
def floor(x):
    if '3rd' in x['Memo']:
        return '3rd floor'
    elif '10th' in x['Memo']:
        return '10th floor'
    else:
        return '4th floor'

rent['Floor'] = rent.apply(floor, axis = 1)
```

```python
rent
```

```python
ax = rent.groupby(['Year and Month','Floor']).agg({'Rent':'sum'}).unstack(level = 1).plot(kind = 'bar', 
                                                                                          figsize = (20,5),
                                                                                         rot = 45)
vals = ax.get_yticks()
ax.set_yticklabels(['${:,.0f}'.format(x) for x in vals])

ax.grid()
plt.ylabel('Rent ($)', fontsize = 15)
plt.xlabel('Year and Month', fontsize = 15)
plt.title('Rent by Floor and by Month', fontsize = 15)
plt.legend(['10th Floor','3rd Floor','4th Floor'], loc = 'best')
plt.show()
```

### Rent by month

```python
rent_by_month = rent.groupby('Year and Month', as_index = False).agg({'Rent':'sum'})
rent_by_month = pd.merge(rent_by_month, sales, on = 'Year and Month')
rent_by_month['% of Revenue'] = rent_by_month['Rent']/rent_by_month['Revenue']
rent_by_month.drop('Revenue', 1, inplace = True)

# format for a human
rent_by_month_format = rent_by_month.copy()
rent_by_month_format['Rent'] = ['${:,.0f}'.format(x) for x in rent_by_month_format['Rent']]
rent_by_month_format['% of Revenue'] = ['{:,.2f}%'.format(x * 100) for x in rent_by_month_format['% of Revenue']]
rent_by_month_format
```

### Plot Rent by month

```python
ax = rent_by_month['Rent'].plot(kind = 'bar', figsize = (20,5), use_index = False)

ax.set_xticks(np.arange(0, len(rent_by_month.index)))
ax.set_xticklabels(rent_by_month['Year and Month'], rotation = 45)

vals = ax.get_yticks()
ax.set_yticklabels(['${:,.0f}'.format(x) for x in vals])

ax.grid()
plt.ylabel('Rent ($)', fontsize = 15)
plt.xlabel('Year and Month', fontsize = 15)
plt.title('Rent by Month', fontsize = 15)
plt.show()
```

### Why the big jump in May 2017?
* Payment for Level 3 Communications was \$2500 in March but \$42,800 in May

# COGS

```python
cogs_raw = excel.parse('COGs')

cogs = cogs_raw.copy()
cogs.reset_index(inplace = True)
cogs.drop(['level_0','level_1','Type','Num','Split','Balance'], 1, inplace = True)
cogs.dropna(how = 'all', inplace = True)
cogs.dropna(axis = 1, how = 'all', inplace = True)
cogs = cogs[cogs['Date'].isnull() == False]
cogs = cogs[cogs['Name'].isnull() == False]


cogs['Year and Month'] = pd.to_datetime(cogs['Date'].dt.year.map(str) + '-' + cogs['Date'].dt.month.map(str))
cogs['Year and Month'] = [str(x)[:7] for x in cogs['Year and Month']]
cogs.rename(columns = {'Amount':'COGS'}, inplace = True)
```

### COGS refund

```python
cogs_refund = cogs.copy()
cogs_refund = cogs_refund[cogs_refund['COGS'] < 0]
cogs_refund = cogs_refund[['Year and Month','Name','Memo','COGS']]
cogs_refund.rename(columns = {'COGS':'COGS Refund'}, inplace = True)

print ('total COGS refund amount: ${:,.0f}'.format(cogs_refund['COGS Refund'].sum() * -1))

cogs_refund
```

### COGS by month

```python
cogs_by_month = cogs.copy()
cogs_by_month = cogs_by_month[cogs_by_month['COGS'] >= 0]
cogs_by_month = cogs_by_month.groupby('Year and Month', as_index = False).agg({'COGS':'sum'})
cogs_by_month = pd.merge(cogs_by_month, sales, on = 'Year and Month')
cogs_by_month['% of Revenue'] = cogs_by_month['COGS']/cogs_by_month['Revenue']
cogs_by_month.drop('Revenue', 1, inplace = True)

# format for a human
cogs_by_month_format = cogs_by_month.copy()
cogs_by_month_format['COGS'] = ['${:,.0f}'.format(x) for x in cogs_by_month_format['COGS']]
cogs_by_month_format['% of Revenue'] = ['{:,.2f}%'.format(x * 100) for x in cogs_by_month_format['% of Revenue']]
cogs_by_month_format
```

```python
cogs_by_month['% of Revenue'].mean()
```

### Plot COGS by month

```python
ax = cogs_by_month.plot(kind = 'bar', figsize = (20,5), use_index = False, color = 'grey')

ax.set_xticks(np.arange(0, len(cogs_by_month.index)))
ax.set_xticklabels(cogs_by_month['Year and Month'], rotation = 45)

vals = ax.get_yticks()
ax.set_yticklabels(['${:,.0f}'.format(x) for x in vals])

ax.legend('')
ax.grid()
plt.ylabel('Cost of Goods Sold ($)', fontsize = 15)
plt.xlabel('Year and Month', fontsize = 15)
plt.title('Cost of Goods (COGS) by Month', fontsize = 15)
plt.show()
```

### COGS by vendor

```python
cogs_by_vendor = cogs.copy()
cogs_by_vendor = cogs_by_vendor[cogs_by_vendor['COGS'] >= 0]

ytd_cogs = cogs_by_vendor['COGS'].sum()

cogs_by_vendor = cogs_by_vendor.groupby('Name', as_index = False).agg({'COGS':'sum'})
cogs_by_vendor.sort_values('COGS', ascending = False, inplace = True)

# merge refunds
cogs_by_vendor = pd.merge(cogs_by_vendor, 
                          cogs_refund.groupby('Name', as_index = False).agg({'COGS Refund':'sum'}), 
                          how = 'left',
                          on = 'Name')

cogs_by_vendor.fillna(0, inplace = True)
cogs_by_vendor.rename(columns = {'COGS':'Vendor COGS YTD'}, inplace = True)

cogs_by_vendor['% of total COGS YTD'] = cogs_by_vendor['Vendor COGS YTD']/ytd_cogs
```

```python
ax = cogs_by_vendor[:25]['Vendor COGS YTD'].plot(kind = 'bar', figsize = (20,5), rot = 270)

vals = ax.get_yticks()
ax.set_yticklabels(['${:,.0f}'.format(x) for x in vals])

ax.set_xticklabels(cogs_by_vendor['Name'][:25], rotation = 270, fontsize = 15)

ax.grid()
plt.ylabel('Cost of Goods Sold ($)', fontsize = 15)
plt.xlabel('Year and Month', fontsize = 15)
plt.title('Top 25 Vendors of 2017 by COGS', fontsize = 15)
plt.show()
```

```python
cogs_by_vendor_format = cogs_by_vendor.copy()
cogs_by_vendor_format['Vendor COGS YTD'] = ['${:,.0f}'.format(x) for x in cogs_by_vendor_format['Vendor COGS YTD']]
cogs_by_vendor_format['COGS Refund'] = ['${:,.0f}'.format(x) for x in cogs_by_vendor_format['COGS Refund']]
cogs_by_vendor_format['% of total COGS YTD'] = ['{:,.2f}%'.format(x * 100) for x in cogs_by_vendor_format['% of total COGS YTD']]
cogs_by_vendor_format.head(25)
```

```python
spent = wages_by_month.merge(rent_by_month, on = 'Year and Month').merge(cont_by_month, on = 'Year and Month').merge(cogs_by_month, on = 'Year and Month').merge(ad_spend_by_month, on = 'Year and Month')
spent.drop(['% of Revenue','% of Revenue_x','% of Revenue_y'], 1, inplace = True)
spent['Total Spent'] = spent.sum(numeric_only = True, axis = 1)
spent = spent.merge(sales, on = 'Year and Month')

total = pd.DataFrame(spent.sum(numeric_only = True)).T
total['Year and Month'] = 'Totals'

spent = pd.concat([spent, total])

spent = spent[['Year and Month','Revenue','Rent','Ad Spend','Wages','Contractor Wages','COGS','Total Spent']]
spent['Total as a % of Revenue'] = spent['Total Spent']/spent['Revenue']
spent.columns = [x.lower() for x in spent.columns]
```
