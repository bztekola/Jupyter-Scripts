### Find and remove rows which contain "total"s
The totals from QuickBooks are the totals over the ENTIRE period for which we pulled data, the entire period often being several years. These QB totals are irrelevant to us because we want the freedom to calculate totals over whatever period we choose (one year, one quarter, 6 months, etc.)

```python
# all columns before "date" are susceptible
n = qb_super_main.columns.get_loc('date')
cols = qb_super_main.columns.tolist()[:n]

# get the index of all "totals"
totals_ix = []
for c in cols:
    df = qb_main[[c]]
    try:
        for i in df[df[c].str.contains('^total', na = False)].index:
            totals_ix.append(i)
    except:
        pass
    
qb_main.drop(totals_ix, inplace = True)
qb_main.reset_index(drop = True, inplace = True)
```

### Structure

```python
# drop these additional rows
ix = []
for col in qb_main.columns:
    try:
        df = qb_main[qb_main[col].str.contains('gross profit', na = False)]
        if df.empty == False:
            for i in df.index.tolist():
                ix.append(i)
    except:
        pass
    
qb_main.drop(ix, inplace = True) 
qb_main.reset_index(drop = True, inplace = True)

# drop these columns; they're not important
for i in [0,1,2]:
    qb_main.drop('unnamed: ' + str(i), 1, inplace = True)

# the format of the CSV output of QB data is such that two columns of nulls
    # separate the transaction labels and the transaction data.
    # below, get this first column of the pair of nulls, so we know where to start labeling for our hierarchy
col = qb_main.loc[:, qb_main.isnull().all()].columns[0]
n = qb_main.columns.get_loc(col)

# fill NANs row wise
for i in np.arange(n-1,0,-1):
    qb_main[qb_main.columns[i]].fillna(qb_main[qb_main.columns[i-1]], inplace = True)
    
# fill NANs column wise
for i in np.arange(0,n):
    qb_main[qb_main.columns[i]].fillna(method = 'ffill', inplace = True)

# drop any rows/cols of all NANs
for i in range(2):
    qb_main.dropna(how = 'all', inplace = True, axis = i)
    qb_main.reset_index(drop = True, inplace = True)
    
# rename columns
cols = ['income/expense',
        'main category',
        'subcategory 01']

for i in range(len(cols)):
    qb_main.rename(columns = {qb_main.columns[i]:cols[i]}, inplace = True)
    qb_main[cols[i]].fillna('none', inplace = True)
    
# correct this entry
qb_main['income/expense'] = qb_main['income/expense'].str.replace('66810 · employer health contrib record','expense')    

# split off but save the accounting codes
for col in ['main category','subcategory 01','split']:
    sp = qb_main[col].str.split('·')
    qb_main[col + ' number'] = qb_main[col]
    qb_main[col] = sp.str[-1]
    qb_main[col] = qb_main[col].str.strip()

# drop where "amount" is zero
# these zero amounts mark an entry where there is no transaction but there is a balance
# for this report we don't care about balances
qb_main.drop(qb_main[qb_main['amount'] == 0].index, inplace = True)
qb_main.reset_index(drop = True, inplace = True)

# drop where date is null
qb_main.drop(qb_main[qb_main['date'].isnull()].index, inplace = True)
qb_main.reset_index(drop = True, inplace = True)
```

### Check totals

```python
# get totals from unstructured data
ls = []
for label in ['total expense','total income']:
    for col in qb_super_main.columns:
        try:
            df = qb_super_main[qb_super_main[col].str.contains(label, na = False)]
            if df.empty == False:
                amt = df['amount'].item()
                ls.append(amt)
        except:
            pass
        
total_expense, total_income = ls

# get totals from structured data
structured_expense, structured_income = qb_main.groupby('income/expense')[['amount']].sum().values.flatten()

# compare expenses
if np.abs(structured_expense - total_expense) > 0.001:
    raise ValueError('expense totals don\'t match')

# compare incomes
if np.abs(structured_income - total_income) > 0.001:
    raise ValueError('income totals don\'t match')
```

# Current period summaries

```python
d = {'year and month':[13,'m2m'],
     'year and quarter':[5,'q2q']}

for k,v in d.items():
    dates = sorted(list(set(qb_main[k])))
    cur_d = dates[-1]
    prev_d = dates[-2]
    last_year_d = dates[-v[0]]
    
    df = qb_main[qb_main[k].isin([cur_d,prev_d,last_year_d])].groupby(['super category',k])[['amount']].sum().unstack(1).fillna(0)
    df.columns = df.columns.droplevel(0)   
    
    # concat profit
    a = qb_main.groupby([k,'income/expense'])[['amount']].sum().unstack(1)
    a.columns = a.columns.droplevel(0)
    a['profit'] = a['income'] - a['expense']

    b = a[a.index.isin([cur_d, prev_d, last_year_d])][['profit']].T
    
    df = pd.concat([df,b])

    # get m2m and yoy
    df[v[1]] = df[cur_d]/df[prev_d] - 1
    df['yoy'] = df[cur_d]/df[last_year_d] - 1
    
    # get % of current expense
    vals = df[cur_d].loc[(~df.index.isin(['profit','income']))].values
    cur_exp = np.sum([v for v in vals if v > 0])
    df['% of current ' + k] = np.where(df[cur_d] > 0, df[cur_d]/cur_exp, np.nan)
    
    for col in ['income','profit']:
        df.loc[col,'% of current ' + k] = np.nan
    
#    if np.abs(1 - df['% of current ' + k].sum()) > 1:
#        raise 0        
    
    check = df.loc[:,df.columns.str.contains('%')].sum().values[0]
    if check < 0.999999 or check > 1.00001:
        raise ValueError('check your % of totals')    
    
    df2 = df.loc[:,cur_d:].copy()    
    df2.sort_values(cur_d, ascending = False, inplace = True)
        
    # pretty dates
    d = str(qb_main['date'].max().date())
    d2 = jb_dates(pd.Series(d), k + ' pretty')[0]
    d3 = df2.columns[0]
    df2.rename(columns = {d3:d2}, inplace = True)
        
    report_dict['current ' + k] = df2
```

### Super overall summary

```python
d = {'year and month':[13,'m2m'],
    'year and quarter':[5,'q2q']}

for k,v in d.items():

    dates = sorted(list(set(qb_main[k])))
    cur_d = dates[-1]
    prev_d = dates[-2]
    last_year_d = dates[-v[0]]
    
    overall = qb_main.groupby([k,'income/expense'])[['amount']].sum().unstack(1)
    overall.columns = overall.columns.droplevel(0)
    overall['profit'] = overall['income'] - overall['expense']
    overall2 = overall[overall.index.isin([cur_d,prev_d,last_year_d])].T

    overall2[v[1]] = np.nan
    overall2['yoy'] = np.nan
    
    for index, row in overall2.iterrows():

        p2p = row[cur_d]/row[prev_d] - 1 if np.logical_and(row[cur_d] > 0, row[prev_d] > 0) else np.nan
        yoy = row[cur_d]/row[last_year_d] - 1 if np.logical_and(row[cur_d] > 0, row[last_year_d] > 0) else np.nan    

        overall2.loc[index, v[1]] = p2p
        overall2.loc[index, 'yoy'] = yoy

    overall2.drop([prev_d, last_year_d], 1, inplace = True)
    
    # pretty dates
    d = str(qb_main['date'].max().date())
    d2 = jb_dates(pd.Series(d), k + ' pretty')[0]
    d3 = overall2.columns[0]
    overall2.rename(columns = {d3:d2}, inplace = True)    

    report_dict['super summary by ' + k] = overall2
```

### Partition data into mutually exclusive sets
* For any given order:
    * Guest with no account (where customer_type = 'guest')
    * Account that's new (where date_created = date_purchased)
    * Account that's existing (where date_created < date_purchased)
    * Account holder who at the time of purchase was a guest (where date_created > date_purchased)

```python
def partition(x):
    
    cust = x['customer type']
    purch = x['date purchased']
    create = x['date created']
    
    if cust == 'guest': 
        ret = 'guest'
    else:
        if create < purch:
            ret = 'existing account'
        elif create == purch:
            ret = 'new account'
        elif create > purch:
            ret = 'guest'
        else:
            ret = 'you missed one'
            
    return ret

acc_main['partition'] = acc_main.apply(partition, axis = 1)
```

```python
d1 = acc_main['date purchased'].min().date()
d2 = acc_main['date purchased'].max().date()

print('from {} to {}'.format(d1,d2))

vc = acc_main.groupby('partition').agg({'subtotal':['count','sum']})
vc.columns = vc.columns.droplevel(0)

print('\ntotals')
display(vc.format_(['n0','m0']))

print('\nproportions')
p = vc.div(vc.sum(), axis = 1)
display(p.format_(['p1','p1']))
```

```python
news_main = pd.read_csv(csv_path + 'campaigns.csv')
news_main.columns = [x.lower() for x in news_main.columns]

print('all lists, before dropping any\n')
print(news_main['list'].value_counts())

news_main.drop(news_main[news_main['list'] == 'Testing Video Thumbs'].index, inplace = True)
news_main.reset_index(drop = True, inplace = True)

news_main['send date'] = pd.to_datetime(news_main['send date'])
news_main['send date'] = [x.date() for x in news_main['send date']]
news_main['send date'] = pd.to_datetime(news_main['send date'])

e = str(qb_main['date'].max() + pd.DateOffset(months = 1))[:7] + '-01'

news_main.drop(news_main[news_main['send date'] >= e].index, inplace = True)
news_main.reset_index(drop = True, inplace = True)

news_main['year and month'] = news_main['send date'].map(str).str[:7]
news_main['year and quarter'] = news_main['send date'].map(str).str[:4] + '-Q' + news_main['send date'].dt.quarter.map(str)
```

```python
d = {'year and month':[13,'m2m',12],
    'year and quarter':[9,'q2q',4]}

for k,v in d.items():
    main = news_main.groupby([k,'list']).agg({'title':'nunique',
                                              'total recipients':'sum',
                                              'successful deliveries':'sum',
                                              'unique opens':'sum',
                                              'unique clicks':'sum',
                                              'unsubscribes':'sum'})

    n = v[0]
    for l in main.index.get_level_values(1):
        df = main[main.index.get_level_values(1) == l].iloc[-n:].copy()
        df['delivery fail rate'] = 1 - df['successful deliveries']/df['total recipients']
        df['open rate'] = df['unique opens']/df['successful deliveries']
        df['click rate'] = df['unique clicks']/df['successful deliveries']
        df['unsubscribe rate'] = df['unsubscribes']/df['total recipients']

        report_dict['newsletter - ' + l.lower() + ' by ' + k] = df
```

# Current totals vs Regression Confidence Interval

```python
from scipy.stats import linregress
# https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.linregress.html
```

```python
for_reg = qb_main.groupby([report_type,'super category'])[['amount']].sum().unstack(1).fillna(0)
for_reg.columns = for_reg.columns.droplevel(0)

for_reg['total expense'] = for_reg.sum(1) - for_reg['income']
for_reg['profit'] = for_reg['income'] - for_reg['total expense']
```

```python
#=========================
# create plot layout
#=========================

m = 3
n = int(np.ceil(len(for_reg.columns)/m))

fig, axes = plt.subplots(nrows = m, 
                         ncols = n,
                         figsize = (20,10))
for i in range(m):
    for j in range(n):
        axes[i,j].set_yticklabels('')
        axes[i,j].set_xticklabels('')

#=========================
# start loop for regression line
#=========================        
        
main_ls = []
x = range(len(for_reg))
i = 0
j = 0
for col in for_reg.columns:
    
    #=========================
    # get slope, intercept, etc.
    #=========================
    
    y = for_reg[col].values
        
    slope, intercept, r_value, p_value, std_err = linregress(x,y)
    
    t = stats.t.ppf(0.975, len(x) - 2) # t-score with n-2 degrees of freedom where alpha = 0.05/2

    lower = intercept + (slope - t * std_err) * x
    upper = intercept + (slope + t * std_err) * x
    
    cur = for_reg.iloc[-1][col]
    
    #=========================
    # store values
    #=========================    
    
    main_ls.append([col,
                    lower[-1],
                    upper[-1],
                    cur])
    
    #=========================
    # plot them
    #=========================
    
    axes[i,j].plot(x, y, 'o')
    axes[i,j].plot(x[-1], y[-1], 'o', color = 'yellow', mew = 10) # plot current value with a big yellow dot
    axes[i,j].plot(x, intercept + slope * x, 'r', lw = 3)
    axes[i,j].fill_between(x, lower, upper, color = 'white')
    axes[i,j].set_title(col.title(), fontsize = 20)
    
    if i == m-1:
        i = 0
    else:
        i += 1
        
    if j < m:
        j += 1
    else:
        j = 0
        
d1 = sorted(list(set(qb_main[report_type])))[-1]
d2 = (jb_dates(pd.Series(d1), report_type + ' pretty')).item()        

plt.suptitle('Regression Line and Confidence Interval for All Categories\nwhere {} is in yellow'.format(d2),
            fontsize = 30, y = 1.05)
axes[-1,-1].axis('off')
plt.savefig(chart_path + 'Regression Line for All Categories', bbox_inches = 'tight')
    
#=========================
# create dataframe
#=========================    
    
cols = ['category','estimated lower bound','estimated upper bound','current value']
results = pd.DataFrame(main_ls, columns = cols)    

def comments(x):
    val = x['current value']
    if val >= x['estimated lower bound'] and val <= x['estimated upper bound']:
        ret = 'within'
    else:
        if val < x['estimated lower bound']:
            ret = 'less than'
        else:
            ret = 'GREATER THAN'
    return ret
    
results['comment on interval'] = results.apply(comments, axis = 1)

#=========================
# arrange "expenses" in descending order by amount
#=========================

order = report_dict['current ' + report_type].index.tolist()
df = pd.DataFrame(order)
df.columns = ['category']

results = pd.merge(df, results, on = 'category')
results['category'] = results['category'].str.title()

fmt = [0,'m0','m0','m0',0]
display(results.format_(fmt))

report_dict['linear regression results'] = results.format_(fmt).to_html(index = False)
```

### Deep dive into Results
Look at the Super Categories which exceed their intervals

Return the names which exceed their own confidence intervals

```python
ls1 = results[(results['comment on interval'] == 'GREATER THAN')
             & (~results['category'].isin(['Income','Profit']))]['category'].tolist()

ls2 = results[(results['comment on interval'] == 'less than')
             & (results['category'].isin(['Income','Profit']))]['category'].tolist()

ls = ls1 + ls2
ls = [x.lower() for x in ls]
try:
    ls.remove('profit')
except:
    pass
for_results = []

for c in ls:
    
    # for eac "super category", groupby period and "name"
    df = qb_main[(qb_main['super category'] == c)].groupby([report_type,'name'])[['amount']].sum().unstack().fillna(0)
    df.columns = df.columns.droplevel(0)
    
    # exclude "names" whose current value is zero
    df2 = df.loc[:,df.loc[df.index[-1]] > 0]    
    
    # for each "name", see if the current value exceeds its upper bound
    for col in df2.columns.tolist():
        
        # get a list of values
        y = df2[col].values.tolist()
        
        # find index of first non-zero entry
        ix = np.nonzero(y)[0][0]
        y = y[ix:]
        x = range(len(y))
        
        # regression line
        slope, intercept, r_value, p_value, std_err = linregress(x,y)

        t = stats.t.ppf(0.975, len(x) - 2) # t-score with n-2 degrees of freedom where alpha = 0.05/2

        lower = (intercept + (slope - t * std_err) * x)[-1]
        upper = (intercept + (slope + t * std_err) * x)[-1]
        cur = y[-1]
        
        if cur > upper:
            for_results.append([c,col,lower,upper,cur])
                       
cols = ['super category',
       'name',
       'lower',
       'upper',
       'current value']
results2 = pd.DataFrame(for_results, columns = cols)
results2.sort_values(['super category','current value'], ascending = [False, False], inplace = True)

for col in ['super category','name']:
    results2[col] = results2[col].str.title()

results2.rename(columns = {'super category':'category'}, inplace = True)
    
fmt = [0,0,'m0','m0','m0']
results2.format_(fmt)
```
