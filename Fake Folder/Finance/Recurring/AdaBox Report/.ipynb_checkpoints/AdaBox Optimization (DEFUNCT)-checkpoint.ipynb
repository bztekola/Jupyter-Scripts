### Get data

```python
subs_main = pd.read_sql(
'''
SELECT
shipment_slug AS 'adabox',
op.part_id AS 'sub part id',
op.orders_id AS 'orders id'
FROM subscriptions_history sh
JOIN subscriptions s ON sh.subscriptions_id = s.subscriptions_id
AND s.subscriptions_type = 'adabox'
JOIN orders_products op ON sh.orders_id = op.orders_id
WHERE sh.action = 'Create Shipment Order'
''', db)
```

```python
ada_main = subs_main.groupby(['adabox','sub part id'], as_index = False)[['orders id']].count()
ada_main.columns = ['adabox','sub part id','subs qty']
```

```python
def name_fix(x):
    
    name = x['adabox'].lower()
    name = name.replace('_',' ')
    
    num = name[-1]
    
    name = name[:7]
    name = name + '00' + num
    
    return name

ada_main['adabox'] = ada_main.apply(name_fix, axis = 1)
```

```python
not_sub_part_id = {3193:3193,
                   3235:3235,
                   3268:3268,
                   3370:3370,
                   3476:3644,
                   3620:3697,
                   3696:np.nan}

ada_main['not subs part id'] = ada_main['sub part id'].map(not_sub_part_id)
```

```python
not_subs_qty = pd.read_sql(
'''
SELECT
part_id AS 'part id',
SUM(products_quantity) AS 'not subs qty',
AVG(products_price) AS 'not subs price'
FROM orders_products
WHERE orders_id NOT IN '''+ str(tuple(subs_main['orders id'])) +'''
AND part_id IN '''+ str(tuple(ada_main.iloc[:-1]['not subs part id'])) +'''
GROUP BY part_id
''', db)
```

```python
ada_main['not subs qty'] = ada_main['not subs part id'].map(dict(zip(not_subs_qty['part id'], not_subs_qty['not subs qty'])))
ada_main['not subs price'] = ada_main['not subs part id'].map(dict(zip(not_subs_qty['part id'], not_subs_qty['not subs price'])))

ada_main['subs price'] = 60.0

ada_main
```

```python
ada_main.set_index('adabox')[['subs qty','not subs qty']].plot(kind = 'bar', rot = 45, title = 'qty sold')
```

### Prepare data

```python
x1 = ada_main['not subs qty'].sum()
y1 = ada_main['not subs price'].mean()

x2 = ada_main['subs qty'].sum()
y2 = ada_main['subs price'].mean()

m = (y2 - y1)/(x2 - x1)

# if y = mx + b, then b = y - mx
b = y1 - m * x1

print('not subs:', (x1,y1))
print('subs:', (x2,y2))
print('slope:', m)
print('y-intercept:', b)
```

### Price elasticity
* see [here](http://www.investinganswers.com/financial-dictionary/economics/price-elasticity-demand-ped-2929)

```python
# elasticity = (% change in qty)/(% change in price)

print('% change in qty: {:,.1f}%'.format((x1/x2-1) * 100))
print('% change in price: {:,.1f}%'.format((y1/y2-1) * 100))

e = (x1/x2-1)/(y1/y2-1)
print('price elasticity is {:,.1f} percentage points'.format(e))
print('so for each percentage point that the adabox price increases, the qty sold changes by {:,.1f}%'.format(e))
```

```python
# for every 1% increase in price, the qty will fall by 3.4%
# 60 to 73.42 is an increase of 22.4%
# so qty should fall by 3.4 * 22.4 = 76.16%
# qty was 17566, now it's 4021, that's a decrease of 77.1%

# the values are not exact because we rounded along the way, but they are correct
```

### Price function

```python
def price(x):
    # price(x) = mx + b
    p = np.array(m*x + b)
    return p

x = np.arange(0,35000)
y = price(x)

fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title('price function')
ax.set_xlabel('qty')
ax.set_ylabel('price')
ax.annotate('X',
           xy = (x2, y2))
ax.annotate('subs qty: {:,.0f}\nsubs avg price: ${:,.2f}'.format(x2,y2),
           xy = (x2 + 1000, y2))

ax.annotate('X',
           xy = (x1, y1))
ax.annotate('standalone qty: {:,.0f}\nstandalone avg price: ${:,.2f}'.format(x1,y1),
           xy = (x1 + 1000, y1))
plt.show()
```

### Revenue function

```python
def revenue(x):
    # if revenue(x) = price(x) * x
    # then revenue(x) = (mx + b)x -> mx^2 + bx
    r = m*x**2 + b*x
    return r

x = np.arange(-500000,500000)
y = revenue(x)

fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title('revenue')
plt.show()
```

### Profit function

```python
c = 39.07

def profit(x):
    # if profit(x) = revenue(x) - cost(x)
    # and cost(x) = 39.07x (shipping = 10.27 + parts = 28.59 + outsource = 0.21 == 39.07)
    # then profit(x) = revenue(x) - cost(x) -> mx^2 + (b - c)x
    profit = m*x**2 + (b - c)*x
    return profit

x = np.arange(-100000,100000)
y = profit(x)

fig, ax = plt.subplots()
ax.plot(x,y)
ax.set_title('profit')
plt.show()
```

### Applied to right now

```python
current_sub_count = pd.read_sql(
'''
SELECT
COUNT(*) AS 'subs'
FROM subscriptions
WHERE subscriptions_status = 'Active'
''', db)

p = np.poly1d([m, b - c, 0]) # of the form f(x) = mx**2 + (b-c)x + 0
der = tuple(np.polyder(p))
# solve for zero
qty_for_max_profit = der[1]/(der[0] * -1)
already_sold = (np.sum(ada_main['subs qty']))
remaining = qty_for_max_profit - already_sold

print('maximized profit is ${:,.0f} with {:,.0f} boxes sold via subs'.format(profit(qty_for_max_profit), qty_for_max_profit))
print('current profit is ${:,.0f} after selling {:,.0f} boxes through subs only, this is up to and including adabox007'.format(profit(already_sold), already_sold))
print('this leaves {:,.0f} more boxes to sell via subs'.format(remaining))
print('subs for adabox008 as of 5/3/2018 is {:,.0f}'.format(current_sub_count['subs'].values[0]))
print('which means when we ship adabox008 we\'ve reached diminishing returns')
```

### Maximize revenue
* i.e. how many units before revenue decreases

```python
# if r(x) = m*x**2 + b*x
# then r'(x) = 2mx + b
# and x = b/-2m

def max_rev(x):
    return 2*m*x + b

max_rev_qty = b/(-2*m)
print('max revenue is ${:,.0f} and occurs when qty equals {:,.0f}'.format(revenue(max_rev_qty), max_rev_qty))
print('profit at {:,.0f} boxes is ${:,.0f} which is a profit margin of {:,.1f}%'.format(max_rev_qty,
                                                                                       profit(max_rev_qty),
                                                                                       profit(max_rev_qty)/max_rev_qty * 100))
```

### Profit loss

```python
# profit is:
# m*x**2 + (b - c)*x
# so find the roots of this polynomial

coeff = [m, (b-c), 0]
loss_qty = np.roots(coeff)[0]
print('{:,.0f} will result in zero profit'.format(loss_qty))
```

### Estimate future subs

```python
deg = 1
estimate_per = 15

y = ada_main[['subs qty','not subs qty']].sum(1).tolist()[1:] # exclude adabox001, for simplicity
x = range(len(y))

# forecast
z = np.polyfit(x,y,deg) # approximate curve to get next few values, i.e. next couple of sub counts
p = np.poly1d(z)

fig, ax = plt.subplots()
ax.plot(x, y, lw = 4, alpha = 0.80, label = 'actual qty')
ax.set_title('units sold for each adabox (subs and standalones)\nwith estimates')
ax.set_xticks(range(15))
ax.set_xticklabels([str(i) for i in np.arange(2, estimate_per + 2)])
ax.set_xlabel('adaboxes')
ax.set_ylabel('qty')
ax.set_ylim(np.min(y) - 100, np.max(y) + 1000)

x_hat = range(estimate_per)
forecast = p(x_hat)
ax.plot(x_hat, forecast, color = 'red', ls = '--', alpha = .80, label = 'estimated count')

ax.legend()
plt.show()

print('mean error %, per box, of actual vs. model is {:,.1f}%'.format(np.mean(np.abs(p(x) - y)/y * 100)))
```

```python
estimates = ada_main.set_index('adabox')[['subs qty','not subs qty']].copy()
estimates.fillna(0, inplace = True)
estimates.drop(estimates[estimates.index == 'adabox 001'].index, inplace = True)
estimates['total qty'] = estimates.sum(1)

future_boxes_length = len(forecast) - len(estimates)
cur_box = 8

index_ls = []
for i in np.arange(cur_box, future_boxes_length + cur_box):
    if i <= 9:
        index_ls.append('adabox 00' + str(i))
    else:
        index_ls.append('adabox 0' + str(i))
        
for i in index_ls:
    estimates.loc[i] = np.nan
    
# add back in adabox001
# we excluded this from our model but we need to put it back in to get the running sum of units sold
for_est = ada_main[ada_main['adabox'] == 'adabox 001'].set_index('adabox')[['subs qty','not subs qty']].copy()
for_est['total qty'] = for_est[['subs qty','not subs qty']].sum(1)

estimates = pd.concat([for_est, estimates], ignore_index = False)

estimates['estimate'] = np.round(for_est['total qty'].tolist() + forecast.tolist())
estimates['% error'] = np.abs((estimates['estimate'] - estimates['total qty']))/estimates['total qty']
estimates['running sum of estimates'] = estimates['estimate'].cumsum()
```

### Intervals for subs per month

```python
#var = forecast
var = ada_main['subs qty']

x_bar = np.mean(var)
s = np.std(var)
n = len(var)
t = stats.t.ppf(0.95, n - 1)
z = 1.96

lower_c = x_bar - t*s/np.sqrt(n)
upper_c = x_bar + t*s/np.sqrt(n)

lower_p = x_bar - t*s
upper_p = x_bar + t*s

print('quarterly subscription count\n')
print('95% confidence interval')
for i in [lower_c, x_bar, upper_c]:
    print('{:,.0f}'.format(np.round(int(i))))

print('\n')    
    
print('95% prediction interval')
for i in [lower_p, x_bar, upper_p]:
    print('{:,.0f}'.format(np.round(int(i))))
```
