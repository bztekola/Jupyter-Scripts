# Get amazon data from the db

```python
orders = pd.read_sql(
'''
SELECT
DATE(date_purchased) AS date_purchased,
DATE_FORMAT(date_purchased, '%Y-%m') AS date_format,
orders_id,
order_total AS o_order_total
FROM orders
WHERE DATE(date_purchased) BETWEEN ' '''+ date_start +''' ' AND ' '''+ date_end +''' '
AND payment_module_code = 'lapwa'
''', db)
```

```python
print ('total count of amazon orders in orders table: {:,.0f}'.format(len(orders.index)))
```

```python
lapwa_log = pd.read_sql(
'''
SELECT
lap.action AS transaction_type,
lap.refund_id,
lap.order_id AS orders_id,
lap.order_reference_id AS amazon_order_reference_id,
lap.amount AS a_order_total,
ot.value AS ot_order_total
FROM lapwa_log lap
JOIN orders_total ot ON lap.order_id = ot.orders_id AND class = 'ot_total'
WHERE order_id IN '''+ str(tuple(orders['orders_id'])) +'''
''', db)
```

# Merge

```python
amazon_db = pd.merge(orders, lapwa_log, how = 'left', on = 'orders_id')

amazon_db['o_order_total'] = [np.round(x, 2) for x in amazon_db['o_order_total']]
amazon_db['ot_order_total'] = [np.round(x, 2) for x in amazon_db['ot_order_total']]
amazon_db['a_order_total'] = [np.round(x, 2) for x in amazon_db['a_order_total']]

print ('total count of orders after merging: {:,.0f}'.format(len(amazon_db['orders_id'].unique())))

if len(amazon_db['orders_id'].unique()) == len(orders.index):
    print ('your orders query has the same count as your amazon merge')
else:
    print ('YOU HAVE MADE A MISTAKE')
```

# Retrieve CSV statements

```python
path = r'/Users/jarad/fake_folder/Finance/Recurring/Amazon Audit/Statements/'
all_files = glob.glob(path + '/*.csv')
frame = pd.DataFrame()
list_ = []

for file_ in all_files:
    df = pd.read_csv(file_, index_col = None, header = 0)
    list_.append(df)
    
csv = pd.concat(list_)    
```

### Clean them

```python
# clean column headers
columns_list = []
for y in csv.columns:
    columns = '_'.join(x.lower() for x in re.findall('[A-Z][^A-Z]*', y))
    columns_list.append(columns)
    
csv.columns = columns_list    

# convert to datetime and get rid of time units
csv['transaction_posted_date'] = pd.to_datetime(csv['transaction_posted_date'])

# exclude data before current month
csv = csv[csv['transaction_posted_date'] <= pd.to_datetime(date_end)]

csv['date_format'] = pd.to_datetime(csv['transaction_posted_date'].dt.year.map(str) + '-' + csv['transaction_posted_date'].dt.month.map(str))
csv['date_format'] = [str(x)[:7] for x in csv['date_format']]
csv['transaction_posted_date'] = pd.to_datetime(csv['transaction_posted_date']).dt.date

# convert to numeric
csv['transaction_amount'] = csv['transaction_amount'].str.replace(',','')
csv['transaction_amount'] = pd.to_numeric(csv['transaction_amount'])

csv['transaction_percentage_fee'] = pd.to_numeric(csv['transaction_percentage_fee'])
csv['transaction_fixed_fee'] = pd.to_numeric(csv['transaction_fixed_fee'])
csv['total_transaction_fee'] = pd.to_numeric(csv['total_transaction_fee'])

csv['net_transaction_amount'] = csv['net_transaction_amount'].str.replace(',','')
csv['net_transaction_amount'] = pd.to_numeric(csv['net_transaction_amount'])

# keep only the columns you want
csv = csv[['transaction_posted_date',
     'date_format',
     'settlement_id',
     'amazon_transaction_id',
     'transaction_type',
     'amazon_order_reference_id',
     'transaction_description',
     'transaction_amount',
     'transaction_percentage_fee',
     'transaction_fixed_fee',
     'total_transaction_fee',
     'net_transaction_amount']]

# change the wording of this label
csv['transaction_type'] = np.where(csv['transaction_type'] == 'Authorization', 'Authorize', csv['transaction_type'])
```

# Audit - Auth and Cap

```python
# adafruit authorize
ada_total_auth = amazon_db[amazon_db['transaction_type'] == 'Authorize'].groupby('date_format', as_index = False).agg({'o_order_total':'sum'}).rename(columns = {'o_order_total':'Ada Total Auth'})
ada_actual_auth = amazon_db[amazon_db['transaction_type'] == 'Authorize'].drop_duplicates('amazon_order_reference_id').groupby('date_format', as_index = False).agg({'o_order_total':'sum'}).rename(columns = {'o_order_total':'Ada Actual Auth'})

# merge
audit = pd.merge(ada_total_auth, ada_actual_auth, on = 'date_format')

# creat column
audit['Ada Declined Auth'] = audit['Ada Total Auth'] - audit['Ada Actual Auth']

# adafruit capture
ada_actual_cap = amazon_db[amazon_db['transaction_type'] == 'Capture'].groupby('date_format', as_index = False).agg({'o_order_total':'sum'}).rename(columns = {'o_order_total':'Ada Actual Cap'})

# merge again
audit = pd.merge(audit, ada_actual_cap, on = 'date_format', copy = False)

# create column
audit['Ada Auth and Cap Diff'] = audit['Ada Actual Auth'] - audit['Ada Actual Cap']

# amazon capture
amazon_actual_cap = pd.merge(amazon_db[['amazon_order_reference_id','date_format']][amazon_db['transaction_type'] == 'Capture'],
                                     csv[['amazon_order_reference_id','transaction_amount']][csv['transaction_type'] == 'Capture'],
                                     on = 'amazon_order_reference_id').groupby('date_format', as_index = False).agg({'transaction_amount':'sum'}).rename(columns = {'transaction_amount':'Amazon Actual Cap'})

# merge again
audit = pd.merge(audit, amazon_actual_cap, on = 'date_format', copy = False)

# create columns
audit['Ada and Amazon Captured Diff'] = audit['Ada Actual Cap'] - audit['Amazon Actual Cap']

audit['-'] = '-'
audit['='] = '='
audit['>'] = '>'

# sort columns
audit = audit[['date_format','Ada Total Auth','-','Ada Declined Auth','=','Ada Actual Auth','-','Ada Actual Cap','=','Ada Auth and Cap Diff',
               '>','Amazon Actual Cap','-','Ada Actual Cap','=','Ada and Amazon Captured Diff']].rename(columns = {'date_format':'Year and Month'})
audit
```
