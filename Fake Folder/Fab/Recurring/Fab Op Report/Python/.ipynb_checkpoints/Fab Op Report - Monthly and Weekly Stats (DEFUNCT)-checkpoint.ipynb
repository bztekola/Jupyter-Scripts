# Fab Op Metric

* This script looks at different parts of the Fab experience, from work order count to machine use, from yield to reworks, etc. The goal here is to develop a single metric that will encapsulate the Fab experience. We will use the [Overall Equipment Effectivness Metric](https://en.wikipedia.org/wiki/Overall_equipment_effectiveness) as our framework for creating this metric.

* script revised around july 2, 2018
* uploaded to jupyter hub on july 13, 2018

```python
import sys
sys.path.insert(0,'/Users/jarad/fake_folder/Python Libraries')

from jb_libraries import *
%matplotlib inline

# do this for monthly, it doesn't matter what's here of it's weekly
date_start = '2017-07-01'
date_end = '2018-07-31'

# do this for weekly, it doesn't matter what's here if it's monthly
week_end = '2018-10-04'

report_type = {'week':[53],
               'year and month':[13]}

display_type = 'week'
#display_type = 'year and month'

print('display type is "%s"' % display_type)

six_months_ago = str((pd.to_datetime(week_end) - pd.DateOffset(months = 6)).date())

tableau_write = 'yes'
```

### Get data
* we use three main tables:
    * sp_actions
    * processes
    * work_orders

```python
sp_actions_main = pd.read_sql(
'''
SELECT
process_id,
sp_id,
IF(spa_qty < spa_qty_reject, spa_qty_reject, spa_qty) AS spa_qty_in,
IF(spa_qty < spa_qty_reject, spa_qty_reject, spa_qty) - spa_qty_reject AS spa_qty_out,
spa_qty_reject,
spa_time_start,
spa_time_end,
spa_admin AS admin
FROM sp_actions
ORDER BY process_id
''', db)

col_fix(sp_actions_main)
sp_actions_main.drop_duplicates(inplace = True)
```

```python
process_class = pd.read_sql(
'''
SELECT
DISTINCT process_class AS process_group
FROM processes
WHERE wo_id IN (SELECT wo_id from work_orders)
''', db)

col_fix(process_class)
ls = process_class['process group'].tolist() + ['processRework']

#=====

processes_super_main = pd.read_sql(
'''
SELECT
process_id,
IF(rework_id > 0 , 'Rework', process_name) AS process_name,
IF(rework_id > 0, rework_id, wo_id) AS wo_id,
time_start AS process_time_start,
time_end AS process_time_end,
qty_in AS process_qty_in,
qty_out AS process_qty_out,
qty_reject AS process_qty_reject,
process_order,
IF(rework_id > 0, 'rework', process_group) AS process_group,
rework_id
FROM processes
WHERE process_class IN '''+ str(tuple(ls)) +'''
''', db)

col_fix(processes_super_main)

processes_main = processes_super_main.copy()
processes_main['process qty in'] = np.where(processes_main['process qty in'] < processes_main['process qty out'],
                                            processes_main['process qty out'],
                                            processes_main['process qty in'])

processes_main['process qty reject'] = processes_main['process qty in'] - processes_main['process qty out']
```

```python
wo_main = pd.read_sql(
'''
SELECT
wo_id,
wo_id_string,
wo_date_created,
wo_date_scheduled,
wo_date_active,
wo_date_completed,
wo.sku_id,
k.part_id,
pd.products_name,
wo_qty AS wo_qty_in,
wo_qty_out,
wo_qty - wo_qty_out AS wo_qty_reject,
wo_status,
wo_substatus,
IF(wo_outsourced > 0, 'yes','no') AS wo_outsourced
FROM work_orders wo
LEFT JOIN skus k ON wo.sku_id = k.sku_id
LEFT JOIN products_description pd ON k.part_id = pd.part_id
''', db)

col_fix(wo_main)
```

### Merge all

```python
main = pd.merge(processes_main, sp_actions_main, how = 'left', on = 'process id').merge(wo_main, how = 'left', on = 'wo id')
main.reset_index(drop = True, inplace = True)

l1 = len(processes_main['process id'].unique())
l2 = len(main['process id'].unique())

if l1 == l2:
    print('match')
else:
    print('mismatch')

# some rework ids are such that they are not WO id
# exclude those weird ones here
main = main[main['wo id'].isin(wo_main['wo id'].tolist())].copy()    
```

### Fix process order of reworks
* and then remove where wo_id = 0 (this indicates a non-fab line, so we don't care about it)

```python
# if its a rework, its process order = 0
# fix this so it equals the max process order of the WO + 1
max_ = main.groupby('wo id', as_index = False)[['process order']].max()

def process_order_fix(x):
    if x['process group'] == 'rework':
        process_id = max_[max_['wo id'] == x['wo id']]['process order']
        process_id = process_id.values[0]
        process_id += 1
    else:
        process_id = x['process order']        
    return process_id

main['process order'] = main.apply(process_order_fix, axis = 1)    

main.drop(main[main['wo id'] == 0].index, inplace = True)
```

### View process group names

```python
print(main['process group'].value_counts())

# fill in the remaining few blanks with "secondary"
main['process group'] = np.where(main['process group'].isin(['none','']), 'secondary', main['process group'])
```

### Clean up columns

```python
# lowercase
for col in main.columns:
    try:
        main[col] = [x.lower() for x in main[col]]
    except:
        pass

# reorder columns
ls1 = wo_main.columns.tolist()

ls2 = processes_main.columns.tolist()
for x in ['wo id','process id']:
    ls2.remove(x)

ls3 = sp_actions_main.columns.tolist()

cols =  ls1 + ls2 + ls3
main = main[cols].copy()

main.sort_values(['wo id','process order','process id'], inplace = True)
```

### Create our custom Fab week, from Fri to Thu

```python
# get all the dates we'll need
main['wo date created'] = pd.to_datetime(main['wo date created'])
min_ = main['wo date created'].min()
max_ = pd.to_datetime(week_end)

# create weeks
weeks = pd.DataFrame({'date':pd.date_range(min_, max_)})
weeks['day'] = weeks['date'].dt.weekday
weeks['day'] = [calendar.day_name[x].lower() for x in weeks['day']]

# flip
weeks = weeks.iloc[::-1]

# we start at the thursday that corresponds to the week_end that you set way at the beginning
# this list is seven zeros, seven ones, seven twos, etc.
ls = []
for i in range(0, int(np.ceil(len(weeks)/7)), 1):
    for j in range(7):
        ls.append(i * -1) # make them negative, like we're counting backwards through time
ls = ls[:len(weeks)] # make sure ls and weeks are the same length
        
weeks['week number'] = ls

# get week ending
weeks['week ending'] = weeks['date'].groupby(weeks['week number']).transform('max')
```

### Create for future left-joins and charts

```python
ym = pd.DataFrame({'year and month': pd.date_range(date_start[:7], date_end[:7], freq = 'MS')})
ym['year and month'] = [str(x)[:7] for x in ym['year and month']]
ym.set_index('year and month', inplace = True)

pretty_ym = [calendar.month_abbr[int(x[-2:])] + '\n' + x[:4] for x in ym.index]

w = pd.DataFrame({'week': np.arange(0,weeks['week number'].min() - 1, -1)})
w.set_index('week', inplace = True)

pretty_w = weeks['week ending'].dt.date.map(str).drop_duplicates().iloc[:53].tolist()
pretty_w = list(reversed(pretty_w))
```

### Create date columns

```python
# convert to datetime and get year and month
ls = ['wo date active',
      'wo date completed',
      'wo date created',
      'wo date scheduled',
      'process time end',
      'process time start',
      'spa time end',
      'spa time start']

for col in ls:
    # to datetime
    main[col] = pd.to_datetime(main[col], errors = 'coerce')
    
    # get year and month
    split = col.split(' ')
    
    new_col = split[0] + ' year and month ' + split[-1]
    main[new_col] = [str(x)[:7] for x in main[col]]
    
    # get week
    new_col2 = split[0] + ' week ' + split[-1]
    main[new_col2] = main[col].dt.date.map(dict(zip(weeks['date'], weeks['week number'])))
```

# Availability Metric

```python
avail_main = main[main['wo outsourced'] == 'no'].drop_duplicates('wo id', keep = 'first').copy()
```

```python
avail_dfs = {}
for t in report_type.keys():
    df1 = avail_main.groupby('wo ' + t + ' scheduled')[['wo id']].count()
    df1.columns = ['scheduled']
    
    df2 = avail_main.groupby('wo ' + t + ' active')[['wo id']].count()
    df2.columns = ['active']
    
    df3 = avail_main.groupby('wo ' + t + ' completed')[['wo id']].count()
    df3.columns = ['completed']
    
    df4 = df1.join(df2).join(df3)
    
    df4['completed as % of scheduled'] = df4['completed']/df4['scheduled']
    df4['completed as % of active'] = df4['completed']/df4['active']
    
    # get all active WOs
    df5 = avail_main[avail_main['wo status'] == 'active'].groupby('wo ' + t + ' scheduled')[['wo id']].nunique()    
    df5.columns = ['open wos']
        
    # join them all
    df = df4.join(df5)
        
    # get Fab Op metric
    w1 = 0.50
    w2 = 1 - w1
    df['availability %'] = w1 * df['completed as % of scheduled'] + w2 * df['completed as % of active']
        
    # restrict by date
    if t == 'year and month':
        df = df[df.index <= date_end[:7]]
    else:
        df['week ending'] = df.index.to_series().map(dict(zip(weeks['week number'], weeks['week ending'])))
    
    restrict = report_type[t][0]
    df = df.iloc[-restrict:]
    df.fillna(0, inplace = True)
    
    avail_dfs[t] = df
    
    fig, ax = plt.subplots(figsize = (20,5))
    chart = df.copy()
    
    if t == 'week':
        chart = chart.iloc[-8:].copy()
    
    width = 0.25
    
    x = np.arange(len(chart))
    y1 = chart['scheduled']
    y2 = chart['active']
    y3 = chart['completed']
    
    y4 = chart['availability %']
            
    ax.bar(x, y1, width = width, color = 'blue', alpha = 0.65, edgecolor = 'black', label = 'scheduled')
    ax.bar(x + width, y2, width = width, color = 'green', alpha = 0.65, edgecolor = 'black', label = 'active')
    ax.bar(x + width * 2, y3, width = width, color = 'purple', alpha = 0.65, edgecolor = 'black', label = 'completed')
    ax.set_xticks(x + width)
    ax.set_title('WOs Scheduled, Active, and Completed by ' + t.title())
    ax.grid(color = 'black', alpha = 0.10)
    ax.legend(loc = 'upper left')
    
    if t == 'year and month':
        ax.set_xticklabels(pretty_ym)
    else:
        ax.set_xticklabels(pretty_w[-8:])
        ax.set_xlabel('week ending')        
        
    ax2 = ax.twinx()
    ax2.plot(x + width, y4, label = y4.name, color = 'orange', ls = '--', lw = 3)
    
    y_max = np.max(ax2.get_yticks())
    ax2.set_ylim(0,y_max + 0.1)
    ax2.set_yticklabels(['{:,.0f}%'.format(p * 100) for p in ax2.get_yticks()])
    ax2.legend(loc = 'upper right')
    
    plt.savefig(ax.get_title(), bbox_inches = 'tight')
    plt.show()
```

# Quality Metric

```python
quality_main = main[(main['rework id'] == 0)].drop_duplicates('process id', keep = 'first').copy()
```

```python
cols = ['fabricate','test','total qty rejected']

quality_dfs = {}

for t in report_type.keys():
       
    df = quality_main.groupby(['process '+ t +' end','process group'])[['process qty reject']].sum().unstack(1).fillna(0)
    df.columns = df.columns.droplevel(0)
    df['total qty rejected'] = df.sum(1)
    
    tot_in = quality_main.groupby('process '+ t +' end')[['process qty in']].sum()
    tot_in.columns = ['total qty in']
    
    df = df.join(tot_in)
    
    step_count_main = quality_main.groupby(['process ' + t + ' end','process group'])[['process group']].count().unstack(1).fillna(0)
    step_count_main.columns = step_count_main.columns.droplevel(0)
    step_count_main['total steps'] = step_count_main.sum(1)    
        
    df = df.join(step_count_main['total steps'])    

    df['reject %'] = df['total qty rejected']/df['total qty in']
    df['wo yield %'] = 1 - df['reject %']

    unique_wos = quality_main.groupby('process '+ t +' end')[['wo id']].nunique()
    unique_wos.columns = ['unique wo count']

    df = df.join(unique_wos)
    
    fab_qty = quality_main[quality_main['process group'] == 'fabricate'].groupby('process ' + t + ' end')[['process qty in']].sum()
    fab_qty.columns = ['fabrication qty in']
    
    df = df.join(fab_qty)
    df['machine yield'] = 1 - (df['fabricate']/df['fabrication qty in'])
    
    if t == 'year and month':
        df = df[df.index <= date_end[:7]]    
    else:
        df['week ending'] = df.index.to_series().map(dict(zip(weeks['week number'], weeks['week ending'])))        
    
    restrict = report_type[t][0]
    df = df.iloc[-restrict:]
    
    quality_dfs[t] = df
    quality_dfs['step count by ' + t] = step_count_main
    
    chart = df.copy()   
    chart = chart.loc[:, (chart != 0).any(0)]
    
    if t == 'week':
        chart.set_index('week ending', inplace = True)        
    
    nrows = len(cols)
    ncols = 1
       
    fig, axes = plt.subplots(nrows = nrows, ncols = ncols, figsize = (20,10))
    title = '     Quality Metric by ' + t.title() 
    plt.suptitle(title + '\nThis shows the qty rejected per step unless otherwise noted', y = 1.65)
    for i in range(len(cols)):
               
        x = range(len(chart))
        y = chart[cols[i]]
        mean = y.values.mean()
        
        if t == 'week':
            color = 'blue'
        else:
            color = 'purple'
        axes[i].plot(x, y, lw = 5, alpha = 0.65, marker = 'o', mew = 2, color = color)
        axes[i].axhline(mean, ls = '--', label = 'average')
        
        axes[i].set_title(cols[i].title())
        axes[i].grid(color = 'black', alpha = 0.1)
        axes[i].set_xlabel('')
        axes[i].set_xticks(x)
        axes[i].set_xticklabels('')
        
        if '%' in cols[i]:
            fmt = '{:,.2f}%'
            mult = 100
        else:
            fmt = '{:,.0f}'
            mult = 1
            
        y_min = y.min()
        y_max = y.max()
            
        axes[i].set_yticks(np.arange(y_min, y_max, (y_max - y_min)/5))
        axes[i].set_yticklabels([fmt.format(x * mult) for x in axes[i].get_yticks()])
        axes[i].legend(loc = 'upper left')
        
    if t == 'week':
        axes[-1].set_xticklabels(pretty_w, rotation = 270)
    else:
        axes[-1].set_xticklabels(pretty_ym)        

    axes[-1].set_xlabel(chart.index.name)
        
    plt.subplots_adjust(top = 1.5)
    plt.savefig(title, bbox_inches = 'tight')
    plt.close()
```

# Rework Metric

```python
reworks_main = main[(main['rework id'] > 0)
                  & (main['wo status'] == 'completed')].copy()
reworks_main['part id'] = [int(x) for x in reworks_main['part id']]
```

```python
rework_dfs = {}

for t in report_type.keys():
    
    df = reworks_main.groupby('spa ' + t + ' end')[['spa qty in','spa qty out','spa qty reject']].sum()
    df.columns = ['qty attempted','qty out','qty rejected']
    
    tot_available = quality_dfs[t][['total qty rejected']]
    tot_available.columns = ['total qty available']
    
    df = df.join(tot_available)
    
    tot_wos = reworks_main.groupby('spa ' + t + ' end')[['rework id']].nunique()
    tot_wos.columns = ['unique wo count']
    
    df = df.join(tot_wos)

    if t == 'year and month':
        ym.index.name = 'spa year and month end'
        df = ym.join(df).fillna(0)
        df = df[df.index <= date_end[:7]]    
    else:
        w.index.name = 'spa week end'
        df = w.join(df).fillna(0)
        df = df.iloc[::-1]
        df['week ending'] = df.index.to_series().map(dict(zip(weeks['week number'], weeks['week ending'])))
        
    df['reject %'] = df['qty rejected']/df['qty attempted']
    df['success %'] = 1 - df['reject %']
    
    restrict = report_type[t][0]
    df = df.iloc[-restrict:]
    
    rework_dfs[t] = df
    
    fig, ax = plt.subplots(figsize = (20,5))
    chart = df.copy()
    
    if t == 'week':
        chart = chart.iloc[-12:].copy()
    
    width = 0.25
    
    x = np.arange(len(chart))
    y1 = chart['qty attempted']
    y2 = chart['success %']
            
    ax.bar(x, y1, color = 'purple', alpha = 0.65, edgecolor = 'black', label = 'qty attempted')
    
    ax.set_xticks(x)
    ax.set_title('Rework Qty Attempted along with Success %\n By ' + t.title())
    ax.set_yticklabels(['{:,.0f}'.format(p) for p in ax.get_yticks()])    
    ax.grid(color = 'black', alpha = 0.10)
    ax.legend(loc = 'upper left')
    
    if t == 'year and month':
        ax.set_xticklabels(pretty_ym)
    else:
        ax.set_xticklabels(pretty_w[-12:])
        ax.set_xlabel('week ending')
        
    ax2 = ax.twinx()
    ax2.plot(x, y2, label = y2.name, color = 'orange', ls = '--', lw = 3)
    
    ax2.set_ylim(0, 1.05)
    ax2.set_yticklabels(['{:,.0f}%'.format(p * 100) for p in ax2.get_yticks()])
    ax2.legend(loc = 'upper right')
    
    plt.savefig(ax.get_title(), bbox_inches = 'tight')
```

# In-Stock Metric

```python
now = str(datetime.datetime.now().date())

stock_super_main = pd.read_sql(
'''
SELECT
DATE(timestamp) AS date,
sku_id,
visible_stock
FROM sku_stock_snapshots
WHERE sku_id IN (SELECT sku_id FROM skus WHERE bom_type = 'pnp' AND sku_status = 'working')
AND DATE(timestamp) BETWEEN ' '''+ date_start +''' ' AND ' '''+ now +''' '
''', db)

col_fix(stock_super_main)
```

```python
# this table shows you a snapshot of each sku every 6 hours or so
# so take the avg per day 
stock_main = stock_super_main.groupby(['date','sku id'], as_index = False)[['visible stock']].mean()

# round it
stock_main['visible stock'] = [int(np.round(x)) for x in stock_main['visible stock']]

# flag the skus and days OOS
stock_main['oos'] = np.where(stock_main['visible stock'] == 0, 1, 0)

# get additional date columns
stock_main['year and month'] = [str(x)[:7] for x in stock_main['date']]
stock_main['week'] = stock_main['date'].map(dict(zip(weeks['date'], weeks['week number'])))
stock_main['week ending'] = stock_main['date'].map(dict(zip(weeks['date'], weeks['week number'])))
```

```python
in_stock_dfs = {}

for t in report_type.keys():
       
    df = stock_main.groupby(t).agg({'oos':'sum','sku id':'count'})
    df.rename(columns = {'sku id':'sku count'}, inplace = True)

    df['out-of-stock %'] = df['oos']/df['sku count']
    df['in-stock %'] = 1 - df['out-of-stock %']
    
    unique_skus = stock_main.groupby(t)[['sku id']].nunique()
    unique_skus.columns = ['count of unique skus']
    
    df = df.join(unique_skus)
    
    if t == 'week':
        df['week ending'] = df.index.to_series().map(dict(zip(weeks['week number'], weeks['week ending'])))
        df = df[df['week ending'] <= week_end]
    else:
        df = df[df.index <= date_end[:7]]
    
    in_stock_dfs[t] = df
```

```python
print(in_stock_dfs[display_type].mean())
in_stock_dfs[display_type].tail(8)
```

# Fab Op Metric Over Time

```python
weights = {'availability':1.5/10,
          'quality'      :2/10,
          'rework'       :0.5/10,
          'in-stock'     :6/10}

if sum(weights.values()) != 1:
    print('check this')

fab_op_dfs = {}
t = display_type    

# merge all metrics
df1 = avail_dfs[t][['availability %']]
df2 = quality_dfs[t][['wo yield %']]
df3 = rework_dfs[t]['success %']
df4 = in_stock_dfs[t]['in-stock %']

df = df1.join(df2).join(df3).join(df4)

# rename columns and index
df.columns = weights.keys()
df.index.name = t

# get each metric times its weight and sum to get overall Fab Op Metric
df['fab op metric'] = np.sum(df * list(weights.values()), axis = 1)
df.fillna(0, inplace = True)

if t == 'week':
    df['week ending'] = df.index.to_series().map(dict(zip(weeks['week number'], weeks['week ending']))) 

fab_op_dfs[t] = df

# get the mean per period
mean = pd.DataFrame(fab_op_dfs[t].mean())
mean.columns = ['avg']

# get the current period
cur_per = fab_op_dfs[t].iloc[[-1]].T
cur_per = cur_per.join(mean)

# merge the weights
cur_per['weights'] = cur_per.index.to_series().map(weights)

# make the headers pretty
if t == 'year and month':
    per = calendar.month_abbr[int(date_end[5:7])] + ' ' + date_end[:4]
else:
    d = str(fab_op_dfs[t].iloc[-1]['week ending'].date())
    y = d[:4]
    m = calendar.month_abbr[int(d[5:7])]
    dy = d[-2:]

    per = 'week ending ' + m + ' ' + dy + ', ' + y     
    cur_per.drop('week ending', inplace = True)

header = str(per)
cur_per.columns.values[0] = header
cur_per.index = [x.title() for x in cur_per.index]
fab_op_dfs['current ' + t] = cur_per

chart = df.copy()
if t == 'week':
    chart.set_index('week ending', inplace = True)

cols = list(chart.columns)
colors = ['purple','blue','green','orange','red']

title = 'Fab Op Metric by ' + t + '\nand over time, along with the average'

fig, axes = plt.subplots(nrows = len(cols), ncols = 1, figsize = (20,20))
plt.suptitle(title, y = 0.92)

for i in range(len(cols)):

    x = range(len(chart))
    y = chart[cols[i]]

    mean = np.mean(y)
    axes[i].axhline(mean, color = 'black', alpha = 0.5, ls = '--')

    axes[i].plot(x, y, color = colors[i], lw = 5, alpha = 0.55)
    axes[i].set_yticklabels(['{:,.2f}%'.format(x * 100) for x in axes[i].get_yticks()])
    axes[i].set_xticks(x)
    axes[i].set_xticklabels('')
    axes[i].grid(color = 'black', alpha = 0.1)
    axes[i].set_title(cols[i].title())

if t == 'week':
    axes[-1].set_xticklabels(pretty_w, rotation = 270)     
    axes[-1].set_xlabel('week ending')
else:
    axes[-1].set_xticklabels(pretty_ym)

plt.savefig(title, bbox_inches = 'tight')
```

```python
# choose your columns

ls = ['wo id',
      'wo date created',
      'wo date scheduled',
      'wo date active',
      'wo date completed',
      'process time start',
      'process time end',
      'spa time start',
      'spa time end']

# exclude outsourced and get only completed WOs
for_time = main[(main['wo outsourced'] == 'no')
              & (main['wo status'] == 'completed')][ls].copy()

# change to datetime
for col in ['created','scheduled','active','completed']:
    for_time['wo date ' + col] = pd.to_datetime(for_time['wo date ' + col], errors = 'coerce')
    
# get data from 6 months ago only    
for_time = for_time[for_time['wo date completed'] >= six_months_ago].copy()
```

### Created/Scheduled elapsed

```python
df1 = for_time.drop_duplicates('wo id').copy()
df1['created to scheduled'] = (df1['wo date scheduled'] - df1['wo date created']).dt.days * 86400
df1['created to scheduled'].dropna(inplace = True)
```

### Scheduled/Active elapsed

```python
df2 = for_time.drop_duplicates('wo id').copy()
df2['scheduled to active'] = (df2['wo date active'] - df2['wo date scheduled']).dt.days * 86400
```

### Active/Completed elapsed

```python
df3 = for_time.drop_duplicates('wo id').copy()
df3['active to completed'] = (df2['wo date completed'] - df2['wo date active']).dt.days * 86400
```

### Time elapsed on WO level

```python
# put it all together
by_wo = pd.merge(df1[['wo id','created to scheduled']],
                 df2[['wo id','scheduled to active']],
                 how = 'left',
                 on = 'wo id').merge(df3[['wo id','active to completed']],
                                    how = 'left',
                                    on = 'wo id')

# map the week completed
by_wo['wo week completed'] = by_wo['wo id'].map(dict(zip(main['wo id'], main['wo week completed'])))
by_wo.sort_values('wo week completed', inplace = True)

t = len(by_wo[by_wo['wo week completed'] == 0])

print('x{} work orders were completed during week ending {}.'.format(t, week_end))

# avg time elapsed per WO for the latest week
a = pd.DataFrame(by_wo[by_wo['wo week completed'] == 0].drop(['wo id','wo week completed'], 1).mean()/86400)

d = week_end
y = d[:4]
m = calendar.month_abbr[int(d[5:7])]
dy = d[-2:]

per = 'week ending ' + m + ' ' + dy + ', ' + y     
a.columns = [per]

# avg time elapsed per WO for all weeks in the last 6 months
b = pd.DataFrame(by_wo.drop(['wo id','wo week completed'], 1).mean()/86400)
b.columns = ['6 month avg']

c = a.join(b)
c.index.names = ['period']
#c.reset_index(inplace = True)

print('The table below is the avg time, per work order, that it takes to go from period to period (measured in days)')
c.format_(['n2','n2'])
```

### To see if the times between processes and sp_actions tables differ

```python
a = pd.read_sql(
'''
SELECT
process_id,
TIMESTAMPDIFF(MINUTE, time_start, time_end) AS p_time 
FROM processes
WHERE process_group IN ('test','prep')
AND time_end != '0000-00-00 00:00:00'
GROUP BY process_id
''', db)

col_fix(a)

b = pd.read_sql(
'''
SELECT
process_id,
SUM(TIMESTAMPDIFF(MINUTE, spa_time_start, spa_time_end)) AS sp_time
FROM sp_actions
WHERE process_id IN '''+ str(tuple(a['process id'].tolist())) +'''
GROUP BY process_id
''', db)

col_fix(b)

c = pd.merge(a, b, how = 'left', on = 'process id')

c.info()
c.dropna(inplace = True)

c['diff (min)'] = np.abs(c['p time'] - c['sp time'])

c.sort_values('diff (min)', ascending = False, inplace = True)
```

### Query for Limor

```python
super_main = pd.read_sql(
'''
SELECT
k.part_id,
pd.products_name,
TIMESTAMPDIFF(MINUTE, spa_time_start, spa_time_end) AS time_elapsed_in_minutes, 
sp.spa_qty,
process_group

FROM sp_actions sp

JOIN processes p ON sp.process_id = p.process_id
AND DATE(time_start) BETWEEN ' '''+ s +''' ' AND ' '''+ e +''' ' # where "s" and "e" are the start and end times, 90 days apart
AND p.wo_id != 0 # to exclude reworks
AND process_group IN ('test','prep') # get only these groups
AND time_end != '0000-00-00 00:00:00' # to exclude open processes

JOIN work_orders wo ON p.wo_id = wo.wo_id
JOIN skus k ON wo.sku_id = k.sku_id
JOIN products_description pd ON k.part_id = pd.part_id
''', db)

col_fix(super_main)

# get the time elapsed over both test and prep steps
total_time = super_main.groupby('part id')[['time elapsed in minutes']].sum()

# get the qty made over only one of the steps
qty_main = super_main.groupby(['part id','process group'])[['spa qty']].sum().unstack(1)
qty_main.columns = qty_main.columns.droplevel(0)
# prep comes after test so if prep is not null, use the prep qty, if prep is null, use the test qty
qty_main['qty made'] = np.where(qty_main['prep'].isnull(), qty_main['test'], qty_main['prep'])
qty_main = qty_main[['qty made']]

# put them together
main = super_main[['part id','products name']].drop_duplicates(['part id','products name'])
main = main.set_index('part id').join(total_time).join(qty_main).reset_index()

# get time per part
main['min per'] = np.round(main['time elapsed in minutes']/main['qty made'], 2)
main.sort_values('time elapsed in minutes', ascending = False, inplace = True)

# get the PLM link
l1 = 'https://volcano.adafruit.com/volcano/ada_plm.php#/Parts/'
l2 = '/PartProfile'
main['link'] = l1 + main['part id'].astype(str) + l2
```
