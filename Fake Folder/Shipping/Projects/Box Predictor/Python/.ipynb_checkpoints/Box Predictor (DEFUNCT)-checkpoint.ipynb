### boxes_clean

```python
df1 = boxes_main[['sku id','name','length','width','height']].copy()
df1['working box'] = 'yes'

df2 = nulls.copy()

# add back into the main box data the freshly cleaned "nulls" df
boxes_clean = pd.concat([df1, df2], sort = True)
boxes_clean.rename(columns = {'length':'outer length',
                              'width':'outer width',
                              'height':'outer height'}, inplace = True)

# to int
boxes_clean['sku id'] = boxes_clean['sku id'].astype(int)

# create inner and outer dimensions, taking into account the thickness of the box
for col in ['length','width','height']:
    boxes_clean['outer ' + col] = pd.to_numeric(boxes_clean['outer ' + col])
    boxes_clean['inner ' + col] = boxes_clean['outer ' + col] - thickness * 2

# get volume of box
boxes_clean['volume'] = boxes_clean[['outer length','outer width','outer height']].prod(1)    

# get capacity of box while taking into account the wall thickness and padding
# from here: https://math.stackexchange.com/questions/371475/calculating-the-internal-volume-of-a-cube-with-wall-thickness
def capacity(x):
    l = x['outer length'] - thickness * 2
    w = x['outer width'] - thickness * 2
    h = x['outer height'] - thickness * 2
    capacity = l * w * h
    return capacity

boxes_clean['capacity'] = boxes_clean.apply(capacity, axis = 1)

# get max interior, for later
boxes_clean['max interior'] = boxes_clean[['inner length','inner height','inner width']].max(1)

# rearrange cols
old_cols = list(boxes_clean.columns)
new_cols = ['sku id','name','working box',
            'outer height','outer length','outer width','volume',
            'inner height','inner length','inner width','capacity',
            'max interior']
s1 = set(old_cols)
s2 = set(new_cols)

if len(s1-s2) + len(s2-s1) == 0:
    print('match')
    boxes_clean = boxes_clean[new_cols]
else:
    print('mismatch')
```

### How to Determine Weights, from [UPS](https://www.ups.com/us/en/help-center/packaging-and-supplies/determine-billable-weight.page)

* Determine the Actual Weight
* Actual weight is the package weight rounded up to the next whole pound.

Use a scale to determine the weight of the package. Round any fraction of a pound to the next whole pound for UPS Express® Envelopes (letters) over eight ounces and all other packages.

*****

* Determine Dimensional Weight
* Dimensional weight reflects package density, which is the amount of space a package occupies in relation to its actual weight. Dimensional weight may apply to all UPS domestic and international package services.

Determine the package dimensions in inches. For each dimension, measure at the longest point, rounding each measurement to the nearest whole number (for example, 1.00 to 1.49 will be considered 1, and 1.50 to 1.99 will be considered 2). {Note that np.round() will do this!}

Multiply the package length (longest side of the package) by the width by the height.  The result is the cubic size in inches.

Divide the cubic size in inches by the divisor to calculate the dimensional weight in pounds. Increase any fraction to the next whole pound.

The divisor varies by rate type: 139 for Daily Rates; 166 for Retail Rates.

Dimensional Weight = (L x W x H) ÷ Divisor

L = Length in inches

W = Width in inches

H = Height in inches

Divisor for Daily Rates = 139

Divisor for Retail Rates = 166

```python
def dim_weight(x):
    l = np.round(x['outer length'])
    w = np.round(x['outer width'])
    h = np.round(x['outer height'])
    dim_weight = np.ceil((l*w*h)/dim_div)
    return dim_weight

boxes_clean['dim weight'] = boxes_clean.apply(dim_weight, axis = 1)
```

```python
ups_date_start = '2017-06-01'
```

```python
ups_super_main = pd.read_sql(
'''
SELECT
*
FROM ups_billing
WHERE orders_id IN (SELECT orders_id FROM orders WHERE DATE(date_purchased) >= ' '''+ ups_date_start +''' ')
''', db)

col_fix(ups_super_main)
ups_super_main['charge description'] = ups_super_main['charge description'].str.lower()
```

```python
ups_main = ups_super_main.copy()

# clean up charge descriptions
df = ups_main['charge description'].str.split(' ', expand = True)
ups_main = ups_main.join(df)

# get the first two words of each charge description
ups_main['jb charge description'] = ups_main[0] + ' ' + ups_main[1]

# manual fix
ups_main['jb charge description'] = np.where(ups_main['charge description'] == 'qst', 'qst', ups_main['jb charge description'])

# clean further
def clean(x):
    if 'ground' in x['jb charge description']:
        return 'ground'
    elif 'return' in x['jb charge description']:
        return 'returns'
    elif 'broker' in x['jb charge description']:
        return 'brokerage fees'
    elif 'shipping charge' in x['jb charge description']:
        return 'shipping charge correction'    
    else:
        return x['jb charge description']
    
ups_main['jb charge description'] = ups_main.apply(clean, axis = 1)
```

```python
ot_main = pd.read_sql(
'''
SELECT
DATE(date_purchased) AS date_purchased,
DATE_FORMAT(date_purchased, '%Y-%m') AS year_and_month,
ot1.orders_id,
ot1.value + IFNULL(ot2.value,0) AS shipping_revenue
FROM orders_total ot1

RIGHT JOIN orders o ON ot1.orders_id = o.orders_id
AND DATE(date_purchased) >= ' '''+ ups_date_start +''' '
AND o.orders_id NOT IN (SELECT orders_id FROM subscriptions_history)

LEFT JOIN orders_total ot2 ON o.orders_id = ot2.orders_id
AND ot2.class = 'ot_ddp'

WHERE ot1.class = 'ot_shipping'
''', db)

col_fix(ot_main)
```

```python
# join the two
df1 = ot_main.set_index('orders id')
df2 = ups_main.groupby('orders id')[['netAmount']].sum()

ups = df1.join(df2)

# clean it up
ups.reset_index(inplace = True)
ups.dropna(inplace = True)
ups.rename(columns = {'netAmount':'ups charge'}, inplace = True)

# drop free shipping
ups.drop(ups[ups['shipping revenue'] == 0].index, inplace = True)

# create cols
ups['overcharge'] = np.where(ups['ups charge'] > ups['shipping revenue'], 'yes', 'no')
ups['shipping profit'] = ups['shipping revenue'] - ups['ups charge']
```

```python
by_month = ups.groupby(['year and month','overcharge'])[['shipping profit']].sum().unstack(1).fillna(0)
by_month.columns = by_month.columns.droplevel(0)
by_month['net'] = by_month.sum(1)

print(by_month.mean())
by_month.format_(['m0'] * 3)
```

```python
# get OIDs that incurred a shipping charge correction AND that we lost money on
oids1 = ups[ups['overcharge'] == 'yes']['orders id'].tolist()
oids2 = ups_main[(ups_main['jb charge description'] == 'shipping charge correction')
                & (ups_main['orders id'].isin(oids1))]['orders id'].tolist()


# get the associated data, but only the service charge and the correction charge
# this is where billed_weight > 0
# because UPS puts zeros in this column for all accessorial charges
df1 = ups_main[(ups_main['orders id'].isin(oids2))
             & (ups_main['billed weight'] > 0)].copy()

# make a binary charge discription, for groupby
df1['jb charge description 02'] = np.where(df1['jb charge description'] != 'shipping charge correction','1st billed weight','2nd billed weight')

# get the billed weights per tracking
df2 = df1.groupby(['tracking','jb charge description 02'])[['billed weight']].sum().unstack(1)
df2.columns = df2.columns.droplevel(0)
df2.reset_index(inplace = True)

# map OIDs
df2['orders id'] = df2['tracking'].map(dict(zip(ups_main['tracking'], ups_main['orders id'])))

# get total weights and counts
df3 = df2.groupby('orders id').agg({'1st billed weight':'sum',
                                    '2nd billed weight':'sum',
                                    'orders id':'count'})
df3.rename(columns = {'orders id':'ups box count'}, inplace = True)
df3.reset_index(inplace = True)

# map part and box data
df4 = pd.merge(df3, by_oid, on = 'orders id', how = 'left')

df4.rename(columns = {'qty':'db box count'}, inplace = True)    
df4['db box count'].fillna(0, inplace = True)
df4['db box count'] = df4['db box count'].astype(int)

# find mismatches
not_equal = df4[df4['ups box count'] != df4['db box count']]
val1 = len(not_equal)
val2 = len(df4)
print('count of mismatch box counts per OID: {:,.0f}'.format(val1))
print('{:,.2f}% of total OID count'.format(val1/val2 * 100))

print('\ndrop these')
df4 = df4[~df4['orders id'].isin(not_equal['orders id'])].copy()

# get total dim weight per OID, for multibox shipments this is the sum of the dim weights of its parts
df5 = otb_main.groupby('orders id', as_index = False)[['dim weight']].sum()

# merhe it
df6 = pd.merge(df4, df5, on = 'orders id', how = 'left')

# check for nulls
if np.sum(df6.isnull().any(1)) == 0:
    print('\nno nulls')
else:
    print('\nnulls')

# get billable weight    
df6['billable weight'] = df6[['dim weight', 'total part weight']].max(1)
```

### The beginning of a Box Predictor program

```python
# get data for currently working boxes
ls = ['name','working box','max interior','capacity','dim weight']
b = boxes_clean[ls][boxes_clean['working box'] == 'yes'].sort_values('capacity').copy()
b.drop('working box',1,inplace = True)
b.reset_index(drop = True, inplace = True)

# handpick some examples
x = df6.iloc[[99]]
```

```python
# find capacity needed, which is basically a little bit more than the total volume of the parts 
# recall that this total part volume has padding factored in
# and the capacity already has the box thickness factored in
c = x['total part volume'].values[0]

# find the max length of the parts, to make sure the longest part will fit inside the box
l = x['max part length'].values[0]

# find not big enough
not_big_enough = b[b['capacity'] < c]['name'].tolist()

# find not long enough
not_long_enough = b[b['max interior'] < l]['name'].tolist()

# see if there are fragile parts, if so, exclude mailers
if x['products fragile'].values[0] > 0:
    do_not_use = b[b['name'].str.contains('Mailer')]['name'].tolist()
    
# get box candidates
remove = not_big_enough + not_long_enough + do_not_use
candidates = b[~b['name'].isin(remove)]
    
# get dim weight of the smallest box
d = candidates['dim weight'].min()

# get the weight of the parts
w = x['total part weight'].values[0]

# with this dim weight and actual weight, get the billable weight
# this is the weight we quote the customer
billable = np.max([d, w])

# any box, when combined with the actual weight, with a billable weight less than or equal to this will break even or make us money
gain = candidates[np.max([candidates['dim weight'], pd.Series([w] * len(candidates))], axis = 0) <= billable]['name'].tolist()

# and the others would cause a loss
loss = candidates[~candidates['name'].isin(gain)]['name'].tolist()

d = {'not suitable':not_big_enough + not_long_enough,
    'do not use - fragile':do_not_use,
    'gain':gain,
    'loss':loss}

display = b.copy()
display['result'] = display['name']

for k,v in d.items():
    for v2 in v:
        display['result'] = display['result'].replace(v2,k)
```
