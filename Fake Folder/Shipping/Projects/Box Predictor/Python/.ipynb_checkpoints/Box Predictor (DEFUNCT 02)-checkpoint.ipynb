### Current state of shipping

```python
ups_super_main = pd.read_sql(
'''
SELECT
*
FROM ups_billing
WHERE orders_id IN (SELECT orders_id FROM orders WHERE DATE(date_purchased) BETWEEN ' '''+ ups_date_start +''' ' AND ' '''+ ups_date_end +''' ')
AND orders_id NOT IN (SELECT orders_id FROM subscriptions_history) # exclude adabox
''', db)

col_fix(ups_super_main)
ups_super_main['charge description'] = ups_super_main['charge description'].str.lower()
```

```python
ups_main = ups_super_main.copy()

# clean up charge descriptions
df = ups_main['charge description'].str.split(' ', expand = True)
ups_main = ups_main.join(df)

# get the first two words of each charge description
ups_main['jb charge description'] = ups_main[0] + ' ' + ups_main[1]

# manual fix
ups_main['jb charge description'] = np.where(ups_main['charge description'] == 'qst', 'qst', ups_main['jb charge description'])

# clean further
def clean(x):
    if 'ground' in x['jb charge description'] and 'return' not in x['jb charge description']:
        return 'ground'
    elif 'return' in x['jb charge description']:
        return 'returns'
    elif 'broker' in x['jb charge description']:
        return 'brokerage fees'
    elif 'shipping charge' in x['jb charge description']:
        return 'shipping charge correction'    
    else:
        return x['jb charge description']
    
# this just shortens the original UPS charge description, doesn't change anything much    
ups_main['jb charge description'] = ups_main.apply(clean, axis = 1)
```

```python
ot_main = pd.read_sql(
'''
SELECT
DATE(date_purchased) AS date_purchased,
DATE_FORMAT(date_purchased, '%Y-%m') AS year_and_month,
ot1.orders_id,
ot1.value + IFNULL(ot2.value,0) AS shipping_revenue
FROM orders_total ot1

RIGHT JOIN orders o ON ot1.orders_id = o.orders_id
AND DATE(date_purchased) BETWEEN ' '''+ ups_date_start +''' ' AND ' '''+ ups_date_end +''' '
AND o.orders_id NOT IN (SELECT orders_id FROM subscriptions_history) # exclude adabox

LEFT JOIN orders_total ot2 ON o.orders_id = ot2.orders_id
AND ot2.class = 'ot_ddp'

WHERE ot1.class = 'ot_shipping'
AND ot1.value > 0 # exclude "free shipping"
''', db)

col_fix(ot_main)
```

```python
# join the two
# not that this is a left join onto orders_total
df1 = ot_main.set_index('orders id')
df2 = ups_main.groupby('orders id')[['netAmount']].sum()

ups = df1.join(df2)

# find nulls
nulls = ups[ups.isnull().any(1)].copy()

# veiw their shipping_modeul_codes
a = pd.read_sql(
'''
SELECT
shipping_module_code,
COUNT(shipping_module_code) AS count
FROM orders
WHERE orders_id IN '''+ str(tuple(nulls.index.tolist())) +'''
GROUP BY shipping_module_code
''', db)

print('shipping module codes of nulls\n')
print(a)

# clean it up
ups.reset_index(inplace = True)
ups.dropna(inplace = True)
ups.rename(columns = {'netAmount':'ups charge'}, inplace = True)

# drop free shipping
# adabox was dropped during SQL queries
ups.drop(ups[ups['shipping revenue'] == 0].index, inplace = True)

# create cols
ups['overcharge'] = np.where(ups['ups charge'] > ups['shipping revenue'], 'yes', 'no')
ups['shipping profit'] = ups['shipping revenue'] - ups['ups charge']

# put all charges on one line per OID
# omit fuel because it's in every order
a = ups_main.groupby(['orders id','jb charge description'], as_index = False)[['tracking']].count()
a.drop(a[a['jb charge description'] == 'fuel surcharge'].index, inplace = True)

b = a.groupby('orders id')['jb charge description'].apply(lambda x: ', '.join(x))

c = pd.DataFrame(b)
c.reset_index(inplace = True)

# this is just a sentence that lists the UPS charges, so no changes to any values
ups = pd.merge(ups, c, on = 'orders id')

ups['date purchased'] = pd.to_datetime(ups['date purchased'])
```

### Current Box Predictor Assessment

```python
# get quotes
sq_main = pd.read_sql(
'''
SELECT
DATE(date_purchased) AS date_purchased,
DATE_FORMAT(date_purchased, '%Y-%m') AS year_and_month,
s.*,
o.shipping_module_code AS service
FROM shipping_quotes s
LEFT JOIN orders o ON s.orders_id = o.orders_id
''', db)

col_fix(sq_main)

# get actual
sl_super_main = pd.read_sql(
'''
SELECT
*
FROM ship_log
WHERE shipped_date > ' '''+ predictor_date_start +''' '
''', db)

col_fix(sl_super_main)
sl_main = sl_super_main[['orders id','sl weight','sl box']].copy()
```

```python
# put the data together
pred_main = pd.merge(sl_main, sq_main, how = 'left', on = 'orders id')
pred_main = pred_main[pred_main['year and month'] >= predictor_date_start[:7]].copy()

# rename cols
pred_main.rename(columns = {'sl weight':'pred weight',
                           'sl box':'pred box',
                           'weight':'actual weight',
                           'box':'actual box'}, inplace = True)

# rearrange them,
ls = pred_main.columns.tolist()
for x in ['pred weight','pred box','actual weight','actual box']:
    ls.remove(x)
ls = ls + ['pred weight','actual weight','pred box','actual box']

pred_main = pred_main[ls]

# find nulls
nulls = pred_main[pred_main.isnull().any(1)]
print('{:,.0f} nulls'.format(len(nulls)))
print('or {:,.2f}% of total lines'.format(len(nulls)/len(pred_main) * 100))
print('remove these')
pred_main.dropna(inplace = True)

# find zeros
zero = pred_main[pred_main['actual weight'] == 0]
print('\n{:,.0f} lines where actual weight is zero'.format(len(zero)))
print('drop these')
pred_main = pred_main[pred_main['actual weight'] > 0].copy()

# to int, round up
for col in ['actual','pred']:
    pred_main[col + ' box'] = pred_main[col + ' box'].astype(int)
    pred_main[col + ' weight rounded'] = np.ceil(pred_main[col + ' weight'])    

# get summary columns    
pred_main['weight error %'] = np.abs((pred_main['actual weight'] - pred_main['pred weight'])/pred_main['actual weight'])
pred_main['weight error % rounded'] = np.abs((pred_main['actual weight rounded'] - pred_main['pred weight rounded'])/pred_main['actual weight rounded'])
pred_main['box result'] = np.where(pred_main['actual box'] == pred_main['pred box'], 1, 0)    

# exclude these services
ls = ['','resellershipping','sameday']
pred_main = pred_main[~pred_main['service'].isin(ls)].copy()
```

### Weight Error %

```python
weight_summary = pred_main.groupby('year and month')[['weight error %']].describe()
print('overall mean is {:,.2f}%'.format(pred_main['weight error %'].mean() * 100))
weight_summary    
```

### Box result

```python
box_summary = pred_main.groupby('year and month').agg({'orders id':'nunique',
                                                       'box result':'sum'}).rename(columns = {'orders id':'unique OID count',
                                                                                             'box result':'correct predictions'})
box_summary['% correct'] = box_summary['correct predictions']/box_summary['unique OID count']

val1 = pred_main['box result'].sum()
val2 = len(pred_main)
print('overall box result is {:,.2f}% correct'.format(val1/val2 * 100))

box_summary
```

### Compare service charges with db

```python
# get all oids from the dataframe you made above
# that is:
# all OIDs that went via worldwide
# AND that incurred a residential charge
# AND that we lost money on
# AND that are from time period P
oids = res_overview['orders id'].tolist()

# breakdown the charges
df = ups_main[ups_main['orders id'].isin(oids)].copy()

# get a list of charge descriptions
service = set(df['jb charge description'])

not_service = ['address correction',
               'residential surcharge',
               'residential adjustment',
               'shipping charge correction',
               'extended area',
               'remote area',
               'electronic export',
               'delivery area',
               'delivery confirmation']
# remove these
for x in not_service:
    try:
        service.remove(x)
    except:
        pass
    
print('there charges are considered "basis charges":\n')
for x in sorted(service):
    print(x)    
    
print('\nthese are considered "accessorial charges":\n')
for x in sorted(not_service):
    print(x)
       
# now, create a second charge description column
# label each row-by-row UPS charge as a "basis" charge or an "accessorial charge"
df['jb charge description 02'] = np.where(df['jb charge description'].isin(service),
                                          'service charge',
                                           df['jb charge description'])

# groupby this new column and sum the UPS charge
res_expand = df.groupby(['orders id','jb charge description 02'])[['netAmount']].sum().unstack(1).fillna(0)
res_expand.columns = res_expand.columns.droplevel(0)
res_expand.reset_index(inplace = True)

# map this stuff
for col in ['shipping revenue','address type','jb charge description']:
    res_expand[col] = res_expand['orders id'].map(dict(zip(res_overview['orders id'], res_overview[col])))
    
# map orders data
for col in ['date purchased','year and month']:
    res_expand[col] = res_expand['orders id'].map(dict(zip(ot_main['orders id'], ot_main[col])))
    
# map total UPS charge
tots = ups_main.groupby('orders id', as_index = False)[['netAmount']].sum()
res_expand['total ups charge'] = res_expand['orders id'].map(dict(zip(tots['orders id'], tots['netAmount'])))    
```
