### Fix where shippers_id equals zero
This occurs when an order is shipped through the gather QA tool or when it's an AdaBox subscription shipment

```python
# this patch will fix most of these lines, but not all because sometimes someone like Nikki will actually ship an order
# see here for an example: https://volcano.adafruit.com/volcano/Order_Status.php?oid=1631974
# we'll keep shippers_id = 0 for OIDs like this

# get OIDs of where shippers_id = 0
oids = list(set(ship_log[ship_log['shippers id'] == 0]['orders id']))

# grabe the shippers name of these OIDs
patch = pd.read_sql(
'''
SELECT orders_id,
LOWER(user_string) AS shippers_name
FROM orders_status_history
WHERE orders_id IN '''+ str(tuple(oids)) +'''
AND user_string != ''
''', db)

col_fix(patch)

# map the shippers id, drop na and dupes
patch['shippers id'] = patch['shippers name'].map(dict(zip(shippers['shippers name'], shippers['shippers id'])))
patch.dropna(inplace = True)
patch.drop_duplicates('orders id', inplace = True)
```

```python
# isolate zeros
zeros = ship_log[ship_log['shippers id'] == 0].copy()
zeros.reset_index(drop = True, inplace = True)

# remove from main df
ship_log.drop(ship_log[ship_log['shippers id'] == 0].index, inplace = True)

# patch it
for index, row in zeros.iterrows():
    oid = row['orders id']
    try:
        new = patch[patch['orders id'] == oid]['shippers id'].values[0]
    except:
        new = 0
    zeros.loc[index, 'shippers id'] = new
    
zeros['shippers id'] = [int(x) for x in zeros['shippers id']]    

ship_log = pd.concat([ship_log, zeros], sort = False)
ship_log.reset_index(drop = True, inplace = True)

print('new count lines where shippers_id equals zero: {:,.0f}'.format(np.sum(ship_log['shippers id'] == 0)))
```

### Check it

```python
d = np.sum(ship_log['umbrella orders id'].duplicated())
t = len(ship_log)
print('count of dupes due to the same partial going in multiple boxes: {:,.0f}'.format(d))
print('which is {:,.2f}% of all lines'.format(d/t * 100))
print('\nassume a single shipper ships multiple boxes within the same partial\ndrop these because so far our ship_log df has no data about qty,\nso we can safely drop dupes and our data will be fine')
ship_log.drop_duplicates(subset = ['umbrella orders id'], inplace = True)
ship_log.reset_index(drop = True, inplace = True)

zero = ship_log[ship_log['orders id'] == 0]
print('\n{:,.0f} OID(s) equal to zero'.format(len(zero)))
print('drop these')
ship_log.drop(ship_log[ship_log['orders id'] == 0].index, inplace = True)
ship_log.reset_index(drop = True, inplace = True)
```

### Get (non partial) gathers data

```python
g1 = pd.read_sql(
'''
SELECT
*
FROM gathers
''', db)

col_fix(g1)
```

```python
# get all OIDs with valid gather_batch ids
for_g1 = orders_main[orders_main['gather batch'] > 0].copy()

# map the qty_total from op_main via OID
for_qty = op_main.groupby('orders id', as_index = False)[['qty total']].sum()
for_g1['qty total'] = for_g1['orders id'].map(dict(zip(for_qty['orders id'], for_qty['qty total'])))

# for both non resellers and reseller/supers get the unique OID count and the qty total per batch
for c in ['non reseller','reseller/super']:
    df = for_g1[for_g1['customer type'] == c]
    df2 = df.groupby('gather batch').agg({'orders id':'nunique',
                                         'qty total':'sum'})
    df2.rename(columns = {'orders id':'unique ' + c + ' shipment count',
                         'qty total':c + ' qty total'}, inplace = True)
    
    # map this data via gather_batch
    for col in df2.columns:
        g1[col] = g1['gather batch'].map(dict(zip(df2.index.to_series(), df2[col])))

# fill nulls        
g1.fillna(0, inplace = True)

# make the numbers look better
for col in g1.columns:
    try:
        g1[col] = [int(x) for x in g1[col]]
    except:
        pass
    
g1['contains partials'] = 'no'
g1['partial gather batch'] = 0
```

### Get (partial) gathers data

```python
g2 = pd.read_sql(
'''
SELECT
partial_gather_id AS partial_gather_batch,
0 AS gather_batch,
pg_gathered_by_id AS gathered_by_id,
pg_time_in AS time_in,
pg_time_out AS time_out,
'yes' AS contains_partials
FROM partial_gathers
''', db)

col_fix(g2)
```

```python
ps_main = pd.read_sql(
'''
SELECT
partial_gather_id AS partial_gather_batch,
ps_parent_id AS orders_id,
ps_partial_id AS partial_id
FROM partial_shipments
''', db)

col_fix(ps_main)

# create umbrella_oid
ps_main['umbrella orders id'] = ps_main['orders id'].map(str) + '-' + ps_main['partial id'].map(str)

# map qty_total
ps_main['qty total'] = ps_main['umbrella orders id'].map(dict(zip(qty_total['umbrella orders id'], qty_total['qty total'])))

# map customer_type
ps_main['customer type'] = ps_main['orders id'].map(dict(zip(orders_main['orders id'], orders_main['customer type'])))

# groupby gather_batch where it's not zero
for_g2 = ps_main[ps_main['partial gather batch'] > 0].groupby('partial gather batch').agg({'umbrella orders id':'count',
                                                                                           'qty total':'sum'})
for_g2.rename(columns = {'umbrella orders id':'unique reseller/super shipment count',
                        'qty total':'reseller/super qty total'}, inplace = True)

# map it via gather_bacth
for col in for_g2.columns:
    g2[col] = g2['partial gather batch'].map(dict(zip(for_g2.index.to_series(), for_g2[col])))
    
# if the only customer type associated with partial orders is "reseller/super", create columns of zeros of non-resellers
if ps_main['customer type'].unique() == 'reseller/super':
    for col in ['unique non reseller shipment count','non reseller qty total']:
        g2[col] = 0
else:
    print('check this')    
```

### 95% confidence intervals for avg count per Nov and Dec

```python
cols = ['non reseller shipments',
        'reseller/super shipments']
# data is oct, nov, dec from 2015 to now
for_conf = by_month[(by_month.index.str.contains('-10|-11|12'))].loc['2017-01':]#[cols]

st = for_conf.index[0]
en = for_conf.index[-1]

main_ls = []

for col in for_conf.columns:
    
    df = for_conf[col]

    m = df.mean()
    s = df.std()
    n = len(df)
    phi = stats.t.ppf(0.95, n-1)
    lower = m - phi * (s/np.sqrt(n))
    upper = m + phi * (s/np.sqrt(n))
    ls = [col, lower, m, upper, st, en, len(df)]
    main_ls.append(ls)
    
conf1 = pd.DataFrame(main_ls, columns = ['data','lower','mean','upper','data start','data end','data points'])    
for col in conf1.columns:
    try:
        conf1[col] = [int(round(x)) for x in conf1[col]]
    except:
        pass
    
print('95% confidence intervals for avg count of shipments per month for Oct, Nov and Dec')    
conf1.format_([0,'n0','n0','n0'])    
```

### 95% confidence intervals for avg shipments per shipper per month

```python
df1 = ship_log[(ship_log['date shipped'].dt.year == 2017)
            & (ship_log['year and month'].str.contains('-10|-11|-12'))]

for_conf2 = pd.DataFrame(index = list(set(df1['year and month'])))
for_conf2.index.names = ['year and month']

for t in ['non reseller','reseller/super']:
    df = df1[df1['customer type'] == t].groupby('year and month').agg({'shippers id':'nunique','umbrella orders id':'count'})
    df.rename(columns = {'shippers id':t + ' shippers',
                         'umbrella orders id':t + ' shipments'}, inplace = True)
    df['avg ' + t] = df[t + ' shipments']/df[t + ' shippers']
    for_conf2 = for_conf2.join(df)

main_ls = []    
for t in ['non reseller','reseller/super']:
    m = for_conf2['avg ' + t].mean()
    s = for_conf2['avg ' + t].std()
    n = len(for_conf2)
    phi = stats.t.ppf(0.95, n-1)

    lower = m - phi * (s/np.sqrt(n))
    upper = m + phi * (s/np.sqrt(n))

    main_ls.append(['avg shipments per ' + t] + [lower,m,upper])
    
conf2 = pd.DataFrame(main_ls, columns = ['data','lower','mean','upper'])
for col in conf2.columns:
    try:
        conf2[col] = [int(round(x)) for x in conf2[col]]
    except:
        pass
print('95% confidence intervals for avg count of shipments per shipper per month')
conf2.format_([0,'n0','n0','n0'])   
```
