### Confidence Interval for weekly shipping profit

```python
if report_type == 'week ending':
        
    for_weekly = main_clean[(main_clean['free shipping'] == 'no')
                          & (main_clean['adabox'] == 'no')
                          & (main_clean['year and month'].str[:4] == '2018')
                          & (main_clean['shipping module code'].isin(['UPS','DHL','USPS']))].copy()
    
    for_weekly['shipping module code'] = [x.upper() for x in for_weekly['shipping module code']]
    sample = for_weekly[for_weekly['shipping module code'].isin(['UPS','DHL'])].groupby('week ending')[['shipping profit']].sum()
    
    conf = jb_conf(sample)
    l = conf.loc['lower'].values[0]
    m = conf.loc['mean'].values[0]
    u = conf.loc['upper'].values[0]
    n = conf.loc['sample size'].values[0]
    
    print('95% confidence interval for avg weekly shipping profit (UPS and DHL combined)\n')
    wrt = 'Lower: ${:,.0f}\nAverage: ${:,.0f}\nUpper: ${:,.0f}\nSample size: {:,.0f} weeks'.format(l,m,u,n)
    weekly_report_dict['confidence interval'] = wrt.replace('\n','<br>')
    print(wrt)
```

### Weekly summary

```python
if report_type == 'week ending': 
    try:
        df = for_weekly[for_weekly['week ending'].isin([week1,week2,week3])].groupby(['week ending','shipping module code']).agg({'orders id':'nunique','shipping profit':'sum'})    
        
        # change to string for future join
        df.reset_index(inplace = True)
        df['week ending'] = [str(x)[:10] for x in df['week ending']]
        df = df.set_index(['week ending','shipping module code'])
        
        # in case of missing data
        a = [[week1, week1, week1,
              week2, week2, week2,
              week3, week3, week3], ['DHL','UPS','USPS'] * 3]
        ix = pd.MultiIndex.from_arrays(a, names = ('week ending','shipping module code'))
        temp = pd.DataFrame(index = ix)
        
        df = temp.join(df)
        df.fillna(0, inplace = True)
    except:
        pass

    df.rename(columns = {'orders id':'oid count',
                        'shipping profit':'shipping profit'}, inplace = True)

    ls = []
    for i in df.index.get_level_values(0).unique():
        for j in df.index.get_level_values(1).unique():
            m = as_p[as_p.index == i][j].values[0]
            ls.append(m)        

    # for a later "check", down below        
    y = df[df.index.get_level_values(1) == 'DHL']['shipping profit'].sum()        

    vals = df.sum().values
    avg = vals[1]/vals[0]
    print('For UPS, DHL, and USPS during weeks ending {} through {}:\n'.format(week1,week3))
    weekly_report_dict['profit summary'] = 'Total OID count: {:,.0f}\nTotal shipping profit: ${:,.0f}\nAvg profit per order: ${:,.2f}'.format(vals[0], vals[1], avg)
    print(weekly_report_dict['profit summary'])

    df['missing as a % of service total'] = ls    
    df.format_(['n0','m2','p2']).replace('nan%','0.00%')
    
    df.sort_index(level = 0, ascending = False, inplace = True)

    t = df[['oid count','shipping profit']].sum().tolist()
    d = pd.DataFrame(data = t).T
    d.columns = ['oid count','shipping profit']
    d['missing as a % of service total'] = np.nan
    d['week ending'] = ''
    d['shipping module code'] = 'Three Week Total'
    d.set_index(['week ending','shipping module code'], inplace = True)

    df = pd.concat([df,d])

    df.index = pd.MultiIndex.from_tuples([(str(x[0])[:11], x[1]) for x in df.index])
    df.index.names = ['Week Ending','Service']
    display(df.format_(['n0','m0','p2']).replace('nan%',''))
    weekly_report_dict['profit summary table'] = df.format_(['n0','m0','p2']).replace('nan%','').to_html(index = True)
else:
    pass
```

### Top DHL losses

```python
if report_type == 'week ending':

    x = 'DHL'
    df = for_weekly[(for_weekly['week ending'].isin([week1,week2,week3]))
                  & (for_weekly['profit loss'] == 'yes')
                  & (for_weekly['shipping module code'] == x)].sort_values('shipping profit')

    # get total count
    tot = len(df)
    tot2 = df['shipping profit'].sum()
    
    # where we did not charge DDP but DHL did
    a = df[(df['ddp revenue'] == 0) & (df['ddp charge'] > 0) 
         & (df['service revenue'] >= df['service charge'])] 
    a_count = len(a)
    a_profit = a['shipping profit'].sum()
    a_title = 'Adafruit did not charge DDP but DHL did'

    # where our DDP was not enough and was greater than zero
    b = df[(df['ddp revenue'] < df['ddp charge']) 
        & (~df['orders id'].isin(a['orders id']))
        & (df['service revenue'] >= df['service charge'])]
    b_count = len(b)
    b_profit = b['shipping profit'].sum()
    b_title = 'DHL DDP greater than Adafruit DDP'

    # excluding DDP, where our service charge not enough
    c = df[(df['service revenue'] < df['service charge'])
         & (~df['orders id'].isin(a['orders id'].tolist() + b['orders id'].tolist()))]
    c_count = len(c)
    c_profit = c['shipping profit'].sum()
    c_title = 'Excluding DDP: DHL service charge greater than Adafruit service charge'

    # put it together
    summ = pd.DataFrame({'count':[a_count, b_count, c_count],
                        'profit loss':[a_profit, b_profit, c_profit]},
                        index = [a_title, b_title, c_title])
    summ.sort_values('count', ascending = False, inplace = True)
    summ.loc['Total'] = summ.sum()
    summ.index.names = ['group']    

    if np.abs(summ.loc['Total','count'] - tot) < .05:
        pass
    else:
        raise ValueError('totals do not match')

    if np.abs(summ.loc['Total','profit loss'] - tot2) < 0.05:
        pass
    else:
        raise ValueError('totals do not match')
        
    summ.reset_index(inplace = True)
        
    print('DHL profit losses for weeks ending {}, {}, and {}'.format(week1,week2,week3))   
    if summ.empty:
        print('no losses')
        weekly_report_dict['dhl losses table'] = 'no losses'
    else:
        display(summ.format_([0,'n0','m0']))
        weekly_report_dict['dhl losses table'] = summ.format_([0,'n0','m0']).to_html(index = False)
```

### DHL avg charge

```python
three_months_ago = str((pd.to_datetime(date_end) - pd.DateOffset(months = 3)).date())

# recall that main_clean excludes missing and dhl_remove 
for_avg_charge = main_clean[(main_clean['date purchased'] >= three_months_ago)
                      & (main_clean['shipping module code'] == 'DHL')].copy()

avg_charge = for_avg_charge.groupby('week ending')[['shipping charge']].mean()

x_bar = np.mean(avg_charge)[0]
s = np.std(avg_charge)[0]
n = len(avg_charge)
deg_free = n - 1

if deg_free >= 30:
    phi = 1.96
else:
    phi = stats.t.ppf(0.95, deg_free)
    
lower = x_bar - phi * (s/np.sqrt(n))
upper = x_bar + phi * (s/np.sqrt(n))

print('95% Confidence Interval for Avg DHL Charge per Week\nlower 95%: ${:,.2f}\nMean: ${:,.2f}\nUpper 95%: ${:,.2f}'.format(lower,x_bar,upper))
```
