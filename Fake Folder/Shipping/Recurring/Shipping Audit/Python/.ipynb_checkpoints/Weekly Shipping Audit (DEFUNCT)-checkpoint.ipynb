# UPS

### Get latest date in ups_billing table
* go back one week 

```python
max_date = pd.read_sql(
'''
SELECT
DATE(MAX(date_purchased)) AS max_date
FROM orders
WHERE orders_id IN (SELECT orders_id FROM ups_billing)
''', db)
```

```python
ups_date_end = str(max_date['max_date'].values[0])
ups_date_start = str((pd.to_datetime(ups_date_end) - pd.DateOffset(weeks = 1)).date())
```

### Get list of UPS ddp charge descriptions

```python
ups_ddp_charge_list = ['Agri Processing',
                      'Broker Fee',
                      'Brokerage Fees',
                      'Brokerage GST',
                      'Ca British Columbia Pst',
                      'Ca Customs Hst',
                      'Complex Entry',
                      'Customs Gst',
                      'Customs Warehouse',
                      'DGoods Air Inaccessible',
                      'Duty Amount',
                      'Pst Quebec',
                      'QST']
```

### Get ups data

```python
ups_db = pd.read_sql(
'''
SELECT
orders_id AS 'orders id',
transaction_date AS 'shipment date',
SUM(CASE WHEN charge_description IN '''+ str(tuple(ups_ddp_charge_list)) +''' THEN netAmount END) AS 'ups ddp charge',
SUM(CASE WHEN charge_description NOT IN '''+ str(tuple(ups_ddp_charge_list)) +''' THEN netAmount END) AS 'ups ship charge'
FROM ups_billing
WHERE orders_id IN (SELECT orders_id FROM orders WHERE DATE(date_purchased) BETWEEN ' '''+ ups_date_start +''' ' AND ' '''+ ups_date_end +''' ')
GROUP BY orders_id
''', db)

ups_db.fillna(0, inplace = True)
ups_db['ups ship charge'] = [np.round(x, 2) for x in ups_db['ups ship charge']]
```

### Get ups/ada shipping data

```python
ups_shipping = pd.read_sql(
'''
SELECT
A.orders_id AS 'orders id',
A.value AS 'ada ship charge',
IFNULL(B.value, 0) AS 'ada ddp charge'
FROM

(SELECT
orders_id,
value
FROM orders_total
WHERE class = 'ot_shipping'
AND orders_id IN '''+ str(tuple(ups_db['orders id'])) +''') A

LEFT JOIN

(SELECT
orders_id,
value
FROM orders_total
WHERE class = 'ot_ddp') B

ON A.orders_id = B.orders_id
''', db)

ups_shipping['ada total charge'] = ups_shipping[['ada ship charge','ada ddp charge']].sum(1)
```

### Check the UPS ddp count real quick

```python
pd.read_sql(
'''
SELECT
COUNT(*) AS 'ada ddp charge count'
FROM orders_total
WHERE class = 'ot_ddp'
AND orders_id IN
(SELECT
orders_id
FROM orders
WHERE shipping_module_code = 'upsxml'
AND DATE(date_purchased) BETWEEN ' '''+ ups_date_start +''' ' AND ' '''+ ups_date_end +''' ')
''', db)
```

### Get other data like delivery country

```python
ups_other = pd.read_sql(
'''
SELECT
orders_id AS 'orders id',
LOWER(delivery_city) AS 'delivery city',
LOWER(delivery_country) AS 'delivery country'
FROM orders 
WHERE orders_id IN '''+ str(tuple(ups_db['orders id'])) +'''
''', db)
```

### Merge

```python
ups_shipping = pd.merge(ups_shipping,
                        ups_db,
                        how = 'left',
                        on = 'orders id').merge(ups_other, on = 'orders id')

ups_shipping = ups_shipping[['shipment date',
                              'orders id',
                              'ada ship charge',
                              'ups ship charge',
                              'ada ddp charge',
                              'ups ddp charge',
                              'delivery city',
                              'delivery country']]

ups_shipping = ups_shipping[ups_shipping['ada ship charge'] != 0]

ups_shipping['ship difference'] = ups_shipping['ada ship charge'] - ups_shipping['ups ship charge']
ups_shipping['ddp difference'] = ups_shipping['ada ddp charge'] - ups_shipping['ups ddp charge']
ups_shipping['ship profit'] = ups_shipping[['ada ship charge','ada ddp charge']].sum(1) - ups_shipping[['ups ship charge','ups ddp charge']].sum(1)
```

```python
print ('UPS data from ' + calendar.month_abbr[int(ups_date_start[5:7])] + ' ' + ups_date_start[-2:] + ', ' + ups_date_start[:4] + ' to ' + calendar.month_abbr[int(ups_date_end[5:7])] + ' ' + ups_date_end[-2:] + ', ' + ups_date_end[:4])
print('Count of orders considered: {:,.0f}'.format(len(ups_shipping)))
print('Adafruit ship charge minus UPS ship charge: ${:,.2f}'.format(ups_shipping['ada ship charge'].sum() - ups_shipping['ups ship charge'].sum()))
print('Adafruit DDP charge minus UPS DDP charge: ${:,.2f}'.format(ups_shipping['ada ddp charge'].sum() - ups_shipping['ups ddp charge'].sum()))
print('Overall UPS shipping profit: ${:,.2f}'.format(ups_shipping['ship profit'].sum()))
```

### UPS Overcharges

```python
ups_over = ups_shipping[ups_shipping['ada ship charge'] < ups_shipping['ups ship charge']].groupby('delivery country')[['delivery country']].count().sort_values('delivery country', ascending = False)
ups_over['% of total'] = ups_over['delivery country']/ups_over['delivery country'].sum()
ups_over.head()
```

# DHL

```python
oids = pd.read_sql(
'''
SELECT
orders_id AS 'orders id'
FROM orders 
WHERE YEAR(date_purchased) >= 2017
AND orders_status != 9
AND orders_status != 10
AND payment_method != 'Replacement Order'
''', db)
```

```python
# get all CSVs
path = r'/Users/jarad/fake_folder/Shipping/Recurring/DHL Tax Analysis/Docs/DHL Invoices/'
all_files = glob.glob(path + '/*.csv')
frame = pd.DataFrame()
list_ = []

for file_ in all_files:
    df = pd.read_csv(file_, index_col = None, header = 0)
    list_.append(df)
    
dhl_super_raw = pd.concat(list_)   

# fix columns
dhl_super_raw.columns = [x.lower() for x in dhl_super_raw]
dhl_super_raw.rename(columns = {'shipment reference 1':'orders id',
                               'dest name':'delivery city',
                               'dest country name':'delivery country'}, inplace = True)

# check for dupes
if dhl_super_raw[dhl_super_raw.duplicated()].empty:
    print('main CSV has no fully duplicated lines')
else:
    print('main CSV does have some fully duplicated lines')

# fix orders id data type
dhl_super_raw['orders id'] = pd.to_numeric(dhl_super_raw['orders id'], errors = 'coerce', downcast = 'integer')
dhl_super_raw['orders id'].fillna(0, inplace = True)
dhl_super_raw['orders id'] = [int(x) for x in dhl_super_raw['orders id']]

# fix weight charge
dhl_super_raw['weight charge'].fillna(0, inplace = True)

# drop any column with all zeros
dhl_super_raw = dhl_super_raw.loc[:, (dhl_super_raw != 0).any(axis = 0)]

# get count of columns with charge names and amounts
# ask jarad for more about why this is the way it is

total_columns = sum(dhl_super_raw.columns.str.contains('xc')) - 1
total_charge_groups = total_columns/3

print('count of charge-type columns: %i' % total_charge_groups)

dhl_clean = pd.DataFrame()

for i in np.arange(1, int(total_charge_groups) + 1):
    
    df = dhl_super_raw.groupby(['shipment number','orders id', 'xc' + str(i) + ' name'])[['xc' + str(i) + ' charge']].sum().unstack(2).fillna(0)
    df.columns = df.columns.get_level_values(1)    
    df.reset_index(inplace = True)
    dhl_clean = dhl_clean.append(df, ignore_index = True)
    
dhl_test = dhl_clean.copy()    
    
# fill nulls
dhl_clean.fillna(0, inplace = True) 

# drop columns with all zeros
dhl_clean = dhl_clean.loc[:, (dhl_clean != 0).any(axis = 0)]

# convert column names to lower case
dhl_clean.columns = [x.lower() for x in dhl_clean]  

# consolidate charges by shipment number
dhl_clean = dhl_clean.groupby(['shipment number','orders id'], as_index = False).sum()

# determine incoming/outgoing
dhl_clean['incoming/outgoing'] = np.where(dhl_clean['orders id'].isin(oids['orders id'].tolist()),
                                         'outgoing',
                                         'incoming')    
# add the weight charge
dhl_clean = pd.merge(dhl_clean,
                    dhl_super_raw.groupby(['shipment number','orders id'], as_index = False)[['weight charge']].sum(),
                    how = 'right',
                    on = ['shipment number','orders id'])

# find and fix shipment number dupes
shipment_number_dupes = dhl_clean.copy()
shipment_number_dupes = shipment_number_dupes[shipment_number_dupes['shipment number'].isin(shipment_number_dupes['shipment number'][shipment_number_dupes['shipment number'].duplicated()].tolist())].sort_values('shipment number')

def shipment_number_dupe_clean(x):
    if x['shipment number'] in shipment_number_dupes['shipment number'].tolist():
        return 0
    else:
        return x['orders id']

dhl_clean['orders id'] = dhl_clean.apply(shipment_number_dupe_clean, axis = 1)

# find and fix orders id dupes
# each set of orders id's have different shipment numbers
# we replace the different shipment numbers with a single shipment number
# we keep track of them to check our work at the end
orders_id_dupes = dhl_clean.copy()
orders_id_dupes = orders_id_dupes[orders_id_dupes['orders id'].isin(orders_id_dupes['orders id'][(orders_id_dupes['orders id'].duplicated()) & (orders_id_dupes['orders id'] != 0)]).tolist()].sort_values('orders id')

dhl_clean = dhl_clean.groupby(['shipment number','orders id'], as_index = False).sum()

def orders_id_dupe_clean(x):
    if x['orders id'] in orders_id_dupes['orders id'].tolist():
        df = orders_id_dupes[orders_id_dupes['orders id'] == x['orders id']]
        return df['shipment number'].values[0]
    else:
        return x['shipment number']
    
dhl_clean['shipment number'] = dhl_clean.apply(orders_id_dupe_clean, axis = 1)

excluded_shipment_numbers = list(set(orders_id_dupes['shipment number'].tolist()) - set(dhl_clean['shipment number'][dhl_clean['orders id'].duplicated()]))

# now we have created a one-to-one between shipment numbers and orders id
dhl_clean = dhl_clean.groupby(['shipment number','orders id'], as_index = False).sum()

# check it out
print('')
print('count of unique shipment numbers in dhl_super_raw: ', len(dhl_super_raw['shipment number'].unique()))
print('count of unique shipment numbers in dhl_clean after dupe clean up: ', len(dhl_clean['shipment number'].unique()))
print('count of shipment numbers we excluded', len(excluded_shipment_numbers))
print('so {} total unique shipment numbers after dupe clean up'.format(len(dhl_clean['shipment number'].unique()) + len(excluded_shipment_numbers)))

if len(dhl_super_raw['shipment number'].unique()) == len(dhl_clean['shipment number'].unique()) + len(excluded_shipment_numbers):
    print('your dupe clean up worked!')
else:
    print('your dupe clean up DID NOT work')
       
# clean columns
dhl_clean['ddp charge'] = dhl_clean[['import / export duties',
                                         'import export duties',
                                         'import export taxes',
                                         'import/ export duties']].sum(1)
dhl_clean.drop(['import / export duties',
                 'import export duties',
                 'import export taxes',
                 'import/ export duties'], 1, inplace = True)

dhl_clean['shipment value protect'] = dhl_clean[['shipment value protection']].sum(1)
dhl_clean.drop(['shipment value protection'], 1, inplace = True)

dhl_clean['licenses and permits'] = dhl_clean[['obtaining permits & licenses',
                                               'obtaining permits &amp; licenses']].sum(1)
dhl_clean.drop(['obtaining permits & licenses',
                'obtaining permits &amp; licenses',], 1, inplace = True)    

# add in weight charge

# create new columns
dhl_clean['ship charge'] = dhl_clean.iloc[:, 2:].sum(1) - dhl_clean['ddp charge']
dhl_clean['service'] = 'dhl'

# check out the totals
print('')
print('dhl_super_raw total: ', dhl_super_raw['total charge'].sum())
print('dhl_clean total: ', dhl_clean['ship charge'].sum() + dhl_clean['ddp charge'].sum())

if np.abs(dhl_super_raw['total charge'].sum() - (dhl_clean['ship charge'].sum() + dhl_clean['ddp charge'].sum())) < 1.00:
    print('total charge from dhl_super_raw MATCHES total charge from dhl_clean')
else:
    print('your totals DON\'T match!!')

# add other data
dhl_clean = pd.merge(dhl_clean,
                    dhl_super_raw.groupby(['shipment number','senders name','senders city','senders country'], as_index = False)[['shipment date']].max(),
                    how = 'left',
                    on = 'shipment number')
dhl_clean['senders name'] = [x.lower() for x in dhl_clean['senders name']]
dhl_clean['senders city'] = [x.lower() for x in dhl_clean['senders city']]
dhl_clean['senders country'] = [x.lower() for x in dhl_clean['senders country']]    

dhl_clean['incoming/outgoing'] = np.where(dhl_clean['orders id'].isin(oids['orders id'].tolist()), 'outgoing','incoming')

dhl_clean['shipment date'] = pd.to_datetime(dhl_clean['shipment date'])
dhl_clean['shipment year and month'] = pd.to_datetime(dhl_clean['shipment date'].dt.year.map(str) + '-' + dhl_clean['shipment date'].dt.month.map(str))
dhl_clean['shipment year and month'] = [str(x)[:7] for x in dhl_clean['shipment year and month']]

dhl_clean['shipment number'] = pd.to_numeric(dhl_clean['shipment number'], downcast = 'integer')
dhl_clean.drop_duplicates('shipment number', keep = 'first',inplace = True)

dhl_main = dhl_clean.copy()    
```

### Get DHL data

```python
dhl_ddp_oids = pd.read_sql(
'''
SELECT
orders_id AS 'orders id'
FROM orders_total
WHERE class = 'ot_ddp'
AND orders_id IN 
(SELECT
orders_id
FROM orders
WHERE shipping_module_code = 'dhlexpress')
''', db)

dhl_shipping = pd.read_sql(
'''
SELECT
A.orders_id AS 'orders id',
A.value AS 'ada ship charge',
IFNULL(B.value, 0) AS 'ada ddp charge'
FROM

(SELECT
orders_id,
value
FROM orders_total
WHERE class = 'ot_shipping'
AND orders_id IN '''+ str(tuple(dhl_clean['orders id'][dhl_clean['incoming/outgoing'] == 'outgoing'])) +''') A

LEFT JOIN

(SELECT
orders_id,
value
FROM orders_total
WHERE class = 'ot_ddp') B

ON A.orders_id = B.orders_id
''', db)

dhl_shipping['ada total charge'] = dhl_shipping[['ada ship charge','ada ddp charge']].sum(1)

dhl_delivery = pd.read_sql(
'''
SELECT
orders_id AS 'orders id',
LOWER (delivery_city) AS 'delivery city',
LOWER(delivery_country) AS'delivery country'
FROM orders
WHERE orders_id IN '''+ str(tuple(dhl_clean['orders id'][dhl_clean['incoming/outgoing'] == 'outgoing'])) +'''
''', db)
```

```python
dhl_shipping = pd.merge(dhl_shipping,
                dhl_clean[['shipment date','orders id','ship charge','ddp charge','incoming/outgoing']],
                how = 'left',
                on = 'orders id').merge(dhl_delivery,
                                       how = 'left',
                                       on = 'orders id')

dhl_shipping['ship profit'] = dhl_shipping[['ada ship charge','ada ddp charge']].sum(1) - dhl_shipping[['ship charge','ddp charge']].sum(1)
dhl_shipping['ddp difference'] = dhl_shipping['ada ddp charge'] - dhl_shipping['ddp charge']
dhl_shipping['did we charge ddp'] = np.where(dhl_shipping['orders id'].isin(dhl_ddp_oids['orders id'].tolist()), 'yes','no')

dhl_date_end = str(dhl_shipping['shipment date'].max().date())
dhl_date_start = str((pd.to_datetime(dhl_date_end) - pd.DateOffset(weeks = 2)).date())
```

```python
# if we charge a DDP then we want the corresponding DHL billing line
# if the corresponding DHL billing line equals zero then we don't have this data yet so exclude the order
# we also want all lines where we did not charge an ddp
dhl_shipping = dhl_shipping[((dhl_shipping['did we charge ddp'] == 'yes') & (dhl_shipping['ddp charge'] != 0))
                            | ((dhl_shipping['did we charge ddp'] == 'no'))]

dhl_percent = dhl_shipping.copy()

dhl_shipping = dhl_shipping[dhl_shipping['shipment date'].between(dhl_date_start, dhl_date_end)]
```

# Order counts and order value per service

```python
count_date_end = count_date_end
count_date_in_between = str((pd.to_datetime(count_date_end) - pd.DateOffset(weeks = 1)).date())
count_date_start = str((pd.to_datetime(count_date_end) - pd.DateOffset(weeks = 2)).date())
```

```python
print(count_date_start)
print(count_date_in_between)
print(count_date_end)
```

```python
summary = pd.read_sql(
'''
SELECT
A.service,
A.recent_week_order_count AS 'order count ending '''+ count_date_in_between +''' ', 
A.recent_week_order_subtotal AS 'order subtotal ending '''+ count_date_in_between +''' ',
B.week_before_order_count AS 'order count ending '''+ count_date_end +''' ',
B.week_before_order_subtotal AS 'order subtotal ending '''+ count_date_end +''' '

FROM

(SELECT
CASE 
WHEN o.shipping_module_code = 'upsxml' THEN 'ups'
WHEN o.shipping_module_code = 'usps' THEN 'usps'
WHEN o.shipping_module_code = 'dhlexpress' THEN 'dhl'
END AS service,
COUNT(o.orders_id) AS recent_week_order_count,
SUM(ot.value) AS recent_week_order_subtotal
FROM orders o
JOIN orders_total ot ON o.orders_id = ot.orders_id
AND ot.class = 'ot_subtotal'
WHERE o.shipping_module_code IN ('upsxml','usps', 'dhlexpress')
AND DATE(o.date_purchased) >= ' '''+ count_date_start +''' ' 
AND DATE(o.date_purchased) < ' '''+ count_date_in_between +''' '
GROUP BY service) A

JOIN

(SELECT
CASE 
WHEN o.shipping_module_code = 'upsxml' THEN 'ups'
WHEN o.shipping_module_code = 'usps' THEN 'usps'
WHEN o.shipping_module_code = 'dhlexpress' THEN 'dhl'
END AS service,
COUNT(o.orders_id) AS week_before_order_count,
SUM(ot.value) AS week_before_order_subtotal
FROM orders o
JOIN orders_total ot ON o.orders_id = ot.orders_id
AND ot.class = 'ot_subtotal'
WHERE o.shipping_module_code IN ('upsxml','usps','dhlexpress')
AND DATE(o.date_purchased) >= ' '''+ count_date_in_between +''' ' 
AND DATE(o.date_purchased) < ' '''+ count_date_end +''' ' 
GROUP BY service) B

ON A.service = B.service
''', db)

summary.sort_values('service', inplace = True)
```

```python
counts = summary.copy()
counts = counts[counts.columns[counts.columns.str.contains('count|service')]]
counts['% change'] = counts.iloc[:, 2]/counts.iloc[:, 1] - 1
counts['service'] = [x.upper() for x in counts['service']]
counts.format_([0,'n0','n0','p2'])
```

```python
subtotals = summary.copy()
subtotals = subtotals[subtotals.columns[subtotals.columns.str.contains('subtotal|service')]]
subtotals['% change'] = subtotals.iloc[:, 2]/subtotals.iloc[:, 1] - 1
subtotals['service'] = [x.upper() for x in counts['service']]
subtotals.format_([0, 'm0','m0','p2'])
```
